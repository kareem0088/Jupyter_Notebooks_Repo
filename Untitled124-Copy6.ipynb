```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import RobustScaler
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm
from xgboost import XGBClassifier, plot_importance

# ========================
# Ø¯Ø§Ù„Ø© Ø¥Ù†Ø´Ø§Ø¡ Ù†Ø§ÙØ°Ø© Ø²Ù…Ù†ÙŠØ©
# ========================
def create_sliding_window(X, y, window_size):
    X_windows, y_windows = [], []
    for i in range(len(X) - window_size + 1):
        X_windows.append(X[i:i + window_size])
        y_windows.append(y[i + window_size - 1])  # Ø¢Ø®Ø± Ø¹Ù†ØµØ± Ù‡Ùˆ Ø§Ù„Ù‡Ø¯Ù
    return np.array(X_windows), np.array(y_windows)

# ========================
# 1. ØªØ­Ù…ÙŠÙ„ ÙˆØªØ¬Ù‡ÙŠØ² Ø§Ù„Ø¨ÙŠØ§Ù†Ø§Øª
# ========================
# df_train = pd.read_csv('your_train_data.csv')
# df_test = pd.read_csv('your_test_data.csv')

features_train = df_train.drop(columns=['Signal','date','timestamp','future_return'])
features_test = df_test.drop(columns=['Signal','date','timestamp','future_return'])

target_train = df_train['Signal']
target_test = df_test['Signal']

# ========================
# 2. Ø¥Ù†Ø´Ø§Ø¡ Ø§Ù„Ù†ÙˆØ§ÙØ° Ø§Ù„Ø²Ù…Ù†ÙŠØ©
# ========================
window_size = 50
X_train_raw, y_train_raw = create_sliding_window(features_train.values, target_train.values, window_size)
X_test_raw, y_test_raw = create_sliding_window(features_test.values, target_test.values, window_size)

# ========================
# 3. ØªØ·Ø¨ÙŠØ¹ Ø§Ù„Ø¨ÙŠØ§Ù†Ø§Øª
# ========================
scaler = RobustScaler()
X_train_scaled = scaler.fit_transform(X_train_raw.reshape(-1, X_train_raw.shape[-1])).reshape(X_train_raw.shape)
X_test_scaled = scaler.transform(X_test_raw.reshape(-1, X_test_raw.shape[-1])).reshape(X_test_raw.shape)

# Flatten Ø¹Ø´Ø§Ù† XGBoost
X_train_flat = X_train_scaled.reshape(X_train_scaled.shape[0], -1)
X_test_flat = X_test_scaled.reshape(X_test_scaled.shape[0], -1)

print(f"ğŸ”¹ Ø´ÙƒÙ„ Ø¨ÙŠØ§Ù†Ø§Øª Ø§Ù„ØªØ¯Ø±ÙŠØ¨: {X_train_flat.shape}")
print(f"ğŸ”¹ Ø´ÙƒÙ„ Ø¨ÙŠØ§Ù†Ø§Øª Ø§Ù„Ø§Ø®ØªØ¨Ø§Ø±: {X_test_flat.shape}")

# ========================
# 4. Ø¨Ù†Ø§Ø¡ Ù†Ù…ÙˆØ°Ø¬ XGBoost
# ========================
model = XGBClassifier(
    n_estimators=300,
    max_depth=8,
    learning_rate=0.05,
    subsample=0.8,
    colsample_bytree=0.8,
    objective="multi:softprob",  # softprob Ù„Ù„Ù€ predict_proba
    num_class=3,
    random_state=42,
    n_jobs=-1,
    eval_metric="mlogloss"
)

# ========================
# 5. ØªØ¯Ø±ÙŠØ¨ Ø§Ù„Ù†Ù…ÙˆØ°Ø¬
# ========================
print("ğŸš€ Training XGBoost...")
model.fit(X_train_flat, y_train_raw,
          eval_set=[(X_test_flat, y_test_raw)],
          verbose=False)

# ========================
# 6. Ø§Ù„ØªÙ‚ÙŠÙŠÙ… Ø§Ù„ØªØ¯Ø±ÙŠØ¬ÙŠ Ù…Ø¹ Threshold
# ========================
y_true, y_pred = [], []
threshold = 0.5
print("ğŸ” Ø¬Ø§Ø±ÙŠ Ø§Ù„ØªÙ‚ÙŠÙŠÙ… Ø§Ù„ØªØ¯Ø±ÙŠØ¬ÙŠ Ù…Ø¹ Threshold...")
for i in tqdm(range(len(X_test_flat))):
    sample_x = X_test_flat[i].reshape(1, -1)
    true_class = y_test_raw[i]
    proba = model.predict_proba(sample_x)[0]

    if proba[1] >= threshold:
        predicted_class = 1
    elif proba[2] >= threshold:
        predicted_class = 2
    else:
        predicted_class = 0

    y_true.append(true_class)
    y_pred.append(predicted_class)

walk_acc = accuracy_score(y_true, y_pred)
print(f"ğŸ§  Walking Forward Accuracy (Threshold 0.5): {walk_acc:.4f}")

# ========================
# 7. Ø§Ù„ØªÙ‚ÙŠÙŠÙ… Ø¹Ù„Ù‰ ÙƒØ§Ù…Ù„ Ø¨ÙŠØ§Ù†Ø§Øª Ø§Ù„Ø§Ø®ØªØ¨Ø§Ø±
# ========================
proba_test = model.predict_proba(X_test_flat)
test_pred = []
for p in proba_test:
    if p[1] >= threshold:
        test_pred.append(1)
    elif p[2] >= threshold:
        test_pred.append(2)
    else:
        test_pred.append(0)

test_acc = accuracy_score(y_test_raw, test_pred)
print(f"ğŸ“Š Test Accuracy (Threshold 0.5): {test_acc:.4f}")

# ========================
# 8. Ù…ØµÙÙˆÙØ© Ø§Ù„Ø§Ù„ØªØ¨Ø§Ø³
# ========================
cm = confusion_matrix(y_test_raw, test_pred)

plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=np.unique(y_train_raw),
            yticklabels=np.unique(y_train_raw))
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix - XGBoost with Threshold")
plt.show()

# ========================
# 9. ØªÙ‚Ø±ÙŠØ± Ø§Ù„Ø£Ø¯Ø§Ø¡
# ========================
print("\nğŸ“ˆ Classification Report:")
print("="*50)
print(classification_report(y_test_raw, test_pred, digits=3))

# ========================
# 10. Ø£Ù‡Ù…ÙŠØ© Ø§Ù„Ù…ÙŠØ²Ø§Øª (Feature Importance)
# ========================
plt.figure(figsize=(10, 6))
plot_importance(model, max_num_features=15, importance_type="gain")
plt.title("Top 15 Feature Importances - XGBoost")
plt.show()

# ========================
# 11. Ø¥Ø­ØµØ§Ø¦ÙŠØ§Øª Ø¥Ø¶Ø§ÙÙŠØ©
# ========================
print(f"\nğŸ“Š Ø¥Ø­ØµØ§Ø¦ÙŠØ§Øª Ø¥Ø¶Ø§ÙÙŠØ©:")
print(f"Ø¹Ø¯Ø¯ Ø¹ÙŠÙ†Ø§Øª Ø§Ù„ØªØ¯Ø±ÙŠØ¨: {len(X_train_flat)}")
print(f"Ø¹Ø¯Ø¯ Ø¹ÙŠÙ†Ø§Øª Ø§Ù„Ø§Ø®ØªØ¨Ø§Ø±: {len(X_test_flat)}")
print(f"Ø¹Ø¯Ø¯ Ø§Ù„Ù…ÙŠØ²Ø§Øª Ø¨Ø¹Ø¯ Flatten: {X_train_flat.shape[1]}")
print(f"Ø­Ø¬Ù… Ø§Ù„Ù†Ø§ÙØ°Ø© Ø§Ù„Ø²Ù…Ù†ÙŠØ©: {window_size}")
print(f"Ø¹Ø¯Ø¯ Ø§Ù„Ø£Ø´Ø¬Ø§Ø±: {model.n_estimators}, Ø¹Ù…Ù‚ Ø§Ù„Ø´Ø¬Ø±Ø©: {model.max_depth}")
```
