```python
import pandas as pd


df = df_train

# التأكد من أن العمود datetime
df['timestamp'] = pd.to_datetime(df['timestamp'])
df['date'] = df['timestamp'].dt.date

# دوال لحساب أعلى/أقل سعر حتى الصف الحالي
def calc_high_so_far(group):
    group['high_today'] = group['high'].cummax()
    idx_high = group['high'].cummax().eq(group['high'])
    group['vol_at_high_today'] = group['volume'].where(idx_high).ffill()
    return group

def calc_low_so_far(group):
    group['low_today'] = group['low'].cummin()
    idx_low = group['low'].cummin().eq(group['low'])
    group['vol_at_low_today'] = group['volume'].where(idx_low).ffill()
    return group

# نخزن عمود date مؤقت
dates = df['date']

# نستبعد date من groupby.apply لتفادي التحذير
df = df.drop(columns=['date'])
df = pd.concat([
    calc_high_so_far(group) 
    for _, group in pd.concat([dates, df], axis=1).groupby(dates)
], axis=0)

df = pd.concat([
    calc_low_so_far(group) 
    for _, group in df.groupby(dates)
], axis=0)

# نرجع العمود date
df['date'] = dates

# حساب بيانات اليوم السابق بدون تسريب
df['high_prev_day'] = df.groupby('date')['high_today'].shift(1)
df['vol_at_high_prev_day'] = df.groupby('date')['vol_at_high_today'].shift(1)
df['low_prev_day'] = df.groupby('date')['low_today'].shift(1)
df['vol_at_low_prev_day'] = df.groupby('date')['vol_at_low_today'].shift(1)

import numpy as np

# Relative Volume (RVOL)
df['RVOL'] = df['volume'] / df['volume'].rolling(window=20).mean()

# Volume Rate of Change (VROC)
df['VROC'] = (df['volume'] - df['volume'].shift(5)) / df['volume'].shift(5)

# Volume Weighted Spread
df['spread_vol'] = (df['high'] - df['low']) * df['RVOL']

# VWAP Deviation (بالانحراف المعياري)
vwap_std = df['vwap'].rolling(window=20).std()
df['vwap_dev'] = (df['close'] - df['vwap']) / vwap_std

# Volume per Transaction
df['vol_per_tx'] = df['volume'] / df['transactions']


# Accumulation/Distribution based on VWAP
df['accum_dist_vwap'] = np.where(df['close'] > df['vwap'], df['volume'], -df['volume'])
df['accum_dist_vwap'] = df['accum_dist_vwap'].cumsum()

# Volume Spike Detector → بدون تسريب
df['vol_spike'] = (
    df['volume'] > (df['volume'].rolling(window=10).mean().shift(1) * 2)
).astype(int)

# Volume Climax Bar → بدون تسريب وبـ tolerance
max_vol_threshold = df['volume'].rolling(window=20).max().shift(1) * 0.95
wide_range_threshold = (df['high'] - df['low']).rolling(window=20).mean().shift(1)

df['climax_bar'] = (
    (df['volume'] >= max_vol_threshold) & 
    ((df['high'] - df['low']) > wide_range_threshold)
).astype(int)

import numpy as np

denominator = (df['high'].rolling(5).max().shift(1) - df['low'].rolling(5).min().shift(1))
denominator = denominator.replace(0, np.nan)  # استبدال المقام إذا كان صفر بـ NaN

df['micro_trend_slope'] = (
    (df['close'] - df['close'].shift(5)) / denominator
)


# Price Change per Second (لو عندك بيانات وقت بالثواني)
# تحويل timestamp إلى datetime
df['timestamp'] = pd.to_datetime(df['timestamp'])
df['seconds_diff'] = (df['timestamp'] - df['timestamp'].shift(1)).dt.total_seconds()
df['price_change_per_sec'] = (df['close'] - df['close'].shift(1)) / df['seconds_diff']

# Position in Range (آخر 20 شمعة) - بدون تسريب
window_range = 20
low_min = df['low'].rolling(window_range).min().shift(1)
high_max = df['high'].rolling(window_range).max().shift(1)
df['pos_in_range'] = (df['close'] - low_min) / (high_max - low_min)

# Micro Support/Resistance Distance - بدون تسريب
df['dist_to_recent_high'] = df['high'].rolling(window_range).max().shift(1) - df['close']
df['dist_to_recent_low'] = df['close'] - df['low'].rolling(window_range).min().shift(1)

# Candle Body-to-Range Ratio (ده طبيعي مفيهوش تسريب)
df['body_to_range'] = abs(df['close'] - df['open']) / (df['high'] - df['low'])

# Transaction Density (طبيعي بدون تسريب)
df['tx_density'] = df['transactions'] / df['volume']

# VWAP Pullback Strength - بدون تسريب
df['vwap_pullback'] = (df['vwap'].shift(1) - df['close'].shift(1)) / df['vwap'].shift(1)

# Volume Shock - بدون تسريب
vol_avg = df['volume'].rolling(20).mean().shift(1)
df['vol_shock'] = (df['volume'] > (vol_avg * 2)).astype(int)

# Price Reaction Next (ده Label للتنبؤ مش مؤشر)
df['price_reaction_next'] = df['close'].shift(-1) - df['close']

# حساب الفرق بالدقائق بين الصف الحالي والسابق
df['minutes_diff'] = (df['timestamp'] - df['timestamp'].shift(1)).dt.total_seconds() / 60

# الفرق بين سعر الفتح الحالي وسعر الإغلاق السابق
df['open_vs_prev_close'] = df['open'] - df['close'].shift(1)

df['vwap_distance'] = df['close'] - df['vwap']

df['vol_change_pct'] = (df['volume'] - df['volume'].shift(1)) / df['volume'].shift(1) * 100

# استخراج الساعة
df['hour'] = df['timestamp'].dt.hour

# استخراج الدقائق
df['minute'] = df['timestamp'].dt.minute

# تحديد الربع من الساعة
def get_quarter(minute):
    if minute < 15:
        return 1
    elif minute < 30:
        return 2
    elif minute < 45:
        return 3
    else:
        return 4

df['quarter_hour'] = df['minute'].apply(get_quarter)

df.dropna(inplace=True)


n_future = 3  # عدد الشموع التي نريد التنبؤ بها

# تأكد من وجود الأعمدة التالية: 'close', 'volume', 'trades'
df['MA9'] = df['close'].rolling(window=9).mean()
df['volume_ma'] = df['volume'].rolling(window=9).mean()
df['volume_ratio'] = df['volume'] / df['volume_ma']  # قوة الحجم الحالي مقارنة بمتوسطه

# عدد الصفقات النسبية (لو عندك عمود مثلا df['trades'])
df['trades_ma'] = df['transactions'].rolling(window=9).mean()
df['trades_ratio'] = df['transactions'] / df['trades_ma']  # حجم النشاط مقارنة بالمعدل

# العائد المستقبلي بناءً على أعلى أو إغلاق السعر بعد n شموع
df['future_return'] = np.log(df['close'].shift(-n_future) / df['close'])

# توليد إشارات ذكية بناءً على سلوك السوق

threshold = df['future_return'].std() * 0.5  # عتبة ديناميكية

df['Signal'] = 0  # افتراضي = محايد

# إشارة شراء:
df.loc[
    (df['future_return'] > threshold) &
    (df['volume_ratio'] > 1.2) &  # حجم تداول مرتفع
    (df['trades_ratio'] > 1.1) &  # نشاط مرتفع
    (df['close'] > df['MA9']),   # السعر فوق المتوسط
    'Signal'
] = 2  # شراء

# إشارة بيع:
df.loc[
    (df['future_return'] < -threshold) &
    (df['volume_ratio'] > 1.2) &
    (df['trades_ratio'] > 1.1) &
    (df['close'] < df['MA9']),
    'Signal'
] = 1  # بيع


df.dropna(inplace=True)
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, RobustScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras import regularizers
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
from tensorflow.keras.optimizers import Adam
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import matplotlib.pyplot as plt
import seaborn as sns
import tqdm
# ========================
# إضافة دالة create_sliding_window المفقودة
# ========================
def create_sliding_window(X, y, window_size):
    """
    إنشاء نوافذ زمنية منزلقة من البيانات
    """
    X_windows = []
    y_windows = []
    
    for i in range(len(X) - window_size + 1):
        X_windows.append(X[i:i + window_size])
        y_windows.append(y[i + window_size - 1])  # آخر هدف في النافذة
    
    return np.array(X_windows), np.array(y_windows)

# ========================
# 1. تحميل وتجهيز البيانات
# ========================
# تأكد من تحميل البيانات أولاً
# df_train = pd.read_csv('your_train_data.csv')
# df_test = pd.read_csv('your_test_data.csv')

# الميزات
features_train = df_train.drop(columns=['Signal','date','timestamp','future_return','price_reaction_next'])
features_test = df_test.drop(columns=['Signal','date','timestamp','future_return','price_reaction_next'])

# الأهداف
target_train = df_train['Signal']
target_test = df_test['Signal']

# إنشاء النوافذ الزمنية
window_size = 25
X_train_raw, y_train_raw = create_sliding_window(features_train.values, target_train.values, window_size)
X_test_raw, y_test_raw = create_sliding_window(features_test.values, target_test.values, window_size)

# تطبيق التطبيع
scaler = RobustScaler()

# تطبيع البيانات
X_train_scaled = scaler.fit_transform(X_train_raw.reshape(-1, X_train_raw.shape[-1])).reshape(X_train_raw.shape)
X_test_scaled = scaler.transform(X_test_raw.reshape(-1, X_test_raw.shape[-1])).reshape(X_test_raw.shape)

# تحويل الأهداف إلى categorical
y_train_cat = to_categorical(y_train_raw, num_classes=3)
y_test_cat = to_categorical(y_test_raw, num_classes=3)

print(f"شكل بيانات التدريب: {X_train_scaled.shape}")
print(f"شكل بيانات الاختبار: {X_test_scaled.shape}")
print(f"شكل أهداف التدريب: {y_train_cat.shape}")
print(f"شكل أهداف الاختبار: {y_test_cat.shape}")

# ========================
# 5. بناء نموذج LSTM
# ========================
model = Sequential()
model.add(LSTM(
    64,
    input_shape=(X_train_scaled.shape[1], X_train_scaled.shape[2]),
    return_sequences=False,
    dropout=0.2,
    kernel_regularizer=regularizers.l2(0.01)
))
model.add(Dense(32, activation='relu', kernel_regularizer=regularizers.l2(0.01)))
model.add(Dropout(0.2))
model.add(Dense(3, activation='softmax'))

# ========================
# 5.1 تحديد نسبة التعلم
# ========================
optimizer = Adam(learning_rate=0.001)
model.compile(optimizer=optimizer, loss='categorical_crossentropy', metrics=['accuracy'])

# عرض ملخص النموذج
model.summary()

# ========================
# 6.1 إعداد التوقف المبكر وحفظ أفضل نموذج
# ========================
early_stop = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)
checkpoint = ModelCheckpoint('best_model.keras', monitor='val_loss', save_best_only=True)

# ========================
# 6. تدريب النموذج
# ========================
history = model.fit(
    X_train_scaled, y_train_cat,  # استخدام المتغيرات الصحيحة
    epochs=20,
    batch_size=100,
    validation_data=(X_test_scaled, y_test_cat),  # استخدام المتغيرات الصحيحة
    callbacks=[early_stop, checkpoint],
    verbose=1
)

# ========================
# 7. تقييم تدريجي خطوة بخطوة (Walking Forward Evaluation)
# ========================
y_true = []
y_pred = []

print("جاري التقييم التدريجي...")
for i in  tqdm.tqdm(range(len(X_test_scaled))):
    sample_x = X_test_scaled[i].reshape(1, X_test_scaled.shape[1], X_test_scaled.shape[2])
    true_class = np.argmax(y_test_cat[i])  # استخدام y_test_cat
    prediction = model.predict(sample_x, verbose=0)
    predicted_class = np.argmax(prediction)
    
    y_true.append(true_class)
    y_pred.append(predicted_class)

walk_acc = accuracy_score(y_true, y_pred)
print(f"🧠 Walking Forward Accuracy: {walk_acc:.4f}")

# ==================================================================================================================
# تقييم النموذج على اختبار حقيقي
# ==================================================================================================================
test_loss, test_accuracy = model.evaluate(X_test_scaled, y_test_cat, verbose=0)  # تصحيح المتغيرات
print(f"📊 Test Loss: {test_loss:.4f}")
print(f"🎯 Test Accuracy: {test_accuracy:.4f}")

# ========================
# رسم منحنى الأداء الكامل
# ========================
epochs_range = range(len(history.history['loss']))

plt.figure(figsize=(12, 6))

# رسم الخسارة
plt.subplot(1, 2, 1)
plt.plot(epochs_range, history.history['loss'], label='Train Loss', linestyle='--')
plt.plot(epochs_range, history.history['val_loss'], label='Validation Loss', linestyle='--')
plt.title('Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)

# رسم الدقة
plt.subplot(1, 2, 2)
plt.plot(epochs_range, history.history['accuracy'], label='Train Accuracy', linestyle='-')
plt.plot(epochs_range, history.history['val_accuracy'], label='Validation Accuracy', linestyle='-')
plt.title('Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()

# ===========
# عمل التنبؤات وتقييم شامل
# ===========
y_pred_probs = model.predict(X_test_scaled)
y_pred = np.argmax(y_pred_probs, axis=1)
y_true = np.argmax(y_test_cat, axis=1)

# ===========
# مصفوفة الالتباس
# ===========
cm = confusion_matrix(y_true, y_pred)

plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
            xticklabels=[0, 1, 2], yticklabels=[0, 1, 2])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()

# ===========
# تقرير الأداء الكامل
# ===========
print("\n📈 Classification Report:")
print("=" * 50)
print(classification_report(y_true, y_pred, digits=3))

# ===========
# إحصائيات إضافية
# ===========
print(f"\n📊 إحصائيات إضافية:")
print(f"عدد عينات التدريب: {len(X_train_scaled)}")
print(f"عدد عينات الاختبار: {len(X_test_scaled)}")
print(f"عدد الميزات: {X_train_scaled.shape[2]}")
print(f"حجم النافزة الزمنية: {window_size}")
print(f"عدد الحقب المكتملة: {len(history.history['loss'])}")
```
