```python
import joblib
from tensorflow import keras
import pandas as pd

# استدعاء النموذج المحفوظ
model = keras.models.load_model(r'C:\Users\Access\Downloads\my_dnn_model.h5')

# استدعاء المقياس
scaler = joblib.load(r'C:\Users\Access\Downloads\scaler.pkl')
```

```python
# تحجيم البيانات الجديدة
data_scaled = scaler.transform(data)

# إجراء التنبؤات
pr = model.predict(data_scaled)
```
