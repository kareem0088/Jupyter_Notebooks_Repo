```python
import pandas as pd 

df = pd.read_csv(r"C:\Users\Access\Documents\AI\gold_heiken_ashi_2020_2025_features.csv")

df = df.drop(columns=['Unnamed: 0'])

df.columns
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras import regularizers
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
from tensorflow.keras.optimizers import Adam
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt

# ========================
# 1. ÿ™ÿ≠ŸÖŸäŸÑ Ÿàÿ™ÿ¨ŸáŸäÿ≤ ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™
# ========================
# df = df_

df['Gmt time'] = pd.to_datetime(df['Gmt time'])
df = df.drop(columns=['Gmt time'])  # <-- ÿ•ÿ∞ÿß ŸÉŸÜÿ™ ŸÑÿß ÿ™ÿ≥ÿ™ÿÆÿØŸÖ ÿßŸÑŸàŸÇÿ™ ŸÉŸÖŸèÿØÿÆŸÑ

features = df.drop(columns=['Signal', 'date'])  # <-- ÿ∫ŸäŸëÿ± 'Signal' ŸÑÿßÿ≥ŸÖ ÿπŸÖŸàÿØ ÿßŸÑŸáÿØŸÅ ŸÑŸà ŸÖÿÆÿ™ŸÑŸÅ
target = df['Signal']

# ========================
# 2. ŸÖŸÇŸäÿßÿ≥ ÿßŸÑŸÖŸäÿ≤ÿßÿ™
# ========================
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)

# ========================
# 3. ÿ•ŸÜÿ¥ÿßÿ° ÿßŸÑŸÜŸàÿßŸÅÿ∞ ÿßŸÑŸÖŸÜÿ≤ŸÑŸÇÿ©
# ========================
def create_sliding_window(X, y, window_size=50):
    Xs, ys = [], []
    for i in range(len(X) - window_size):
        if y[i + window_size] != 0:  # ŸÜÿ≠ÿ™ŸÅÿ∏ ŸÅŸÇÿ∑ ÿ®ÿ•ÿ¥ÿßÿ±ÿßÿ™ 1 Ÿà 2
            Xs.append(X[i:i+window_size])
            ys.append(y[i + window_size])
    return np.array(Xs), np.array(ys)

window_size = 50
X, y = create_sliding_window(features_scaled, target.values, window_size=window_size)
y_cat = to_categorical(y, num_classes=3)

# ========================
# 4. ÿ™ŸÇÿ≥ŸäŸÖ ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™
# ========================
X_train, X_test, y_train, y_test = train_test_split(X, y_cat, test_size=0.2, shuffle=False)

# ========================
# 5. ÿ®ŸÜÿßÿ° ŸÜŸÖŸàÿ∞ÿ¨ LSTM
# ========================

model = Sequential()
model.add(LSTM(
    16,
    input_shape=(X.shape[1], X.shape[2]),
    return_sequences=False,
    dropout=0.2,  # ŸÑÿ≠ÿ∞ŸÅ ÿ®ÿπÿ∂ ÿßŸÑŸÖÿØÿÆŸÑÿßÿ™ ÿπÿ¥Ÿàÿßÿ¶ŸäŸãÿß
    kernel_regularizer=regularizers.l2(0.01)  # ÿ™ŸÇŸÑŸäŸÑ ÿßŸÑÿ≠ŸÅÿ∏
))
model.add(Dense(8, activation='relu', kernel_regularizer=regularizers.l2(0.01)))
model.add(Dropout(0.2))  # ÿ®ÿπÿØ Dense
model.add(Dense(3, activation='softmax'))

# ========================
# 5.1 ÿ™ÿ≠ÿØŸäÿØ ŸÜÿ≥ÿ®ÿ© ÿßŸÑÿ™ÿπŸÑŸÖ
# ========================
optimizer = Adam(learning_rate=0.001)
model.compile(optimizer=optimizer, loss='categorical_crossentropy', metrics=['accuracy'])

# ========================
# 6.1 ÿ•ÿπÿØÿßÿØ ÿßŸÑÿ™ŸàŸÇŸÅ ÿßŸÑŸÖÿ®ŸÉÿ± Ÿàÿ≠ŸÅÿ∏ ÿ£ŸÅÿ∂ŸÑ ŸÜŸÖŸàÿ∞ÿ¨
# ========================
early_stop = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
checkpoint = ModelCheckpoint('best_model.keras', monitor='val_loss', save_best_only=True)

# ========================
# 6. ÿ™ÿØÿ±Ÿäÿ® ÿßŸÑŸÜŸÖŸàÿ∞ÿ¨
# ========================
history = model.fit(
    X_train, y_train,
    epochs=200,
    batch_size=100,
    validation_data=(X_test, y_test),
    callbacks=[early_stop, checkpoint]
)

# ========================
# 7. ÿ™ŸÇŸäŸäŸÖ ÿ™ÿØÿ±Ÿäÿ¨Ÿä ÿÆÿ∑Ÿàÿ© ÿ®ÿÆÿ∑Ÿàÿ© (Walking Forward Evaluation)
# ========================
y_true = []
y_pred = []

for i in range(len(X_test)):
    sample_x = X_test[i].reshape(1, X_test.shape[1], X_test.shape[2])
    true_class = np.argmax(y_test[i])
    prediction = model.predict(sample_x, verbose=0)
    predicted_class = np.argmax(prediction)
    
    y_true.append(true_class)
    y_pred.append(predicted_class)

walk_acc = accuracy_score(y_true, y_pred)
print(f"üß† Walking Forward Accuracy: {walk_acc:.2f}")

# ==================================================================================================================
# ÿ™ŸÇŸäŸäŸÖ ÿßŸÑŸÜŸÖŸàÿ∞ÿ¨ ÿπŸÑŸâ ÿßÿÆÿ™ÿ®ÿßÿ± ÿ≠ŸÇŸäŸÇŸä
# ==================================================================================================================
test_loss, test_accuracy = model.evaluate(X_test, y_test, verbose=0)

# ========================
#  ÿ±ÿ≥ŸÖ ŸÖŸÜÿ≠ŸÜŸâ ÿßŸÑÿ£ÿØÿßÿ° ÿßŸÑŸÉÿßŸÖŸÑ
# ========================
epochs_range = range(len(history.history['loss']))

plt.figure(figsize=(12, 6))

# ÿ±ÿ≥ŸÖ ÿßŸÑÿÆÿ≥ÿßÿ±ÿ©
plt.plot(epochs_range, history.history['loss'], label='Train Loss', linestyle='--')
plt.plot(epochs_range, history.history['val_loss'], label='Validation Loss', linestyle='--')

# ÿ±ÿ≥ŸÖ ÿßŸÑÿØŸÇÿ©
plt.plot(epochs_range, history.history['accuracy'], label='Train Accuracy', linestyle='-')
plt.plot(epochs_range, history.history['val_accuracy'], label='Validation Accuracy', linestyle='-')

plt.title('Training, Validation and Test Metrics')
plt.xlabel('Epoch')
plt.ylabel('Value')
plt.legend(loc='best')
plt.grid(True)
plt.tight_layout()
plt.show()

import numpy as np
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns
import matplotlib.pyplot as plt

# ===========
#  ÿπŸÖŸÑ ÿßŸÑÿ™ŸÜÿ®ÿ§ÿßÿ™
# ===========
y_pred_probs = model.predict(X_test)
y_pred = np.argmax(y_pred_probs, axis=1)
y_true = np.argmax(y_test, axis=1)

# ===========
#  ŸÖÿµŸÅŸàŸÅÿ© ÿßŸÑÿßŸÑÿ™ÿ®ÿßÿ≥
# ===========
cm = confusion_matrix(y_true, y_pred)

# ===========
#  ÿ±ÿ≥ŸÖ ÿßŸÑŸÖÿµŸÅŸàŸÅÿ©
# ===========
plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=[1, 2], yticklabels=[1, 2])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()

# ===========
#  ÿ™ŸÇÿ±Ÿäÿ± ÿßŸÑÿ£ÿØÿßÿ° ÿßŸÑŸÉÿßŸÖŸÑ
# ===========
print("Classification Report:\n")
print(classification_report(y_true, y_pred, digits=3))


```

```python
import numpy as np
import pandas as pd
import os
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
import matplotlib.pyplot as plt

# ========================
# 1. ÿ™ÿ≠ŸÖŸäŸÑ Ÿàÿ™ÿ¨ŸáŸäÿ≤ ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™
# ========================
# df = df_

# ÿ™ÿ≠ŸàŸäŸÑ ÿßŸÑŸàŸÇÿ™ ŸÖŸÜ String ÿ•ŸÑŸâ datetime
df['Gmt time'] = pd.to_datetime(df['Gmt time'])

# ÿ≠ÿ∞ŸÅ ÿßŸÑÿ£ÿπŸÖÿØÿ© ÿ∫Ÿäÿ± ÿßŸÑŸÖÿ∑ŸÑŸàÿ®ÿ©
df = df.drop(columns=['Gmt time'])  # <-- ÿ•ÿ∞ÿß ŸÉŸÜÿ™ ŸÑÿß ÿ™ÿ≥ÿ™ÿÆÿØŸÖ ÿßŸÑŸàŸÇÿ™ ŸÉŸÖŸèÿØÿÆŸÑ

# ÿßŸÑŸÖŸÖŸäÿ≤ÿßÿ™ ŸàÿßŸÑŸáÿØŸÅ
features = df.drop(columns=['Signal', 'date'])
target = df['Signal']

# ========================
# 2. ŸÖŸÇŸäÿßÿ≥ ÿßŸÑŸÖŸäÿ≤ÿßÿ™
# ========================
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)

# ========================
# 3. ÿ•ŸÜÿ¥ÿßÿ° ÿßŸÑŸÜŸàÿßŸÅÿ∞ ÿßŸÑŸÖŸÜÿ≤ŸÑŸÇÿ©
# ========================
def create_sliding_window(X, y, window_size=50):
    Xs, ys = [], []
    for i in range(len(X) - window_size):
        if y[i + window_size] != 0:  # ŸÜÿ≠ÿ™ŸÅÿ∏ ŸÅŸÇÿ∑ ÿ®ÿ•ÿ¥ÿßÿ±ÿßÿ™ 1 Ÿà 2
            Xs.append(X[i:i+window_size])
            ys.append(y[i + window_size])
    return np.array(Xs), np.array(ys)

window_size = 50
X, y = create_sliding_window(features_scaled, target.values, window_size=window_size)

# ÿ™ÿ≠ŸàŸäŸÑ ÿßŸÑŸáÿØŸÅ ÿ•ŸÑŸâ one-hot encoding
y_cat = to_categorical(y, num_classes=3)

# ========================
# 4. ÿ™ŸÇÿ≥ŸäŸÖ ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™
# ========================
X_train, X_test, y_train, y_test = train_test_split(X, y_cat, test_size=0.1, shuffle=False)

# ========================
# 5. ÿ®ŸÜÿßÿ° ŸÜŸÖŸàÿ∞ÿ¨ LSTM
# ========================
model = Sequential()
model.add(LSTM(64, input_shape=(X.shape[1], X.shape[2]), return_sequences=False))
model.add(Dense(32, activation='relu'))
model.add(Dense(3, activation='softmax'))  # ÿ™ÿµŸÜŸäŸÅ 3 ŸÅÿ¶ÿßÿ™

# ========================
# 6. ÿ•ÿπÿØÿßÿØ ŸÜÿ≥ÿ®ÿ© ÿßŸÑÿ™ÿπŸÑŸÖ ŸàÿßŸÑÿ™ÿ¨ŸÖŸäÿπ
# ========================
learning_rate = 0.001  # <-- ÿ∫ŸäŸëÿ± ÿßŸÑŸÇŸäŸÖÿ© ÿ•ÿ∞ÿß ÿ£ÿ±ÿØÿ™
optimizer = Adam(learning_rate=learning_rate)
model.compile(optimizer=optimizer, loss='categorical_crossentropy', metrics=['accuracy'])

# ========================
# 7. ÿßŸÑÿ™ŸàŸÇŸÅ ÿßŸÑŸÖÿ®ŸÉÿ± Ÿàÿ≠ŸÅÿ∏ ÿßŸÑŸÜŸÖŸàÿ∞ÿ¨
# ========================
early_stop = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

checkpoint_dir = "model_checkpoints"
os.makedirs(checkpoint_dir, exist_ok=True)

checkpoint_callback = ModelCheckpoint(
    filepath=os.path.join(checkpoint_dir, "epoch_{epoch:02d}.weights.h5"),  # ‚úÖ ÿ™ÿ∫ŸäŸäÿ± ÿßŸÑÿßŸÖÿ™ÿØÿßÿØ ŸáŸÜÿß
    save_weights_only=True,
    save_freq='epoch'
)

# ========================
# 8. ÿ™ÿØÿ±Ÿäÿ® ÿßŸÑŸÜŸÖŸàÿ∞ÿ¨
# ========================
epochs = 30  # <-- ÿ∫ŸäŸëÿ± ÿπÿØÿØ ÿßŸÑÿ•Ÿäÿ®ŸàŸÉ ŸÉŸÖÿß ÿ™ÿ±ŸäÿØ

history = model.fit(
    X_train, y_train,
    epochs=epochs,
    batch_size=100,
    validation_data=(X_test, y_test),
    callbacks=[early_stop, checkpoint_callback]
)

# ========================
# 9. ÿ™ÿ≠ŸÖŸäŸÑ ÿßŸÑŸÜŸÖŸàÿ∞ÿ¨ ŸÖŸÜ ÿ•Ÿäÿ®ŸàŸÉ ÿ™ÿÆÿ™ÿßÿ±Ÿá
# ========================
chosen_epoch = 2  # <-- ÿ∫ŸäŸëÿ± ÿ±ŸÇŸÖ ÿßŸÑÿ•Ÿäÿ®ŸàŸÉ ÿ≠ÿ≥ÿ® ŸÖÿß ÿ™ÿ±ŸäÿØ ÿ™ÿ¨ÿ±ÿ®ÿ™Ÿá
model.load_weights(os.path.join(checkpoint_dir, f"epoch_{chosen_epoch:02d}.weights.h5"))  # ‚úÖ ÿ™ÿ∫ŸäŸäÿ± ÿßŸÑÿßŸÖÿ™ÿØÿßÿØ ŸáŸÜÿß

# ========================
# 10. ÿ™ŸÇŸäŸäŸÖ ÿßŸÑŸÜŸÖŸàÿ∞ÿ¨
# ========================
loss, acc = model.evaluate(X_test, y_test)
print(f"Test Accuracy from epoch {chosen_epoch}: {acc:.2f}")

# ========================
# 11. ÿ±ÿ≥ŸÖ ÿßŸÑÿ™ÿØÿ±Ÿäÿ®
# ========================
plt.plot(history.history['accuracy'], label='train acc')
plt.plot(history.history['val_accuracy'], label='val acc')
plt.legend()
plt.title("LSTM Accuracy")
plt.show()
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from tensorflow.keras.utils import to_categorical
import matplotlib.pyplot as plt

# ========================
# 1. ÿ™ÿ≠ŸÖŸäŸÑ Ÿàÿ™ÿ¨ŸáŸäÿ≤ ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™
# ========================
# ÿ™ÿ£ŸÉÿØ ÿ£ŸÜ df ŸáŸà DataFrame Ÿäÿ≠ÿ™ŸàŸä ÿπŸÑŸâ ÿßŸÑÿ£ÿπŸÖÿØÿ©: 'Gmt time', 'Signal', 'date', Ÿàÿ®ŸÇŸäÿ© ÿßŸÑŸÖŸäÿ≤ÿßÿ™

# ÿ™ÿ≠ŸàŸäŸÑ ÿßŸÑŸàŸÇÿ™ ŸÖŸÜ String ÿ•ŸÑŸâ datetime
df['Gmt time'] = pd.to_datetime(df['Gmt time'])

# ÿ≠ÿ∞ŸÅ ÿßŸÑÿ£ÿπŸÖÿØÿ© ÿ∫Ÿäÿ± ÿßŸÑŸÖÿ∑ŸÑŸàÿ®ÿ©
df = df.drop(columns=['Gmt time'])  # ÿ•ÿ∞ÿß ŸÑŸÖ ÿ™ŸÉŸÜ ÿ™ÿ≥ÿ™ÿÆÿØŸÖ ÿßŸÑŸàŸÇÿ™ ŸÉŸÖŸèÿØÿÆŸÑ

# ŸÅÿµŸÑ ÿßŸÑŸÖŸäÿ≤ÿßÿ™ ŸàÿßŸÑŸáÿØŸÅ
features = df.drop(columns=['Signal', 'date'])  # ‚Üê ÿπÿØŸëŸÑ ÿ≠ÿ≥ÿ® ÿßÿ≥ŸÖ ÿπŸÖŸàÿØ ÿßŸÑŸáÿØŸÅ
target = df['Signal']

# ========================
# 2. ÿ™ŸÇÿ≥ŸäŸÖ ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™ ŸÇÿ®ŸÑ ÿ£Ÿä Scaling
# ========================
features_train, features_test, target_train, target_test = train_test_split(
    features, target, test_size=0.1, shuffle=False
)

# ========================
# 3. ŸÖŸÇŸäÿßÿ≥ ÿßŸÑŸÖŸäÿ≤ÿßÿ™ (ÿ®ÿØŸàŸÜ ÿ™ÿ≥ÿ±Ÿäÿ®)
# ========================
scaler = StandardScaler()
features_train_scaled = scaler.fit_transform(features_train)
features_test_scaled = scaler.transform(features_test)

# ========================
# 4. ÿ•ŸÜÿ¥ÿßÿ° ÿßŸÑŸÜŸàÿßŸÅÿ∞ ÿßŸÑŸÖŸÜÿ≤ŸÑŸÇÿ©
# ========================
def create_sliding_window(X, y, window_size=50):
    Xs, ys = [], []
    for i in range(len(X) - window_size):
        if y[i + window_size] != 0:  # ŸÅŸÇÿ∑ ÿ•ÿ¥ÿßÿ±ÿßÿ™ ÿßŸÑÿ¥ÿ±ÿßÿ°/ÿßŸÑÿ®Ÿäÿπ
            Xs.append(X[i:i+window_size])
            ys.append(y[i + window_size])
    return np.array(Xs), np.array(ys)

window_size = 50
X_train, y_train = create_sliding_window(features_train_scaled, target_train.values, window_size)
X_test, y_test = create_sliding_window(features_test_scaled, target_test.values, window_size)

# ========================
# 5. ÿ™ÿ≠ŸàŸäŸÑ ÿßŸÑŸáÿØŸÅ ÿ•ŸÑŸâ One-hot
# ========================
y_train = to_categorical(y_train, num_classes=3)
y_test = to_categorical(y_test, num_classes=3)

# ========================
# 6. ÿ®ŸÜÿßÿ° ŸÜŸÖŸàÿ∞ÿ¨ LSTM
# ========================
model = Sequential()
model.add(LSTM(64, input_shape=(X_train.shape[1], X_train.shape[2]), return_sequences=False))
model.add(Dense(32, activation='relu'))
model.add(Dense(3, activation='softmax'))  # ÿ™ÿµŸÜŸäŸÅ 3 ŸÅÿ¶ÿßÿ™

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# ========================
# 7. ÿ™ÿØÿ±Ÿäÿ® ÿßŸÑŸÜŸÖŸàÿ∞ÿ¨
# ========================
history = model.fit(X_train, y_train, epochs=2, batch_size=100, validation_data=(X_test, y_test))

# ========================
# 8. ÿ™ŸÇŸäŸäŸÖ ÿßŸÑŸÜŸÖŸàÿ∞ÿ¨
# ========================
loss, acc = model.evaluate(X_test, y_test)
print(f"Test Accuracy: {acc:.2f}")

# ========================
# 9. ÿ±ÿ≥ŸÖ ÿßŸÑÿ™ÿØÿ±Ÿäÿ®
# ========================
plt.plot(history.history['accuracy'], label='train acc')
plt.plot(history.history['val_accuracy'], label='val acc')
plt.legend()
plt.title("LSTM Accuracy")
plt.show()
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from tensorflow.keras.utils import to_categorical
import matplotlib.pyplot as plt

# ========================
# 1. ÿ™ÿ≠ŸÖŸäŸÑ Ÿàÿ™ÿ¨ŸáŸäÿ≤ ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™
# ========================
# ÿ™ÿ£ŸÉÿØ ÿ£ŸÜ df ÿ™ŸÖ ÿ™ÿ≠ŸÖŸäŸÑŸá ŸÖÿ≥ÿ®ŸÇŸãÿß ŸàŸÅŸäŸá ÿßŸÑÿ£ÿπŸÖÿØÿ© ÿßŸÑÿ™ÿßŸÑŸäÿ©: Gmt time, Signal, date, + ÿ®ÿßŸÇŸä ÿßŸÑŸÖŸÖŸäÿ≤ÿßÿ™
# df = pd.read_csv("your_data.csv")  # ŸÇŸÖ ÿ®ÿ™ÿ≠ŸÖŸäŸÑ ÿ®ŸäÿßŸÜÿßÿ™ŸÉ

# ÿ™ÿ≠ŸàŸäŸÑ ÿßŸÑŸàŸÇÿ™ ŸÖŸÜ String ÿ•ŸÑŸâ datetime
df['Gmt time'] = pd.to_datetime(df['Gmt time'])

# ÿ≠ÿ∞ŸÅ ÿßŸÑÿ£ÿπŸÖÿØÿ© ÿ∫Ÿäÿ± ÿßŸÑŸÖŸÅŸäÿØÿ©
df = df.drop(columns=['Gmt time'])

# ÿßÿÆÿ™Ÿäÿßÿ± ÿßŸÑŸÖŸäÿ≤ÿßÿ™ ŸàÿßŸÑŸáÿØŸÅ
features = df.drop(columns=['Signal', 'date'])
target = df['Signal']

# ========================
# 2. ŸÖŸÇŸäÿßÿ≥ ÿßŸÑŸÖŸäÿ≤ÿßÿ™
# ========================
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)

# ========================
# 3. ÿ•ŸÜÿ¥ÿßÿ° ÿßŸÑŸÜŸàÿßŸÅÿ∞ ÿßŸÑŸÖŸÜÿ≤ŸÑŸÇÿ©
# ========================
def create_sliding_window(X, y, window_size=50):
    Xs, ys = [], []
    for i in range(len(X) - window_size):
        Xs.append(X[i:i + window_size])
        ys.append(y[i + window_size])
    return np.array(Xs), np.array(ys)

window_size = 50
X, y = create_sliding_window(features_scaled, target.values, window_size=window_size)

# ÿ™ÿ≠ŸàŸäŸÑ ÿßŸÑŸáÿØŸÅ ÿ•ŸÑŸâ one-hot encoding
y_cat = to_categorical(y, num_classes=3)

# ========================
# 4. ÿ™ŸÇÿ≥ŸäŸÖ ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™
# ========================
X_train, X_test, y_train, y_test = train_test_split(X, y_cat, test_size=0.2, shuffle=False)

# ========================
# 5. ÿ≠ÿ≥ÿßÿ® sample weights
# ========================
# ŸÜÿ≠ŸàŸÑ y_train ŸÖŸÜ one-hot ÿ•ŸÑŸâ ÿ£ÿ±ŸÇÿßŸÖ
y_train_labels = np.argmax(y_train, axis=1)

# ŸÜÿ≠ÿØÿØ ÿßŸÑÿ£Ÿàÿ≤ÿßŸÜ: ÿßŸÑŸÅÿ¶ÿ© 0 = 0ÿå ÿßŸÑŸÅÿ¶ÿßÿ™ 1 Ÿà 2 = 1
weights_dict = {0: 0.0, 1: 100.0, 2: 100.0}
sample_weights = np.array([weights_dict[label] for label in y_train_labels])

# ========================
# 6. ÿ®ŸÜÿßÿ° ŸÜŸÖŸàÿ∞ÿ¨ LSTM
# ========================
model = Sequential()
model.add(LSTM(64, input_shape=(X.shape[1], X.shape[2]), return_sequences=False))
model.add(Dense(32, activation='relu'))
model.add(Dense(3, activation='softmax'))  # 3 ŸÅÿ¶ÿßÿ™

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# ========================
# 7. ÿ™ÿØÿ±Ÿäÿ® ÿßŸÑŸÜŸÖŸàÿ∞ÿ¨ ÿ®ÿßÿ≥ÿ™ÿÆÿØÿßŸÖ sample_weight
# ========================
history = model.fit(
    X_train,
    y_train,
    sample_weight=sample_weights,
    epochs=20,
    batch_size=32,
    validation_data=(X_test, y_test)
)

# ========================
# 8. ÿ™ŸÇŸäŸäŸÖ ÿßŸÑŸÜŸÖŸàÿ∞ÿ¨
# ========================
loss, acc = model.evaluate(X_test, y_test)
print(f"Test Accuracy: {acc:.2f}")

# ========================
# 9. ÿ±ÿ≥ŸÖ ÿßŸÑÿ£ÿØÿßÿ°
# ========================
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title("LSTM Accuracy")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend()
plt.show()
```
