```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras import regularizers
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
from tensorflow.keras.optimizers import Adam
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt

# ========================
# 1. تحميل وتجهيز البيانات
# ========================
# df = df_

df['Gmt time'] = pd.to_datetime(df['Gmt time'])
df = df.drop(columns=['Gmt time'])  # <-- إذا كنت لا تستخدم الوقت كمُدخل

features = df.drop(columns=['Signal', 'date'])  # <-- غيّر 'Signal' لاسم عمود الهدف لو مختلف
target = df['Signal']

# ========================
# 2. مقياس الميزات
# ========================
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)

# ========================
# 3. إنشاء النوافذ المنزلقة
# ========================
def create_sliding_window(X, y, window_size=50):
    Xs, ys = [], []
    for i in range(len(X) - window_size):
        if y[i + window_size] != 0:  # نحتفظ فقط بإشارات 1 و 2
            Xs.append(X[i:i+window_size])
            ys.append(y[i + window_size])
    return np.array(Xs), np.array(ys)

window_size = 50
X, y = create_sliding_window(features_scaled, target.values, window_size=window_size)
y_cat = to_categorical(y, num_classes=3)

# ========================
# 4. تقسيم البيانات
# ========================
X_train, X_test, y_train, y_test = train_test_split(X, y_cat, test_size=0.2, shuffle=False)

# ========================
# 5. بناء نموذج LSTM
# ========================

model = Sequential()
model.add(LSTM(
    16,
    input_shape=(X.shape[1], X.shape[2]),
    return_sequences=False,
    dropout=0.2,  # لحذف بعض المدخلات عشوائيًا
    kernel_regularizer=regularizers.l2(0.01)  # تقليل الحفظ
))
model.add(Dense(8, activation='relu', kernel_regularizer=regularizers.l2(0.01)))
model.add(Dropout(0.2))  # بعد Dense
model.add(Dense(3, activation='softmax'))

# ========================
# 5.1 تحديد نسبة التعلم
# ========================
optimizer = Adam(learning_rate=0.001)
model.compile(optimizer=optimizer, loss='categorical_crossentropy', metrics=['accuracy'])

# ========================
# 6.1 إعداد التوقف المبكر وحفظ أفضل نموذج
# ========================
early_stop = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
checkpoint = ModelCheckpoint('best_model.keras', monitor='val_loss', save_best_only=True)

# ========================
# 6. تدريب النموذج
# ========================
history = model.fit(
    X_train, y_train,
    epochs=200,
    batch_size=100,
    validation_data=(X_test, y_test),
    callbacks=[early_stop, checkpoint]
)

# ========================
# 7. تقييم تدريجي خطوة بخطوة (Walking Forward Evaluation)
# ========================
y_true = []
y_pred = []

for i in range(len(X_test)):
    sample_x = X_test[i].reshape(1, X_test.shape[1], X_test.shape[2])
    true_class = np.argmax(y_test[i])
    prediction = model.predict(sample_x, verbose=0)
    predicted_class = np.argmax(prediction)
    
    y_true.append(true_class)
    y_pred.append(predicted_class)

walk_acc = accuracy_score(y_true, y_pred)
print(f"🧠 Walking Forward Accuracy: {walk_acc:.2f}")

# ==================================================================================================================
# تقييم النموذج على اختبار حقيقي
# ==================================================================================================================
test_loss, test_accuracy = model.evaluate(X_test, y_test, verbose=0)

# ========================
#  رسم منحنى الأداء الكامل
# ========================
epochs_range = range(len(history.history['loss']))

plt.figure(figsize=(12, 6))

# رسم الخسارة
plt.plot(epochs_range, history.history['loss'], label='Train Loss', linestyle='--')
plt.plot(epochs_range, history.history['val_loss'], label='Validation Loss', linestyle='--')

# رسم الدقة
plt.plot(epochs_range, history.history['accuracy'], label='Train Accuracy', linestyle='-')
plt.plot(epochs_range, history.history['val_accuracy'], label='Validation Accuracy', linestyle='-')

plt.title('Training, Validation and Test Metrics')
plt.xlabel('Epoch')
plt.ylabel('Value')
plt.legend(loc='best')
plt.grid(True)
plt.tight_layout()
plt.show()

import numpy as np
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns
import matplotlib.pyplot as plt

# ===========
#  عمل التنبؤات
# ===========
y_pred_probs = model.predict(X_test)
y_pred = np.argmax(y_pred_probs, axis=1)
y_true = np.argmax(y_test, axis=1)

# ===========
#  مصفوفة الالتباس
# ===========
cm = confusion_matrix(y_true, y_pred)

# ===========
#  رسم المصفوفة
# ===========
plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=[1, 2], yticklabels=[1, 2])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()

# ===========
#  تقرير الأداء الكامل
# ===========
print("Classification Report:\n")
print(classification_report(y_true, y_pred, digits=3))


```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from tensorflow.keras.utils import to_categorical
import matplotlib.pyplot as plt

# ========================
# 1. تحميل وتجهيز البيانات
# ========================
# تأكد أن df تم تحميله مسبقًا وفيه الأعمدة التالية: Gmt time, Signal, date, + باقي المميزات
# df = pd.read_csv("your_data.csv")  # قم بتحميل بياناتك

# تحويل الوقت من String إلى datetime
df['Gmt time'] = pd.to_datetime(df['Gmt time'])

# حذف الأعمدة غير المفيدة
df = df.drop(columns=['Gmt time'])

# اختيار الميزات والهدف
features = df.drop(columns=['Signal', 'date'])
target = df['Signal']

# ========================
# 2. مقياس الميزات
# ========================
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)

# ========================
# 3. إنشاء النوافذ المنزلقة
# ========================
def create_sliding_window(X, y, window_size=50):
    Xs, ys = [], []
    for i in range(len(X) - window_size):
        Xs.append(X[i:i + window_size])
        ys.append(y[i + window_size])
    return np.array(Xs), np.array(ys)

window_size = 50
X, y = create_sliding_window(features_scaled, target.values, window_size=window_size)

# تحويل الهدف إلى one-hot encoding
y_cat = to_categorical(y, num_classes=3)

# ========================
# 4. تقسيم البيانات
# ========================
X_train, X_test, y_train, y_test = train_test_split(X, y_cat, test_size=0.2, shuffle=False)

# ========================
# 5. حساب sample weights
# ========================
# نحول y_train من one-hot إلى أرقام
y_train_labels = np.argmax(y_train, axis=1)

# نحدد الأوزان: الفئة 0 = 0، الفئات 1 و 2 = 1
weights_dict = {0: 0.0, 1: 100.0, 2: 100.0}
sample_weights = np.array([weights_dict[label] for label in y_train_labels])

# ========================
# 6. بناء نموذج LSTM
# ========================
model = Sequential()
model.add(LSTM(64, input_shape=(X.shape[1], X.shape[2]), return_sequences=False))
model.add(Dense(32, activation='relu'))
model.add(Dense(3, activation='softmax'))  # 3 فئات

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# ========================
# 7. تدريب النموذج باستخدام sample_weight
# ========================
history = model.fit(
    X_train,
    y_train,
    sample_weight=sample_weights,
    epochs=20,
    batch_size=32,
    validation_data=(X_test, y_test)
)

# ========================
# 8. تقييم النموذج
# ========================
loss, acc = model.evaluate(X_test, y_test)
print(f"Test Accuracy: {acc:.2f}")

# ========================
# 9. رسم الأداء
# ========================
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title("LSTM Accuracy")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend()
plt.show()
```
