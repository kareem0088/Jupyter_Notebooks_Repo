```python
import asyncio
import nest_asyncio
nest_asyncio.apply()

from twisted.internet import asyncioreactor
asyncioreactor.install(asyncio.get_event_loop())

from twisted.internet import reactor
from ctrader_open_api import Client, Protobuf, TcpProtocol, EndPoints
from ctrader_open_api.messages.OpenApiMessages_pb2 import (
    ProtoOAApplicationAuthReq, ProtoOAAccountAuthReq,
    ProtoOASymbolsListReq, ProtoOASymbolsListRes,
    ProtoOAGetTrendbarsReq, ProtoOAGetTrendbarsRes
)
from ctrader_open_api.messages.OpenApiModelMessages_pb2 import (
    ProtoOATrendbarPeriod
)
import pandas as pd
import lightgbm as lgb
import pandas_ta as ta

# معلومات الاتصال بـ cTrader
hostType = "Live"
host = EndPoints.PROTOBUF_LIVE_HOST if hostType.lower() == "live" else EndPoints.PROTOBUF_DEMO_HOST
client = Client(host, EndPoints.PROTOBUF_PORT, TcpProtocol)

def onError(error):
    print("Error: ", error)

def connected():
    print("Connected")
    auth_request = ProtoOAApplicationAuthReq()
    auth_request.clientId = "12063_H9AEW0UKNXCY9zkUMVZ6ikKRyRzkS4HflhEfTAe7M1GnPyRDrm"
    auth_request.clientSecret = "XD25AP1bl0Ssmm8sKKXS3cxoTcws2SY4Li5k3wrxcpY3uOl5z7"
    deferred = client.send(auth_request)
    deferred.addCallback(authenticate_trader)
    deferred.addErrback(onError)

def authenticate_trader(response):
    acc_auth_request = ProtoOAAccountAuthReq()
    acc_auth_request.ctidTraderAccountId = 4745363
    deferred = client.send(acc_auth_request)
    deferred.addCallback(get_symbols_list)
    deferred.addErrback(onError)

def get_symbols_list(response):
    symbols_request = ProtoOASymbolsListReq()
    symbols_request.ctidTraderAccountId = 4745363
    deferred = client.send(symbols_request)
    deferred.addCallback(handle_symbols_list)
    deferred.addErrback(onError)

def handle_symbols_list(message):
    message_payload = Protobuf.extract(message)
    if isinstance(message_payload, ProtoOASymbolsListRes):
        symbols = message_payload.symbol
        symbol_id = None
        for symbol in symbols:
            if symbol.symbolName == "EURUSD":
                symbol_id = symbol.symbolId
                print(f"Found EURUSD, symbolId: {symbol_id}")
                break
        if symbol_id is not None:
            get_trendbars(symbol_id)
        else:
            print("EURUSD not found in the symbol list.")
            reactor.stop()
    else:
        print("Error receiving symbol list.")
        reactor.stop()

def get_trendbars(symbol_id):
    request = ProtoOAGetTrendbarsReq()
    request.ctidTraderAccountId = 6054628
    request.symbolId = symbol_id
    request.period = ProtoOATrendbarPeriod.M1
    end_time = int(pd.Timestamp.utcnow().timestamp() * 1000)
    start_time = end_time - (100 * 60 * 1000)
    request.fromTimestamp = start_time
    request.toTimestamp = end_time
    deferred = client.send(request)
    deferred.addCallback(onMessageReceived)
    deferred.addErrback(onError)

def onMessageReceived(message):
    message_payload = Protobuf.extract(message)
    if isinstance(message_payload, ProtoOAGetTrendbarsRes):
        handle_candle_data(message_payload)
    else:
        print("Message received: \n", message_payload)

def handle_candle_data(message):
    trendbars = message.trendbar
    candles = []
    for trendbar in trendbars:
        candles.append({
            'timestamp': trendbar.utcTimestamp,
            'open': trendbar.open,
            'high': trendbar.high,
            'low': trendbar.low,
            'close': trendbar.close,
            'volume': trendbar.volume,
        })
    df = pd.DataFrame(candles)
    df['timestamp'] = pd.to_datetime(df['timestamp'], unit='ms')
    df.set_index('timestamp', inplace=True)
    df['EMA_100'] = ta.ema(df['close'], length=100)
    last_row = df.dropna().iloc[-1]
    input_data = last_row[['open', 'high', 'low', 'close', 'volume', 'EMA_100']].to_frame().T
    model = lgb.Booster(model_file='model_path')
    prediction = model.predict(input_data)
    print(f'Prediction: {prediction[0]}')
    reactor.stop()

def disconnected(reason):
    print("Disconnected: ", reason)

client.setConnectedCallback(connected)
client.setDisconnectedCallback(disconnected)

def start_service():
    if not reactor.running:
        client.startService()
        reactor.run()

start_service()
```
