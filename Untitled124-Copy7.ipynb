```python
def fut_data(List):
    for df in List:
        
        #df = df_train
        
        # ÿßŸÑÿ™ÿ£ŸÉÿØ ŸÖŸÜ ÿ£ŸÜ ÿßŸÑÿπŸÖŸàÿØ datetime
        df['timestamp'] = pd.to_datetime(df['timestamp'])
        df['date'] = df['timestamp'].dt.date
        
        # ÿØŸàÿßŸÑ ŸÑÿ≠ÿ≥ÿßÿ® ÿ£ÿπŸÑŸâ/ÿ£ŸÇŸÑ ÿ≥ÿπÿ± ÿ≠ÿ™Ÿâ ÿßŸÑÿµŸÅ ÿßŸÑÿ≠ÿßŸÑŸä
        def calc_high_so_far(group):
            group['high_today'] = group['high'].cummax()
            idx_high = group['high'].cummax().eq(group['high'])
            group['vol_at_high_today'] = group['volume'].where(idx_high).ffill()
            return group
        
        def calc_low_so_far(group):
            group['low_today'] = group['low'].cummin()
            idx_low = group['low'].cummin().eq(group['low'])
            group['vol_at_low_today'] = group['volume'].where(idx_low).ffill()
            return group
        
        # ŸÜÿÆÿ≤ŸÜ ÿπŸÖŸàÿØ date ŸÖÿ§ŸÇÿ™
        dates = df['date']
        
        # ŸÜÿ≥ÿ™ÿ®ÿπÿØ date ŸÖŸÜ groupby.apply ŸÑÿ™ŸÅÿßÿØŸä ÿßŸÑÿ™ÿ≠ÿ∞Ÿäÿ±
        df = df.drop(columns=['date'])
        df = pd.concat([
            calc_high_so_far(group) 
            for _, group in pd.concat([dates, df], axis=1).groupby(dates)
        ], axis=0)
        
        df = pd.concat([
            calc_low_so_far(group) 
            for _, group in df.groupby(dates)
        ], axis=0)
        
        # ŸÜÿ±ÿ¨ÿπ ÿßŸÑÿπŸÖŸàÿØ date
        df['date'] = dates
        
        # ÿ≠ÿ≥ÿßÿ® ÿ®ŸäÿßŸÜÿßÿ™ ÿßŸÑŸäŸàŸÖ ÿßŸÑÿ≥ÿßÿ®ŸÇ ÿ®ÿØŸàŸÜ ÿ™ÿ≥ÿ±Ÿäÿ®
        df['high_prev_day'] = df.groupby('date')['high_today'].shift(1)
        df['vol_at_high_prev_day'] = df.groupby('date')['vol_at_high_today'].shift(1)
        df['low_prev_day'] = df.groupby('date')['low_today'].shift(1)
        df['vol_at_low_prev_day'] = df.groupby('date')['vol_at_low_today'].shift(1)
        
        
        # Relative Volume (RVOL)
        df['RVOL'] = df['volume'] / df['volume'].rolling(window=20).mean()
        
        # Volume Rate of Change (VROC)
        df['VROC'] = (df['volume'] - df['volume'].shift(5)) / df['volume'].shift(5)
        
        # Volume Weighted Spread
        df['spread_vol'] = (df['high'] - df['low']) * df['RVOL']
        
        # VWAP Deviation (ÿ®ÿßŸÑÿßŸÜÿ≠ÿ±ÿßŸÅ ÿßŸÑŸÖÿπŸäÿßÿ±Ÿä)
        vwap_std = df['vwap'].rolling(window=20).std()
        df['vwap_dev'] = (df['close'] - df['vwap']) / vwap_std
        
        # Volume per Transaction
        df['vol_per_tx'] = df['volume'] / df['transactions']
        
        
        # Accumulation/Distribution based on VWAP
        df['accum_dist_vwap'] = np.where(df['close'] > df['vwap'], df['volume'], -df['volume'])
        df['accum_dist_vwap'] = df['accum_dist_vwap'].cumsum()
        
        # Volume Spike Detector ‚Üí ÿ®ÿØŸàŸÜ ÿ™ÿ≥ÿ±Ÿäÿ®
        df['vol_spike'] = (
            df['volume'] > (df['volume'].rolling(window=10).mean().shift(1) * 2)
        ).astype(int)
        
        # Volume Climax Bar ‚Üí ÿ®ÿØŸàŸÜ ÿ™ÿ≥ÿ±Ÿäÿ® Ÿàÿ®ŸÄ tolerance
        max_vol_threshold = df['volume'].rolling(window=20).max().shift(1) * 0.95
        wide_range_threshold = (df['high'] - df['low']).rolling(window=20).mean().shift(1)
        
        df['climax_bar'] = (
            (df['volume'] >= max_vol_threshold) & 
            ((df['high'] - df['low']) > wide_range_threshold)
        ).astype(int)
        
        
        denominator = (df['high'].rolling(5).max().shift(1) - df['low'].rolling(5).min().shift(1))
        denominator = denominator.replace(0, np.nan)  # ÿßÿ≥ÿ™ÿ®ÿØÿßŸÑ ÿßŸÑŸÖŸÇÿßŸÖ ÿ•ÿ∞ÿß ŸÉÿßŸÜ ÿµŸÅÿ± ÿ®ŸÄ NaN
        
        df['micro_trend_slope'] = (
            (df['close'] - df['close'].shift(5)) / denominator
        )
        
        
        # Price Change per Second (ŸÑŸà ÿπŸÜÿØŸÉ ÿ®ŸäÿßŸÜÿßÿ™ ŸàŸÇÿ™ ÿ®ÿßŸÑÿ´ŸàÿßŸÜŸä)
        # ÿ™ÿ≠ŸàŸäŸÑ timestamp ÿ•ŸÑŸâ datetime
        df['timestamp'] = pd.to_datetime(df['timestamp'])
        df['seconds_diff'] = (df['timestamp'] - df['timestamp'].shift(1)).dt.total_seconds()
        df['price_change_per_sec'] = (df['close'] - df['close'].shift(1)) / df['seconds_diff']
        
        # Position in Range (ÿ¢ÿÆÿ± 20 ÿ¥ŸÖÿπÿ©) - ÿ®ÿØŸàŸÜ ÿ™ÿ≥ÿ±Ÿäÿ®
        window_range = 20
        low_min = df['low'].rolling(window_range).min().shift(1)
        high_max = df['high'].rolling(window_range).max().shift(1)
        df['pos_in_range'] = (df['close'] - low_min) / (high_max - low_min)
        
        # Micro Support/Resistance Distance - ÿ®ÿØŸàŸÜ ÿ™ÿ≥ÿ±Ÿäÿ®
        df['dist_to_recent_high'] = df['high'].rolling(window_range).max().shift(1) - df['close']
        df['dist_to_recent_low'] = df['close'] - df['low'].rolling(window_range).min().shift(1)
        
        # Candle Body-to-Range Ratio (ÿØŸá ÿ∑ÿ®ŸäÿπŸä ŸÖŸÅŸäŸáŸàÿ¥ ÿ™ÿ≥ÿ±Ÿäÿ®)
        df['body_to_range'] = abs(df['close'] - df['open']) / (df['high'] - df['low'])
        
        # Transaction Density (ÿ∑ÿ®ŸäÿπŸä ÿ®ÿØŸàŸÜ ÿ™ÿ≥ÿ±Ÿäÿ®)
        df['tx_density'] = df['transactions'] / df['volume']
        
        # VWAP Pullback Strength - ÿ®ÿØŸàŸÜ ÿ™ÿ≥ÿ±Ÿäÿ®
        df['vwap_pullback'] = (df['vwap'].shift(1) - df['close'].shift(1)) / df['vwap'].shift(1)
        
        # Volume Shock - ÿ®ÿØŸàŸÜ ÿ™ÿ≥ÿ±Ÿäÿ®
        vol_avg = df['volume'].rolling(20).mean().shift(1)
        df['vol_shock'] = (df['volume'] > (vol_avg * 2)).astype(int)
        
        # Price Reaction Next (ÿØŸá Label ŸÑŸÑÿ™ŸÜÿ®ÿ§ ŸÖÿ¥ ŸÖÿ§ÿ¥ÿ±)
        df['price_reaction_next'] = df['close'].shift(-1) - df['close']
        
        # ÿ≠ÿ≥ÿßÿ® ÿßŸÑŸÅÿ±ŸÇ ÿ®ÿßŸÑÿØŸÇÿßÿ¶ŸÇ ÿ®ŸäŸÜ ÿßŸÑÿµŸÅ ÿßŸÑÿ≠ÿßŸÑŸä ŸàÿßŸÑÿ≥ÿßÿ®ŸÇ
        df['minutes_diff'] = (df['timestamp'] - df['timestamp'].shift(1)).dt.total_seconds() / 60
        
        # ÿßŸÑŸÅÿ±ŸÇ ÿ®ŸäŸÜ ÿ≥ÿπÿ± ÿßŸÑŸÅÿ™ÿ≠ ÿßŸÑÿ≠ÿßŸÑŸä Ÿàÿ≥ÿπÿ± ÿßŸÑÿ•ÿ∫ŸÑÿßŸÇ ÿßŸÑÿ≥ÿßÿ®ŸÇ
        df['open_vs_prev_close'] = df['open'] - df['close'].shift(1)
        
        df['vwap_distance'] = df['close'] - df['vwap']
        
        df['vol_change_pct'] = (df['volume'] - df['volume'].shift(1)) / df['volume'].shift(1) * 100
        
        # ÿßÿ≥ÿ™ÿÆÿ±ÿßÿ¨ ÿßŸÑÿ≥ÿßÿπÿ©
        df['hour'] = df['timestamp'].dt.hour
        
        # ÿßÿ≥ÿ™ÿÆÿ±ÿßÿ¨ ÿßŸÑÿØŸÇÿßÿ¶ŸÇ
        df['minute'] = df['timestamp'].dt.minute
        
        # ÿ™ÿ≠ÿØŸäÿØ ÿßŸÑÿ±ÿ®ÿπ ŸÖŸÜ ÿßŸÑÿ≥ÿßÿπÿ©
        def get_quarter(minute):
            if minute < 15:
                return 1
            elif minute < 30:
                return 2
            elif minute < 45:
                return 3
            else:
                return 4
        
        df['quarter_hour'] = df['minute'].apply(get_quarter)
        
        df.dropna(inplace=True)
        
        n_future = 3  # 6 ÿ≥ÿßÿπÿßÿ™ ŸÑŸÇÿØŸëÿßŸÖ
        
        df['future_return'] = df['close'].shift(-n_future) / df['close'] - 1
        
        # ŸÜÿ≠ÿØÿØ ÿßŸÑÿ•ÿ¥ÿßÿ±ÿßÿ™ ÿ®ŸÜÿßÿ°Ÿã ÿπŸÑŸâ ÿπÿ™ÿ®ÿ© ŸÖÿπŸäŸÜÿ© ŸÑŸÑÿ≠ÿ±ŸÉÿ©
        threshold = 0.0003  
        
        df['Signal'] = 0
        df.loc[df['future_return'] > threshold, 'Signal'] = 2
        df.loc[df['future_return'] < -threshold, 'Signal'] = 1
        
        df.dropna(inplace=True)
        #df_train = df
        
```

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import RobustScaler
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import matplotlib.pyplot as plt
import seaborn as sns
import tqdm
from xgboost import XGBClassifier, plot_importance

# ========================
# ÿØÿßŸÑÿ© ÿ•ŸÜÿ¥ÿßÿ° ŸÜÿßŸÅÿ∞ÿ© ÿ≤ŸÖŸÜŸäÿ©
# ========================
def create_sliding_window(X, y, window_size):
    X_windows, y_windows = [], []
    for i in range(len(X) - window_size + 1):
        X_windows.append(X[i:i + window_size])
        y_windows.append(y[i + window_size - 1])  # ÿ¢ÿÆÿ± ÿπŸÜÿµÿ± ŸáŸà ÿßŸÑŸáÿØŸÅ
    return np.array(X_windows), np.array(y_windows)


# ========================
# 1. ÿ™ÿ≠ŸÖŸäŸÑ Ÿàÿ™ÿ¨ŸáŸäÿ≤ ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™
# ========================
# df_train = pd.read_csv('your_train_data.csv')
# df_test = pd.read_csv('your_test_data.csv')

features_train = df_train.drop(columns=['Signal','date','timestamp','future_return'])
features_test = df_test.drop(columns=['Signal','date','timestamp','future_return'])

target_train = df_train['Signal']
target_test = df_test['Signal']

# ========================
# 2. ÿ•ŸÜÿ¥ÿßÿ° ÿßŸÑŸÜŸàÿßŸÅÿ∞ ÿßŸÑÿ≤ŸÖŸÜŸäÿ©
# ========================
window_size = 50
X_train_raw, y_train_raw = create_sliding_window(features_train.values, target_train.values, window_size)
X_test_raw, y_test_raw = create_sliding_window(features_test.values, target_test.values, window_size)

# ========================
# 3. ÿ™ÿ∑ÿ®Ÿäÿπ ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™
# ========================
scaler = RobustScaler()

X_train_scaled = scaler.fit_transform(X_train_raw.reshape(-1, X_train_raw.shape[-1])).reshape(X_train_raw.shape)
X_test_scaled = scaler.transform(X_test_raw.reshape(-1, X_test_raw.shape[-1])).reshape(X_test_raw.shape)

# Flatten ÿπÿ¥ÿßŸÜ XGBoost ŸÖÿß ÿ®ŸäŸÇÿ®ŸÑÿ¥ ÿ®ŸäÿßŸÜÿßÿ™ 3D
X_train_flat = X_train_scaled.reshape(X_train_scaled.shape[0], -1)
X_test_flat = X_test_scaled.reshape(X_test_scaled.shape[0], -1)

print(f"üîπ ÿ¥ŸÉŸÑ ÿ®ŸäÿßŸÜÿßÿ™ ÿßŸÑÿ™ÿØÿ±Ÿäÿ®: {X_train_flat.shape}")
print(f"üîπ ÿ¥ŸÉŸÑ ÿ®ŸäÿßŸÜÿßÿ™ ÿßŸÑÿßÿÆÿ™ÿ®ÿßÿ±: {X_test_flat.shape}")

# ========================
# 4. ÿ®ŸÜÿßÿ° ŸÜŸÖŸàÿ∞ÿ¨ XGBoost
# ========================
model = XGBClassifier(
    n_estimators=90,
    max_depth=8,
    learning_rate=0.05,
    subsample=0.8,
    colsample_bytree=0.8,
    objective="multi:probsoftmax",  # ÿ™ÿµŸÜŸäŸÅ ŸÖÿ™ÿπÿØÿØ ÿßŸÑŸÅÿ¶ÿßÿ™
    num_class=3,
    random_state=42,
    n_jobs=-1,
    eval_metric="mlogloss"   # ‚úÖ ŸáŸÜÿß ÿ®ÿØŸÑ ŸÖÿß ŸäŸÉŸàŸÜ ÿ¨ŸàŸá fit
)

# ========================
# 5. ÿ™ÿØÿ±Ÿäÿ® ÿßŸÑŸÜŸÖŸàÿ∞ÿ¨
# ========================
print("üöÄ Training XGBoost...")
model.fit(X_train_flat, y_train_raw,
          eval_set=[(X_test_flat, y_test_raw)],
          verbose=True)

# ========================
# 6. ÿßŸÑÿ™ŸÇŸäŸäŸÖ ÿßŸÑÿ™ÿØÿ±Ÿäÿ¨Ÿä Walking Forward
# ========================
y_true, y_pred = [], []
print("üîç ÿ¨ÿßÿ±Ÿä ÿßŸÑÿ™ŸÇŸäŸäŸÖ ÿßŸÑÿ™ÿØÿ±Ÿäÿ¨Ÿä...")
for i in tqdm.tqdm(range(len(X_test_flat))):
    sample_x = X_test_flat[i].reshape(1, -1)
    true_class = y_test_raw[i]
    predicted_class = model.predict(sample_x)[0]

    y_true.append(true_class)
    y_pred.append(predicted_class)

walk_acc = accuracy_score(y_true, y_pred)
print(f"üß† Walking Forward Accuracy: {walk_acc:.4f}")

# ========================
# 7. ÿßŸÑÿ™ŸÇŸäŸäŸÖ ÿπŸÑŸâ ÿ®ŸäÿßŸÜÿßÿ™ ÿßŸÑÿßÿÆÿ™ÿ®ÿßÿ± ŸÉÿßŸÖŸÑÿ©
# ========================
test_pred = model.predict(X_test_flat)
test_acc = accuracy_score(y_test_raw, test_pred)

print(f"üìä Test Accuracy: {test_acc:.4f}")

# ========================
# 8. ŸÖÿµŸÅŸàŸÅÿ© ÿßŸÑÿßŸÑÿ™ÿ®ÿßÿ≥
# ========================
cm = confusion_matrix(y_test_raw, test_pred)

plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=np.unique(y_train_raw),
            yticklabels=np.unique(y_train_raw))
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix - XGBoost")
plt.show()

# ========================
# 9. ÿ™ŸÇÿ±Ÿäÿ± ÿßŸÑÿ£ÿØÿßÿ°
# ========================
print("\nüìà Classification Report:")
print("="*50)
print(classification_report(y_test_raw, test_pred, digits=3))

# ========================
# 10. ÿ£ŸáŸÖŸäÿ© ÿßŸÑŸÖŸäÿ≤ÿßÿ™ (Feature Importance)
# ========================
plt.figure(figsize=(10, 6))
plot_importance(model, max_num_features=15, importance_type="gain")
plt.title("Top 15 Feature Importances - XGBoost")
plt.show()

# ========================
# 11. ÿ•ÿ≠ÿµÿßÿ¶Ÿäÿßÿ™ ÿ•ÿ∂ÿßŸÅŸäÿ©
# ========================
print(f"\nüìä ÿ•ÿ≠ÿµÿßÿ¶Ÿäÿßÿ™ ÿ•ÿ∂ÿßŸÅŸäÿ©:")
print(f"ÿπÿØÿØ ÿπŸäŸÜÿßÿ™ ÿßŸÑÿ™ÿØÿ±Ÿäÿ®: {len(X_train_flat)}")
print(f"ÿπÿØÿØ ÿπŸäŸÜÿßÿ™ ÿßŸÑÿßÿÆÿ™ÿ®ÿßÿ±: {len(X_test_flat)}")
print(f"ÿπÿØÿØ ÿßŸÑŸÖŸäÿ≤ÿßÿ™ ÿ®ÿπÿØ Flatten: {X_train_flat.shape[1]}")
print(f"ÿ≠ÿ¨ŸÖ ÿßŸÑŸÜÿßŸÅÿ∞ÿ© ÿßŸÑÿ≤ŸÖŸÜŸäÿ©: {window_size}")
print(f"ÿπÿØÿØ ÿßŸÑÿ£ÿ¥ÿ¨ÿßÿ±: {model.n_estimators}, ÿπŸÖŸÇ ÿßŸÑÿ¥ÿ¨ÿ±ÿ©: {model.max_depth}")


# ========================
# 12. ÿ±ÿ≥ŸÖ ÿßÿ≠ÿ™ŸÖÿßŸÑÿßÿ™ ÿßŸÑÿ´ŸÇÿ© ŸÑŸÉŸÑ ŸÅÿ¶ÿ©
# ========================

# ÿßŸÑÿ≠ÿµŸàŸÑ ÿπŸÑŸâ ÿßÿ≠ÿ™ŸÖÿßŸÑÿßÿ™ ŸÉŸÑ ÿπŸäŸÜÿ©
proba_test = model.predict_proba(X_test_flat)

# ÿ™ÿ≠ŸàŸäŸÑŸáÿß ÿ•ŸÑŸâ DataFrame ŸÑÿ™ÿ≥ŸáŸäŸÑ ÿßŸÑÿ±ÿ≥ŸÖ
df_proba = pd.DataFrame(proba_test, columns=['Class_0','Sell_1','Buy_2'])
df_proba['True_Label'] = y_test_raw


import matplotlib.pyplot as plt
import numpy as np

# ====== ÿ•ÿπÿØÿßÿØ ÿßŸÑŸÄ bins ======
bins = np.linspace(0, 1, 21)  # ÿ™ŸÇÿ≥ŸäŸÖ ÿßŸÑÿßÿ≠ÿ™ŸÖÿßŸÑ ŸÖŸÜ 0 ÿ•ŸÑŸâ 1

# ====== Sell Probability Histogram ======
plt.figure(figsize=(10,5))
plt.hist([df_proba['Sell_1'][df_proba['True_Label']==0],
          df_proba['Sell_1'][df_proba['True_Label']==1],
          df_proba['Sell_1'][df_proba['True_Label']==2]],
         bins=bins,
         color=['blue','green','red'],
         label=['No Trade (0)','Actual Sell (1)','Actual Buy (2)'],
         stacked=True, alpha=0.7)
plt.axvline(0.5, color='black', linestyle='--', label='Threshold 0.5')
plt.title("Sell Probabilities Histogram Colored by True Label")
plt.xlabel("Predicted Sell Probability")
plt.ylabel("Number of Signals")
plt.legend()
plt.show()

# ====== Buy Probability Histogram ======
plt.figure(figsize=(10,5))
plt.hist([df_proba['Buy_2'][df_proba['True_Label']==0],
          df_proba['Buy_2'][df_proba['True_Label']==1],
          df_proba['Buy_2'][df_proba['True_Label']==2]],
         bins=bins,
         color=['blue','green','red'],
         label=['No Trade (0)','Actual Sell (1)','Actual Buy (2)'],
         stacked=True, alpha=0.7)
plt.axvline(0.5, color='black', linestyle='--', label='Threshold 0.5')
plt.title("Buy Probabilities Histogram Colored by True Label")
plt.xlabel("Predicted Buy Probability")
plt.ylabel("Number of Signals")
plt.legend()
plt.show()

# ====== No Trade Probability Histogram ======
plt.figure(figsize=(10,5))
plt.hist([df_proba['Class_0'][df_proba['True_Label']==0],
          df_proba['Class_0'][df_proba['True_Label']==1],
          df_proba['Class_0'][df_proba['True_Label']==2]],
         bins=bins,
         color=['blue','green','red'],
         label=['Actual No Trade (0)','Actual Sell (1)','Actual Buy (2)'],
         stacked=True, alpha=0.7)
plt.axvline(0.5, color='black', linestyle='--', label='Threshold 0.5')
plt.title("No Trade Probabilities Histogram Colored by True Label")
plt.xlabel("Predicted No Trade Probability")
plt.ylabel("Number of Signals")
plt.legend()
plt.show()

```

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import RobustScaler
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import matplotlib.pyplot as plt
import seaborn as sns
import tqdm
from xgboost import XGBClassifier, plot_importance

# ========================
# ÿØÿßŸÑÿ© ÿ•ŸÜÿ¥ÿßÿ° ŸÜÿßŸÅÿ∞ÿ© ÿ≤ŸÖŸÜŸäÿ©
# ========================
def create_sliding_window(X, y, window_size):
    X_windows, y_windows = [], []
    for i in range(len(X) - window_size + 1):
        X_windows.append(X[i:i + window_size])
        y_windows.append(y[i + window_size - 1])  # ÿ¢ÿÆÿ± ÿπŸÜÿµÿ± ŸáŸà ÿßŸÑŸáÿØŸÅ
    return np.array(X_windows), np.array(y_windows)

# ========================
# 1. ÿ™ÿ≠ŸÖŸäŸÑ Ÿàÿ™ÿ¨ŸáŸäÿ≤ ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™
# ========================
# df_train = pd.read_csv('your_train_data.csv')
# df_test = pd.read_csv('your_test_data.csv')

features_train = df_train.drop(columns=['Signal','date','timestamp','future_return'])
features_test = df_test.drop(columns=['Signal','date','timestamp','future_return'])

target_train = df_train['Signal']
target_test = df_test['Signal']

# ========================
# 2. ÿ•ŸÜÿ¥ÿßÿ° ÿßŸÑŸÜŸàÿßŸÅÿ∞ ÿßŸÑÿ≤ŸÖŸÜŸäÿ©
# ========================
window_size = 50
X_train_raw, y_train_raw = create_sliding_window(features_train.values, target_train.values, window_size)
X_test_raw, y_test_raw = create_sliding_window(features_test.values, target_test.values, window_size)

# ========================
# 3. ÿ™ÿ∑ÿ®Ÿäÿπ ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™
# ========================
scaler = RobustScaler()

X_train_scaled = scaler.fit_transform(X_train_raw.reshape(-1, X_train_raw.shape[-1])).reshape(X_train_raw.shape)
X_test_scaled = scaler.transform(X_test_raw.reshape(-1, X_test_raw.shape[-1])).reshape(X_test_raw.shape)

# Flatten ÿπÿ¥ÿßŸÜ XGBoost ŸÖÿß ÿ®ŸäŸÇÿ®ŸÑÿ¥ ÿ®ŸäÿßŸÜÿßÿ™ 3D
X_train_flat = X_train_scaled.reshape(X_train_scaled.shape[0], -1)
X_test_flat = X_test_scaled.reshape(X_test_scaled.shape[0], -1)

print(f"üîπ ÿ¥ŸÉŸÑ ÿ®ŸäÿßŸÜÿßÿ™ ÿßŸÑÿ™ÿØÿ±Ÿäÿ®: {X_train_flat.shape}")
print(f"üîπ ÿ¥ŸÉŸÑ ÿ®ŸäÿßŸÜÿßÿ™ ÿßŸÑÿßÿÆÿ™ÿ®ÿßÿ±: {X_test_flat.shape}")

# ========================
# 4. ÿ®ŸÜÿßÿ° ŸÜŸÖŸàÿ∞ÿ¨ XGBoost ŸÖÿπ Early Stopping
# ========================
model = XGBClassifier(
    n_estimators=1000,
    max_depth=8,
    learning_rate=0.05,
    subsample=0.8,
    colsample_bytree=0.8,
    objective="multi:softprob",   # ÿßÿ≠ÿ™ŸÖÿßŸÑÿßÿ™ ŸÑŸÑÿ™ÿµŸÜŸäŸÅ ÿßŸÑŸÖÿ™ÿπÿØÿØ
    num_class=3,
    random_state=42,
    n_jobs=-1,
    eval_metric=["mlogloss", "merror"]  # ŸÜÿ≠ÿ∑Ÿáÿß ŸáŸÜÿß ŸÖÿ¥ ÿ¨ŸàŸá fit
)

# ========================
# 5. ÿ™ÿØÿ±Ÿäÿ® ÿßŸÑŸÜŸÖŸàÿ∞ÿ¨ ŸÖÿπ Early Stopping
# ========================
print("üöÄ Training XGBoost with Early Stopping...")
evals_result = {}
model.fit(
    X_train_flat, y_train_raw,
    eval_set=[(X_train_flat, y_train_raw), (X_test_flat, y_test_raw)],
    early_stopping_rounds=20,
    verbose=True,
    evals_result=evals_result
)

# ========================
# 6. ÿ±ÿ≥ŸÖ ŸÖŸÜÿ≠ŸÜŸäÿßÿ™ ÿßŸÑÿ™ÿØÿ±Ÿäÿ® (ÿÆÿ≥ÿßÿ±ÿ© ŸàÿØŸÇÿ©)
# ========================
epochs = len(evals_result['validation_0']['mlogloss'])
x_axis = range(epochs)

plt.figure(figsize=(12,5))
plt.plot(x_axis, evals_result['validation_0']['mlogloss'], label='Train Log Loss')
plt.plot(x_axis, evals_result['validation_1']['mlogloss'], label='Test Log Loss')
plt.xlabel('Epochs')
plt.ylabel('Log Loss')
plt.title('Training and Test Log Loss')
plt.legend()
plt.show()

plt.figure(figsize=(12,5))
plt.plot(x_axis, evals_result['validation_0']['merror'], label='Train Error')
plt.plot(x_axis, evals_result['validation_1']['merror'], label='Test Error')
plt.xlabel('Epochs')
plt.ylabel('Error')
plt.title('Training and Test Error')
plt.legend()
plt.show()

# ========================
# 7. ÿßŸÑÿ™ŸÇŸäŸäŸÖ Walking Forward
# ========================
y_true, y_pred = [], []
print("üîç ÿ¨ÿßÿ±Ÿä ÿßŸÑÿ™ŸÇŸäŸäŸÖ ÿßŸÑÿ™ÿØÿ±Ÿäÿ¨Ÿä...")
for i in tqdm.tqdm(range(len(X_test_flat))):
    sample_x = X_test_flat[i].reshape(1, -1)
    true_class = y_test_raw[i]
    predicted_class = model.predict(sample_x)[0]

    y_true.append(true_class)
    y_pred.append(predicted_class)

walk_acc = accuracy_score(y_true, y_pred)
print(f"üß† Walking Forward Accuracy: {walk_acc:.4f}")

# ========================
# 8. ÿßŸÑÿ™ŸÇŸäŸäŸÖ ÿπŸÑŸâ ÿ®ŸäÿßŸÜÿßÿ™ ÿßŸÑÿßÿÆÿ™ÿ®ÿßÿ± ŸÉÿßŸÖŸÑÿ©
# ========================
test_pred = model.predict(X_test_flat)
test_acc = accuracy_score(y_test_raw, test_pred)

print(f"üìä Test Accuracy: {test_acc:.4f}")

# ========================
# 9. ŸÖÿµŸÅŸàŸÅÿ© ÿßŸÑÿßŸÑÿ™ÿ®ÿßÿ≥
# ========================
cm = confusion_matrix(y_test_raw, test_pred)

plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=np.unique(y_train_raw),
            yticklabels=np.unique(y_train_raw))
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix - XGBoost")
plt.show()

# ========================
# 10. ÿ™ŸÇÿ±Ÿäÿ± ÿßŸÑÿ£ÿØÿßÿ°
# ========================
print("\nüìà Classification Report:")
print("="*50)
print(classification_report(y_test_raw, test_pred, digits=3))

# ========================
# 11. ÿ£ŸáŸÖŸäÿ© ÿßŸÑŸÖŸäÿ≤ÿßÿ™
# ========================
plt.figure(figsize=(10, 6))
plot_importance(model, max_num_features=15, importance_type="gain")
plt.title("Top 15 Feature Importances - XGBoost")
plt.show()

# ========================
# 12. ÿßÿ≠ÿ™ŸÖÿßŸÑÿßÿ™ ÿßŸÑŸÅÿ¶ÿßÿ™
# ========================
proba_test = model.predict_proba(X_test_flat)

df_proba = pd.DataFrame(proba_test, columns=['Class_0','Sell_1','Buy_2'])
df_proba['True_Label'] = y_test_raw

bins = np.linspace(0, 1, 21)

# Sell Histogram
plt.figure(figsize=(10,5))
plt.hist([df_proba['Sell_1'][df_proba['True_Label']==0],
          df_proba['Sell_1'][df_proba['True_Label']==1],
          df_proba['Sell_1'][df_proba['True_Label']==2]],
         bins=bins, color=['blue','green','red'],
         label=['No Trade (0)','Actual Sell (1)','Actual Buy (2)'],
         stacked=True, alpha=0.7)
plt.axvline(0.5, color='black', linestyle='--', label='Threshold 0.5')
plt.title("Sell Probabilities Histogram Colored by True Label")
plt.xlabel("Predicted Sell Probability")
plt.ylabel("Number of Signals")
plt.legend()
plt.show()

# Buy Histogram
plt.figure(figsize=(10,5))
plt.hist([df_proba['Buy_2'][df_proba['True_Label']==0],
          df_proba['Buy_2'][df_proba['True_Label']==1],
          df_proba['Buy_2'][df_proba['True_Label']==2]],
         bins=bins, color=['blue','green','red'],
         label=['No Trade (0)','Actual Sell (1)','Actual Buy (2)'],
         stacked=True, alpha=0.7)
plt.axvline(0.5, color='black', linestyle='--', label='Threshold 0.5')
plt.title("Buy Probabilities Histogram Colored by True Label")
plt.xlabel("Predicted Buy Probability")
plt.ylabel("Number of Signals")
plt.legend()
plt.show()

# No Trade Histogram
plt.figure(figsize=(10,5))
plt.hist([df_proba['Class_0'][df_proba['True_Label']==0],
          df_proba['Class_0'][df_proba['True_Label']==1],
          df_proba['Class_0'][df_proba['True_Label']==2]],
         bins=bins, color=['blue','green','red'],
         label=['Actual No Trade (0)','Actual Sell (1)','Actual Buy (2)'],
         stacked=True, alpha=0.7)
plt.axvline(0.5, color='black', linestyle='--', label='Threshold 0.5')
plt.title("No Trade Probabilities Histogram Colored by True Label")
plt.xlabel("Predicted No Trade Probability")
plt.ylabel("Number of Signals")
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import RobustScaler
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import matplotlib.pyplot as plt
import seaborn as sns
import tqdm
import xgboost as xgb
from xgboost import plot_importance

# ========================
# ÿØÿßŸÑÿ© ÿ•ŸÜÿ¥ÿßÿ° ŸÜÿßŸÅÿ∞ÿ© ÿ≤ŸÖŸÜŸäÿ©
# ========================
def create_sliding_window(X, y, window_size):
    X_windows, y_windows = [], []
    for i in range(len(X) - window_size + 1):
        X_windows.append(X[i:i + window_size])
        y_windows.append(y[i + window_size - 1])  # ÿ¢ÿÆÿ± ÿπŸÜÿµÿ± ŸáŸà ÿßŸÑŸáÿØŸÅ
    return np.array(X_windows), np.array(y_windows)

# ========================
# 1. ÿ™ÿ≠ŸÖŸäŸÑ Ÿàÿ™ÿ¨ŸáŸäÿ≤ ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™
# ========================
# df_train = pd.read_csv('your_train_data.csv')
# df_test = pd.read_csv('your_test_data.csv')

features_train = df_train.drop(columns=['Signal','date','timestamp','future_return'])
features_test = df_test.drop(columns=['Signal','date','timestamp','future_return'])

target_train = df_train['Signal']
target_test = df_test['Signal']

# ========================
# 2. ÿ•ŸÜÿ¥ÿßÿ° ÿßŸÑŸÜŸàÿßŸÅÿ∞ ÿßŸÑÿ≤ŸÖŸÜŸäÿ©
# ========================
window_size = 50
X_train_raw, y_train_raw = create_sliding_window(features_train.values, target_train.values, window_size)
X_test_raw, y_test_raw = create_sliding_window(features_test.values, target_test.values, window_size)

# ========================
# 3. ÿ™ÿ∑ÿ®Ÿäÿπ ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™
# ========================
scaler = RobustScaler()
X_train_scaled = scaler.fit_transform(X_train_raw.reshape(-1, X_train_raw.shape[-1])).reshape(X_train_raw.shape)
X_test_scaled = scaler.transform(X_test_raw.reshape(-1, X_test_raw.shape[-1])).reshape(X_test_raw.shape)

# Flatten ÿπÿ¥ÿßŸÜ XGBoost ŸÖÿß ÿ®ŸäŸÇÿ®ŸÑÿ¥ ÿ®ŸäÿßŸÜÿßÿ™ 3D
X_train_flat = X_train_scaled.reshape(X_train_scaled.shape[0], -1)
X_test_flat = X_test_scaled.reshape(X_test_scaled.shape[0], -1)

print(f"üîπ ÿ¥ŸÉŸÑ ÿ®ŸäÿßŸÜÿßÿ™ ÿßŸÑÿ™ÿØÿ±Ÿäÿ®: {X_train_flat.shape}")
print(f"üîπ ÿ¥ŸÉŸÑ ÿ®ŸäÿßŸÜÿßÿ™ ÿßŸÑÿßÿÆÿ™ÿ®ÿßÿ±: {X_test_flat.shape}")

# ========================
# 4. ÿ®ŸÜÿßÿ° ŸÜŸÖŸàÿ∞ÿ¨ XGBoost
# ========================
model = xgb.XGBClassifier(
    n_estimators=300,
    max_depth=8,
    learning_rate=0.05,
    subsample=0.8,
    colsample_bytree=0.8,
    objective="multi:softprob",  # softprob ÿπÿ¥ÿßŸÜ ŸÜÿ¨Ÿäÿ® ÿßŸÑÿßÿ≠ÿ™ŸÖÿßŸÑÿßÿ™
    num_class=3,
    random_state=42,
    n_jobs=-1
)

# ========================
# 5. ÿ™ÿØÿ±Ÿäÿ® ÿßŸÑŸÜŸÖŸàÿ∞ÿ¨ ŸÖÿπ Early Stopping
# ========================
print("üöÄ Training with early stopping...")
eval_set = [(X_train_flat, y_train_raw), (X_test_flat, y_test_raw)]
model.fit(
    X_train_flat, y_train_raw,
    eval_set=eval_set,
    early_stopping_rounds=20,
    verbose=True
)

# ÿßŸÑÿ≠ÿµŸàŸÑ ÿπŸÑŸâ ÿ≥ÿ¨ŸÑ ÿßŸÑÿ™ÿØÿ±Ÿäÿ®
history = model.evals_result()

# ========================
# 6. ÿ±ÿ≥ŸÖ ŸÖŸÜÿ≠ŸÜŸäÿßÿ™ ÿßŸÑÿ™ÿØÿ±Ÿäÿ® (ÿÆÿ≥ÿßÿ±ÿ© ŸàÿØŸÇÿ©)
# ========================
epochs = len(history['validation_0']['mlogloss'])
x_axis = range(epochs)

plt.figure(figsize=(10,5))
plt.plot(x_axis, history['validation_0']['mlogloss'], label='Train Log Loss')
plt.plot(x_axis, history['validation_1']['mlogloss'], label='Test Log Loss')
plt.legend()
plt.xlabel("Epochs")
plt.ylabel("Log Loss")
plt.title("Training vs Validation Log Loss")
plt.show()

plt.figure(figsize=(10,5))
plt.plot(x_axis, history['validation_0']['merror'], label='Train Error')
plt.plot(x_axis, history['validation_1']['merror'], label='Test Error')
plt.legend()
plt.xlabel("Epochs")
plt.ylabel("Classification Error")
plt.title("Training vs Validation Error")
plt.show()

# ========================
# 7. ÿßŸÑÿ™ŸÇŸäŸäŸÖ ÿßŸÑÿ™ÿØÿ±Ÿäÿ¨Ÿä Walking Forward
# ========================
y_true, y_pred = [], []
print("üîç ÿ¨ÿßÿ±Ÿä ÿßŸÑÿ™ŸÇŸäŸäŸÖ ÿßŸÑÿ™ÿØÿ±Ÿäÿ¨Ÿä...")
for i in tqdm.tqdm(range(len(X_test_flat))):
    sample_x = X_test_flat[i].reshape(1, -1)
    true_class = y_test_raw[i]
    predicted_class = model.predict(sample_x)[0]
    y_true.append(true_class)
    y_pred.append(predicted_class)

walk_acc = accuracy_score(y_true, y_pred)
print(f"üß† Walking Forward Accuracy: {walk_acc:.4f}")

# ========================
# 8. ÿßŸÑÿ™ŸÇŸäŸäŸÖ ÿπŸÑŸâ ÿ®ŸäÿßŸÜÿßÿ™ ÿßŸÑÿßÿÆÿ™ÿ®ÿßÿ± ŸÉÿßŸÖŸÑÿ©
# ========================
test_pred = model.predict(X_test_flat)
test_acc = accuracy_score(y_test_raw, test_pred)
print(f"üìä Test Accuracy: {test_acc:.4f}")

# ========================
# 9. ŸÖÿµŸÅŸàŸÅÿ© ÿßŸÑÿßŸÑÿ™ÿ®ÿßÿ≥
# ========================
cm = confusion_matrix(y_test_raw, test_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=np.unique(y_train_raw),
            yticklabels=np.unique(y_train_raw))
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix - XGBoost")
plt.show()

# ========================
# 10. ÿ™ŸÇÿ±Ÿäÿ± ÿßŸÑÿ£ÿØÿßÿ°
# ========================
print("\nüìà Classification Report:")
print("="*50)
print(classification_report(y_test_raw, test_pred, digits=3))

# ========================
# 11. ÿ£ŸáŸÖŸäÿ© ÿßŸÑŸÖŸäÿ≤ÿßÿ™
# ========================
plt.figure(figsize=(10, 6))
plot_importance(model, max_num_features=15, importance_type="gain")
plt.title("Top 15 Feature Importances - XGBoost")
plt.show()

# ========================
# 12. ÿ•ÿ≠ÿµÿßÿ¶Ÿäÿßÿ™ ÿ•ÿ∂ÿßŸÅŸäÿ©
# ========================
print(f"\nüìä ÿ•ÿ≠ÿµÿßÿ¶Ÿäÿßÿ™ ÿ•ÿ∂ÿßŸÅŸäÿ©:")
print(f"ÿπÿØÿØ ÿπŸäŸÜÿßÿ™ ÿßŸÑÿ™ÿØÿ±Ÿäÿ®: {len(X_train_flat)}")
print(f"ÿπÿØÿØ ÿπŸäŸÜÿßÿ™ ÿßŸÑÿßÿÆÿ™ÿ®ÿßÿ±: {len(X_test_flat)}")
print(f"ÿπÿØÿØ ÿßŸÑŸÖŸäÿ≤ÿßÿ™ ÿ®ÿπÿØ Flatten: {X_train_flat.shape[1]}")
print(f"ÿ≠ÿ¨ŸÖ ÿßŸÑŸÜÿßŸÅÿ∞ÿ© ÿßŸÑÿ≤ŸÖŸÜŸäÿ©: {window_size}")
print(f"ÿπÿØÿØ ÿßŸÑÿ£ÿ¥ÿ¨ÿßÿ±: {model.n_estimators}, ÿπŸÖŸÇ ÿßŸÑÿ¥ÿ¨ÿ±ÿ©: {model.max_depth}")

# ========================
# 13. ÿ±ÿ≥ŸÖ ÿßÿ≠ÿ™ŸÖÿßŸÑÿßÿ™ ÿßŸÑÿ´ŸÇÿ© ŸÑŸÉŸÑ ŸÅÿ¶ÿ©
# ========================
proba_test = model.predict_proba(X_test_flat)
df_proba = pd.DataFrame(proba_test, columns=['Class_0','Sell_1','Buy_2'])
df_proba['True_Label'] = y_test_raw

bins = np.linspace(0, 1, 21)

# Sell
plt.figure(figsize=(10,5))
plt.hist([df_proba['Sell_1'][df_proba['True_Label']==0],
          df_proba['Sell_1'][df_proba['True_Label']==1],
          df_proba['Sell_1'][df_proba['True_Label']==2]],
         bins=bins, color=['blue','green','red'],
         label=['No Trade (0)','Actual Sell (1)','Actual Buy (2)'],
         stacked=True, alpha=0.7)
plt.axvline(0.5, color='black', linestyle='--', label='Threshold 0.5')
plt.title("Sell Probabilities Histogram")
plt.xlabel("Predicted Sell Probability")
plt.ylabel("Number of Signals")
plt.legend()
plt.show()

# Buy
plt.figure(figsize=(10,5))
plt.hist([df_proba['Buy_2'][df_proba['True_Label']==0],
          df_proba['Buy_2'][df_proba['True_Label']==1],
          df_proba['Buy_2'][df_proba['True_Label']==2]],
         bins=bins, color=['blue','green','red'],
         label=['No Trade (0)','Actual Sell (1)','Actual Buy (2)'],
         stacked=True, alpha=0.7)
plt.axvline(0.5, color='black', linestyle='--', label='Threshold 0.5')
plt.title("Buy Probabilities Histogram")
plt.xlabel("Predicted Buy Probability")
plt.ylabel("Number of Signals")
plt.legend()
plt.show()

# No Trade
plt.figure(figsize=(10,5))
plt.hist([df_proba['Class_0'][df_proba['True_Label']==0],
          df_proba['Class_0'][df_proba['True_Label']==1],
          df_proba['Class_0'][df_proba['True_Label']==2]],
         bins=bins, color=['blue','green','red'],
         label=['Actual No Trade (0)','Actual Sell (1)','Actual Buy (2)'],
         stacked=True, alpha=0.7)
plt.axvline(0.5, color='black', linestyle='--', label='Threshold 0.5')
plt.title("No Trade Probabilities Histogram")
plt.xlabel("Predicted No Trade Probability")
plt.ylabel("Number of Signals")
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import RobustScaler
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm
from xgboost import XGBClassifier, plot_importance

# ========================
# ÿØÿßŸÑÿ© ÿ•ŸÜÿ¥ÿßÿ° ŸÜÿßŸÅÿ∞ÿ© ÿ≤ŸÖŸÜŸäÿ©
# ========================
def create_sliding_window(X, y, window_size):
    X_windows, y_windows = [], []
    for i in range(len(X) - window_size + 1):
        X_windows.append(X[i:i + window_size])
        y_windows.append(y[i + window_size - 1])  # ÿ¢ÿÆÿ± ÿπŸÜÿµÿ± ŸáŸà ÿßŸÑŸáÿØŸÅ
    return np.array(X_windows), np.array(y_windows)

# ========================
# 1. ÿ™ÿ≠ŸÖŸäŸÑ Ÿàÿ™ÿ¨ŸáŸäÿ≤ ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™
# ========================
# df_train = pd.read_csv('your_train_data.csv')
# df_test = pd.read_csv('your_test_data.csv')

features_train = df_train.drop(columns=['Signal','date','timestamp','future_return'])
features_test = df_test.drop(columns=['Signal','date','timestamp','future_return'])

target_train = df_train['Signal']
target_test = df_test['Signal']

# ========================
# 2. ÿ•ŸÜÿ¥ÿßÿ° ÿßŸÑŸÜŸàÿßŸÅÿ∞ ÿßŸÑÿ≤ŸÖŸÜŸäÿ©
# ========================
window_size = 50
X_train_raw, y_train_raw = create_sliding_window(features_train.values, target_train.values, window_size)
X_test_raw, y_test_raw = create_sliding_window(features_test.values, target_test.values, window_size)

# ========================
# 3. ÿ™ÿ∑ÿ®Ÿäÿπ ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™
# ========================
scaler = RobustScaler()
X_train_scaled = scaler.fit_transform(X_train_raw.reshape(-1, X_train_raw.shape[-1])).reshape(X_train_raw.shape)
X_test_scaled = scaler.transform(X_test_raw.reshape(-1, X_test_raw.shape[-1])).reshape(X_test_raw.shape)

# Flatten ÿπÿ¥ÿßŸÜ XGBoost
X_train_flat = X_train_scaled.reshape(X_train_scaled.shape[0], -1)
X_test_flat = X_test_scaled.reshape(X_test_scaled.shape[0], -1)

print(f"üîπ ÿ¥ŸÉŸÑ ÿ®ŸäÿßŸÜÿßÿ™ ÿßŸÑÿ™ÿØÿ±Ÿäÿ®: {X_train_flat.shape}")
print(f"üîπ ÿ¥ŸÉŸÑ ÿ®ŸäÿßŸÜÿßÿ™ ÿßŸÑÿßÿÆÿ™ÿ®ÿßÿ±: {X_test_flat.shape}")

# ========================
# 4. ÿ®ŸÜÿßÿ° ŸÜŸÖŸàÿ∞ÿ¨ XGBoost
# ========================
model = XGBClassifier(
    n_estimators=300,
    max_depth=8,
    learning_rate=0.05,
    subsample=0.8,
    colsample_bytree=0.8,
    objective="multi:softprob",  # softprob ŸÑŸÑŸÄ predict_proba
    num_class=3,
    random_state=42,
    n_jobs=-1,
    eval_metric="mlogloss"
)

# ========================
# 5. ÿ™ÿØÿ±Ÿäÿ® ÿßŸÑŸÜŸÖŸàÿ∞ÿ¨
# ========================
print("üöÄ Training XGBoost...")
model.fit(X_train_flat, y_train_raw,
          eval_set=[(X_test_flat, y_test_raw)],
          verbose=False)

# ========================
# 6. ÿßŸÑÿ™ŸÇŸäŸäŸÖ ÿßŸÑÿ™ÿØÿ±Ÿäÿ¨Ÿä ŸÖÿπ Threshold
# ========================
y_true, y_pred = [], []
threshold = 0.5
print("üîç ÿ¨ÿßÿ±Ÿä ÿßŸÑÿ™ŸÇŸäŸäŸÖ ÿßŸÑÿ™ÿØÿ±Ÿäÿ¨Ÿä ŸÖÿπ Threshold...")
for i in tqdm(range(len(X_test_flat))):
    sample_x = X_test_flat[i].reshape(1, -1)
    true_class = y_test_raw[i]
    proba = model.predict_proba(sample_x)[0]

    if proba[1] >= threshold:
        predicted_class = 1
    elif proba[2] >= threshold:
        predicted_class = 2
    else:
        predicted_class = 0

    y_true.append(true_class)
    y_pred.append(predicted_class)

walk_acc = accuracy_score(y_true, y_pred)
print(f"üß† Walking Forward Accuracy (Threshold 0.5): {walk_acc:.4f}")

# ========================
# 7. ÿßŸÑÿ™ŸÇŸäŸäŸÖ ÿπŸÑŸâ ŸÉÿßŸÖŸÑ ÿ®ŸäÿßŸÜÿßÿ™ ÿßŸÑÿßÿÆÿ™ÿ®ÿßÿ±
# ========================
proba_test = model.predict_proba(X_test_flat)
test_pred = []
for p in proba_test:
    if p[1] >= threshold:
        test_pred.append(1)
    elif p[2] >= threshold:
        test_pred.append(2)
    else:
        test_pred.append(0)

test_acc = accuracy_score(y_test_raw, test_pred)
print(f"üìä Test Accuracy (Threshold 0.5): {test_acc:.4f}")

# ========================
# 8. ŸÖÿµŸÅŸàŸÅÿ© ÿßŸÑÿßŸÑÿ™ÿ®ÿßÿ≥
# ========================
cm = confusion_matrix(y_test_raw, test_pred)

plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=np.unique(y_train_raw),
            yticklabels=np.unique(y_train_raw))
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix - XGBoost with Threshold")
plt.show()

# ========================
# 9. ÿ™ŸÇÿ±Ÿäÿ± ÿßŸÑÿ£ÿØÿßÿ°
# ========================
print("\nüìà Classification Report:")
print("="*50)
print(classification_report(y_test_raw, test_pred, digits=3))

# ========================
# 10. ÿ£ŸáŸÖŸäÿ© ÿßŸÑŸÖŸäÿ≤ÿßÿ™ (Feature Importance)
# ========================
plt.figure(figsize=(10, 6))
plot_importance(model, max_num_features=15, importance_type="gain")
plt.title("Top 15 Feature Importances - XGBoost")
plt.show()

# ========================
# 11. ÿ•ÿ≠ÿµÿßÿ¶Ÿäÿßÿ™ ÿ•ÿ∂ÿßŸÅŸäÿ©
# ========================
print(f"\nüìä ÿ•ÿ≠ÿµÿßÿ¶Ÿäÿßÿ™ ÿ•ÿ∂ÿßŸÅŸäÿ©:")
print(f"ÿπÿØÿØ ÿπŸäŸÜÿßÿ™ ÿßŸÑÿ™ÿØÿ±Ÿäÿ®: {len(X_train_flat)}")
print(f"ÿπÿØÿØ ÿπŸäŸÜÿßÿ™ ÿßŸÑÿßÿÆÿ™ÿ®ÿßÿ±: {len(X_test_flat)}")
print(f"ÿπÿØÿØ ÿßŸÑŸÖŸäÿ≤ÿßÿ™ ÿ®ÿπÿØ Flatten: {X_train_flat.shape[1]}")
print(f"ÿ≠ÿ¨ŸÖ ÿßŸÑŸÜÿßŸÅÿ∞ÿ© ÿßŸÑÿ≤ŸÖŸÜŸäÿ©: {window_size}")
print(f"ÿπÿØÿØ ÿßŸÑÿ£ÿ¥ÿ¨ÿßÿ±: {model.n_estimators}, ÿπŸÖŸÇ ÿßŸÑÿ¥ÿ¨ÿ±ÿ©: {model.max_depth}")
```

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import RobustScaler
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import matplotlib.pyplot as plt
import seaborn as sns
import tqdm
from xgboost import XGBClassifier, plot_importance

# ========================
# ÿØÿßŸÑÿ© ÿ•ŸÜÿ¥ÿßÿ° ŸÜÿßŸÅÿ∞ÿ© ÿ≤ŸÖŸÜŸäÿ©
# ========================
def create_sliding_window(X, y, window_size):
    X_windows, y_windows = [], []
    for i in range(len(X) - window_size + 1):
        X_windows.append(X[i:i + window_size])
        y_windows.append(y[i + window_size - 1])  # ÿ¢ÿÆÿ± ÿπŸÜÿµÿ± ŸáŸà ÿßŸÑŸáÿØŸÅ
    return np.array(X_windows), np.array(y_windows)


# ========================
# 1. ÿ™ÿ≠ŸÖŸäŸÑ Ÿàÿ™ÿ¨ŸáŸäÿ≤ ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™
# ========================
# df_train = pd.read_csv('your_train_data.csv')
# df_test = pd.read_csv('your_test_data.csv')

features_train = df_train.drop(columns=['Signal','date','timestamp','future_return','price_reaction_next','accum_dist_vwap'])
features_test = df_test.drop(columns=['Signal','date','timestamp','future_return','price_reaction_next','accum_dist_vwap'])

target_train = df_train['Signal']
target_test = df_test['Signal']

# ========================
# 2. ÿ•ŸÜÿ¥ÿßÿ° ÿßŸÑŸÜŸàÿßŸÅÿ∞ ÿßŸÑÿ≤ŸÖŸÜŸäÿ©
# ========================
window_size = 50
X_train_raw, y_train_raw = create_sliding_window(features_train.values, target_train.values, window_size)
X_test_raw, y_test_raw = create_sliding_window(features_test.values, target_test.values, window_size)

# ========================
# 3. ÿ™ÿ∑ÿ®Ÿäÿπ ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™
# ========================
scaler = RobustScaler()

X_train_scaled = scaler.fit_transform(X_train_raw.reshape(-1, X_train_raw.shape[-1])).reshape(X_train_raw.shape)
X_test_scaled = scaler.transform(X_test_raw.reshape(-1, X_test_raw.shape[-1])).reshape(X_test_raw.shape)

# Flatten ÿπÿ¥ÿßŸÜ XGBoost ŸÖÿß ÿ®ŸäŸÇÿ®ŸÑÿ¥ ÿ®ŸäÿßŸÜÿßÿ™ 3D
X_train_flat = X_train_scaled.reshape(X_train_scaled.shape[0], -1)
X_test_flat = X_test_scaled.reshape(X_test_scaled.shape[0], -1)

print(f"üîπ ÿ¥ŸÉŸÑ ÿ®ŸäÿßŸÜÿßÿ™ ÿßŸÑÿ™ÿØÿ±Ÿäÿ®: {X_train_flat.shape}")
print(f"üîπ ÿ¥ŸÉŸÑ ÿ®ŸäÿßŸÜÿßÿ™ ÿßŸÑÿßÿÆÿ™ÿ®ÿßÿ±: {X_test_flat.shape}")

# ========================
# 4. ÿ®ŸÜÿßÿ° ŸÜŸÖŸàÿ∞ÿ¨ XGBoost
# ========================
model = XGBClassifier(
    n_estimators=90,
    max_depth=8,
    learning_rate=0.05,
    subsample=0.8,
    colsample_bytree=0.8,
    objective="multi:probsoftmax",  # ÿ™ÿµŸÜŸäŸÅ ŸÖÿ™ÿπÿØÿØ ÿßŸÑŸÅÿ¶ÿßÿ™
    num_class=3,
    random_state=42,
    n_jobs=-1,
    eval_metric="mlogloss"   # ‚úÖ ŸáŸÜÿß ÿ®ÿØŸÑ ŸÖÿß ŸäŸÉŸàŸÜ ÿ¨ŸàŸá fit
)

# ========================
# 5. ÿ™ÿØÿ±Ÿäÿ® ÿßŸÑŸÜŸÖŸàÿ∞ÿ¨
# ========================
print("üöÄ Training XGBoost...")
model.fit(X_train_flat, y_train_raw,
          eval_set=[(X_test_flat, y_test_raw)],
          verbose=True)

# ========================
# 6. ÿßŸÑÿ™ŸÇŸäŸäŸÖ ÿßŸÑÿ™ÿØÿ±Ÿäÿ¨Ÿä Walking Forward
# ========================
y_true, y_pred = [], []
print("üîç ÿ¨ÿßÿ±Ÿä ÿßŸÑÿ™ŸÇŸäŸäŸÖ ÿßŸÑÿ™ÿØÿ±Ÿäÿ¨Ÿä...")
for i in tqdm.tqdm(range(len(X_test_flat))):
    sample_x = X_test_flat[i].reshape(1, -1)
    true_class = y_test_raw[i]
    predicted_class = model.predict(sample_x)[0]

    y_true.append(true_class)
    y_pred.append(predicted_class)

walk_acc = accuracy_score(y_true, y_pred)
print(f"üß† Walking Forward Accuracy: {walk_acc:.4f}")

# ========================
# 7. ÿßŸÑÿ™ŸÇŸäŸäŸÖ ÿπŸÑŸâ ÿ®ŸäÿßŸÜÿßÿ™ ÿßŸÑÿßÿÆÿ™ÿ®ÿßÿ± ŸÉÿßŸÖŸÑÿ©
# ========================
test_pred = model.predict(X_test_flat)
test_acc = accuracy_score(y_test_raw, test_pred)

print(f"üìä Test Accuracy: {test_acc:.4f}")

# ========================
# 8. ŸÖÿµŸÅŸàŸÅÿ© ÿßŸÑÿßŸÑÿ™ÿ®ÿßÿ≥
# ========================
cm = confusion_matrix(y_test_raw, test_pred)

plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=np.unique(y_train_raw),
            yticklabels=np.unique(y_train_raw))
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix - XGBoost")
plt.show()

# ========================
# 9. ÿ™ŸÇÿ±Ÿäÿ± ÿßŸÑÿ£ÿØÿßÿ°
# ========================
print("\nüìà Classification Report:")
print("="*50)
print(classification_report(y_test_raw, test_pred, digits=3))

# ========================
# 10. ÿ£ŸáŸÖŸäÿ© ÿßŸÑŸÖŸäÿ≤ÿßÿ™ (Feature Importance)
# ========================
plt.figure(figsize=(10, 6))
plot_importance(model, max_num_features=15, importance_type="gain")
plt.title("Top 15 Feature Importances - XGBoost")
plt.show()

# ========================
# 11. ÿ•ÿ≠ÿµÿßÿ¶Ÿäÿßÿ™ ÿ•ÿ∂ÿßŸÅŸäÿ©
# ========================
print(f"\nüìä ÿ•ÿ≠ÿµÿßÿ¶Ÿäÿßÿ™ ÿ•ÿ∂ÿßŸÅŸäÿ©:")
print(f"ÿπÿØÿØ ÿπŸäŸÜÿßÿ™ ÿßŸÑÿ™ÿØÿ±Ÿäÿ®: {len(X_train_flat)}")
print(f"ÿπÿØÿØ ÿπŸäŸÜÿßÿ™ ÿßŸÑÿßÿÆÿ™ÿ®ÿßÿ±: {len(X_test_flat)}")
print(f"ÿπÿØÿØ ÿßŸÑŸÖŸäÿ≤ÿßÿ™ ÿ®ÿπÿØ Flatten: {X_train_flat.shape[1]}")
print(f"ÿ≠ÿ¨ŸÖ ÿßŸÑŸÜÿßŸÅÿ∞ÿ© ÿßŸÑÿ≤ŸÖŸÜŸäÿ©: {window_size}")
print(f"ÿπÿØÿØ ÿßŸÑÿ£ÿ¥ÿ¨ÿßÿ±: {model.n_estimators}, ÿπŸÖŸÇ ÿßŸÑÿ¥ÿ¨ÿ±ÿ©: {model.max_depth}")


# ========================
# 12. ÿ±ÿ≥ŸÖ ÿßÿ≠ÿ™ŸÖÿßŸÑÿßÿ™ ÿßŸÑÿ´ŸÇÿ© ŸÑŸÉŸÑ ŸÅÿ¶ÿ©
# ========================

# ÿßŸÑÿ≠ÿµŸàŸÑ ÿπŸÑŸâ ÿßÿ≠ÿ™ŸÖÿßŸÑÿßÿ™ ŸÉŸÑ ÿπŸäŸÜÿ©
proba_test = model.predict_proba(X_test_flat)

# ÿ™ÿ≠ŸàŸäŸÑŸáÿß ÿ•ŸÑŸâ DataFrame ŸÑÿ™ÿ≥ŸáŸäŸÑ ÿßŸÑÿ±ÿ≥ŸÖ
df_proba = pd.DataFrame(proba_test, columns=['Class_0','Sell_1','Buy_2'])
df_proba['True_Label'] = y_test_raw


import matplotlib.pyplot as plt
import numpy as np

# ====== ÿ•ÿπÿØÿßÿØ ÿßŸÑŸÄ bins ======
bins = np.linspace(0, 1, 21)  # ÿ™ŸÇÿ≥ŸäŸÖ ÿßŸÑÿßÿ≠ÿ™ŸÖÿßŸÑ ŸÖŸÜ 0 ÿ•ŸÑŸâ 1

# ====== Sell Probability Histogram ======
plt.figure(figsize=(10,5))
plt.hist([df_proba['Sell_1'][df_proba['True_Label']==0],
          df_proba['Sell_1'][df_proba['True_Label']==1],
          df_proba['Sell_1'][df_proba['True_Label']==2]],
         bins=bins,
         color=['blue','green','red'],
         label=['No Trade (0)','Actual Sell (1)','Actual Buy (2)'],
         stacked=True, alpha=0.7)
plt.axvline(0.5, color='black', linestyle='--', label='Threshold 0.5')
plt.title("Sell Probabilities Histogram Colored by True Label")
plt.xlabel("Predicted Sell Probability")
plt.ylabel("Number of Signals")
plt.legend()
plt.show()

# ====== Buy Probability Histogram ======
plt.figure(figsize=(10,5))
plt.hist([df_proba['Buy_2'][df_proba['True_Label']==0],
          df_proba['Buy_2'][df_proba['True_Label']==1],
          df_proba['Buy_2'][df_proba['True_Label']==2]],
         bins=bins,
         color=['blue','green','red'],
         label=['No Trade (0)','Actual Sell (1)','Actual Buy (2)'],
         stacked=True, alpha=0.7)
plt.axvline(0.5, color='black', linestyle='--', label='Threshold 0.5')
plt.title("Buy Probabilities Histogram Colored by True Label")
plt.xlabel("Predicted Buy Probability")
plt.ylabel("Number of Signals")
plt.legend()
plt.show()

# ====== No Trade Probability Histogram ======
plt.figure(figsize=(10,5))
plt.hist([df_proba['Class_0'][df_proba['True_Label']==0],
          df_proba['Class_0'][df_proba['True_Label']==1],
          df_proba['Class_0'][df_proba['True_Label']==2]],
         bins=bins,
         color=['blue','green','red'],
         label=['Actual No Trade (0)','Actual Sell (1)','Actual Buy (2)'],
         stacked=True, alpha=0.7)
plt.axvline(0.5, color='black', linestyle='--', label='Threshold 0.5')
plt.title("No Trade Probabilities Histogram Colored by True Label")
plt.xlabel("Predicted No Trade Probability")
plt.ylabel("Number of Signals")
plt.legend()
plt.show()

```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, RobustScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras import regularizers
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
from tensorflow.keras.optimizers import Adam
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import matplotlib.pyplot as plt
import seaborn as sns
import tqdm
# ========================
# ÿ•ÿ∂ÿßŸÅÿ© ÿØÿßŸÑÿ© create_sliding_window ÿßŸÑŸÖŸÅŸÇŸàÿØÿ©
# ========================
def create_sliding_window(X, y, window_size):
    """
    ÿ•ŸÜÿ¥ÿßÿ° ŸÜŸàÿßŸÅÿ∞ ÿ≤ŸÖŸÜŸäÿ© ŸÖŸÜÿ≤ŸÑŸÇÿ© ŸÖŸÜ ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™
    """
    X_windows = []
    y_windows = []
    
    for i in range(len(X) - window_size + 1):
        X_windows.append(X[i:i + window_size])
        y_windows.append(y[i + window_size - 1])  # ÿ¢ÿÆÿ± ŸáÿØŸÅ ŸÅŸä ÿßŸÑŸÜÿßŸÅÿ∞ÿ©
    
    return np.array(X_windows), np.array(y_windows)

# ========================
# 1. ÿ™ÿ≠ŸÖŸäŸÑ Ÿàÿ™ÿ¨ŸáŸäÿ≤ ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™
# ========================
# ÿ™ÿ£ŸÉÿØ ŸÖŸÜ ÿ™ÿ≠ŸÖŸäŸÑ ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™ ÿ£ŸàŸÑÿßŸã
# df_train = pd.read_csv('your_train_data.csv')
# df_test = pd.read_csv('your_test_data.csv')

# ÿßŸÑŸÖŸäÿ≤ÿßÿ™
features_train = df_train.drop(columns=['Signal','date','timestamp','future_return'])
features_test = df_test.drop(columns=['Signal','date','timestamp','future_return'])

# ÿßŸÑÿ£ŸáÿØÿßŸÅ
target_train = df_train['Signal']
target_test = df_test['Signal']

# ÿ•ŸÜÿ¥ÿßÿ° ÿßŸÑŸÜŸàÿßŸÅÿ∞ ÿßŸÑÿ≤ŸÖŸÜŸäÿ©
window_size = 50
X_train_raw, y_train_raw = create_sliding_window(features_train.values, target_train.values, window_size)
X_test_raw, y_test_raw = create_sliding_window(features_test.values, target_test.values, window_size)

# ÿ™ÿ∑ÿ®ŸäŸÇ ÿßŸÑÿ™ÿ∑ÿ®Ÿäÿπ
scaler = RobustScaler()

# ÿ™ÿ∑ÿ®Ÿäÿπ ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™
X_train_scaled = scaler.fit_transform(X_train_raw.reshape(-1, X_train_raw.shape[-1])).reshape(X_train_raw.shape)
X_test_scaled = scaler.transform(X_test_raw.reshape(-1, X_test_raw.shape[-1])).reshape(X_test_raw.shape)

# ÿ™ÿ≠ŸàŸäŸÑ ÿßŸÑÿ£ŸáÿØÿßŸÅ ÿ•ŸÑŸâ categorical
y_train_cat = to_categorical(y_train_raw, num_classes=3)
y_test_cat = to_categorical(y_test_raw, num_classes=3)

print(f"ÿ¥ŸÉŸÑ ÿ®ŸäÿßŸÜÿßÿ™ ÿßŸÑÿ™ÿØÿ±Ÿäÿ®: {X_train_scaled.shape}")
print(f"ÿ¥ŸÉŸÑ ÿ®ŸäÿßŸÜÿßÿ™ ÿßŸÑÿßÿÆÿ™ÿ®ÿßÿ±: {X_test_scaled.shape}")
print(f"ÿ¥ŸÉŸÑ ÿ£ŸáÿØÿßŸÅ ÿßŸÑÿ™ÿØÿ±Ÿäÿ®: {y_train_cat.shape}")
print(f"ÿ¥ŸÉŸÑ ÿ£ŸáÿØÿßŸÅ ÿßŸÑÿßÿÆÿ™ÿ®ÿßÿ±: {y_test_cat.shape}")

# ========================
# 5. ÿ®ŸÜÿßÿ° ŸÜŸÖŸàÿ∞ÿ¨ LSTM
# ========================
model = Sequential()
model.add(LSTM(
    64,
    input_shape=(X_train_scaled.shape[1], X_train_scaled.shape[2]),
    return_sequences=False,
    dropout=0.2,
    kernel_regularizer=regularizers.l2(0.01)
))
model.add(Dense(32, activation='relu', kernel_regularizer=regularizers.l2(0.01)))
model.add(Dropout(0.2))
model.add(Dense(3, activation='softmax'))

# ========================
# 5.1 ÿ™ÿ≠ÿØŸäÿØ ŸÜÿ≥ÿ®ÿ© ÿßŸÑÿ™ÿπŸÑŸÖ
# ========================
optimizer = Adam(learning_rate=0.001)
model.compile(optimizer=optimizer, loss='categorical_crossentropy', metrics=['accuracy'])

# ÿπÿ±ÿ∂ ŸÖŸÑÿÆÿµ ÿßŸÑŸÜŸÖŸàÿ∞ÿ¨
model.summary()

# ========================
# 6.1 ÿ•ÿπÿØÿßÿØ ÿßŸÑÿ™ŸàŸÇŸÅ ÿßŸÑŸÖÿ®ŸÉÿ± Ÿàÿ≠ŸÅÿ∏ ÿ£ŸÅÿ∂ŸÑ ŸÜŸÖŸàÿ∞ÿ¨
# ========================
early_stop = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
checkpoint = ModelCheckpoint('best_model.keras', monitor='val_loss', save_best_only=True)

# ========================
# 6. ÿ™ÿØÿ±Ÿäÿ® ÿßŸÑŸÜŸÖŸàÿ∞ÿ¨
# ========================
history = model.fit(
    X_train_scaled, y_train_cat,  # ÿßÿ≥ÿ™ÿÆÿØÿßŸÖ ÿßŸÑŸÖÿ™ÿ∫Ÿäÿ±ÿßÿ™ ÿßŸÑÿµÿ≠Ÿäÿ≠ÿ©
    epochs=200,
    batch_size=100,
    validation_data=(X_test_scaled, y_test_cat),  # ÿßÿ≥ÿ™ÿÆÿØÿßŸÖ ÿßŸÑŸÖÿ™ÿ∫Ÿäÿ±ÿßÿ™ ÿßŸÑÿµÿ≠Ÿäÿ≠ÿ©
    callbacks=[early_stop, checkpoint],
    verbose=1
)

# ========================
# 7. ÿ™ŸÇŸäŸäŸÖ ÿ™ÿØÿ±Ÿäÿ¨Ÿä ÿÆÿ∑Ÿàÿ© ÿ®ÿÆÿ∑Ÿàÿ© (Walking Forward Evaluation)
# ========================
y_true = []
y_pred = []

print("ÿ¨ÿßÿ±Ÿä ÿßŸÑÿ™ŸÇŸäŸäŸÖ ÿßŸÑÿ™ÿØÿ±Ÿäÿ¨Ÿä...")
for i in  tqdm.tqdm(range(len(X_test_scaled))):
    sample_x = X_test_scaled[i].reshape(1, X_test_scaled.shape[1], X_test_scaled.shape[2])
    true_class = np.argmax(y_test_cat[i])  # ÿßÿ≥ÿ™ÿÆÿØÿßŸÖ y_test_cat
    prediction = model.predict(sample_x, verbose=0)
    predicted_class = np.argmax(prediction)
    
    y_true.append(true_class)
    y_pred.append(predicted_class)

walk_acc = accuracy_score(y_true, y_pred)
print(f"üß† Walking Forward Accuracy: {walk_acc:.4f}")

# ==================================================================================================================
# ÿ™ŸÇŸäŸäŸÖ ÿßŸÑŸÜŸÖŸàÿ∞ÿ¨ ÿπŸÑŸâ ÿßÿÆÿ™ÿ®ÿßÿ± ÿ≠ŸÇŸäŸÇŸä
# ==================================================================================================================
test_loss, test_accuracy = model.evaluate(X_test_scaled, y_test_cat, verbose=0)  # ÿ™ÿµÿ≠Ÿäÿ≠ ÿßŸÑŸÖÿ™ÿ∫Ÿäÿ±ÿßÿ™
print(f"üìä Test Loss: {test_loss:.4f}")
print(f"üéØ Test Accuracy: {test_accuracy:.4f}")

# ========================
# ÿ±ÿ≥ŸÖ ŸÖŸÜÿ≠ŸÜŸâ ÿßŸÑÿ£ÿØÿßÿ° ÿßŸÑŸÉÿßŸÖŸÑ
# ========================
epochs_range = range(len(history.history['loss']))

plt.figure(figsize=(12, 6))

# ÿ±ÿ≥ŸÖ ÿßŸÑÿÆÿ≥ÿßÿ±ÿ©
plt.subplot(1, 2, 1)
plt.plot(epochs_range, history.history['loss'], label='Train Loss', linestyle='--')
plt.plot(epochs_range, history.history['val_loss'], label='Validation Loss', linestyle='--')
plt.title('Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)

# ÿ±ÿ≥ŸÖ ÿßŸÑÿØŸÇÿ©
plt.subplot(1, 2, 2)
plt.plot(epochs_range, history.history['accuracy'], label='Train Accuracy', linestyle='-')
plt.plot(epochs_range, history.history['val_accuracy'], label='Validation Accuracy', linestyle='-')
plt.title('Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()

# ===========
# ÿπŸÖŸÑ ÿßŸÑÿ™ŸÜÿ®ÿ§ÿßÿ™ Ÿàÿ™ŸÇŸäŸäŸÖ ÿ¥ÿßŸÖŸÑ
# ===========
y_pred_probs = model.predict(X_test_scaled)
y_pred = np.argmax(y_pred_probs, axis=1)
y_true = np.argmax(y_test_cat, axis=1)

# ===========
# ŸÖÿµŸÅŸàŸÅÿ© ÿßŸÑÿßŸÑÿ™ÿ®ÿßÿ≥
# ===========
cm = confusion_matrix(y_true, y_pred)

plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
            xticklabels=[0, 1, 2], yticklabels=[0, 1, 2])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()

# ===========
# ÿ™ŸÇÿ±Ÿäÿ± ÿßŸÑÿ£ÿØÿßÿ° ÿßŸÑŸÉÿßŸÖŸÑ
# ===========
print("\nüìà Classification Report:")
print("=" * 50)
print(classification_report(y_true, y_pred, digits=3))

# ===========
# ÿ•ÿ≠ÿµÿßÿ¶Ÿäÿßÿ™ ÿ•ÿ∂ÿßŸÅŸäÿ©
# ===========
print(f"\nüìä ÿ•ÿ≠ÿµÿßÿ¶Ÿäÿßÿ™ ÿ•ÿ∂ÿßŸÅŸäÿ©:")
print(f"ÿπÿØÿØ ÿπŸäŸÜÿßÿ™ ÿßŸÑÿ™ÿØÿ±Ÿäÿ®: {len(X_train_scaled)}")
print(f"ÿπÿØÿØ ÿπŸäŸÜÿßÿ™ ÿßŸÑÿßÿÆÿ™ÿ®ÿßÿ±: {len(X_test_scaled)}")
print(f"ÿπÿØÿØ ÿßŸÑŸÖŸäÿ≤ÿßÿ™: {X_train_scaled.shape[2]}")
print(f"ÿ≠ÿ¨ŸÖ ÿßŸÑŸÜÿßŸÅÿ≤ÿ© ÿßŸÑÿ≤ŸÖŸÜŸäÿ©: {window_size}")
print(f"ÿπÿØÿØ ÿßŸÑÿ≠ŸÇÿ® ÿßŸÑŸÖŸÉÿ™ŸÖŸÑÿ©: {len(history.history['loss'])}")
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import RobustScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import matplotlib.pyplot as plt
import seaborn as sns
import tqdm

# ========================
# ÿØÿßŸÑÿ© ÿ•ŸÜÿ¥ÿßÿ° ÿßŸÑŸÜŸàÿßŸÅÿ∞ ÿßŸÑÿ≤ŸÖŸÜŸäÿ©
# ========================
def create_sliding_window(X, y, window_size):
    X_windows = []
    y_windows = []
    for i in range(len(X) - window_size + 1):
        X_windows.append(X[i:i + window_size])
        y_windows.append(y[i + window_size - 1])  # ÿ¢ÿÆÿ± ŸáÿØŸÅ ŸÅŸä ÿßŸÑŸÜÿßŸÅÿ∞ÿ©
    return np.array(X_windows), np.array(y_windows)

# ========================
# 1. ÿ™ÿ≠ŸÖŸäŸÑ Ÿàÿ™ÿ¨ŸáŸäÿ≤ ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™
# ========================
# df_train = pd.read_csv('your_train_data.csv')
# df_test = pd.read_csv('your_test_data.csv')

# ÿßŸÑŸÖŸäÿ≤ÿßÿ™
features_train = df_train.drop(columns=['Signal','date','timestamp','future_return'])
features_test = df_test.drop(columns=['Signal','date','timestamp','future_return'])

# ÿßŸÑÿ£ŸáÿØÿßŸÅ
target_train = df_train['Signal']
target_test = df_test['Signal']

# ÿ•ŸÜÿ¥ÿßÿ° ÿßŸÑŸÜŸàÿßŸÅÿ∞ ÿßŸÑÿ≤ŸÖŸÜŸäÿ©
window_size = 50
X_train_raw, y_train_raw = create_sliding_window(features_train.values, target_train.values, window_size)
X_test_raw, y_test_raw = create_sliding_window(features_test.values, target_test.values, window_size)

# ÿ™ÿ∑ÿ®ŸäŸÇ ÿßŸÑÿ™ÿ∑ÿ®Ÿäÿπ
scaler = RobustScaler()
X_train_scaled = scaler.fit_transform(X_train_raw.reshape(X_train_raw.shape[0], -1))
X_test_scaled = scaler.transform(X_test_raw.reshape(X_test_raw.shape[0], -1))

print(f"ÿ¥ŸÉŸÑ ÿ®ŸäÿßŸÜÿßÿ™ ÿßŸÑÿ™ÿØÿ±Ÿäÿ®: {X_train_scaled.shape}")
print(f"ÿ¥ŸÉŸÑ ÿ®ŸäÿßŸÜÿßÿ™ ÿßŸÑÿßÿÆÿ™ÿ®ÿßÿ±: {X_test_scaled.shape}")

# ========================
# 2. ÿ®ŸÜÿßÿ° ŸÜŸÖŸàÿ∞ÿ¨ Decision Tree
# ========================
model = DecisionTreeClassifier(
    criterion="gini",   # ÿ£Ÿà "entropy"
    max_depth=30,
    random_state=42
)

# ÿ™ÿØÿ±Ÿäÿ® ÿßŸÑŸÜŸÖŸàÿ∞ÿ¨
model.fit(X_train_scaled, y_train_raw)

# ========================
# 3. ÿßŸÑÿ™ŸÇŸäŸäŸÖ ÿÆÿ∑Ÿàÿ© ÿ®ÿÆÿ∑Ÿàÿ© (Walking Forward Evaluation)
# ========================
y_true = []
y_pred = []

print("ÿ¨ÿßÿ±Ÿä ÿßŸÑÿ™ŸÇŸäŸäŸÖ ÿßŸÑÿ™ÿØÿ±Ÿäÿ¨Ÿä...")
for i in tqdm.tqdm(range(len(X_test_scaled))):
    sample_x = X_test_scaled[i].reshape(1, -1)
    true_class = y_test_raw[i]
    predicted_class = model.predict(sample_x)[0]
    
    y_true.append(true_class)
    y_pred.append(predicted_class)

walk_acc = accuracy_score(y_true, y_pred)
print(f"üß† Walking Forward Accuracy: {walk_acc:.4f}")

# ==================================================================================================================
# ÿ™ŸÇŸäŸäŸÖ ÿßŸÑŸÜŸÖŸàÿ∞ÿ¨ ÿπŸÑŸâ ÿßÿÆÿ™ÿ®ÿßÿ± ÿ≠ŸÇŸäŸÇŸä
# ==================================================================================================================
test_accuracy = accuracy_score(y_test_raw, model.predict(X_test_scaled))
print(f"üéØ Test Accuracy: {test_accuracy:.4f}")

# ========================
# ŸÖÿµŸÅŸàŸÅÿ© ÿßŸÑÿßŸÑÿ™ÿ®ÿßÿ≥
# ========================
cm = confusion_matrix(y_test_raw, model.predict(X_test_scaled))

plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
            xticklabels=[0, 1, 2], yticklabels=[0, 1, 2])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()

# ========================
# ÿ™ŸÇÿ±Ÿäÿ± ÿßŸÑÿ£ÿØÿßÿ° ÿßŸÑŸÉÿßŸÖŸÑ
# ========================
print("\nüìà Classification Report:")
print("=" * 50)
print(classification_report(y_test_raw, model.predict(X_test_scaled), digits=3))

# ========================
# ÿ•ÿ≠ÿµÿßÿ¶Ÿäÿßÿ™ ÿ•ÿ∂ÿßŸÅŸäÿ©
# ========================
print(f"\nüìä ÿ•ÿ≠ÿµÿßÿ¶Ÿäÿßÿ™ ÿ•ÿ∂ÿßŸÅŸäÿ©:")
print(f"ÿπÿØÿØ ÿπŸäŸÜÿßÿ™ ÿßŸÑÿ™ÿØÿ±Ÿäÿ®: {len(X_train_scaled)}")
print(f"ÿπÿØÿØ ÿπŸäŸÜÿßÿ™ ÿßŸÑÿßÿÆÿ™ÿ®ÿßÿ±: {len(X_test_scaled)}")
print(f"ÿπÿØÿØ ÿßŸÑŸÖŸäÿ≤ÿßÿ™ ÿ®ÿπÿØ ŸÅÿ±ÿØ ÿßŸÑŸÜÿßŸÅÿ∞ÿ©: {X_train_scaled.shape[1]}")
print(f"ÿ≠ÿ¨ŸÖ ÿßŸÑŸÜÿßŸÅÿ∞ÿ© ÿßŸÑÿ≤ŸÖŸÜŸäÿ©: {window_size}")
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import RobustScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import matplotlib.pyplot as plt
import seaborn as sns
import tqdm

# ========================
# ÿØÿßŸÑÿ© ÿ•ŸÜÿ¥ÿßÿ° ŸÜÿßŸÅÿ∞ÿ© ÿ≤ŸÖŸÜŸäÿ©
# ========================
def create_sliding_window(X, y, window_size):
    X_windows, y_windows = [], []
    for i in range(len(X) - window_size + 1):
        X_windows.append(X[i:i + window_size])
        y_windows.append(y[i + window_size - 1])  # ÿ¢ÿÆÿ± ÿπŸÜÿµÿ± ŸáŸà ÿßŸÑŸáÿØŸÅ
    return np.array(X_windows), np.array(y_windows)


# ========================
# 1. ÿ™ÿ≠ŸÖŸäŸÑ Ÿàÿ™ÿ¨ŸáŸäÿ≤ ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™
# ========================
# df_train = pd.read_csv('your_train_data.csv')
# df_test = pd.read_csv('your_test_data.csv')

features_train = df_train.drop(columns=['Signal','date','timestamp','future_return'])
features_test = df_test.drop(columns=['Signal','date','timestamp','future_return'])

target_train = df_train['Signal']
target_test = df_test['Signal']

# ========================
# 2. ÿ•ŸÜÿ¥ÿßÿ° ÿßŸÑŸÜŸàÿßŸÅÿ∞ ÿßŸÑÿ≤ŸÖŸÜŸäÿ©
# ========================
window_size = 50
X_train_raw, y_train_raw = create_sliding_window(features_train.values, target_train.values, window_size)
X_test_raw, y_test_raw = create_sliding_window(features_test.values, target_test.values, window_size)

# ========================
# 3. ÿ™ÿ∑ÿ®Ÿäÿπ ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™
# ========================
scaler = RobustScaler()

X_train_scaled = scaler.fit_transform(X_train_raw.reshape(-1, X_train_raw.shape[-1])).reshape(X_train_raw.shape)
X_test_scaled = scaler.transform(X_test_raw.reshape(-1, X_test_raw.shape[-1])).reshape(X_test_raw.shape)

# Flatten ŸÑŸÑŸÜŸÖÿßÿ∞ÿ¨ ÿßŸÑŸÉŸÑÿßÿ≥ŸäŸÉŸäÿ©
X_train_flat = X_train_scaled.reshape(X_train_scaled.shape[0], -1)
X_test_flat = X_test_scaled.reshape(X_test_scaled.shape[0], -1)

print(f"üîπ ÿ¥ŸÉŸÑ ÿ®ŸäÿßŸÜÿßÿ™ ÿßŸÑÿ™ÿØÿ±Ÿäÿ®: {X_train_flat.shape}")
print(f"üîπ ÿ¥ŸÉŸÑ ÿ®ŸäÿßŸÜÿßÿ™ ÿßŸÑÿßÿÆÿ™ÿ®ÿßÿ±: {X_test_flat.shape}")

# ========================
# 4. ÿ®ŸÜÿßÿ° ŸÜŸÖŸàÿ∞ÿ¨ Random Forest
# ========================
model = RandomForestClassifier(
    n_estimators=300,       # ÿπÿØÿØ ÿßŸÑÿ£ÿ¥ÿ¨ÿßÿ±
    max_depth=15,           # ÿ£ŸÇÿµŸâ ÿπŸÖŸÇ ŸÑŸÑÿ¥ÿ¨ÿ±ÿ©
    random_state=42,
    n_jobs=-1,              # ÿßÿ≥ÿ™ÿÆÿØÿßŸÖ ŸÉŸÑ ÿßŸÑŸÖÿπÿßŸÑÿ¨ÿßÿ™
    class_weight="balanced" # ŸÖÿπÿßŸÑÿ¨ÿ© ÿπÿØŸÖ ÿ™Ÿàÿßÿ≤ŸÜ ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™
)

# ========================
# 5. ÿ™ÿØÿ±Ÿäÿ® ÿßŸÑŸÜŸÖŸàÿ∞ÿ¨
# ========================
print("üöÄ Training Random Forest...")
model.fit(X_train_flat, y_train_raw)

# ========================
# 6. ÿßŸÑÿ™ŸÇŸäŸäŸÖ ÿßŸÑÿ™ÿØÿ±Ÿäÿ¨Ÿä Walking Forward
# ========================
y_true, y_pred = [], []
print("üîç ÿ¨ÿßÿ±Ÿä ÿßŸÑÿ™ŸÇŸäŸäŸÖ ÿßŸÑÿ™ÿØÿ±Ÿäÿ¨Ÿä...")
for i in tqdm.tqdm(range(len(X_test_flat))):
    sample_x = X_test_flat[i].reshape(1, -1)
    true_class = y_test_raw[i]
    predicted_class = model.predict(sample_x)[0]

    y_true.append(true_class)
    y_pred.append(predicted_class)

walk_acc = accuracy_score(y_true, y_pred)
print(f"üß† Walking Forward Accuracy: {walk_acc:.4f}")

# ========================
# 7. ÿßŸÑÿ™ŸÇŸäŸäŸÖ ÿπŸÑŸâ ÿ®ŸäÿßŸÜÿßÿ™ ÿßŸÑÿßÿÆÿ™ÿ®ÿßÿ± ŸÉÿßŸÖŸÑÿ©
# ========================
test_pred = model.predict(X_test_flat)
test_acc = accuracy_score(y_test_raw, test_pred)

print(f"üìä Test Accuracy: {test_acc:.4f}")

# ========================
# 8. ŸÖÿµŸÅŸàŸÅÿ© ÿßŸÑÿßŸÑÿ™ÿ®ÿßÿ≥
# ========================
cm = confusion_matrix(y_test_raw, test_pred)

plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=np.unique(y_train_raw),
            yticklabels=np.unique(y_train_raw))
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix - Random Forest")
plt.show()

# ========================
# 9. ÿ™ŸÇÿ±Ÿäÿ± ÿßŸÑÿ£ÿØÿßÿ°
# ========================
print("\nüìà Classification Report:")
print("="*50)
print(classification_report(y_test_raw, test_pred, digits=3))

# ========================
# 10. ÿ•ÿ≠ÿµÿßÿ¶Ÿäÿßÿ™ ÿ•ÿ∂ÿßŸÅŸäÿ©
# ========================
print(f"\nüìä ÿ•ÿ≠ÿµÿßÿ¶Ÿäÿßÿ™ ÿ•ÿ∂ÿßŸÅŸäÿ©:")
print(f"ÿπÿØÿØ ÿπŸäŸÜÿßÿ™ ÿßŸÑÿ™ÿØÿ±Ÿäÿ®: {len(X_train_flat)}")
print(f"ÿπÿØÿØ ÿπŸäŸÜÿßÿ™ ÿßŸÑÿßÿÆÿ™ÿ®ÿßÿ±: {len(X_test_flat)}")
print(f"ÿπÿØÿØ ÿßŸÑŸÖŸäÿ≤ÿßÿ™ ÿ®ÿπÿØ Flatten: {X_train_flat.shape[1]}")
print(f"ÿ≠ÿ¨ŸÖ ÿßŸÑŸÜÿßŸÅÿ∞ÿ© ÿßŸÑÿ≤ŸÖŸÜŸäÿ©: {window_size}")
print(f"ÿπÿØÿØ ÿßŸÑÿ£ÿ¥ÿ¨ÿßÿ±: {model.n_estimators}, ÿπŸÖŸÇ ÿßŸÑÿ¥ÿ¨ÿ±ÿ©: {model.max_depth}")
```
