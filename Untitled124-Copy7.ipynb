```python
import pandas as pd
ٌُُdf = pd.read_csv(r"C:\Users\Access\Documents\DATA\AAPL\df-23-8-25-8.csv" )

df_train = df[:70000]

df_test = df[70000:]

#================================================================================================================

df = df_train

# التأكد من أن العمود datetime
df['timestamp'] = pd.to_datetime(df['timestamp'])
df['date'] = df['timestamp'].dt.date

# دوال لحساب أعلى/أقل سعر حتى الصف الحالي
def calc_high_so_far(group):
    group['high_today'] = group['high'].cummax()
    idx_high = group['high'].cummax().eq(group['high'])
    group['vol_at_high_today'] = group['volume'].where(idx_high).ffill()
    return group

def calc_low_so_far(group):
    group['low_today'] = group['low'].cummin()
    idx_low = group['low'].cummin().eq(group['low'])
    group['vol_at_low_today'] = group['volume'].where(idx_low).ffill()
    return group

# نخزن عمود date مؤقت
dates = df['date']

# نستبعد date من groupby.apply لتفادي التحذير
df = df.drop(columns=['date'])
df = pd.concat([
    calc_high_so_far(group) 
    for _, group in pd.concat([dates, df], axis=1).groupby(dates)
], axis=0)

df = pd.concat([
    calc_low_so_far(group) 
    for _, group in df.groupby(dates)
], axis=0)

# نرجع العمود date
df['date'] = dates

# حساب بيانات اليوم السابق بدون تسريب
df['high_prev_day'] = df.groupby('date')['high_today'].shift(1)
df['vol_at_high_prev_day'] = df.groupby('date')['vol_at_high_today'].shift(1)
df['low_prev_day'] = df.groupby('date')['low_today'].shift(1)
df['vol_at_low_prev_day'] = df.groupby('date')['vol_at_low_today'].shift(1)

import numpy as np

# Relative Volume (RVOL)
df['RVOL'] = df['volume'] / df['volume'].rolling(window=20).mean()

# Volume Rate of Change (VROC)
df['VROC'] = (df['volume'] - df['volume'].shift(5)) / df['volume'].shift(5)

# Volume Weighted Spread
df['spread_vol'] = (df['high'] - df['low']) * df['RVOL']

# VWAP Deviation (بالانحراف المعياري)
vwap_std = df['vwap'].rolling(window=20).std()
df['vwap_dev'] = (df['close'] - df['vwap']) / vwap_std

# Volume per Transaction
df['vol_per_tx'] = df['volume'] / df['transactions']


# Accumulation/Distribution based on VWAP
df['accum_dist_vwap'] = np.where(df['close'] > df['vwap'], df['volume'], -df['volume'])
df['accum_dist_vwap'] = df['accum_dist_vwap'].cumsum()

# Volume Spike Detector → بدون تسريب
df['vol_spike'] = (
    df['volume'] > (df['volume'].rolling(window=10).mean().shift(1) * 2)
).astype(int)

# Volume Climax Bar → بدون تسريب وبـ tolerance
max_vol_threshold = df['volume'].rolling(window=20).max().shift(1) * 0.95
wide_range_threshold = (df['high'] - df['low']).rolling(window=20).mean().shift(1)

df['climax_bar'] = (
    (df['volume'] >= max_vol_threshold) & 
    ((df['high'] - df['low']) > wide_range_threshold)
).astype(int)

import numpy as np

denominator = (df['high'].rolling(5).max().shift(1) - df['low'].rolling(5).min().shift(1))
denominator = denominator.replace(0, np.nan)  # استبدال المقام إذا كان صفر بـ NaN

df['micro_trend_slope'] = (
    (df['close'] - df['close'].shift(5)) / denominator
)


# Price Change per Second (لو عندك بيانات وقت بالثواني)
# تحويل timestamp إلى datetime
df['timestamp'] = pd.to_datetime(df['timestamp'])
df['seconds_diff'] = (df['timestamp'] - df['timestamp'].shift(1)).dt.total_seconds()
df['price_change_per_sec'] = (df['close'] - df['close'].shift(1)) / df['seconds_diff']

# Position in Range (آخر 20 شمعة) - بدون تسريب
window_range = 20
low_min = df['low'].rolling(window_range).min().shift(1)
high_max = df['high'].rolling(window_range).max().shift(1)
df['pos_in_range'] = (df['close'] - low_min) / (high_max - low_min)

# Micro Support/Resistance Distance - بدون تسريب
df['dist_to_recent_high'] = df['high'].rolling(window_range).max().shift(1) - df['close']
df['dist_to_recent_low'] = df['close'] - df['low'].rolling(window_range).min().shift(1)

# Candle Body-to-Range Ratio (ده طبيعي مفيهوش تسريب)
df['body_to_range'] = abs(df['close'] - df['open']) / (df['high'] - df['low'])

# Transaction Density (طبيعي بدون تسريب)
df['tx_density'] = df['transactions'] / df['volume']

# VWAP Pullback Strength - بدون تسريب
df['vwap_pullback'] = (df['vwap'].shift(1) - df['close'].shift(1)) / df['vwap'].shift(1)

# Volume Shock - بدون تسريب
vol_avg = df['volume'].rolling(20).mean().shift(1)
df['vol_shock'] = (df['volume'] > (vol_avg * 2)).astype(int)

# Price Reaction Next (ده Label للتنبؤ مش مؤشر)
df['price_reaction_next'] = df['close'].shift(-1) - df['close']

# حساب الفرق بالدقائق بين الصف الحالي والسابق
df['minutes_diff'] = (df['timestamp'] - df['timestamp'].shift(1)).dt.total_seconds() / 60

# الفرق بين سعر الفتح الحالي وسعر الإغلاق السابق
df['open_vs_prev_close'] = df['open'] - df['close'].shift(1)

df['vwap_distance'] = df['close'] - df['vwap']

df['vol_change_pct'] = (df['volume'] - df['volume'].shift(1)) / df['volume'].shift(1) * 100

# استخراج الساعة
df['hour'] = df['timestamp'].dt.hour

# استخراج الدقائق
df['minute'] = df['timestamp'].dt.minute

# تحديد الربع من الساعة
def get_quarter(minute):
    if minute < 15:
        return 1
    elif minute < 30:
        return 2
    elif minute < 45:
        return 3
    else:
        return 4

df['quarter_hour'] = df['minute'].apply(get_quarter)

df.dropna(inplace=True)

n_future = 3  # 6 ساعات لقدّام

df['future_return'] = df['close'].shift(-n_future) / df['close'] - 1

# نحدد الإشارات بناءً على عتبة معينة للحركة
threshold = 0.0003  

df['Signal'] = 0
df.loc[df['future_return'] > threshold, 'Signal'] = 2
df.loc[df['future_return'] < -threshold, 'Signal'] = 1

df.dropna(inplace=True)

#===========================================================================================

df_train = df
```

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import RobustScaler
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import matplotlib.pyplot as plt
import seaborn as sns
import tqdm
from xgboost import XGBClassifier, plot_importance

# ========================
# دالة إنشاء نافذة زمنية
# ========================
def create_sliding_window(X, y, window_size):
    X_windows, y_windows = [], []
    for i in range(len(X) - window_size + 1):
        X_windows.append(X[i:i + window_size])
        y_windows.append(y[i + window_size - 1])  # آخر عنصر هو الهدف
    return np.array(X_windows), np.array(y_windows)


# ========================
# 1. تحميل وتجهيز البيانات
# ========================
# df_train = pd.read_csv('your_train_data.csv')
# df_test = pd.read_csv('your_test_data.csv')

features_train = df_train.drop(columns=['Signal','date','timestamp','future_return','price_reaction_next','accum_dist_vwap'])
features_test = df_test.drop(columns=['Signal','date','timestamp','future_return','price_reaction_next','accum_dist_vwap'])

target_train = df_train['Signal']
target_test = df_test['Signal']

# ========================
# 2. إنشاء النوافذ الزمنية
# ========================
window_size = 50
X_train_raw, y_train_raw = create_sliding_window(features_train.values, target_train.values, window_size)
X_test_raw, y_test_raw = create_sliding_window(features_test.values, target_test.values, window_size)

# ========================
# 3. تطبيع البيانات
# ========================
scaler = RobustScaler()

X_train_scaled = scaler.fit_transform(X_train_raw.reshape(-1, X_train_raw.shape[-1])).reshape(X_train_raw.shape)
X_test_scaled = scaler.transform(X_test_raw.reshape(-1, X_test_raw.shape[-1])).reshape(X_test_raw.shape)

# Flatten عشان XGBoost ما بيقبلش بيانات 3D
X_train_flat = X_train_scaled.reshape(X_train_scaled.shape[0], -1)
X_test_flat = X_test_scaled.reshape(X_test_scaled.shape[0], -1)

print(f"🔹 شكل بيانات التدريب: {X_train_flat.shape}")
print(f"🔹 شكل بيانات الاختبار: {X_test_flat.shape}")

# ========================
# 4. بناء نموذج XGBoost
# ========================
model = XGBClassifier(
    n_estimators=90,
    max_depth=8,
    learning_rate=0.05,
    subsample=0.8,
    colsample_bytree=0.8,
    objective="multi:probsoftmax",  # تصنيف متعدد الفئات
    num_class=3,
    random_state=42,
    n_jobs=-1,
    eval_metric="mlogloss"   # ✅ هنا بدل ما يكون جوه fit
)

# ========================
# 5. تدريب النموذج
# ========================
print("🚀 Training XGBoost...")
model.fit(X_train_flat, y_train_raw,
          eval_set=[(X_test_flat, y_test_raw)],
          verbose=True)

# ========================
# 6. التقييم التدريجي Walking Forward
# ========================
y_true, y_pred = [], []
print("🔍 جاري التقييم التدريجي...")
for i in tqdm.tqdm(range(len(X_test_flat))):
    sample_x = X_test_flat[i].reshape(1, -1)
    true_class = y_test_raw[i]
    predicted_class = model.predict(sample_x)[0]

    y_true.append(true_class)
    y_pred.append(predicted_class)

walk_acc = accuracy_score(y_true, y_pred)
print(f"🧠 Walking Forward Accuracy: {walk_acc:.4f}")

# ========================
# 7. التقييم على بيانات الاختبار كاملة
# ========================
test_pred = model.predict(X_test_flat)
test_acc = accuracy_score(y_test_raw, test_pred)

print(f"📊 Test Accuracy: {test_acc:.4f}")

# ========================
# 8. مصفوفة الالتباس
# ========================
cm = confusion_matrix(y_test_raw, test_pred)

plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=np.unique(y_train_raw),
            yticklabels=np.unique(y_train_raw))
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix - XGBoost")
plt.show()

# ========================
# 9. تقرير الأداء
# ========================
print("\n📈 Classification Report:")
print("="*50)
print(classification_report(y_test_raw, test_pred, digits=3))

# ========================
# 10. أهمية الميزات (Feature Importance)
# ========================
plt.figure(figsize=(10, 6))
plot_importance(model, max_num_features=15, importance_type="gain")
plt.title("Top 15 Feature Importances - XGBoost")
plt.show()

# ========================
# 11. إحصائيات إضافية
# ========================
print(f"\n📊 إحصائيات إضافية:")
print(f"عدد عينات التدريب: {len(X_train_flat)}")
print(f"عدد عينات الاختبار: {len(X_test_flat)}")
print(f"عدد الميزات بعد Flatten: {X_train_flat.shape[1]}")
print(f"حجم النافذة الزمنية: {window_size}")
print(f"عدد الأشجار: {model.n_estimators}, عمق الشجرة: {model.max_depth}")



# ====== bins ======
bins = np.linspace(0, 1, 21)  # من 0 الى 1

# ====== دالة لحساب الدقة وعدد الصح داخل bin ======
def bin_accuracy(df, prob_col, true_class, low, high):
    mask = (df[prob_col] >= low) & (df[prob_col] < high)
    total = mask.sum()
    if total == 0:
        return None, 0, 0
    correct = (df["True_Label"][mask] == true_class).sum()
    acc = correct / total
    return acc, correct, total

# ====== دالة للرسم ======
def plot_hist_with_bin_acc(df, prob_col, class_id, class_name):
    plt.figure(figsize=(12,6))

    # histogram stacked by true labels
    counts, edges, bars = plt.hist([
            df[prob_col][df['True_Label']==0],
            df[prob_col][df['True_Label']==1],
            df[prob_col][df['True_Label']==2]
        ],
        bins=bins,
        stacked=True,
        color=['blue','green','red'],
        label=['No Trade (0)','Sell (1)','Buy (2)'],
        alpha=0.6
    )

    # إضافة الدقة + عدد الصح داخل كل بار
    for i in range(len(edges)-1):
        acc, correct, total = bin_accuracy(df, prob_col, class_id, edges[i], edges[i+1])
        height = sum(counts[:,i])
        if acc is not None and height > 0:
            plt.text((edges[i]+edges[i+1])/2,
                     height/2,
                     f"{acc*100:.0f}% ({correct})", 
                     ha='center', va='center',
                     fontsize=9, color="black")

    plt.axvline(0.5, color='black', linestyle='--', label='Threshold 0.5')
    plt.title(f"{class_name} Probabilities Histogram + Accuracy per Bin")
    plt.xlabel("Predicted Probability (Confidence)")
    plt.ylabel("Number of Signals")
    plt.legend()
    plt.show()

# ====== الرسومات ======
plot_hist_with_bin_acc(df_proba, "Sell_1", 1, "Sell")
plot_hist_with_bin_acc(df_proba, "Buy_2", 2, "Buy")
plot_hist_with_bin_acc(df_proba, "Class_0", 0, "No Trade")


```

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import RobustScaler
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import matplotlib.pyplot as plt
import seaborn as sns
import tqdm
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout

# ========================
# دالة إنشاء نافذة زمنية
# ========================
def create_sliding_window(X, y, window_size):
    X_windows, y_windows = [], []
    for i in range(len(X) - window_size + 1):
        X_windows.append(X[i:i + window_size])
        y_windows.append(y[i + window_size - 1])  # آخر عنصر هو الهدف
    return np.array(X_windows), np.array(y_windows)


# ========================
# 1. تحميل وتجهيز البيانات
# ========================
features_train = df_train.drop(columns=['Signal','date','timestamp','future_return','price_reaction_next','accum_dist_vwap'])
features_test = df_test.drop(columns=['Signal','date','timestamp','future_return','price_reaction_next','accum_dist_vwap'])

target_train = df_train['Signal']
target_test = df_test['Signal']

# ========================
# 2. إنشاء النوافذ الزمنية
# ========================
window_size = 50
X_train_raw, y_train_raw = create_sliding_window(features_train.values, target_train.values, window_size)
X_test_raw, y_test_raw = create_sliding_window(features_test.values, target_test.values, window_size)

# ========================
# 3. تطبيع البيانات
# ========================
scaler = RobustScaler()

X_train_scaled = scaler.fit_transform(X_train_raw.reshape(-1, X_train_raw.shape[-1])).reshape(X_train_raw.shape)
X_test_scaled = scaler.transform(X_test_raw.reshape(-1, X_test_raw.shape[-1])).reshape(X_test_raw.shape)

print(f"🔹 شكل بيانات التدريب: {X_train_scaled.shape}")
print(f"🔹 شكل بيانات الاختبار: {X_test_scaled.shape}")

# ========================
# 4. بناء نموذج LSTM
# ========================
model = Sequential([
    LSTM(64, input_shape=(X_train_scaled.shape[1], X_train_scaled.shape[2])),
    Dropout(0.3),
    Dense(32, activation='relu'),
    Dense(3, activation='softmax')   # multi-class
])

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# ========================
# 5. تدريب النموذج
# ========================
print("🚀 Training LSTM...")
history = model.fit(
    X_train_scaled, y_train_raw,
    validation_data=(X_test_scaled, y_test_raw),
    epochs=15,
    batch_size=64,
    verbose=1
)

# ========================
# 6. التقييم التدريجي Walking Forward
# ========================
y_true, y_pred = [], []
print("🔍 جاري التقييم التدريجي...")
for i in tqdm.tqdm(range(len(X_test_scaled))):
    sample_x = X_test_scaled[i].reshape(1, X_test_scaled.shape[1], X_test_scaled.shape[2])
    true_class = y_test_raw[i]
    predicted_class = np.argmax(model.predict(sample_x, verbose=0), axis=1)[0]

    y_true.append(true_class)
    y_pred.append(predicted_class)

walk_acc = accuracy_score(y_true, y_pred)
print(f"🧠 Walking Forward Accuracy: {walk_acc:.4f}")

# ========================
# 7. التقييم على بيانات الاختبار كاملة
# ========================
test_pred = np.argmax(model.predict(X_test_scaled, verbose=0), axis=1)
test_acc = accuracy_score(y_test_raw, test_pred)

print(f"📊 Test Accuracy: {test_acc:.4f}")

# ========================
# 8. مصفوفة الالتباس
# ========================
cm = confusion_matrix(y_test_raw, test_pred)

plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=np.unique(y_train_raw),
            yticklabels=np.unique(y_train_raw))
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix - LSTM")
plt.show()

# ========================
# 9. تقرير الأداء
# ========================
print("\n📈 Classification Report:")
print("="*50)
print(classification_report(y_test_raw, test_pred, digits=3))

# ====== bins ======
bins = np.linspace(0, 1, 21)  # من 0 الى 1

# ====== دالة لحساب الدقة وعدد الصح داخل bin ======
def bin_accuracy(df, prob_col, true_class, low, high):
    mask = (df[prob_col] >= low) & (df[prob_col] < high)
    total = mask.sum()
    if total == 0:
        return None, 0, 0
    correct = (df["True_Label"][mask] == true_class).sum()
    acc = correct / total
    return acc, correct, total

# ====== دالة للرسم ======
def plot_hist_with_bin_acc(df, prob_col, class_id, class_name):
    plt.figure(figsize=(12,6))

    counts, edges, bars = plt.hist([
            df[prob_col][df['True_Label']==0],
            df[prob_col][df['True_Label']==1],
            df[prob_col][df['True_Label']==2]
        ],
        bins=bins,
        stacked=True,
        color=['blue','green','red'],
        label=['No Trade (0)','Sell (1)','Buy (2)'],
        alpha=0.6
    )

    for i in range(len(edges)-1):
        acc, correct, total = bin_accuracy(df, prob_col, class_id, edges[i], edges[i+1])
        height = sum(counts[:,i])
        if acc is not None and height > 0:
            plt.text((edges[i]+edges[i+1])/2,
                     height/2,
                     f"{acc*100:.0f}% ({correct})", 
                     ha='center', va='center',
                     fontsize=9, color="black")

    plt.axvline(0.5, color='black', linestyle='--', label='Threshold 0.5')
    plt.title(f"{class_name} Probabilities Histogram + Accuracy per Bin")
    plt.xlabel("Predicted Probability (Confidence)")
    plt.ylabel("Number of Signals")
    plt.legend()
    plt.show()

# ====== التنبؤ بالاحتمالات لعمل histogram ======
y_proba = model.predict(X_test_scaled, verbose=0)
df_proba = pd.DataFrame({
    "True_Label": y_test_raw,
    "Class_0": y_proba[:,0],
    "Sell_1": y_proba[:,1],
    "Buy_2": y_proba[:,2]
})

# ====== الرسومات ======
plot_hist_with_bin_acc(df_proba, "Sell_1", 1, "Sell")
plot_hist_with_bin_acc(df_proba, "Buy_2", 2, "Buy")
plot_hist_with_bin_acc(df_proba, "Class_0", 0, "No Trade")
```

```python
def fut_data(List):
    for df in List:
        
        #df = df_train
        
        # التأكد من أن العمود datetime
        df['timestamp'] = pd.to_datetime(df['timestamp'])
        df['date'] = df['timestamp'].dt.date
        
        # دوال لحساب أعلى/أقل سعر حتى الصف الحالي
        def calc_high_so_far(group):
            group['high_today'] = group['high'].cummax()
            idx_high = group['high'].cummax().eq(group['high'])
            group['vol_at_high_today'] = group['volume'].where(idx_high).ffill()
            return group
        
        def calc_low_so_far(group):
            group['low_today'] = group['low'].cummin()
            idx_low = group['low'].cummin().eq(group['low'])
            group['vol_at_low_today'] = group['volume'].where(idx_low).ffill()
            return group
        
        # نخزن عمود date مؤقت
        dates = df['date']
        
        # نستبعد date من groupby.apply لتفادي التحذير
        df = df.drop(columns=['date'])
        df = pd.concat([
            calc_high_so_far(group) 
            for _, group in pd.concat([dates, df], axis=1).groupby(dates)
        ], axis=0)
        
        df = pd.concat([
            calc_low_so_far(group) 
            for _, group in df.groupby(dates)
        ], axis=0)
        
        # نرجع العمود date
        df['date'] = dates
        
        # حساب بيانات اليوم السابق بدون تسريب
        df['high_prev_day'] = df.groupby('date')['high_today'].shift(1)
        df['vol_at_high_prev_day'] = df.groupby('date')['vol_at_high_today'].shift(1)
        df['low_prev_day'] = df.groupby('date')['low_today'].shift(1)
        df['vol_at_low_prev_day'] = df.groupby('date')['vol_at_low_today'].shift(1)
        
        
        # Relative Volume (RVOL)
        df['RVOL'] = df['volume'] / df['volume'].rolling(window=20).mean()
        
        # Volume Rate of Change (VROC)
        df['VROC'] = (df['volume'] - df['volume'].shift(5)) / df['volume'].shift(5)
        
        # Volume Weighted Spread
        df['spread_vol'] = (df['high'] - df['low']) * df['RVOL']
        
        # VWAP Deviation (بالانحراف المعياري)
        vwap_std = df['vwap'].rolling(window=20).std()
        df['vwap_dev'] = (df['close'] - df['vwap']) / vwap_std
        
        # Volume per Transaction
        df['vol_per_tx'] = df['volume'] / df['transactions']
        
        
        # Accumulation/Distribution based on VWAP
        df['accum_dist_vwap'] = np.where(df['close'] > df['vwap'], df['volume'], -df['volume'])
        df['accum_dist_vwap'] = df['accum_dist_vwap'].cumsum()
        
        # Volume Spike Detector → بدون تسريب
        df['vol_spike'] = (
            df['volume'] > (df['volume'].rolling(window=10).mean().shift(1) * 2)
        ).astype(int)
        
        # Volume Climax Bar → بدون تسريب وبـ tolerance
        max_vol_threshold = df['volume'].rolling(window=20).max().shift(1) * 0.95
        wide_range_threshold = (df['high'] - df['low']).rolling(window=20).mean().shift(1)
        
        df['climax_bar'] = (
            (df['volume'] >= max_vol_threshold) & 
            ((df['high'] - df['low']) > wide_range_threshold)
        ).astype(int)
        
        
        denominator = (df['high'].rolling(5).max().shift(1) - df['low'].rolling(5).min().shift(1))
        denominator = denominator.replace(0, np.nan)  # استبدال المقام إذا كان صفر بـ NaN
        
        df['micro_trend_slope'] = (
            (df['close'] - df['close'].shift(5)) / denominator
        )
        
        
        # Price Change per Second (لو عندك بيانات وقت بالثواني)
        # تحويل timestamp إلى datetime
        df['timestamp'] = pd.to_datetime(df['timestamp'])
        df['seconds_diff'] = (df['timestamp'] - df['timestamp'].shift(1)).dt.total_seconds()
        df['price_change_per_sec'] = (df['close'] - df['close'].shift(1)) / df['seconds_diff']
        
        # Position in Range (آخر 20 شمعة) - بدون تسريب
        window_range = 20
        low_min = df['low'].rolling(window_range).min().shift(1)
        high_max = df['high'].rolling(window_range).max().shift(1)
        df['pos_in_range'] = (df['close'] - low_min) / (high_max - low_min)
        
        # Micro Support/Resistance Distance - بدون تسريب
        df['dist_to_recent_high'] = df['high'].rolling(window_range).max().shift(1) - df['close']
        df['dist_to_recent_low'] = df['close'] - df['low'].rolling(window_range).min().shift(1)
        
        # Candle Body-to-Range Ratio (ده طبيعي مفيهوش تسريب)
        df['body_to_range'] = abs(df['close'] - df['open']) / (df['high'] - df['low'])
        
        # Transaction Density (طبيعي بدون تسريب)
        df['tx_density'] = df['transactions'] / df['volume']
        
        # VWAP Pullback Strength - بدون تسريب
        df['vwap_pullback'] = (df['vwap'].shift(1) - df['close'].shift(1)) / df['vwap'].shift(1)
        
        # Volume Shock - بدون تسريب
        vol_avg = df['volume'].rolling(20).mean().shift(1)
        df['vol_shock'] = (df['volume'] > (vol_avg * 2)).astype(int)
        
        # Price Reaction Next (ده Label للتنبؤ مش مؤشر)
        df['price_reaction_next'] = df['close'].shift(-1) - df['close']
        
        # حساب الفرق بالدقائق بين الصف الحالي والسابق
        df['minutes_diff'] = (df['timestamp'] - df['timestamp'].shift(1)).dt.total_seconds() / 60
        
        # الفرق بين سعر الفتح الحالي وسعر الإغلاق السابق
        df['open_vs_prev_close'] = df['open'] - df['close'].shift(1)
        
        df['vwap_distance'] = df['close'] - df['vwap']
        
        df['vol_change_pct'] = (df['volume'] - df['volume'].shift(1)) / df['volume'].shift(1) * 100
        
        # استخراج الساعة
        df['hour'] = df['timestamp'].dt.hour
        
        # استخراج الدقائق
        df['minute'] = df['timestamp'].dt.minute
        
        # تحديد الربع من الساعة
        def get_quarter(minute):
            if minute < 15:
                return 1
            elif minute < 30:
                return 2
            elif minute < 45:
                return 3
            else:
                return 4
        
        df['quarter_hour'] = df['minute'].apply(get_quarter)
        
        df.dropna(inplace=True)
        
        n_future = 3  # 6 ساعات لقدّام
        
        df['future_return'] = df['close'].shift(-n_future) / df['close'] - 1
        
        # نحدد الإشارات بناءً على عتبة معينة للحركة
        threshold = 0.0003  
        
        df['Signal'] = 0
        df.loc[df['future_return'] > threshold, 'Signal'] = 2
        df.loc[df['future_return'] < -threshold, 'Signal'] = 1
        
        df.dropna(inplace=True)
        #df_train = df
        
```

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import RobustScaler
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import matplotlib.pyplot as plt
import seaborn as sns
import tqdm
from xgboost import XGBClassifier, plot_importance

# ========================
# دالة إنشاء نافذة زمنية
# ========================
def create_sliding_window(X, y, window_size):
    X_windows, y_windows = [], []
    for i in range(len(X) - window_size + 1):
        X_windows.append(X[i:i + window_size])
        y_windows.append(y[i + window_size - 1])  # آخر عنصر هو الهدف
    return np.array(X_windows), np.array(y_windows)

# ========================
# 1. تحميل وتجهيز البيانات
# ========================
# df_train = pd.read_csv('your_train_data.csv')
# df_test = pd.read_csv('your_test_data.csv')

features_train = df_train.drop(columns=['Signal','date','timestamp','future_return'])
features_test = df_test.drop(columns=['Signal','date','timestamp','future_return'])

target_train = df_train['Signal']
target_test = df_test['Signal']

# ========================
# 2. إنشاء النوافذ الزمنية
# ========================
window_size = 50
X_train_raw, y_train_raw = create_sliding_window(features_train.values, target_train.values, window_size)
X_test_raw, y_test_raw = create_sliding_window(features_test.values, target_test.values, window_size)

# ========================
# 3. تطبيع البيانات
# ========================
scaler = RobustScaler()
X_train_scaled = scaler.fit_transform(X_train_raw.reshape(-1, X_train_raw.shape[-1])).reshape(X_train_raw.shape)
X_test_scaled = scaler.transform(X_test_raw.reshape(-1, X_test_raw.shape[-1])).reshape(X_test_raw.shape)

# Flatten عشان XGBoost ما بيقبلش بيانات 3D
X_train_flat = X_train_scaled.reshape(X_train_scaled.shape[0], -1)
X_test_flat = X_test_scaled.reshape(X_test_scaled.shape[0], -1)

print(f"🔹 شكل بيانات التدريب: {X_train_flat.shape}")
print(f"🔹 شكل بيانات الاختبار: {X_test_flat.shape}")

# ========================
# 4. بناء نموذج XGBoost مع التوقف المبكر
# ========================
model = XGBClassifier(
    n_estimators=1000,
    max_depth=8,
    learning_rate=0.05,
    subsample=0.8,
    colsample_bytree=0.8,
    objective="multi:softprob",  # نستخدم softprob لاحتمالات
    num_class=3,
    random_state=42,
    n_jobs=-1,
    eval_metric=["mlogloss", "merror"],  # هنا
    early_stopping_rounds=10             # هنا
)

# ========================
# 5. تدريب النموذج
# ========================
print("🚀 Training XGBoost with Early Stopping...")
evals_result = {}
model.fit(
    X_train_flat, y_train_raw,
    eval_set=[(X_train_flat, y_train_raw), (X_test_flat, y_test_raw)],
    verbose=True,
    evals_result=evals_result
)

# ========================
# 6. رسم منحنيات التدريب (خسارة ودقة)
# ========================
epochs = len(evals_result['validation_0']['mlogloss'])
x_axis = range(epochs)

plt.figure(figsize=(12,5))
plt.plot(x_axis, evals_result['validation_0']['mlogloss'], label='Train Log Loss')
plt.plot(x_axis, evals_result['validation_1']['mlogloss'], label='Test Log Loss')
plt.xlabel("Epochs")
plt.ylabel("Log Loss")
plt.title("XGBoost Log Loss")
plt.legend()
plt.show()

plt.figure(figsize=(12,5))
plt.plot(x_axis, evals_result['validation_0']['merror'], label='Train Error')
plt.plot(x_axis, evals_result['validation_1']['merror'], label='Test Error')
plt.xlabel("Epochs")
plt.ylabel("Classification Error")
plt.title("XGBoost Classification Error")
plt.legend()
plt.show()

# ========================
# 7. التقييم التدريجي Walking Forward
# ========================
y_true, y_pred = [], []
print("🔍 جاري التقييم التدريجي...")
for i in tqdm.tqdm(range(len(X_test_flat))):
    sample_x = X_test_flat[i].reshape(1, -1)
    true_class = y_test_raw[i]
    predicted_class = model.predict(sample_x)[0]

    y_true.append(true_class)
    y_pred.append(predicted_class)

walk_acc = accuracy_score(y_true, y_pred)
print(f"🧠 Walking Forward Accuracy: {walk_acc:.4f}")

# ========================
# 8. التقييم على بيانات الاختبار كاملة
# ========================
test_pred = model.predict(X_test_flat)
test_acc = accuracy_score(y_test_raw, test_pred)
print(f"📊 Test Accuracy: {test_acc:.4f}")

# ========================
# 9. مصفوفة الالتباس
# ========================
cm = confusion_matrix(y_test_raw, test_pred)

plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=np.unique(y_train_raw),
            yticklabels=np.unique(y_train_raw))
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix - XGBoost")
plt.show()

# ========================
# 10. تقرير الأداء
# ========================
print("\n📈 Classification Report:")
print("="*50)
print(classification_report(y_test_raw, test_pred, digits=3))

# ========================
# 11. أهمية الميزات (Feature Importance)
# ========================
plt.figure(figsize=(10, 6))
plot_importance(model, max_num_features=15, importance_type="gain")
plt.title("Top 15 Feature Importances - XGBoost")
plt.show()

# ========================
# 12. رسم احتمالات الثقة لكل فئة
# ========================
proba_test = model.predict_proba(X_test_flat)
df_proba = pd.DataFrame(proba_test, columns=['Class_0','Sell_1','Buy_2'])
df_proba['True_Label'] = y_test_raw

bins = np.linspace(0, 1, 21)

# Sell
plt.figure(figsize=(10,5))
plt.hist([df_proba['Sell_1'][df_proba['True_Label']==0],
          df_proba['Sell_1'][df_proba['True_Label']==1],
          df_proba['Sell_1'][df_proba['True_Label']==2]],
         bins=bins,
         color=['blue','green','red'],
         label=['No Trade (0)','Actual Sell (1)','Actual Buy (2)'],
         stacked=True, alpha=0.7)
plt.axvline(0.5, color='black', linestyle='--', label='Threshold 0.5')
plt.title("Sell Probabilities Histogram")
plt.xlabel("Predicted Sell Probability")
plt.ylabel("Number of Signals")
plt.legend()
plt.show()

# Buy
plt.figure(figsize=(10,5))
plt.hist([df_proba['Buy_2'][df_proba['True_Label']==0],
          df_proba['Buy_2'][df_proba['True_Label']==1],
          df_proba['Buy_2'][df_proba['True_Label']==2]],
         bins=bins,
         color=['blue','green','red'],
         label=['No Trade (0)','Actual Sell (1)','Actual Buy (2)'],
         stacked=True, alpha=0.7)
plt.axvline(0.5, color='black', linestyle='--', label='Threshold 0.5')
plt.title("Buy Probabilities Histogram")
plt.xlabel("Predicted Buy Probability")
plt.ylabel("Number of Signals")
plt.legend()
plt.show()

# No Trade
plt.figure(figsize=(10,5))
plt.hist([df_proba['Class_0'][df_proba['True_Label']==0],
          df_proba['Class_0'][df_proba['True_Label']==1],
          df_proba['Class_0'][df_proba['True_Label']==2]],
         bins=bins,
         color=['blue','green','red'],
         label=['Actual No Trade (0)','Actual Sell (1)','Actual Buy (2)'],
         stacked=True, alpha=0.7)
plt.axvline(0.5, color='black', linestyle='--', label='Threshold 0.5')
plt.title("No Trade Probabilities Histogram")
plt.xlabel("Predicted No Trade Probability")
plt.ylabel("Number of Signals")
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, RobustScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras import regularizers
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
from tensorflow.keras.optimizers import Adam
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import matplotlib.pyplot as plt
import seaborn as sns
import tqdm
# ========================
# إضافة دالة create_sliding_window المفقودة
# ========================
def create_sliding_window(X, y, window_size):
    """
    إنشاء نوافذ زمنية منزلقة من البيانات
    """
    X_windows = []
    y_windows = []
    
    for i in range(len(X) - window_size + 1):
        X_windows.append(X[i:i + window_size])
        y_windows.append(y[i + window_size - 1])  # آخر هدف في النافذة
    
    return np.array(X_windows), np.array(y_windows)

# ========================
# 1. تحميل وتجهيز البيانات
# ========================
# تأكد من تحميل البيانات أولاً
# df_train = pd.read_csv('your_train_data.csv')
# df_test = pd.read_csv('your_test_data.csv')

# الميزات
features_train = df_train.drop(columns=['Signal','date','timestamp','future_return'])
features_test = df_test.drop(columns=['Signal','date','timestamp','future_return'])

# الأهداف
target_train = df_train['Signal']
target_test = df_test['Signal']

# إنشاء النوافذ الزمنية
window_size = 50
X_train_raw, y_train_raw = create_sliding_window(features_train.values, target_train.values, window_size)
X_test_raw, y_test_raw = create_sliding_window(features_test.values, target_test.values, window_size)

# تطبيق التطبيع
scaler = RobustScaler()

# تطبيع البيانات
X_train_scaled = scaler.fit_transform(X_train_raw.reshape(-1, X_train_raw.shape[-1])).reshape(X_train_raw.shape)
X_test_scaled = scaler.transform(X_test_raw.reshape(-1, X_test_raw.shape[-1])).reshape(X_test_raw.shape)

# تحويل الأهداف إلى categorical
y_train_cat = to_categorical(y_train_raw, num_classes=3)
y_test_cat = to_categorical(y_test_raw, num_classes=3)

print(f"شكل بيانات التدريب: {X_train_scaled.shape}")
print(f"شكل بيانات الاختبار: {X_test_scaled.shape}")
print(f"شكل أهداف التدريب: {y_train_cat.shape}")
print(f"شكل أهداف الاختبار: {y_test_cat.shape}")

# ========================
# 5. بناء نموذج LSTM
# ========================
model = Sequential()
model.add(LSTM(
    64,
    input_shape=(X_train_scaled.shape[1], X_train_scaled.shape[2]),
    return_sequences=False,
    dropout=0.2,
    kernel_regularizer=regularizers.l2(0.01)
))
model.add(Dense(32, activation='relu', kernel_regularizer=regularizers.l2(0.01)))
model.add(Dropout(0.2))
model.add(Dense(3, activation='softmax'))

# ========================
# 5.1 تحديد نسبة التعلم
# ========================
optimizer = Adam(learning_rate=0.001)
model.compile(optimizer=optimizer, loss='categorical_crossentropy', metrics=['accuracy'])

# عرض ملخص النموذج
model.summary()

# ========================
# 6.1 إعداد التوقف المبكر وحفظ أفضل نموذج
# ========================
early_stop = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
checkpoint = ModelCheckpoint('best_model.keras', monitor='val_loss', save_best_only=True)

# ========================
# 6. تدريب النموذج
# ========================
history = model.fit(
    X_train_scaled, y_train_cat,  # استخدام المتغيرات الصحيحة
    epochs=200,
    batch_size=100,
    validation_data=(X_test_scaled, y_test_cat),  # استخدام المتغيرات الصحيحة
    callbacks=[early_stop, checkpoint],
    verbose=1
)

# ========================
# 7. تقييم تدريجي خطوة بخطوة (Walking Forward Evaluation)
# ========================
y_true = []
y_pred = []

print("جاري التقييم التدريجي...")
for i in  tqdm.tqdm(range(len(X_test_scaled))):
    sample_x = X_test_scaled[i].reshape(1, X_test_scaled.shape[1], X_test_scaled.shape[2])
    true_class = np.argmax(y_test_cat[i])  # استخدام y_test_cat
    prediction = model.predict(sample_x, verbose=0)
    predicted_class = np.argmax(prediction)
    
    y_true.append(true_class)
    y_pred.append(predicted_class)

walk_acc = accuracy_score(y_true, y_pred)
print(f"🧠 Walking Forward Accuracy: {walk_acc:.4f}")

# ==================================================================================================================
# تقييم النموذج على اختبار حقيقي
# ==================================================================================================================
test_loss, test_accuracy = model.evaluate(X_test_scaled, y_test_cat, verbose=0)  # تصحيح المتغيرات
print(f"📊 Test Loss: {test_loss:.4f}")
print(f"🎯 Test Accuracy: {test_accuracy:.4f}")

# ========================
# رسم منحنى الأداء الكامل
# ========================
epochs_range = range(len(history.history['loss']))

plt.figure(figsize=(12, 6))

# رسم الخسارة
plt.subplot(1, 2, 1)
plt.plot(epochs_range, history.history['loss'], label='Train Loss', linestyle='--')
plt.plot(epochs_range, history.history['val_loss'], label='Validation Loss', linestyle='--')
plt.title('Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)

# رسم الدقة
plt.subplot(1, 2, 2)
plt.plot(epochs_range, history.history['accuracy'], label='Train Accuracy', linestyle='-')
plt.plot(epochs_range, history.history['val_accuracy'], label='Validation Accuracy', linestyle='-')
plt.title('Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()

# ===========
# عمل التنبؤات وتقييم شامل
# ===========
y_pred_probs = model.predict(X_test_scaled)
y_pred = np.argmax(y_pred_probs, axis=1)
y_true = np.argmax(y_test_cat, axis=1)

# ===========
# مصفوفة الالتباس
# ===========
cm = confusion_matrix(y_true, y_pred)

plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
            xticklabels=[0, 1, 2], yticklabels=[0, 1, 2])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()

# ===========
# تقرير الأداء الكامل
# ===========
print("\n📈 Classification Report:")
print("=" * 50)
print(classification_report(y_true, y_pred, digits=3))

# ===========
# إحصائيات إضافية
# ===========
print(f"\n📊 إحصائيات إضافية:")
print(f"عدد عينات التدريب: {len(X_train_scaled)}")
print(f"عدد عينات الاختبار: {len(X_test_scaled)}")
print(f"عدد الميزات: {X_train_scaled.shape[2]}")
print(f"حجم النافزة الزمنية: {window_size}")
print(f"عدد الحقب المكتملة: {len(history.history['loss'])}")
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import RobustScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import matplotlib.pyplot as plt
import seaborn as sns
import tqdm

# ========================
# دالة إنشاء نافذة زمنية
# ========================
def create_sliding_window(X, y, window_size):
    X_windows, y_windows = [], []
    for i in range(len(X) - window_size + 1):
        X_windows.append(X[i:i + window_size])
        y_windows.append(y[i + window_size - 1])  # آخر عنصر هو الهدف
    return np.array(X_windows), np.array(y_windows)


# ========================
# 1. تحميل وتجهيز البيانات
# ========================
# df_train = pd.read_csv('your_train_data.csv')
# df_test = pd.read_csv('your_test_data.csv')

features_train = df_train.drop(columns=['Signal','date','timestamp','future_return'])
features_test = df_test.drop(columns=['Signal','date','timestamp','future_return'])

target_train = df_train['Signal']
target_test = df_test['Signal']

# ========================
# 2. إنشاء النوافذ الزمنية
# ========================
window_size = 50
X_train_raw, y_train_raw = create_sliding_window(features_train.values, target_train.values, window_size)
X_test_raw, y_test_raw = create_sliding_window(features_test.values, target_test.values, window_size)

# ========================
# 3. تطبيع البيانات
# ========================
scaler = RobustScaler()

X_train_scaled = scaler.fit_transform(X_train_raw.reshape(-1, X_train_raw.shape[-1])).reshape(X_train_raw.shape)
X_test_scaled = scaler.transform(X_test_raw.reshape(-1, X_test_raw.shape[-1])).reshape(X_test_raw.shape)

# Flatten للنماذج الكلاسيكية
X_train_flat = X_train_scaled.reshape(X_train_scaled.shape[0], -1)
X_test_flat = X_test_scaled.reshape(X_test_scaled.shape[0], -1)

print(f"🔹 شكل بيانات التدريب: {X_train_flat.shape}")
print(f"🔹 شكل بيانات الاختبار: {X_test_flat.shape}")

# ========================
# 4. بناء نموذج Random Forest
# ========================
model = RandomForestClassifier(
    n_estimators=300,       # عدد الأشجار
    max_depth=15,           # أقصى عمق للشجرة
    random_state=42,
    n_jobs=-1,              # استخدام كل المعالجات
    class_weight="balanced" # معالجة عدم توازن البيانات
)

# ========================
# 5. تدريب النموذج
# ========================
print("🚀 Training Random Forest...")
model.fit(X_train_flat, y_train_raw)

# ========================
# 6. التقييم التدريجي Walking Forward
# ========================
y_true, y_pred = [], []
print("🔍 جاري التقييم التدريجي...")
for i in tqdm.tqdm(range(len(X_test_flat))):
    sample_x = X_test_flat[i].reshape(1, -1)
    true_class = y_test_raw[i]
    predicted_class = model.predict(sample_x)[0]

    y_true.append(true_class)
    y_pred.append(predicted_class)

walk_acc = accuracy_score(y_true, y_pred)
print(f"🧠 Walking Forward Accuracy: {walk_acc:.4f}")

# ========================
# 7. التقييم على بيانات الاختبار كاملة
# ========================
test_pred = model.predict(X_test_flat)
test_acc = accuracy_score(y_test_raw, test_pred)

print(f"📊 Test Accuracy: {test_acc:.4f}")

# ========================
# 8. مصفوفة الالتباس
# ========================
cm = confusion_matrix(y_test_raw, test_pred)

plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=np.unique(y_train_raw),
            yticklabels=np.unique(y_train_raw))
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix - Random Forest")
plt.show()

# ========================
# 9. تقرير الأداء
# ========================
print("\n📈 Classification Report:")
print("="*50)
print(classification_report(y_test_raw, test_pred, digits=3))

# ========================
# 10. إحصائيات إضافية
# ========================
print(f"\n📊 إحصائيات إضافية:")
print(f"عدد عينات التدريب: {len(X_train_flat)}")
print(f"عدد عينات الاختبار: {len(X_test_flat)}")
print(f"عدد الميزات بعد Flatten: {X_train_flat.shape[1]}")
print(f"حجم النافذة الزمنية: {window_size}")
print(f"عدد الأشجار: {model.n_estimators}, عمق الشجرة: {model.max_depth}")
```
