```python
import pandas as pd

# المسار الخاص بالملف
file_path = r"C:\Users\Access\Downloads\TEST\EURUSD_Ticks_20.09.2024-20.09.2024.csv"

# قراءة البيانات من الملف
df = pd.read_csv(file_path)
```

```python
import pandas as pd

# قراءة البيانات من الملف
data = df

# إنشاء قائمة لتخزين الشموع اليابانية
candles = []

# كل 3 أسطر يتم تحويلها إلى شمعة
for i in range(0, len(data), 3):
    subset = data.iloc[i:i+3]
    
    if len(subset) == 3:
        open_price = subset['Ask'].iloc[0]  # أول سعر
        close_price = subset['Ask'].iloc[2]  # آخر سعر
        high_price = subset['Ask'].max()  # أعلى سعر
        low_price = subset['Ask'].min()  # أقل سعر
        
        open_time = subset['Gmt time'].iloc[0]  # أول وقت
        
        # حساب الفوليوم الإجمالي
        total_volume = subset['AskVolume'].sum() + subset['BidVolume'].sum()
        
        # إضافة الشمعة إلى القائمة
        candles.append({
            'Time': open_time,
            'Open': open_price,
            'High': high_price,
            'Low': low_price,
            'Close': close_price,
            'Volume': total_volume
        })

# تحويل قائمة الشموع إلى DataFrame
candles_df = pd.DataFrame(candles)

# حفظ الشموع في ملف جديد
candles_df.to_csv(r"C:\Users\Access\Downloads\TEST\test.csv", index=False)
```

```python
from sklearn.preprocessing import MinMaxScaler
df_test=candles_df
# تحديد الأعمدة التي سيتم تطبيعها
columns_to_normalize = ['Open', 'High', 'Low', 'Close', 'Volume']

# نسخ البيانات الأصلية حتى لا تتأثر
df_test_normalized = df_test.copy()

# تعريف الـ Scaler
scaler = MinMaxScaler()

# تطبيق الـ Scaler على الأعمدة المحددة
df_test_normalized[columns_to_normalize] = scaler.fit_transform(df_test[columns_to_normalize])

# عرض البيانات المطبعة
print(df_test_normalized.head())
```
