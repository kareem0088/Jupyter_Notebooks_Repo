```python
import asyncio
import json
import websockets
import pandas as pd
from collections import defaultdict
import time

# إعداد متغيرات لتجميع البيانات
agg_data = defaultdict(list)
interval = 60  # ثانية = دقيقة واحدة

async def collect_trades():
    url = "wss://stream.binance.com:9443/ws/btcusdt@trade"
    async with websockets.connect(url) as ws:
        start_time = time.time()

        while True:
            msg = await ws.recv()
            data = json.loads(msg)

            trade_time = data['T']  # وقت الصفقة (ms)
            price = float(data['p'])
            qty = float(data['q'])
            is_buyer_maker = data['m']  # true = بيع، false = شراء

            # تصنيف شراء / بيع
            side = "sell" if is_buyer_maker else "buy"

            agg_data["prices"].append(price)
            agg_data["volume"].append(qty)
            agg_data["trades"].append(1)
            agg_data[f"{side}_volume"].append(qty)
            agg_data[f"{side}_trades"].append(1)

            # تحقق من انتهاء الفترة (1 دقيقة)
            if time.time() - start_time >= interval:
                df = pd.DataFrame({
                    "time": [pd.to_datetime(trade_time, unit='ms')],
                    "open": [agg_data["prices"][0]],
                    "high": [max(agg_data["prices"])],
                    "low": [min(agg_data["prices"])],
                    "close": [agg_data["prices"][-1]],
                    "volume": [sum(agg_data["volume"])],
                    "trades": [sum(agg_data["trades"])],
                    "buy_volume": [sum(agg_data["buy_volume"])],
                    "sell_volume": [sum(agg_data["sell_volume"])],
                    "buy_trades": [sum(agg_data["buy_trades"])],
                    "sell_trades": [sum(agg_data["sell_trades"])],
                })

                # مشتقات إضافية
                df["delta_volume"] = df["buy_volume"] - df["sell_volume"]
                df["avg_trade_size"] = df["volume"] / df["trades"]
                df["imbalance"] = (df["buy_volume"] - df["sell_volume"]) / df["volume"]
                df["vwap"] = sum([p*q for p, q in zip(agg_data["prices"], agg_data["volume"])]) / df["volume"]
                df["volatility"] = (df["high"] - df["low"]) / df["open"]

                print(df)

                # إعادة التهيئة للفترة التالية
                agg_data.clear()
                start_time = time.time()

asyncio.run(collect_trades())
```

```python
import pandas as pd
import numpy as np
from binance.client import Client
from datetime import datetime, timedelta, timezone
from tqdm import tqdm   # مكتبة مؤشر التحميل

# اتصال بباينانس
client = Client("", "")  # مش محتاج API Keys للبيانات العامة

symbol = "BTCUSDT"
interval = Client.KLINE_INTERVAL_1MINUTE
end = datetime.now(timezone.utc)
start = end - timedelta(days=1)

# تحميل الشموع التاريخية (OHLCV)
klines = client.get_historical_klines(
    symbol,
    interval,
    start.strftime("%Y-%m-%d %H:%M:%S"),
    end.strftime("%Y-%m-%d %H:%M:%S")
)

# DataFrame للشموع
df = pd.DataFrame(klines, columns=[
    "open_time","open","high","low","close","volume",
    "close_time","quote_av","trades","taker_buy_base_av","taker_buy_quote_av","ignore"
])

# تحويل الأنواع
float_cols = ["open","high","low","close","volume","taker_buy_base_av","taker_buy_quote_av","quote_av"]
df[float_cols] = df[float_cols].astype(float)
df["trades"] = df["trades"].astype(int)
df["time"] = pd.to_datetime(df["open_time"], unit="ms")

# ===========================================================
# هنا نجيب بيانات الصفقات (trades) ونحسب buy/sell بالظبط
# ===========================================================

buy_trades_list = []
sell_trades_list = []
buy_volume_list = []
sell_volume_list = []

# إضافة tqdm كـ Progress Bar
for i, row in tqdm(df.iterrows(), total=len(df), desc="Processing candles"):
    start_t = int(row["open_time"])
    end_t = int(row["close_time"])
    
    trades = client.get_aggregate_trades(symbol=symbol, startTime=start_t, endTime=end_t, limit=1000)
    
    # تصنيف الصفقات
    buys = [float(t["q"]) for t in trades if not t["m"]]   # m=False => المشتري هو taker
    sells = [float(t["q"]) for t in trades if t["m"]]      # m=True  => البائع هو taker
    
    buy_trades_list.append(len(buys))
    sell_trades_list.append(len(sells))
    buy_volume_list.append(sum(buys))
    sell_volume_list.append(sum(sells))

df["buy_trades"] = buy_trades_list
df["sell_trades"] = sell_trades_list
df["buy_volume"] = buy_volume_list
df["sell_volume"] = sell_volume_list

# المشتقات
df["delta_volume"] = df["buy_volume"] - df["sell_volume"]
df["avg_trade_size"] = df["volume"] / df["trades"]
df["imbalance"] = df["buy_volume"] / df["volume"]
df["vwap"] = df["quote_av"] / df["volume"]
df["volatility"] = df["high"] - df["low"]

# ترتيب الأعمدة
cols = ["time","open","high","low","close","volume","trades",
        "buy_volume","sell_volume","buy_trades","sell_trades",
        "delta_volume","avg_trade_size","imbalance","vwap","volatility"]
df = df[cols]

print(df.head())
```

```python
import pandas as pd
from binance.client import Client
from datetime import datetime, timedelta, timezone

client = Client("", "")  

symbol = "BTCUSDT"
end = datetime.now(timezone.utc)
start = end - timedelta(days=10)

# سحب كل الصفقات (Trades) من الفترة كلها
trades = client.get_aggregate_trades(
    symbol=symbol,
    startTime=int(start.timestamp()*1000),
    endTime=int(end.timestamp()*1000)
)

# نحول DataFrame
df_trades = pd.DataFrame(trades)
df_trades["time"] = pd.to_datetime(df_trades["T"], unit="ms")
df_trades["price"] = df_trades["p"].astype(float)
df_trades["qty"] = df_trades["q"].astype(float)

# نفرق بين شراء وبيع
df_trades["is_buyer"] = ~df_trades["m"]   # m=False => buyer taker
df_trades["buy_qty"] = df_trades.apply(lambda x: x["qty"] if x["is_buyer"] else 0, axis=1)
df_trades["sell_qty"] = df_trades.apply(lambda x: x["qty"] if not x["is_buyer"] else 0, axis=1)
df_trades["buy_trades"] = df_trades["is_buyer"].astype(int)
df_trades["sell_trades"] = (~df_trades["is_buyer"]).astype(int)

# تجميع بالدقيقة
agg = df_trades.resample("1min", on="time").agg({
    "price":"ohlc",
    "qty":"sum",
    "buy_qty":"sum",
    "sell_qty":"sum",
    "buy_trades":"sum",
    "sell_trades":"sum",
    "is_buyer":"count"
})

agg.columns = ["open","high","low","close","volume","buy_volume","sell_volume","buy_trades","sell_trades","trades"]

# حساب المشتقات
agg["delta_volume"] = agg["buy_volume"] - agg["sell_volume"]
agg["avg_trade_size"] = agg["volume"] / agg["trades"]
agg["imbalance"] = agg["buy_volume"] / agg["volume"]
agg["vwap"] = agg["volume"] / agg["trades"]
agg["volatility"] = agg["high"] - agg["low"]

print(agg.head())
```
