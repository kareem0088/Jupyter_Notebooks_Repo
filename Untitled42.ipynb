```python
# التأكد من وجود الأعمدة المطلوبة
if not all(col in df.columns for col in ['High', 'Low', 'Close']):
    raise ValueError("البيانات يجب أن تحتوي على الأعمدة 'High', 'Low', و 'Close' لحساب ATR.")

# حساب التذبذب المتوسط (ATR) لفترات مختلفة
def calculate_atr(df, period):
    df['H-L'] = df['High'] - df['Low']
    df['H-C'] = abs(df['High'] - df['Close'].shift(1))
    df['L-C'] = abs(df['Low'] - df['Close'].shift(1))
    df['TR'] = df[['H-L', 'H-C', 'L-C']].max(axis=1)
    atr = df['TR'].rolling(window=period).mean()
    return atr

# إضافة ATR للفترات المطلوبة
df['atr_1'] = calculate_atr(df, 1)
df['atr_5'] = calculate_atr(df, 5)
df['atr_14'] = calculate_atr(df, 14)
```

```python
import numpy as np
import pandas as pd
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import classification_report

# افتراض أن البيانات تم تحميلها في DataFrame (bs)
# إزالة القيم 0 من عمود "trade_result" (تصفية البيانات)
# استبدال جميع القيم -1 بـ 0 في عمود trade_result
#gg['trade_result'] = gg['trade_result'].replace(-1, 0)


# إعداد البيانات (إزالة الأعمدة غير المطلوبة)
#X = bs.drop(columns=['Is_Hammer','trade_result',"trade_type", 'tp_buy', 'sl_buy', 'tp_sell', 'sl_sell', 'fvg_hammer_buy', 'fvg_hammer_sell','Bullish_Hammer','Bearish_Hammer','Bullish','Bearish','Change','Change_Percentage'])
X = dff[['Gmt time','Open','High','Low','Close','Upper_Wick','Lower_Wick','Wick_Size','Is_Hammer','Z_Score_Upper_Wick','Z_Score_Lower_Wick','Candle_Body','Volatility_Index','Bullish_Hammer','Bearish_Hammer','Bullish','Bearish']]
y = dff["trade_result"]

# ترميز التصنيفات (None, Buy, Sell) إلى قيم عددية
label_encoder = LabelEncoder()
y = label_encoder.fit_transform(y)

# تطبيع البيانات (Normalization) للميزات
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# تقسيم البيانات إلى تدريب واختبار
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# إعادة تشكيل البيانات لتناسب تنسيق LSTM [samples, time_steps, features]
X_train = X_train.reshape(X_train.shape[0], 1, X_train.shape[1])  # 1 time step
X_test = X_test.reshape(X_test.shape[0], 1, X_test.shape[1])  # 1 time step

# بناء نموذج LSTM متعدد الطبقات
model = Sequential()

# الطبقة الأولى LSTM
model.add(LSTM(units=128, return_sequences=True, input_shape=(X_train.shape[1], X_train.shape[2])))
model.add(Dropout(0.3))  # تقليل الإفراط في التكيف

# الطبقة الثانية LSTM
model.add(LSTM(units=64, return_sequences=False))
model.add(Dropout(0.3))

# طبقة Fully Connected إضافية
model.add(Dense(units=32, activation='relu'))
model.add(Dropout(0.2))

# طبقة الإخراج
model.add(Dense(units=3, activation='softmax'))  # ثلاث فئات للإخراج (None, Buy, Sell)

# تجميع النموذج
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# تدريب النموذج
model.fit(X_train, y_train, epochs=2, batch_size=1000, validation_data=(X_test, y_test))

# تقييم النموذج على بيانات الاختبار
loss, accuracy = model.evaluate(X_test, y_test)
print(f"Accuracy on test data: {accuracy * 100:.2f}%")

# التنبؤ على بيانات الاختبار
y_pred = model.predict(X_test)
y_pred_classes = np.argmax(y_pred, axis=1)  # تحويل الاحتمالات إلى التصنيفات النهائية

# طباعة تقرير التصنيف باستخدام الأرقام
print("\nClassification Report:")
print(classification_report(y_test, y_pred_classes))

import numpy as np
import pandas as pd
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns
import matplotlib.pyplot as plt

# حساب مصفوفة التشويش
cm = confusion_matrix(y_test, y_pred_classes)

# طباعة المصفوفة
print("Confusion Matrix:")
print(cm)

# رسم المصفوفة باستخدام seaborn
plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=label_encoder.classes_, yticklabels=label_encoder.classes_)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()

```
