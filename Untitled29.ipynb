```python
import pandas as pd
import numpy as np

data['Result'] = data['Result'].map({'WIN': 1, 'LOSS': -1, 'NONE': 0})  # تحويل النصوص إلى أرقام

import gym
from gym import spaces

class TradingEnv(gym.Env):
    def __init__(self, data):
        super(TradingEnv, self).__init__()
        self.data = data
        self.current_step = 0

        # مساحة الحالة: جميع المؤشرات والأسعار
        self.observation_space = spaces.Box(low=-np.inf, high=np.inf, shape=(data.shape[1]-3,), dtype=np.float32)

        # مساحة الإجراءات: [0: لا صفقة, 1: شراء, -1: بيع]
        self.action_space = spaces.Discrete(3)

    def reset(self):
        self.current_step = 0
        return self.data.iloc[self.current_step, 1:-3].values  # بدون الأعمدة Gmt time, Signal, Take Profit, Stop Loss, Result

    def step(self, action):
        done = False
        self.current_step += 1
        if self.current_step >= len(self.data) - 1:
            done = True

        # المكافأة بناءً على الإجراء ونتيجة الصفقة
        reward = 0
        if action == 1 and self.data.iloc[self.current_step]['Signal'] == 1:
            reward = self.data.iloc[self.current_step]['Result']
        elif action == -1 and self.data.iloc[self.current_step]['Signal'] == -1:
            reward = self.data.iloc[self.current_step]['Result']
        elif action == 0:
            reward = 0

        # الحالة التالية
        next_state = self.data.iloc[self.current_step, 1:-3].values
        return next_state, reward, done, {}

from stable_baselines3 import PPO

# إنشاء البيئة
env = TradingEnv(data)

# إنشاء نموذج PPO
model = PPO("MlpPolicy", env, verbose=1)

# تدريب النموذج
model.learn(total_timesteps=100000)


for step in range(len(test_data)):
    observation = test_data.iloc[step, 1:-3].values

    # توقع الإجراء
    action, _ = model.predict(observation)

    if action == 1:
        print(f"شراء عند {test_data.iloc[step]['Open']}")
    elif action == -1:
        print(f"بيع عند {test_data.iloc[step]['Open']}")
    else:
        print("لا توجد صفقة")
```
