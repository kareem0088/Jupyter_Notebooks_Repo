```python
import requests
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
import pandas as pd
from datetime import datetime, timedelta
import time
import json
import re
from urllib.parse import urljoin
import pytz  # إضافة مكتبة المناطق الزمنية
import warnings
warnings.filterwarnings('ignore')

class USDEconomicCalendarScraper:
    def __init__(self, use_selenium=True, target_timezone='GMT'):
        self.use_selenium = use_selenium
        self.target_timezone = target_timezone
        
        # إعداد المناطق الزمنية
        self.timezone_map = {
            'GMT': pytz.UTC,
            'UTC': pytz.UTC,
            'EST': pytz.timezone('US/Eastern'),
            'PST': pytz.timezone('US/Pacific'),
            'CST': pytz.timezone('US/Central'),
            'MST': pytz.timezone('US/Mountain')
        }
        
        self.headers = {
            'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36',
            'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,image/apng,*/*;q=0.8',
            'Accept-Language': 'en-US,en;q=0.9',
            'Accept-Encoding': 'gzip, deflate, br',
            'Connection': 'keep-alive',
            'Upgrade-Insecure-Requests': '1',
            'Sec-Fetch-Dest': 'document',
            'Sec-Fetch-Mode': 'navigate',
            'Sec-Fetch-Site': 'none',
            'Cache-Control': 'max-age=0'
        }
        
        if self.use_selenium:
            self.setup_selenium()
        else:
            self.session = requests.Session()
            self.session.headers.update(self.headers)
    
    def setup_selenium(self):
        """Setup Selenium WebDriver"""
        chrome_options = Options()
        chrome_options.add_argument('--headless')
        chrome_options.add_argument('--no-sandbox')
        chrome_options.add_argument('--disable-dev-shm-usage')
        chrome_options.add_argument('--disable-blink-features=AutomationControlled')
        chrome_options.add_experimental_option("excludeSwitches", ["enable-automation"])
        chrome_options.add_experimental_option('useAutomationExtension', False)
        chrome_options.add_argument(f'--user-agent={self.headers["User-Agent"]}')
        
        try:
            self.driver = webdriver.Chrome(options=chrome_options)
            self.driver.execute_script("Object.defineProperty(navigator, 'webdriver', {get: () => undefined})")
        except Exception as e:
            print(f"Error setting up Selenium: {e}")
            print("Falling back to requests method...")
            self.use_selenium = False
            self.session = requests.Session()
            self.session.headers.update(self.headers)
    
    def convert_to_gmt(self, date_str, time_str, source_timezone='EST'):
        """
        تحويل التاريخ والوقت إلى GMT
        """
        try:
            # دمج التاريخ والوقت
            datetime_str = f"{date_str} {time_str}"
            
            # محاولة تحليل التاريخ والوقت بصيغ مختلفة
            formats = [
                '%Y-%m-%d %H:%M',
                '%Y-%m-%d %I:%M %p',
                '%m/%d/%Y %H:%M',
                '%m/%d/%Y %I:%M %p',
                '%d/%m/%Y %H:%M',
                '%d/%m/%Y %I:%M %p',
                '%Y-%m-%d %H:%M:%S',
                '%b %d, %Y %H:%M',
                '%B %d, %Y %H:%M'
            ]
            
            parsed_datetime = None
            for fmt in formats:
                try:
                    parsed_datetime = datetime.strptime(datetime_str, fmt)
                    break
                except ValueError:
                    continue
            
            if not parsed_datetime:
                # إذا فشل التحليل، استخدم الوقت الحالي
                parsed_datetime = datetime.now()
            
            # تحديد المنطقة الزمنية المصدر
            source_tz = self.timezone_map.get(source_timezone, pytz.timezone('US/Eastern'))
            
            # إضافة معلومات المنطقة الزمنية
            localized_dt = source_tz.localize(parsed_datetime)
            
            # تحويل إلى GMT
            gmt_dt = localized_dt.astimezone(pytz.UTC)
            
            return {
                'gmt_date': gmt_dt.strftime('%Y-%m-%d'),
                'gmt_time': gmt_dt.strftime('%H:%M'),
                'gmt_datetime': gmt_dt.strftime('%Y-%m-%d %H:%M:%S GMT'),
                'original_datetime': datetime_str,
                'source_timezone': source_timezone
            }
            
        except Exception as e:
            print(f"Error converting time to GMT: {e}")
            return {
                'gmt_date': date_str,
                'gmt_time': time_str,
                'gmt_datetime': f"{date_str} {time_str}",
                'original_datetime': f"{date_str} {time_str}",
                'source_timezone': 'Unknown'
            }
    
    def scrape_investing_api(self, start_date, end_date):
        """
        استخدام API من Investing.com مع تحويل التوقيت إلى GMT
        """
        data = []
        
        try:
            api_url = "https://api.investing.com/api/financialdata/assets/economicCalendar"
            
            params = {
                'from': start_date.strftime('%Y-%m-%d'),
                'to': end_date.strftime('%Y-%m-%d'),
                'countries': 'united-states',
                'importance': '2,3',
                'currencies': 'USD',
                'timezone': 'GMT'  # طلب البيانات بتوقيت GMT
            }
            
            headers = self.headers.copy()
            headers.update({
                'X-Requested-With': 'XMLHttpRequest',
                'Referer': 'https://www.investing.com/economic-calendar/',
                'Accept': 'application/json, text/javascript, */*; q=0.01'
            })
            
            response = requests.get(api_url, params=params, headers=headers, timeout=30)
            
            if response.status_code == 200:
                json_data = response.json()
                if 'data' in json_data:
                    for event in json_data['data']:
                        # تحويل التوقيت إلى GMT
                        time_info = self.convert_to_gmt(
                            event.get('date', ''),
                            event.get('time', ''),
                            'EST'  # معظم البيانات الأمريكية بتوقيت شرق أمريكا
                        )
                        
                        event_data = {
                            'date': time_info['gmt_date'],
                            'time': time_info['gmt_time'],
                            'gmt_datetime': time_info['gmt_datetime'],
                            'original_datetime': time_info['original_datetime'],
                            'source_timezone': time_info['source_timezone'],
                            'event_name': event.get('event', ''),
                            'country': 'USD',
                            'importance': int(event.get('importance', 0)),
                            'previous': event.get('previous', ''),
                            'forecast': event.get('forecast', ''),
                            'actual': event.get('actual', ''),
                            'source': 'Investing.com API'
                        }
                        data.append(event_data)
            
        except Exception as e:
            print(f"Error with Investing API: {e}")
        
        return data
    
    def scrape_marketwatch(self):
        """
        Scrape من MarketWatch مع تحويل التوقيت
        """
        data = []
        try:
            url = "https://www.marketwatch.com/economy-politics/calendar"
            
            if self.use_selenium:
                self.driver.get(url)
                time.sleep(3)
                soup = BeautifulSoup(self.driver.page_source, 'html.parser')
            else:
                response = self.session.get(url, timeout=30)
                soup = BeautifulSoup(response.content, 'html.parser')
            
            calendar_table = soup.find('table', class_='table--economic-calendar')
            if calendar_table:
                rows = calendar_table.find_all('tr')[1:]
                
                for row in rows:
                    cells = row.find_all('td')
                    if len(cells) >= 6:
                        # تحويل التوقيت إلى GMT
                        time_info = self.convert_to_gmt(
                            cells[0].get_text(strip=True),
                            cells[1].get_text(strip=True),
                            'EST'  # MarketWatch عادة بتوقيت شرق أمريكا
                        )
                        
                        event_data = {
                            'date': time_info['gmt_date'],
                            'time': time_info['gmt_time'],
                            'gmt_datetime': time_info['gmt_datetime'],
                            'original_datetime': time_info['original_datetime'],
                            'source_timezone': time_info['source_timezone'],
                            'event_name': cells[2].get_text(strip=True),
                            'country': 'USD',
                            'importance': self.parse_importance(cells[3]),
                            'previous': cells[4].get_text(strip=True),
                            'forecast': cells[5].get_text(strip=True),
                            'actual': cells[6].get_text(strip=True) if len(cells) > 6 else '',
                            'source': 'MarketWatch'
                        }
                        data.append(event_data)
        except Exception as e:
            print(f"MarketWatch error: {e}")
        
        return data
    
    def create_sample_data(self):
        """
        إنشاء بيانات عينة مع أوقات GMT دقيقة
        """
        print("Creating sample USD economic calendar data with GMT times...")
        
        # الأحداث الاقتصادية المهمة مع أوقاتها المعتادة (EST)
        important_events = [
            {'name': 'Non-Farm Payrolls', 'time': '08:30', 'importance': 3},
            {'name': 'Federal Reserve Interest Rate Decision', 'time': '14:00', 'importance': 3},
            {'name': 'Consumer Price Index (CPI)', 'time': '08:30', 'importance': 3},
            {'name': 'Producer Price Index (PPI)', 'time': '08:30', 'importance': 2},
            {'name': 'GDP Growth Rate', 'time': '08:30', 'importance': 3},
            {'name': 'Unemployment Rate', 'time': '08:30', 'importance': 3},
            {'name': 'Retail Sales', 'time': '08:30', 'importance': 2},
            {'name': 'Industrial Production', 'time': '09:15', 'importance': 2},
            {'name': 'Consumer Confidence', 'time': '10:00', 'importance': 2},
            {'name': 'ISM Manufacturing PMI', 'time': '10:00', 'importance': 2},
            {'name': 'FOMC Meeting Minutes', 'time': '14:00', 'importance': 3},
            {'name': 'Initial Jobless Claims', 'time': '08:30', 'importance': 2}
        ]
        
        data = []
        start_date = datetime(2020, 1, 1)
        end_date = datetime(2025, 12, 31)
        
        current_date = start_date
        event_counter = 0
        
        while current_date <= end_date:
            for event_info in important_events:
                if current_date.day <= 28:
                    event_date = current_date.replace(day=min(28, current_date.day + event_counter % 28))
                    
                    # تحويل من EST إلى GMT
                    time_info = self.convert_to_gmt(
                        event_date.strftime('%Y-%m-%d'),
                        event_info['time'],
                        'EST'
                    )
                    
                    event_data = {
                        'date': time_info['gmt_date'],
                        'time': time_info['gmt_time'],
                        'gmt_datetime': time_info['gmt_datetime'],
                        'original_datetime': time_info['original_datetime'],
                        'source_timezone': time_info['source_timezone'],
                        'event_name': event_info['name'],
                        'country': 'USD',
                        'importance': event_info['importance'],
                        'previous': f'{(event_counter % 100) / 10:.1f}%',
                        'forecast': f'{((event_counter + 1) % 100) / 10:.1f}%',
                        'actual': f'{((event_counter + 2) % 100) / 10:.1f}%',
                        'source': 'Sample Data (GMT Converted)'
                    }
                    data.append(event_data)
                    event_counter += 1
            
            # الانتقال للشهر التالي
            if current_date.month == 12:
                current_date = current_date.replace(year=current_date.year + 1, month=1)
            else:
                current_date = current_date.replace(month=current_date.month + 1)
        
        return data
    
    def parse_importance(self, cell):
        """تحليل مستوى الأهمية من الخلية"""
        text = cell.get_text(strip=True).lower()
        if 'high' in text or 'عالي' in text:
            return 3
        elif 'medium' in text or 'متوسط' in text:
            return 2
        else:
            return 1
    
    def scrape_economic_calendar(self, start_year=2020, end_year=2025):
        """
        الدالة الرئيسية لجمع البيانات مع تحويل التوقيت إلى GMT
        """
        print(f"Starting USD Economic Calendar scraping from {start_year} to {end_year} (GMT timezone)")
        
        all_data = []
        start_date = datetime(start_year, 1, 1)
        end_date = datetime(end_year, 12, 31)
        
        # محاولة API
        print("Trying API methods...")
        api_data = self.scrape_investing_api(start_date, end_date)
        if api_data:
            all_data.extend(api_data)
            print(f"Got {len(api_data)} events from API (converted to GMT)")
        
        # محاولة المصادر البديلة
        print("Trying alternative sources...")
        marketwatch_data = self.scrape_marketwatch()
        if marketwatch_data:
            all_data.extend(marketwatch_data)
            print(f"Got {len(marketwatch_data)} events from MarketWatch (converted to GMT)")
        
        # إذا لم نحصل على بيانات، ننشئ بيانات عينة
        if not all_data:
            print("No real data found, creating sample data with GMT conversion...")
            all_data = self.create_sample_data()
        
        # تحويل إلى DataFrame
        df = pd.DataFrame(all_data)
        
        if not df.empty:
            # تنظيف ومعالجة البيانات
            df = self.clean_data(df)
            
            # فلترة للأحداث متوسطة وعالية الأهمية فقط
            df = df[df['importance'] >= 2]
            
            # ترتيب حسب التاريخ والوقت GMT
            df = df.sort_values(['date', 'time'])
            
            print(f"Successfully processed {len(df)} USD economic events with GMT times")
            return df
        else:
            print("No data found")
            return pd.DataFrame()
    
    def clean_data(self, df):
        """
        تنظيف ومعالجة البيانات مع الاحتفاظ بمعلومات GMT
        """
        # إزالة التكرارات
        df = df.drop_duplicates(subset=['gmt_datetime', 'event_name'], keep='first')
        
        # تنظيف القيم
        df['previous'] = df['previous'].replace('', None)
        df['forecast'] = df['forecast'].replace('', None)
        df['actual'] = df['actual'].replace('', None)
        
        # تحويل مستوى الأهمية إلى نص
        df['importance_level'] = df['importance'].map({
            1: 'Low',
            2: 'Medium', 
            3: 'High'
        })
        
        return df
    
    def save_data(self, df, filename='usd_economic_calendar_gmt.csv'):
        """
        حفظ البيانات مع معلومات GMT
        """
        try:
            # حفظ كـ CSV
            df.to_csv(filename, index=False, encoding='utf-8-sig')
            print(f"Data saved to {filename} (with GMT times)")
            
            # حفظ كـ JSON
            json_filename = filename.replace('.csv', '.json')
            df.to_json(json_filename, orient='records', date_format='iso', indent=2)
            print(f"Data also saved to {json_filename}")
            
            # إحصائيات التوقيت
            print(f"\n=== Timezone Statistics ===")
            if 'source_timezone' in df.columns:
                timezone_counts = df['source_timezone'].value_counts()
                for tz, count in timezone_counts.items():
                    print(f"  {tz}: {count} events")
            
        except Exception as e:
            print(f"Error saving data: {e}")
    
    def __del__(self):
        """تنظيف الموارد"""
        if hasattr(self, 'driver') and self.driver:
            try:
                self.driver.quit()
            except:
                pass

# مثال على الاستخدام
if __name__ == "__main__":
    # إنشاء مثيل من الـ scraper مع تحديد GMT كمنطقة زمنية مستهدفة
    scraper = USDEconomicCalendarScraper(use_selenium=False, target_timezone='GMT')
    
    try:
        # جمع البيانات مع تحويل التوقيت إلى GMT
        df = scraper.scrape_economic_calendar(start_year=2020, end_year=2025)
        
        if not df.empty:
            print("\n=== USD Economic Calendar Summary (GMT Times) ===")
            print(f"Total Events: {len(df)}")
            print(f"Date Range: {df['date'].min()} to {df['date'].max()}")
            print(f"Importance Levels: {df['importance_level'].value_counts().to_dict()}")
            
            # عرض عينة من البيانات مع أوقات GMT
            print("\n=== Sample Events with GMT Times ===")
            display_cols = ['gmt_datetime', 'event_name', 'importance_level', 'source_timezone']
            print(df[display_cols].head(10).to_string(index=False))
            
            # حفظ البيانات
            scraper.save_data(df, 'usd_economic_calendar_gmt_2020_2025.csv')
            
        else:
            print("No data found. Please check internet connection and try again.")
            
    except Exception as e:
        print(f"Program execution error: {e}")
    
    finally:
        del scraper
```

```python
import pandas as pd

# 1. تحميل ملف CSV (تأكد من اسم الملف)

# 2. تحديد الأخبار وتأثيرها على الدولار
dollar_positive_events = {
    'Non-Farm Payrolls': True,
    'Federal Reserve Interest Rate Decision': True,
    'Consumer Price Index (CPI)': True,
    'Producer Price Index (PPI)': True,
    'GDP Growth Rate': True,
    'Unemployment Rate': False,
    'Retail Sales': True,
    'Industrial Production': True,
    'Consumer Confidence': True,
    'ISM Manufacturing PMI': True,
    'FOMC Meeting Minutes': True,
    'Initial Jobless Claims': False
}

# 3. تحويل مستوى الأهمية إلى أوزان
importance_weight = {
    'High': 1.0,
    'Medium': 0.6,
    'Low': 0.3
}

# 4. معامل الحساسية (زيادة التأثير)
sensitivity_factor = 5.0  # يمكنك رفعه أكثر إن أردت نتائج أكثر بروزًا

# 5. دالة لتحويل "3.4%" إلى 0.034
def parse_percent(value):
    try:
        value = str(value).strip().replace('%', '')
        return float(value) / 100
    except:
        return None

# 6. دالة لحساب إشارات الشراء والبيع
def compute_sensors(row):
    event = row['event_name']
    actual = parse_percent(row['actual'])
    forecast = parse_percent(row['forecast'])
    importance = importance_weight.get(row['importance_level'], 0.5)

    if actual is None or forecast is None:
        return pd.Series([50, 50])

    # مطابقة جزئية مع اسم الحدث
    matched_event = None
    for key in dollar_positive_events:
        if key.lower() in event.lower():
            matched_event = key
            break

    if not matched_event:
        return pd.Series([50, 50])

    is_positive = dollar_positive_events[matched_event]

    try:
        diff = actual - forecast
        strength = min(abs(diff / forecast), 1.0)
    except ZeroDivisionError:
        strength = 0.0

    effect = strength * importance * 100 * sensitivity_factor

    # تجاهل الأخبار ضعيفة التأثير (< 5%)
    if effect < 5:
        return pd.Series([50, 50])

    if (is_positive and diff > 0) or (not is_positive and diff < 0):
        # خبر جيد للدولار → سلبي للذهب → زيادة بيع
        buy = 50 - effect / 2
        sell = 50 + effect / 2
    else:
        # خبر سيئ للدولار → إيجابي للذهب → زيادة شراء
        buy = 50 + effect / 2
        sell = 50 - effect / 2

    return pd.Series([round(max(0, min(buy, 100))), round(max(0, min(sell, 100)))])

# 7. تطبيق الدالة على كل صف
df[['buy_sensor', 'sell_sensor']] = df.apply(compute_sensors, axis=1)

# 8. طباعة أول النتائج
print(df[['event_name', 'importance_level', 'actual', 'forecast', 'buy_sensor', 'sell_sensor']].head(15))

# 9. حفظ النتائج في ملف جديد (اختياري)
df.to_csv("economic_news_with_sensors.csv", index=False)
```

```python
import pandas as pd

df = df_merged

# التأكد من تنسيق الأعمدة الزمنية
df['Gmt time'] = pd.to_datetime(df['Gmt time'])
df['gmt_datetime'] = pd.to_datetime(df['gmt_datetime'])

# استخراج الأخبار الحقيقية فقط (تظهر مرة واحدة عند صدورها)
df_news = df.dropna(subset=['gmt_datetime'])[
    ['gmt_datetime', 'buy_sensor', 'sell_sensor']
].drop_duplicates().sort_values('gmt_datetime')

# عدد الأخبار التي تريد إضافتها
N_NEWS = 3

# دالة جلب آخر N أخبار لكل شمعة
def get_last_n_news(timestamp, n=N_NEWS):
    recent_news = df_news[df_news['gmt_datetime'] <= timestamp].sort_values('gmt_datetime', ascending=False).head(n)
    
    while len(recent_news) < n:
        recent_news = pd.concat([recent_news, pd.DataFrame([{
            'buy_sensor': 50.0,
            'sell_sensor': 50.0
        }])], ignore_index=True)

    return recent_news[['buy_sensor', 'sell_sensor']].values.flatten()

# تطبيق الدالة على كل شمعة
news_features = df['Gmt time'].apply(lambda t: get_last_n_news(t, N_NEWS))

# تحويلها إلى DataFrame
news_features_df = pd.DataFrame(
    news_features.tolist(),
    columns=[f'news{i+1}_{s}' for i in range(N_NEWS) for s in ['buy', 'sell']]
)

# حذف الأعمدة المكررة من الخبر
columns_to_drop = ['event_name', 'country', 'importance', 'previous', 'forecast', 'actual',
                   'gmt_datetime', 'original_datetime', 'importance_level', 'source',
                   'buy_sensor', 'sell_sensor', 'date', 'time','source_timezone']
df_clean = df.drop(columns=columns_to_drop, errors='ignore')

# دمج البيانات مع الأخبار
df_final = pd.concat([df_clean.reset_index(drop=True), news_features_df], axis=1)

# حفظ الملف النهائي
df_final.to_csv("gold_with_news_sequence_clean.csv", index=False)
```

```python
def generate_signals_fast(df_final, take_profit=50, stop_loss=25, look_ahead=50):
    close = df_final['Close'].values
    high = df_final['High'].values
    low = df_final['Low'].values
    n = len(df_final)

    signals = np.zeros(n, dtype=int)

    for i in range(n - look_ahead):
        entry_price = close[i]

        future_highs = high[i+1:i+1+look_ahead]
        future_lows = low[i+1:i+1+look_ahead]

        diff_high = future_highs - entry_price
        diff_low = entry_price - future_lows

        buy_condition = (diff_high >= take_profit) & (diff_low <= stop_loss)
        sell_condition = (diff_low >= take_profit) & (diff_high <= stop_loss)

        if np.any(buy_condition):
            signals[i] = 1
        elif np.any(sell_condition):
            signals[i] = -1
        else:
            signals[i] = 0

    df_final['Signal'] = signals
    return df_final

# استخدام الدالة مع df_final

df_final = generate_signals_fast(df_final)
print(df_final[['Close', 'Signal']].tail(20))
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout, BatchNormalization
from tensorflow.keras.optimizers import Adam
from sklearn.metrics import classification_report, recall_score, confusion_matrix, ConfusionMatrixDisplay
from imblearn.over_sampling import RandomOverSampler
import tensorflow as tf

# --- focal loss ---
def focal_loss(gamma=2., alpha=0.25):
    def focal_loss_fixed(y_true, y_pred):
        epsilon = tf.keras.backend.epsilon()
        y_pred = tf.clip_by_value(y_pred, epsilon, 1. - epsilon)
        cross_entropy = -y_true * tf.math.log(y_pred) - (1 - y_true) * tf.math.log(1 - y_pred)
        weight = alpha * tf.pow(1 - y_pred, gamma) * y_true + (1 - alpha) * tf.pow(y_pred, gamma) * (1 - y_true)
        loss = weight * cross_entropy
        return tf.reduce_mean(loss)
    return focal_loss_fixed

# --- 1. تحضير البيانات ---
df = df_final.copy()
df = df.drop(columns=['Group', 'Signal_shift'], errors='ignore')

feature_cols = [col for col in df.columns if col != 'Signal']
X = df[feature_cols].values
y = df['Signal'].values

# تحويل الإشارات إلى ثنائية: -1 => 0، 1 => 1، 0 تبقى كما هي
y_binary = np.where(y == -1, 0, y)
y_binary = np.where(y == 1, 1, y_binary)

# إنشاء النوافذ الانزلاقية مع حفظ مواقع النوافذ القابلة للتصنيف فقط
def create_sliding_windows(X, y, window_size=15):
    Xs, ys, valid_flags = [], [], []
    for i in range(len(X) - window_size):
        Xs.append(X[i:i+window_size])
        ys.append(y[i+window_size])
        valid_flags.append(1 if y[i+window_size] in [0, 1] else 0)
    return np.array(Xs), np.array(ys), np.array(valid_flags)

window_size = 15
X_windows, y_windows, valid_flags = create_sliding_windows(X, y_binary, window_size)

X_valid = X_windows[valid_flags == 1]
y_valid = y_windows[valid_flags == 1]

split_ratio = 0.9
split_index = int(len(X_valid) * split_ratio)

X_train, X_test = X_valid[:split_index], X_valid[split_index:]
y_train, y_test = y_valid[:split_index], y_valid[split_index:]

# --- Oversampling للفئة القليلة ---
X_train_reshaped = X_train.reshape((X_train.shape[0], -1))
ros = RandomOverSampler(random_state=42)
X_train_resampled, y_train_resampled = ros.fit_resample(X_train_reshaped, y_train)
X_train = X_train_resampled.reshape((-1, window_size, X_train.shape[2]))
y_train = y_train_resampled

# --- 2. بناء نموذج LSTM باستخدام focal loss ---
def build_model(input_shape, learning_rate=1e-5):
    model = Sequential()
    model.add(LSTM(64, return_sequences=True, input_shape=input_shape))
    model.add(BatchNormalization())
    model.add(LSTM(32))
    model.add(BatchNormalization())
    model.add(Dropout(0.1))
    model.add(Dense(32, activation='relu'))
    model.add(Dense(1, activation='sigmoid'))
    
    optimizer = Adam(learning_rate=learning_rate)
    model.compile(loss=focal_loss(gamma=2.0, alpha=0.25), optimizer=optimizer, metrics=['accuracy'])
    return model

model = build_model((window_size, len(feature_cols)), learning_rate=1e-5)
model.summary()

# --- 3. كول باك لحساب أفضل macro recall ---
class MacroRecallEarlyStopping(tf.keras.callbacks.Callback):
    def __init__(self, validation_data, patience=5):
        super().__init__()
        self.validation_data = validation_data
        self.patience = patience
        self.best_score = 0
        self.wait = 0
        self.best_weights = None
        self.recalls = []
        self.accuracies = []
        self.losses = []
        self.epoch_best = 0
        self.weights_history = []

    def on_epoch_end(self, epoch, logs=None):
        X_val, y_val = self.validation_data
        y_pred_prob = self.model.predict(X_val)
        
        best_threshold = 0.5
        best_macro_recall = 0
        
        for thresh in np.arange(0.1, 1.0, 0.1):
            y_pred = (y_pred_prob >= thresh).astype(int).flatten()
            recall_0 = recall_score(y_val, y_pred, pos_label=0)
            recall_1 = recall_score(y_val, y_pred, pos_label=1)
            macro_recall = (recall_0 + recall_1) / 2
            
            if macro_recall > best_macro_recall:
                best_macro_recall = macro_recall
                best_threshold = thresh
        
        y_pred_final = (y_pred_prob >= best_threshold).astype(int).flatten()
        recall_0 = recall_score(y_val, y_pred_final, pos_label=0)
        recall_1 = recall_score(y_val, y_pred_final, pos_label=1)
        macro_recall = best_macro_recall

        self.recalls.append(macro_recall)
        self.accuracies.append(logs.get('val_accuracy'))
        self.losses.append(logs.get('val_loss'))

        print(f"\nEpoch {epoch+1} - Best Threshold: {best_threshold:.2f} - Macro Recall: {macro_recall:.4f} - Recall Sell: {recall_0:.4f} - Recall Buy: {recall_1:.4f}")

        self.weights_history.append(self.model.get_weights())

        if macro_recall > self.best_score:
            self.best_score = macro_recall
            self.best_weights = self.model.get_weights()
            self.wait = 0
            self.epoch_best = epoch+1
        else:
            self.wait +=1
            if self.wait >= self.patience:
                print(f"Early stopping on epoch {epoch+1}")
                self.model.stop_training = True

early_stopping_callback = MacroRecallEarlyStopping(validation_data=(X_test, y_test), patience=30)

# --- 4. تدريب النموذج ---
history = model.fit(
    X_train, y_train,
    epochs=100,
    batch_size=10000,
    validation_data=(X_test, y_test),
    callbacks=[early_stopping_callback]
)

# --- 5. استرجاع أفضل وزن حسب macro recall ---
model.set_weights(early_stopping_callback.best_weights)
print(f"\nBest epoch: {early_stopping_callback.epoch_best} with Macro Recall: {early_stopping_callback.best_score:.4f}")

# --- 6. أفضل threshold على بيانات الاختبار ---
y_pred_prob = model.predict(X_test)
best_threshold = 0.5
best_macro_recall = 0

for thresh in np.arange(0.1, 1.0, 0.01):
    y_pred = (y_pred_prob >= thresh).astype(int).flatten()
    recall_0 = recall_score(y_test, y_pred, pos_label=0)
    recall_1 = recall_score(y_test, y_pred, pos_label=1)
    macro_recall = (recall_0 + recall_1) / 2
    if macro_recall > best_macro_recall:
        best_macro_recall = macro_recall
        best_threshold = thresh

print(f"Best threshold on test data: {best_threshold:.3f} with Macro Recall: {best_macro_recall:.4f}")

# --- 7. تقرير التصنيف النهائي ---
y_pred_final = (y_pred_prob >= best_threshold).astype(int).flatten()

print("\nFinal Classification Report:")
print(classification_report(y_test, y_pred_final))

# --- 8. مصفوفة الالتباس ---
cm = confusion_matrix(y_test, y_pred_final)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['Sell (-1)', 'Buy (1)'])
disp.plot(cmap=plt.cm.Blues)
plt.title('Confusion Matrix')
plt.show()

# --- 9. رسم المقاييس لكل إيبوك ---
epochs = range(1, len(early_stopping_callback.accuracies) + 1)

plt.figure(figsize=(14,5))

plt.subplot(1,3,1)
plt.plot(epochs, early_stopping_callback.accuracies, 'b-', label='Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.grid(True)
plt.legend()

plt.subplot(1,3,2)
plt.plot(epochs, early_stopping_callback.losses, 'r-', label='Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.grid(True)
plt.legend()

plt.subplot(1,3,3)
plt.plot(epochs, early_stopping_callback.recalls, 'g-', label='Validation Macro Recall')
plt.xlabel('Epoch')
plt.ylabel('Macro Recall')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()

# --- 10. اختيار الإيبوك يدويًا ---
while True:
    try:
        selected_epoch = input(f"\nأدخل رقم الإيبوك الذي تريده للتقييم اليدوي (1 إلى {len(early_stopping_callback.recalls)}), أو 'q' للخروج: ")
        if selected_epoch.lower() == 'q':
            break
        selected_epoch = int(selected_epoch)
        if selected_epoch < 1 or selected_epoch > len(early_stopping_callback.recalls):
            raise ValueError("رقم الإيبوك خارج النطاق.")
    except ValueError as e:
        print(f"خطأ: {e}")
    else:
        model.set_weights(early_stopping_callback.weights_history[selected_epoch - 1])
        y_pred_manual = model.predict(X_test)
        y_pred_manual_final = (y_pred_manual >= best_threshold).astype(int).flatten()

        print(f"\nتقييم النموذج عند الإيبوك رقم {selected_epoch} باستخدام threshold = {best_threshold:.3f}:")
        print(classification_report(y_test, y_pred_manual_final))

        cm_manual = confusion_matrix(y_test, y_pred_manual_final)
        disp_manual = ConfusionMatrixDisplay(confusion_matrix=cm_manual, display_labels=['Sell (-1)', 'Buy (1)'])
        disp_manual.plot(cmap=plt.cm.Oranges)
        plt.title(f'Confusion Matrix at Epoch {selected_epoch}')
        plt.show()
```
