```python
import pandas as pd
df= pd.read_csv(r"C:\Users\Access\Documents\AI ROBOT\XAUUSD_Candlestick_1_M_BID_2020-1-1_2024-12-31.csv")
df=df.drop(columns='Volume')
import pandas as pd

# تحويل العمود الزمني إلى نوع datetime وتعيينه كمؤشر
df['Gmt time'] = pd.to_datetime(df['Gmt time'])
df.set_index('Gmt time', inplace=True)

# تحويل البيانات من 1 دقيقة إلى 5 دقائق
df_5m = df.resample('5T').agg({
    'Open': 'first',   # أول سعر افتتاح في الفاصل الزمني
    'High': 'max',     # أعلى سعر في الفاصل الزمني
    'Low': 'min',      # أقل سعر في الفاصل الزمني
    'Close': 'last'    # آخر سعر إغلاق في الفاصل الزمني
})

# عرض النتيجة
df_5m=df_5m.reset_index(drop=False) 
```

```python
import pandas as pd
import matplotlib.pyplot as plt


# تعريف القوائم للمعاملات
BB_list = [0.1, 0.5, 1, 1.5, 2, 2.5, 3, 3.5, 4, 4.5, 5]
XX_list = [0.1, 0.5, 1, 1.5, 2, 2.5, 3, 3.5, 4, 4.5, 5]
VV_list = [0.1, 0.5, 1, 1.5, 2, 2.5, 3, 3.5, 4, 4.5, 5]
TP_list = [2, 2.5, 3, 3.5, 4]
SL_list = [1.5, 2, 2.5, 3 ]

# قائمة لتخزين رقم العملية والقيمة النهائية (final_balance) للسطر الأخير لكل عملية
operations = []

# متغير لترقيم العمليات
op_number = 1

# التكرار عبر جميع التركيبات
for BB in BB_list:
    for XX in XX_list:
        for VV in VV_list:
            for TP in TP_list:
                for SL in SL_list:
                    # استخدام نسخة جديدة من البيانات لكل تركيبة
                    df = df_5m[288:576].copy()
                    
                    # دالة التحقق من الشمعة الكبيرة باستخدام BB
                    def is_large_body(candle):
                        body = abs(candle['Open'] - candle['Close'])
                        lower_shadow = abs(candle['Low'] - min(candle['Open'], candle['Close']))
                        upper_shadow = abs(candle['High'] - max(candle['Open'], candle['Close']))
                        return (body * BB) and (lower_shadow * BB) and (upper_shadow * BB)
                    
                    # دالة التحقق من شمعة الهامر باستخدام XX و VV
                    def is_hammer(candle):
                        body = abs(candle['Open'] - candle['Close'])
                        lower_shadow = abs(candle['Low'] - min(candle['Open'], candle['Close']))
                        upper_shadow = abs(candle['High'] - max(candle['Open'], candle['Close']))
                        return (body * BB) and (lower_shadow * BB) and (upper_shadow * BB)
                    
                    # دالة حساب صفقة شراء باستخدام الهامر
                    def calculate_buy_trade_hammer(row, df, tp_multiplier=TP, sl_multiplier=SL):
                        idx = row.name  # الفهرس الزمني
                        if idx == df.index[0]:
                            return 0, None, None  # لا توجد شمعة سابقة
                        first_candle = df.loc[df.index[df.index.get_loc(idx) - 1]]
                        second_candle = row
                        if (is_large_body(first_candle) and 
                            first_candle['Open'] > first_candle['Close'] and 
                            is_hammer(second_candle) and 
                            second_candle['Open'] < second_candle['Close'] and 
                            (abs(first_candle['Open'] - first_candle['Close']) >= 0.75)):
                            
                            points = abs(first_candle['Open'] - first_candle['Close'])
                            stop_loss = second_candle['Close'] - (points * sl_multiplier)
                            take_profit = second_candle['Close'] + (tp_multiplier * points)
                            return 1, take_profit, stop_loss
                        return 0, None, None
                    
                    # تطبيق صفقة الشراء
                    df['buy_trade_hammer'], df['tp_buy_hammer'], df['sl_buy_hammer'] = zip(*df.apply(lambda row: calculate_buy_trade_hammer(row, df), axis=1))
                    df['trade_type_hammer'] = df.apply(lambda row: 1 if row['buy_trade_hammer'] == 1 else 0, axis=1)
                    df['tp_hammer'] = df.apply(lambda row: row['tp_buy_hammer'] if row['trade_type_hammer'] == 1 else None, axis=1)
                    df['sl_hammer'] = df.apply(lambda row: row['sl_buy_hammer'] if row['trade_type_hammer'] == 1 else None, axis=1)
                    
                    def determine_trade_result_hammer(row, df):
                        idx = row.name
                        if row['trade_type_hammer'] == 1:
                            for i in range(df.index.get_loc(idx) + 1, len(df)):
                                future_row = df.iloc[i]
                                if future_row['High'] >= row['tp_hammer']:
                                    return 2  # رابحة
                                elif future_row['Low'] <= row['sl_hammer']:
                                    return 1  # خاسرة
                        return 0
                    
                    df['trade_result_hammer'] = df.apply(lambda row: determine_trade_result_hammer(row, df), axis=1)
                    
                    # دالة التحقق من شمعة Shooting Star باستخدام XX و VV
                    def is_shooting_star(candle):
                        body = abs(candle['Open'] - candle['Close'])
                        upper_shadow = abs(candle['High'] - max(candle['Open'], candle['Close']))
                        lower_shadow = abs(candle['Low'] - min(candle['Open'], candle['Close']))
                        return (body * BB) , (lower_shadow * BB) , (upper_shadow * BB)
                    
                    # دالة حساب صفقة بيع باستخدام شمعة Shooting Star
                    def calculate_sell_trade_shooting_star(row, df, tp_multiplier=TP, sl_multiplier=SL):
                        idx = row.name
                        if idx == df.index[0]:
                            return 0, None, None
                        first_candle = df.loc[df.index[df.index.get_loc(idx) - 1]]
                        second_candle = row
                        if (is_large_body(first_candle) and 
                            first_candle['Open'] < first_candle['Close'] and 
                            is_shooting_star(second_candle) and 
                            second_candle['Open'] > second_candle['Close'] and 
                            (abs(first_candle['Open'] - first_candle['Close']) >= 0.75)):
                            
                            points = abs(first_candle['Open'] - first_candle['Close'])
                            stop_loss = second_candle['Close'] + (points * sl_multiplier)
                            take_profit = second_candle['Close'] - (tp_multiplier * points)
                            return 1, take_profit, stop_loss
                        return 0, None, None
                    
                    # تطبيق صفقة البيع
                    df['sell_trade_shooting'], df['tp_sell_shooting'], df['sl_sell_shooting'] = zip(*df.apply(lambda row: calculate_sell_trade_shooting_star(row, df), axis=1))
                    df['trade_type_shooting'] = df.apply(lambda row: -1 if row['sell_trade_shooting'] == 1 else 0, axis=1)
                    df['tp_shooting'] = df.apply(lambda row: row['tp_sell_shooting'] if row['trade_type_shooting'] == -1 else None, axis=1)
                    df['sl_shooting'] = df.apply(lambda row: row['sl_sell_shooting'] if row['trade_type_shooting'] == -1 else None, axis=1)
                    
                    def determine_trade_result_shooting(row, df):
                        idx = row.name
                        if row['trade_type_shooting'] == -1:
                            for i in range(df.index.get_loc(idx) + 1, len(df)):
                                future_row = df.iloc[i]
                                if future_row['Low'] <= row['tp_shooting']:
                                    return 2  # رابحة
                                elif future_row['High'] >= row['sl_shooting']:
                                    return 1  # خاسرة
                        return 0
                    
                    df['trade_result_shooting'] = df.apply(lambda row: determine_trade_result_shooting(row, df), axis=1)
                    
                    # حساب الرصيد النهائي بعد تطبيق نسب الربح/الخسارة والعمولة
                    initial_balance = 1000
                    final_balance = [initial_balance] * len(df)
                    
                    for i in range(1, len(df)):
                        previous_balance = final_balance[i - 1]
                        current_row = df.iloc[i]
                        current_balance = previous_balance
                        # نتائج صفقة الشراء
                        if current_row['trade_result_hammer'] == 2:
                            current_balance += current_balance * (TP/100 - (0.00406*SL))
                        elif current_row['trade_result_hammer'] == 1:
                            current_balance -= current_balance * (SL/100 + (0.00406*SL))
                        # نتائج صفقة البيع
                        if current_row['trade_result_shooting'] == 2:
                            current_balance += current_balance * (TP/100 - (0.00406*SL))
                        elif current_row['trade_result_shooting'] == 1:
                            current_balance -= current_balance * (SL/100 + (0.00406*SL))
                        final_balance[i] = current_balance
                    
                    df['final_balance'] = final_balance
                    
                    # إعادة ترتيب الأعمدة المطلوبة
                    df = df[['Gmt time', 'Open', 'High', 'Low', 'Close', 
                             'trade_type_hammer', 'trade_result_hammer', 'tp_hammer', 'sl_hammer',
                             'trade_type_shooting', 'trade_result_shooting', 'tp_shooting', 'sl_shooting',
                             'final_balance']]
                    
                    # استخراج السطر الأخير وقيمة final_balance له
                    last_row = df.iloc[-1]
                    fb = last_row['final_balance']
                    
                    # طباعة السطر الأخير مع رقم العملية والمعاملات المستخدمة
                    print(f"Operation {op_number}: BB = {BB}, XX = {XX}, VV = {VV}, TP = {TP}, SL = {SL}")
                    print("Last row:")
                    print(last_row)
                    print("-" * 80)
                    
                    # تخزين رقم العملية وقيمة final_balance في القائمة
                    operations.append((op_number, fb))
                    
                    # زيادة رقم العملية
                    op_number += 1

# الآن رسم بياني نقطي باستخدام البيانات المخزنة في operations
ops, balances = zip(*operations)  # يفصل رقم العملية والقيم

plt.figure(figsize=(12, 6))
plt.scatter(ops, balances, color='blue')
plt.xlabel('Operation Number')
plt.ylabel('Final Balance')
plt.title('Scatter Plot of Final Balance vs Operation Number')

# إضافة تعليقات لكل نقطة توضح رقم العملية
for op, bal in operations:
    plt.annotate(str(op), (op, bal), textcoords="offset points", xytext=(0,5), ha='center')

plt.show()
```
