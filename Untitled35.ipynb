```python
# حساب حجم الشمعة (Candle Range)
df['Candle_Range'] = df['High'] - df['Low']
```

```python
# حساب نوع الشمعة
def determine_candle_type(row):
    if row['Close'] > row['Open']:
        return 'Bullish'  # صعودية
    elif row['Close'] < row['Open']:
        return 'Bearish'  # هبوطية
    else:
        return 'Doji'  # متعادلة

df['Candle_Type'] = df.apply(determine_candle_type, axis=1)
```

```python
# حساب حجم الذيل العلوي والسفلي
df['Upper_Wick'] = df.apply(lambda row: row['High'] - max(row['Open'], row['Close']), axis=1)
df['Lower_Wick'] = df.apply(lambda row: min(row['Open'], row['Close']) - row['Low'], axis=1)
```

```python
#نسبة الذيل العلوي والسفلي إلى نطاق الشمعة (Upper/Lower Wick to Candle Range Ratio)
df['Candle_Range'] = df['High'] - df['Low']
df['Upper_Wick_Ratio'] = df['Upper_Wick'] / df['Candle_Range']
df['Lower_Wick_Ratio'] = df['Lower_Wick'] / df['Candle_Range']
```

```python
#حجم الشمعة مقارنة بحجم الذيل العلوي والسفلي (Candle Body to Wick Size Ratio)
df['Candle_Body'] = abs(df['Close'] - df['Open'])
df['Wick_Size'] = df['Upper_Wick'] + df['Lower_Wick']
df['Body_to_Wick_Ratio'] = df['Candle_Body'] / df['Wick_Size']
#النتيجة:
#نسبة أكبر من 1 تشير إلى أن الجسم أكبر من الذيل، مما يعكس اتجاه قوي.
#نسبة أصغر من 1 تشير إلى أن الذيل أكبر من الجسم، مما يعكس ترددًا في السوق أو عدم يقين.
```

```python
#تحديد مناطق الدعم والمقاومة باستخدام الذيل
df['Support_Zone'] = df['Low'] - df['Lower_Wick']
df['Resistance_Zone'] = df['High'] + df['Upper_Wick']
#مناطق الدعم: الذيل السفلي الطويل قد يشير إلى أن السعر لا يستطيع الهبوط بشكل أكبر عند نقطة معينة.
#مناطق المقاومة: الذيل العلوي الطويل قد يشير إلى أن السعر لا يستطيع الصعود بشكل أكبر عند نقطة معينة.
```

```python
#مؤشر القيم المعيارية (Z-score) للذيل العلوي والسفلي
upper_wick_mean = df['Upper_Wick'].mean()
upper_wick_std = df['Upper_Wick'].std()
df['Z_Score_Upper_Wick'] = (df['Upper_Wick'] - upper_wick_mean) / upper_wick_std

lower_wick_mean = df['Lower_Wick'].mean()
lower_wick_std = df['Lower_Wick'].std()
df['Z_Score_Lower_Wick'] = (df['Lower_Wick'] - lower_wick_mean) / lower_wick_std

#يمكنك حساب Z-Score لقياس الذيل العلوي والسفلي مقارنة بالمتوسط التاريخي. هذا يساعدك في تحديد ما إذا كانت الذيل تتجاوز المعدلات الطبيعية أو أنها تقع ضمن النطاق الطبيعي.
#النتيجة:
#قيمة Z كبيرة (> 2 أو < -2) تشير إلى أن الذيل العلوي أو السفلي قد يكون مفرطًا مقارنة بالمعدل الطبيعي، مما قد يعني حدوث حدث غير عادي أو تغيير في الاتجاه.
```

```python
#مؤشر التقلب المعياري (Volatility Index)
#يمكن حساب مؤشر التقلب باستخدام نطاق الشمعة (Candle Range) لفهم مدى تقلبات السوق.

df['Volatility_Index'] = ((df['High'] - df['Low']) / df['Close']) * 100

#النتيجة:
#كلما ارتفع مؤشر التقلب، زادت تقلبات السوق. هذه المعلومة مفيدة عند تحديد فرص البيع أو الشراء بناءً على مستويات التقلب.
```

```python
#حساب قوة الاتجاه باستخدام الذيل العلوي والسفلي

df['Bullish_Power'] = (df['Lower_Wick'] / df['Candle_Range']) * 100
df['Bearish_Power'] = (df['Upper_Wick'] / df['Candle_Range']) * 100
```

```python
# حساب الأعمدة الأساسية
df['Candle_Body'] = abs(df['Close'] - df['Open'])

# التعرف على شمعة المطرقة
df['Is_Hammer'] = (df['Lower_Wick'] > 2 * df['Candle_Body']) & (df['Upper_Wick'] < 0.5 * df['Candle_Body'])

# التعرف على شمعة دوجي
df['Is_Doji'] = (df['Candle_Body'] < 0.1 * (df['High'] - df['Low'])) & (df['Upper_Wick'] > 0.5 * df['Candle_Body']) & (df['Lower_Wick'] > 0.5 * df['Candle_Body'])

# التعرف على شمعة الابتلاع الصعودي
df['Is_Bullish_Engulfing'] = (df['Close'] > df['Open']) & (df['Open'].shift(1) > df['Close'].shift(1)) & (df['Close'] > df['Open'].shift(1)) & (df['Open'] < df['Close'].shift(1))

# التعرف على شمعة الابتلاع الهبوطي
df['Is_Bearish_Engulfing'] = (df['Close'] < df['Open']) & (df['Open'].shift(1) < df['Close'].shift(1)) & (df['Close'] < df['Open'].shift(1)) & (df['Open'] > df['Close'].shift(1))
```

```python
# تحويل أعمدة الأنماط إلى قيم 1 و 0
df['Is_Hammer'] = df['Is_Hammer'].astype(int)
df['Is_Doji'] = df['Is_Doji'].astype(int)
df['Is_Bullish_Engulfing'] = df['Is_Bullish_Engulfing'].astype(int)
df['Is_Bearish_Engulfing'] = df['Is_Bearish_Engulfing'].astype(int)
```

```python
# تحديد الهامر الشرائي والبيعي
df['Bullish_Hammer'] = ((df['Is_Hammer']) & (df['Close'] > df['Open'])).astype(int)
df['Bearish_Hammer'] = ((df['Is_Hammer']) & (df['Close'] < df['Open'])).astype(int)
```

```python

# تعريف دالة التحقق من شمعة الدوجي
def is_doji(candle):
    body = abs(candle['Open'] - candle['Close'])
    total_range = candle['High'] - candle['Low']
    return body < 0.1 * total_range

# حساب FVG 
def calculate_fvg_sell(row, df):
    idx = row.name
    if idx < 2:  # تأكد أن هناك ثلاث شمعات قبل حساب FVG
        return 0

    first_candle = df.iloc[idx - 2]
    third_candle = row

    # تحقق من FVG الشرائي
    if first_candle['Low'] > third_candle['High'] and is_doji(third_candle):
        return 1
    return 0

# حساب FVG 
def calculate_fvg_buy(row, df):
    idx = row.name
    if idx < 2:  # تأكد أن هناك ثلاث شمعات قبل حساب FVG
        return 0

    first_candle = df.iloc[idx - 2]
    third_candle = row

    # تحقق من FVG البيعي
    if first_candle['High'] < third_candle['Low'] and is_doji(third_candle):
        return 1
    return 0

# تطبيق العمودين الجديدين
df['fvg_doji_buy'] = df.apply(lambda row: calculate_fvg_buy(row, df), axis=1)
df['fvg_doji_sell'] = df.apply(lambda row: calculate_fvg_sell(row, df), axis=1)
```

```python
# تعريف دالة التحقق من شمعة المطرقة (Hammer)
def is_hammer(candle):
    body = abs(candle['Open'] - candle['Close'])
    lower_shadow = abs(candle['Low'] - min(candle['Open'], candle['Close']))
    upper_shadow = abs(candle['High'] - max(candle['Open'], candle['Close']))
    
    # الشرط الأول: الظل السفلي طويل والظل العلوي قصير أو معدوم
    condition_1 = lower_shadow > 2 * body and upper_shadow < body
    
    # الشرط الثاني: الظل العلوي طويل والظل السفلي قصير أو معدوم
    condition_2 = upper_shadow > 2 * body and lower_shadow < body

    # تحقق إذا كانت الشمعة تطابق أي من الشرطين
    return condition_1 or condition_2

# حساب FVG 
def calculate_fvg_buy(row, df):
    idx = row.name
    if idx < 2:  # تأكد أن هناك ثلاث شمعات قبل حساب FVG
        return 0

    first_candle = df.iloc[idx - 2]
    third_candle = row

    # تحقق من FVG البيعي مع شرط الهامر
    if first_candle['High'] < third_candle['Low'] and is_hammer(third_candle):
        return 1
    return 0

# حساب FVG 
def calculate_fvg_sell(row, df):
    idx = row.name
    if idx < 2:  # تأكد أن هناك ثلاث شمعات قبل حساب FVG
        return 0

    first_candle = df.iloc[idx - 2]
    third_candle = row

    # تحقق من FVG الشرائي مع شرط الهامر
    if first_candle['Low'] > third_candle['High'] and is_hammer(third_candle):
        return 1
    return 0

# تطبيق العمودين الجديدين
df['fvg_hammer_buy'] = df.apply(lambda row: calculate_fvg_buy(row, df), axis=1)
df['fvg_hammer_sell'] = df.apply(lambda row: calculate_fvg_sell(row, df), axis=1)

```

```python
# تعريف دالة التحقق من الابتلاع الشرائي (Bullish Engulfing)
def is_bullish_engulfing(candle3, candle4):
    # الشمعة الرابعة تبتلع الشمعة الثالثة مع شرط أن الشمعة الرابعة صعودية
    return candle4['Open'] < candle3['Close'] and candle4['Close'] > candle3['Open'] and candle4['Close'] > candle4['Open']

# تعريف دالة التحقق من الابتلاع البيعي (Bearish Engulfing)
def is_bearish_engulfing(candle3, candle4):
    # الشمعة الرابعة تبتلع الشمعة الثالثة مع شرط أن الشمعة الرابعة هبوطية
    return candle4['Open'] > candle3['Close'] and candle4['Close'] < candle3['Open'] and candle4['Close'] < candle4['Open']

# حساب FVG البيعي باستخدام الابتلاع الشرائي
def calculate_fvg_buy(row, df):
    idx = row.name
    if idx < 3:  # تأكد أن هناك ثلاث شمعات قبل حساب FVG
        return 0

    first_candle = df.iloc[idx - 3]
    second_candle = df.iloc[idx - 2]
    third_candle = df.iloc[idx - 1]
    fourth_candle = row

    if first_candle['Low'] > third_candle['High'] and is_bullish_engulfing(third_candle, fourth_candle):
        return 1
    return 0

# حساب FVG الشرائي باستخدام الابتلاع البيعي
def calculate_fvg_sell(row, df):
    idx = row.name
    if idx < 3:  # تأكد أن هناك ثلاث شمعات قبل حساب FVG
        return 0

    first_candle = df.iloc[idx - 3]
    second_candle = df.iloc[idx - 2]
    third_candle = df.iloc[idx - 1]
    fourth_candle = row

    if first_candle['High'] < third_candle['Low'] and is_bearish_engulfing(third_candle, fourth_candle):
        return 1
    return 0

# تطبيق العمودين الجديدين
df['fvg_bearish_sell'] = df.apply(lambda row: calculate_fvg_sell(row, df), axis=1)
df['fvg_bullish_buy'] = df.apply(lambda row: calculate_fvg_buy(row, df), axis=1)


```

```python
# حساب قوة FVG البيعي
def calculate_fvg_sell_strength(row, df):
    idx = row.name
    if idx < 3:  # تأكد أن هناك ثلاث شمعات قبل حساب FVG
        return 0

    first_candle = df.iloc[idx - 3]
    second_candle = df.iloc[idx - 2]
    third_candle = df.iloc[idx - 1]
    fourth_candle = row

    # حساب قوة FVG البيعي
    fvg_sell_strength = 0
    if first_candle['Low'] > third_candle['High']:
        # المسافة بين الشمعة الأولى والشمعة الثالثة
        fvg_sell_strength = first_candle['Low'] - third_candle['High']

    return fvg_sell_strength

# حساب قوة FVG الشرائي
def calculate_fvg_buy_strength(row, df):
    idx = row.name
    if idx < 3:  # تأكد أن هناك ثلاث شمعات قبل حساب FVG
        return 0

    first_candle = df.iloc[idx - 3]
    second_candle = df.iloc[idx - 2]
    third_candle = df.iloc[idx - 1]
    fourth_candle = row

    # حساب قوة FVG الشرائي
    fvg_buy_strength = 0
    if first_candle['High'] < third_candle['Low']:
        # المسافة بين الشمعة الأولى والشمعة الثالثة
        fvg_buy_strength = third_candle['Low'] - first_candle['High']

    return fvg_buy_strength

# تطبيق الأعمدة الجديدة لحساب قوة FVG البيعي والشرائي
df['fvg_sell_strength'] = df.apply(lambda row: calculate_fvg_sell_strength(row, df), axis=1)
df['fvg_buy_strength'] = df.apply(lambda row: calculate_fvg_buy_strength(row, df), axis=1)
```

```python
# حساب قرب FVG البيعي
def calculate_fvg_sell_proximity(row, df):
    idx = row.name
    if idx < 3:  # تأكد أن هناك ثلاث شمعات قبل حساب FVG
        return 0

    first_candle = df.iloc[idx - 3]
    second_candle = df.iloc[idx - 2]
    third_candle = df.iloc[idx - 1]
    fourth_candle = row

    # حساب المسافة بين الشمعة الحالية وحدود الفجوة البيعية
    proximity_sell = 0
    if first_candle['Low'] > third_candle['High']:
        proximity_sell = row['Close'] - third_candle['High']  # المسافة من السعر الحالي إلى الحد الأعلى للفجوة البيعية

    return proximity_sell

# حساب قرب FVG الشرائي
def calculate_fvg_buy_proximity(row, df):
    idx = row.name
    if idx < 3:  # تأكد أن هناك ثلاث شمعات قبل حساب FVG
        return 0

    first_candle = df.iloc[idx - 3]
    second_candle = df.iloc[idx - 2]
    third_candle = df.iloc[idx - 1]
    fourth_candle = row

    # حساب المسافة بين الشمعة الحالية وحدود الفجوة الشرائية
    proximity_buy = 0
    if first_candle['High'] < third_candle['Low']:
        proximity_buy = third_candle['Low'] - row['Close']  # المسافة من السعر الحالي إلى الحد الأدنى للفجوة الشرائية

    return proximity_buy

# تطبيق الأعمدة الجديدة لحساب قرب FVG البيعي والشرائي
df['fvg_sell_proximity'] = df.apply(lambda row: calculate_fvg_sell_proximity(row, df), axis=1)
df['fvg_buy_proximity'] = df.apply(lambda row: calculate_fvg_buy_proximity(row, df), axis=1)
```

```python
def calculate_fvg_buy(row, df):
    idx = row.name
    if idx < 2:  # تأكد من وجود ثلاث شمعات على الأقل
        return 0, None, None

    first_candle = df.iloc[idx - 2]  # الشمعة الأولى
    third_candle = row  # الشمعة الثالثة

    # تحقق من FVG الشرائي مع شرط الهامر
    if first_candle['High'] < third_candle['Low'] and is_hammer(third_candle):
        take_profit = first_candle['High']  # أعلى سعر للشمعة الأولى
        stop_loss = third_candle['Low']  # أقل سعر للشمعة الثالثة
        return 1, take_profit, stop_loss

    return 0, None, None


def calculate_fvg_sell(row, df):
    idx = row.name
    if idx < 2:  # تأكد من وجود ثلاث شمعات على الأقل
        return 0, None, None

    first_candle = df.iloc[idx - 2]  # الشمعة الأولى
    third_candle = row  # الشمعة الثالثة

    # تحقق من FVG البيعي مع شرط الهامر
    if first_candle['Low'] > third_candle['High'] and is_hammer(third_candle):
        take_profit = first_candle['Low']  # أقل سعر للشمعة الأولى
        stop_loss = third_candle['High']  # أعلى سعر للشمعة الثالثة
        return 1, take_profit, stop_loss

    return 0, None, None

# تطبيق الأعمدة الجديدة على DataFrame
df['fvg_hammer_buy'], df['tp_buy'], df['sl_buy'] = zip(*df.apply(lambda row: calculate_fvg_buy(row, df), axis=1))
df['fvg_hammer_sell'], df['tp_sell'], df['sl_sell'] = zip(*df.apply(lambda row: calculate_fvg_sell(row, df), axis=1))

# تحديد نوع الصفقة
def determine_trade_type(row):
    if row['fvg_hammer_buy'] == 1:
        return '1'
    elif row['fvg_hammer_sell'] == 1:
        return '-1'
    else:
        return '0'

df['trade_type'] = df.apply(determine_trade_type, axis=1)
```

```python
# تجهيز البيانات لتدريب نموذج LSTM لتصنيف نوع الصفقة
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout

# الميزات المستخدمة
features = [
    'Gmt time', 'Open', 'High', 'Low', 'Close', 'Change_Percentage', 'Change', 'Candle_Range',
    'Upper_Wick', 'Lower_Wick', 'Upper_Wick_Ratio', 'Lower_Wick_Ratio', 'Candle_Body', 'Wick_Size',
    'Body_to_Wick_Ratio', 'Support_Zone', 'Resistance_Zone', 'Z_Score_Upper_Wick', 'Z_Score_Lower_Wick',
    'Volatility_Index', 'Bullish_Power', 'Bearish_Power', 'Is_Hammer', 'Is_Doji', 'Is_Bullish_Engulfing',
    'Is_Bearish_Engulfing', 'Bullish_Hammer', 'Bearish_Hammer', 'Bullish', 'Bearish', 'fvg_doji_buy',
    'fvg_doji_sell', 'fvg_hammer_buy', 'fvg_hammer_sell', 'fvg_sell_strength', 'fvg_buy_strength',
    'fvg_sell_proximity', 'fvg_buy_proximity'
]

# تجهيز البيانات
X = df[features].values

def preprocess_labels(df, label_col):
    # تطبيع الميزات
    scaler = MinMaxScaler()
    X_scaled = scaler.fit_transform(X)

    # تجهيز التسميات
    label_encoder = LabelEncoder()
    y = label_encoder.fit_transform(df[label_col])

    # تحويل البيانات إلى تسلسلات
    sequence_length = 10  # عدد الشموع لكل تسلسل
    X_sequences, y_sequences = [], []

    for i in range(len(X_scaled) - sequence_length):
        X_sequences.append(X_scaled[i:i + sequence_length])
        y_sequences.append(y[i + sequence_length - 1])

    return np.array(X_sequences), np.array(y_sequences), label_encoder, scaler

X_sequences, y_sequences, label_encoder, scaler = preprocess_labels(df, 'trade_type')

# تقسيم البيانات
X_train, X_test, y_train, y_test = train_test_split(X_sequences, y_sequences, test_size=0.2, random_state=42)

# بناء النموذج
model = Sequential([
    LSTM(50, return_sequences=True, input_shape=(X_sequences.shape[1], X_sequences.shape[2])),
    Dropout(0.2),
    LSTM(50, return_sequences=False),
    Dropout(0.2),
    Dense(25, activation='relu'),
    Dense(len(label_encoder.classes_), activation='softmax')  # عدد الأنواع المختلفة من الصفقات
])

# تجميع النموذج
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# تدريب النموذج
history = model.fit(X_train, y_train, epochs=20, batch_size=32, validation_data=(X_test, y_test))

# تقييم النموذج
loss, accuracy = model.evaluate(X_test, y_test)
print(f"Test Loss: {loss:.4f}, Test Accuracy: {accuracy:.4f}")

# التنبؤ
predictions = model.predict(X_test)
predicted_labels = np.argmax(predictions, axis=1)
predicted_trade_types = label_encoder.inverse_transform(predicted_labels)

# طباعة أول 10 نتائج
print("Predicted trade types:", predicted_trade_types[:10])
```

```python
import numpy as np
import pandas as pd
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler

# افتراض أن البيانات تم تحميلها في DataFrame (df)
# يتم استخدام الأعمدة لتحديد X (الخصائص) و y (التصنيفات)

# إعداد البيانات (إزالة العمود غير المرغوب فيه "Gmt time" إذا كان ليس جزءًا من الميزات)
X = df.drop(columns=["Gmt time", "trade_type",'tp_buy','sl_buy','tp_sell','sl_sell'])  # حذف العمود الزمني والعمود المستهدف
y = df["trade_type"]  # العمود المستهدف

# ترميز التصنيفات (None, Buy, Sell) إلى قيم عددية
label_encoder = LabelEncoder()
y = label_encoder.fit_transform(y)  # تحويل None, Buy, Sell إلى أرقام

# تطبيع البيانات (Normalization) للميزات
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# تقسيم البيانات إلى تدريب واختبار
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# إعادة تشكيل البيانات لتناسب تنسيق LSTM [samples, time_steps, features]
X_train = X_train.reshape(X_train.shape[0], 1, X_train.shape[1])  # 1 time step
X_test = X_test.reshape(X_test.shape[0], 1, X_test.shape[1])  # 1 time step

# بناء نموذج LSTM
model = Sequential()

# إضافة طبقة LSTM
model.add(LSTM(units=50, return_sequences=False, input_shape=(X_train.shape[1], X_train.shape[2])))

# إضافة طبقة Dropout لتقليل الإفراط في التكيف
model.add(Dropout(0.2))

# إضافة طبقة Fully Connected
model.add(Dense(units=3, activation='softmax'))  # ثلاث فئات للإخراج (None, Buy, Sell)

# تجميع النموذج
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# تدريب النموذج
model.fit(X_train, y_train, epochs=10, batch_size=32, validation_data=(X_test, y_test))

# تقييم النموذج
loss, accuracy = model.evaluate(X_test, y_test)
print(f"Accuracy: {accuracy * 100:.2f}%")
```
