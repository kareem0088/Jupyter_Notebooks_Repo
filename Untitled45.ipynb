```python
import pandas as pd

# تحويل العمود الزمني إلى نوع datetime وتعيينه كمؤشر
df['Gmt time'] = pd.to_datetime(df['Gmt time'])
df.set_index('Gmt time', inplace=True)

# تحويل البيانات من 1 دقيقة إلى 5 دقائق
df_5m = df.resample('5T').agg({
    'Open': 'first',   # أول سعر افتتاح في الفاصل الزمني
    'High': 'max',     # أعلى سعر في الفاصل الزمني
    'Low': 'min',      # أقل سعر في الفاصل الزمني
    'Close': 'last'    # آخر سعر إغلاق في الفاصل الزمني
})

# عرض النتيجة
print(df_5m)
```

```python
df_5m=df_5m.reset_index(drop=False) 
```

```python
import pandas as pd

df = df_5m

# دالة التحقق من الشمعة الكبيرة (كما هي)
def is_large_body(candle):
    body = abs(candle['Open'] - candle['Close'])
    lower_shadow = abs(candle['Low'] - min(candle['Open'], candle['Close']))
    upper_shadow = abs(candle['High'] - max(candle['Open'], candle['Close']))
    return body > (lower_shadow + upper_shadow)

# دالة التحقق من شمعة الدوجي مع معالجة القسمة على الصفر
def is_doji(candle, doji_threshold=0.1):
    body_size = abs(candle['Open'] - candle['Close'])
    total_range = candle['High'] - candle['Low']
    
    # تجنب القسمة على الصفر
    if total_range == 0:
        return False
    
    return (body_size / total_range) < doji_threshold

# دالة حساب صفقة البيع باستخدام الدوجي (كما هي)
def calculate_sell_trade_doji(row, df, tp_multiplier=1.5, sl_multiplier=1):
    idx = row.name
    if idx == df.index[0]:
        return 0, None, None
    
    first_candle = df.loc[df.index[df.index.get_loc(idx) - 1]]
    second_candle = row
    
    # الشرط: شمعة صاعدة كبيرة تليها دوجي
    if is_large_body(first_candle) and first_candle['Close'] > first_candle['Open'] and is_doji(second_candle):
        entry_price = second_candle['Close']
        points = abs(first_candle['High'] - first_candle['Low'])  # نطاق الشمعة
        take_profit = entry_price - (points * tp_multiplier)
        stop_loss = entry_price + (points * sl_multiplier)
        return 1, take_profit, stop_loss
    
    return 0, None, None

# تطبيق الأعمدة الجديدة
df['sell_trade_doji'], df['tp_sell_doji'], df['sl_sell_doji'] = zip(*df.apply(
    lambda row: calculate_sell_trade_doji(row, df), axis=1))

# تحديد نوع الصفقة
df['trade_type_doji'] = df.apply(lambda row: -1 if row['sell_trade_doji'] == 1 else 0, axis=1)

# دمج أعمدة TP و SL
df['tp_doji'] = df.apply(lambda row: row['tp_sell_doji'] if row['trade_type_doji'] == -1 else None, axis=1)
df['sl_doji'] = df.apply(lambda row: row['sl_sell_doji'] if row['trade_type_doji'] == -1 else None, axis=1)

# دالة تحديد نتيجة الصفقة البيعية
def determine_trade_result_doji(row, df):
    idx = row.name
    if row['trade_type_doji'] == -1:
        for i in range(df.index.get_loc(idx) + 1, len(df)):
            future_row = df.iloc[i]
            if future_row['Low'] <= row['tp_doji']:
                return 2  # رابحة
            elif future_row['High'] >= row['sl_doji']:
                return 1  # خاسرة
    return 0  # لا نتيجة

# تطبيق النتائج
df['trade_result_doji'] = df.apply(lambda row: determine_trade_result_doji(row, df), axis=1)

# عرض النتائج
print(df[['Gmt time','Open', 'High', 'Low', 'Close', 
          'trade_type_doji', 'trade_result_doji', 
          'tp_doji', 'sl_doji']])
```
