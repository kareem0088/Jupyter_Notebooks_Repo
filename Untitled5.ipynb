```python
import pandas as pd

# المسار الخاص بالملف
file_path = r"C:\Users\Access\Downloads\EURUSD_Ticks.csv"

# قراءة البيانات من الملف
df = pd.read_csv(file_path)
```

```python
df[:]
```

```python
drrr = df 
drrr['Open'] = drrr['Ask']  # أو استخدم سعر الـ ask في بداية الفترة
drrr['Close'] = (drrr['Ask'] + drrr['Bid']) / 2  # أو استخدم df['ask'] أو df['bid'] حسب الحاجة
drrr['High'] = drrr[['Ask', 'Bid']].max(axis=1)  # أعلى سعر
drrr['Low'] = drrr[['Ask', 'Bid']].min(axis=1)    # أقل سعر
```
