```python
from tkinter import *
from tkinter import filedialog, simpledialog
from PIL import Image, ImageTk

window = Tk()
window.title("Simple Image Processing")
window.geometry("800x600")

img = None  
img_display = None

def open_image():
    global img, img_display
    filepath = filedialog.askopenfilename(filetypes=[("Image files", "*.jpg;*.jpeg;*.png;*.bmp")])
    if filepath:
        img = Image.open(filepath)
        img_display = ImageTk.PhotoImage(img.resize((400, 400)))
        canvas.create_image(0, 0, anchor=NW, image=img_display)

def show_channel(channel):
    global img, img_display  
    if img is not None:  
        if img.mode != 'RGB':
            print("Image is not in RGB mode.")
            return
        
        r, g, b = img.split()  
        if channel == 'R':
            img_channel = Image.merge("RGB", (r, Image.new("L", r.size), Image.new("L", r.size)))
        elif channel == 'G':
            img_channel = Image.merge("RGB", (Image.new("L", g.size), g, Image.new("L", g.size)))
        elif channel == 'B':
            img_channel = Image.merge("RGB", (Image.new("L", b.size), Image.new("L", b.size), b))
        
        img_display = ImageTk.PhotoImage(img_channel.resize((400, 400))) 
        canvas.create_image(0, 0, anchor=NW, image=img_display)
    else:
        print("No image loaded. Please open an image first.")

def convert_to_gray():
    global img, img_display  
    if img is not None:  
        img_gray = img.convert("L")  
        img_display = ImageTk.PhotoImage(img_gray.resize((400, 400)))
        canvas.create_image(0, 0, anchor=NW, image=img_display)
    else:
        print("No image loaded. Please open an image first.")

def add_constant():
    global img, img_display
    if img is not None:
        constant = simpledialog.askinteger("Input", "Enter a constant to add (0-255):", minvalue=0, maxvalue=255)
        if constant is not None:
            img_with_constant = img.point(lambda p: min(p + constant, 255))  
            img_display = ImageTk.PhotoImage(img_with_constant.resize((400, 400)))
            canvas.create_image(0, 0, anchor=NW, image=img_display)

def subtract_constant():
    global img, img_display
    if img is not None:
        constant = simpledialog.askinteger("Input", "Enter a constant to subtract (0-255):", minvalue=0, maxvalue=255)
        if constant is not None:
            img_with_constant = img.point(lambda p: max(p - constant, 0))  
            img_display = ImageTk.PhotoImage(img_with_constant.resize((400, 400)))
            canvas.create_image(0, 0, anchor=NW, image=img_display)

def multiply_constant():
    global img, img_display
    if img is not None:
        constant = simpledialog.askfloat("Input", "Enter a constant to multiply (e.g., 0.5 for half brightness):")
        if constant is not None:
            img_with_constant = img.point(lambda p: min(int(p * constant), 255))  
            img_display = ImageTk.PhotoImage(img_with_constant.resize((400, 400)))
            canvas.create_image(0, 0, anchor=NW, image=img_display)

def complement():
    global img, img_display
    if img is not None:
        img_complement = img.point(lambda p: 255 - p) 
        img_display = ImageTk.PhotoImage(img_complement.resize((400, 400)))
        canvas.create_image(0, 0, anchor=NW, image=img_display)
    else:
        print("No image loaded. Please open an image first.")

btn_open = Button(window, text="Open Image", command=open_image)
btn_open.pack()

btn_red = Button(window, text="Show R Channel", command=lambda: show_channel('R'))
btn_red.pack()

btn_green = Button(window, text="Show G Channel", command=lambda: show_channel('G'))
btn_green.pack()

btn_blue = Button(window, text="Show B Channel", command=lambda: show_channel('B'))
btn_blue.pack()

btn_gray = Button(window, text="Convert to Grayscale", command=convert_to_gray)
btn_gray.pack()

btn_add_constant = Button(window, text="Add Constant", command=add_constant)
btn_add_constant.pack()

btn_subtract_constant = Button(window, text="Subtract Constant", command=subtract_constant)
btn_subtract_constant.pack()

btn_multiply_constant = Button(window, text="Multiply Constant", command=multiply_constant)
btn_multiply_constant.pack()

btn_complement = Button(window, text="Complement", command=complement)
btn_complement.pack()

canvas = Canvas(window, width=400, height=400)
canvas.pack()

window.mainloop()
```
