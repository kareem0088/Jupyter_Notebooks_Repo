```python
import pandas 
dd = pandas.read_csv(r"C:\Users\Access\Downloads\EURUSD-2019\EURUSD_Candlestick_1_M_BID_2019-1-1_2024-11-17.csv")
import pandas as pd

# افترض أن الداتا فريم الخاص بك يسمى df
# لاختيار الأعمدة المطلوبة فقط
df = dd[['Gmt time', 'Open', 'High', 'Low', 'Close']]
```

```python
# تحويل العمود إلى نوع datetime
df['Gmt time'] = pd.to_datetime(df['Gmt time'])

# استخراج الشهور والأيام والساعات والدقائق والثواني
df['Month'] = df['Gmt time'].dt.month
df['Day'] = df['Gmt time'].dt.day
df['Hour'] = df['Gmt time'].dt.hour
df['Minute'] = df['Gmt time'].dt.minute
df['Second'] = df['Gmt time'].dt.second

# عرض النتيجة
print(df)
```
