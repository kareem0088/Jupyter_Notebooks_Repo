```python
import pandas as pd

# تعريف دالة التحقق من الشمعة الكبيرة مقارنة بالظلال
def is_large_body(candle):
    body = abs(candle['Open'] - candle['Close'])
    lower_shadow = abs(candle['Low'] - min(candle['Open'], candle['Close']))
    upper_shadow = abs(candle['High'] - max(candle['Open'], candle['Close']))
    return body > (lower_shadow + upper_shadow)

# تعريف دالة التحقق من شمعة الهامر أو الدوجي
def is_hammer_or_doji(candle):
    body = abs(candle['Open'] - candle['Close'])
    lower_shadow = abs(candle['Low'] - min(candle['Open'], candle['Close']))
    upper_shadow = abs(candle['High'] - max(candle['Open'], candle['Close']))
    is_hammer = lower_shadow > 2 * body and upper_shadow < body
    is_doji = body <= (0.1 * (candle['High'] - candle['Low']))
    return is_hammer or is_doji

# دالة حساب صفقة شراء
def calculate_buy_trade(row, df, tp_multiplier=2, sl_multiplier=1):
    idx = row.name
    if idx < 1:
        return 0, None, None
    first_candle = df.iloc[idx - 1]
    second_candle = row
    if is_large_body(first_candle) and first_candle['Close'] < first_candle['Open'] and is_hammer_or_doji(second_candle):
        points = abs(second_candle['Close'] - second_candle['Low'])
        take_profit = second_candle['Close'] + (points * tp_multiplier)
        stop_loss = second_candle['Low'] - (points * sl_multiplier)
        return 1, take_profit, stop_loss
    return 0, None, None

# دالة حساب صفقة بيع
def calculate_sell_trade(row, df, tp_multiplier=2, sl_multiplier=1):
    idx = row.name
    if idx < 1:
        return 0, None, None
    first_candle = df.iloc[idx - 1]
    second_candle = row
    if is_large_body(first_candle) and first_candle['Close'] > first_candle['Open'] and is_hammer_or_doji(second_candle):
        points = abs(second_candle['High'] - second_candle['Close'])
        take_profit = second_candle['Close'] - (points * tp_multiplier)
        stop_loss = second_candle['High'] + (points * sl_multiplier)
        return 1, take_profit, stop_loss
    return 0, None, None

# تطبيق الأعمدة الجديدة على DataFrame الخاص بك
df['buy_trade'], df['tp_buy'], df['sl_buy'] = zip(*df.apply(lambda row: calculate_buy_trade(row, df), axis=1))
df['sell_trade'], df['tp_sell'], df['sl_sell'] = zip(*df.apply(lambda row: calculate_sell_trade(row, df), axis=1))

# تحديد نوع الصفقة
def determine_trade_type(row):
    if row['buy_trade'] == 1:
        return 1
    elif row['sell_trade'] == 1:
        return -1
    else:
        return 0

df['trade_type'] = df.apply(determine_trade_type, axis=1)

# دالة لتحديد حالة الصفقة (رابحة 1، خاسرة -1، أو لا توجد صفقة 0)
def determine_trade_result(row, df):
    idx = row.name
    if row['trade_type'] == 1:  # صفقة شراء
        for i in range(idx + 1, len(df)):
            future_row = df.iloc[i]
            if future_row['High'] >= row['tp_buy']:
                return 1  # رابحة
            elif future_row['Low'] <= row['sl_buy']:
                return -1  # خاسرة
    elif row['trade_type'] == -1:  # صفقة بيع
        for i in range(idx + 1, len(df)):
            future_row = df.iloc[i]
            if future_row['Low'] <= row['tp_sell']:
                return 1  # رابحة
            elif future_row['High'] >= row['sl_sell']:
                return -1  # خاسرة
    return 0  # لا توجد نتيجة

# تطبيق دالة تحديد حالة الصفقة
df['trade_result'] = df.apply(lambda row: determine_trade_result(row, df), axis=1)

# عرض الأعمدة المطلوبة
print(df[['Open', 'High', 'Low', 'Close', 'trade_type', 'trade_result', 'tp_buy', 'sl_buy', 'tp_sell', 'sl_sell']])
```
