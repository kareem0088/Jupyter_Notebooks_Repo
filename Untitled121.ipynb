```python
import pandas as pd
import numpy as np

# Relative Volume (RVOL)
df['RVOL'] = df['volume'] / df['volume'].rolling(window=20).mean()

# Volume Rate of Change (VROC)
df['VROC'] = (df['volume'] - df['volume'].shift(5)) / df['volume'].shift(5)

# Cumulative Delta Volume (لو عندك بيانات ask_volume و bid_volume)
# df['cum_delta_vol'] = (df['ask_volume'] - df['bid_volume']).cumsum()
```

```python
df
```

```python
# Volume Weighted Spread
df['spread_vol'] = (df['high'] - df['low']) * df['RVOL']

# VWAP Deviation (بالانحراف المعياري)
vwap_std = df['vwap'].rolling(window=20).std()
df['vwap_dev'] = (df['close'] - df['vwap']) / vwap_std

# Volume per Transaction
df['vol_per_tx'] = df['volume'] / df['transactions']
```

```python
df
```

```python
# Volume Pressure Index (لو عندك بيانات اتجاه الصفقات)
green_vol = df.loc[df['close'] > df['open'], 'volume']
red_vol   = df.loc[df['close'] < df['open'], 'volume']
df['VPI'] = (green_vol.sum() - red_vol.sum()) / (green_vol.sum() + red_vol.sum())

# Accumulation/Distribution based on VWAP
df['accum_dist_vwap'] = np.where(df['close'] > df['vwap'], df['volume'], -df['volume'])
df['accum_dist_vwap'] = df['accum_dist_vwap'].cumsum()
```

```python
df
```

```python
# Volume Spike Detector → 1 أو 0
df['vol_spike'] = (df['volume'] > (df['volume'].rolling(window=10).mean() * 2)).astype(int)

# Volume Climax Bar → 1 أو 0
max_vol = df['volume'].rolling(window=20).max()
wide_range = (df['high'] - df['low']) > (df['high'] - df['low']).rolling(window=20).mean()
df['climax_bar'] = ((df['volume'] == max_vol) & wide_range).astype(int)
```

```python
import pandas as pd
import numpy as np

# Micro Trend Slope (5 شموع)
df['micro_trend_slope'] = (df['close'] - df['close'].shift(5)) / \
                          (df['high'].rolling(5).max() - df['low'].rolling(5).min())

# Price Change per Second (لو عندك بيانات وقت بالثواني)
# تحويل timestamp إلى datetime
df['timestamp'] = pd.to_datetime(df['timestamp'])
df['seconds_diff'] = (df['timestamp'] - df['timestamp'].shift(1)).dt.total_seconds()
df['price_change_per_sec'] = (df['close'] - df['close'].shift(1)) / df['seconds_diff']
```

```python
# Position in Range (آخر 20 شمعة)
window_range = 20
low_min = df['low'].rolling(window_range).min()
high_max = df['high'].rolling(window_range).max()
df['pos_in_range'] = (df['close'] - low_min) / (high_max - low_min)

# Micro Support/Resistance Distance
df['dist_to_recent_high'] = df['high'].rolling(window_range).max() - df['close']
df['dist_to_recent_low'] = df['close'] - df['low'].rolling(window_range).min()

# Candle Body-to-Range Ratio
df['body_to_range'] = abs(df['close'] - df['open']) / (df['high'] - df['low'])
```

```python
# Transaction Density
df['tx_density'] = df['transactions'] / df['volume']

# VWAP Pullback Strength
df['vwap_pullback'] = (df['vwap'] - df['close'].shift(1)) / df['vwap']

# Volume Shock + Price Reaction
vol_avg = df['volume'].rolling(20).mean()
df['vol_shock'] = (df['volume'] > (vol_avg * 2)).astype(int)
df['price_reaction_next'] = df['close'].shift(-1) - df['close']
```

```python
# استخراج التاريخ من التايم ستامب
df['date'] = df['timestamp'].dt.date

# أعلى سعر لليوم الحالي + حجم الشمعة عنده
high_today_info = df.loc[df.groupby('date')['high'].idxmax(), ['date', 'high', 'volume']]
high_today_info = high_today_info.rename(columns={'high': 'high_today', 'volume': 'vol_at_high_today'})

# أقل سعر لليوم الحالي + حجم الشمعة عنده
low_today_info = df.loc[df.groupby('date')['low'].idxmin(), ['date', 'low', 'volume']]
low_today_info = low_today_info.rename(columns={'low': 'low_today', 'volume': 'vol_at_low_today'})

# دمج الأعمدة مع الجدول الأصلي
df = df.merge(high_today_info, on='date', how='left')
df = df.merge(low_today_info, on='date', how='left')

# الحصول على بيانات اليوم السابق
df['prev_date'] = df['date'].shift(1)
df['high_prev_day'] = df.groupby('date')['high_today'].shift(1)
df['vol_at_high_prev_day'] = df.groupby('date')['vol_at_high_today'].shift(1)
df['low_prev_day'] = df.groupby('date')['low_today'].shift(1)
df['vol_at_low_prev_day'] = df.groupby('date')['vol_at_low_today'].shift(1)
```
