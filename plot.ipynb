```python
import matplotlib.pyplot as plt
import numpy as np

# 
trading_days = range(1, 251)  # 250  
base_balance = [100 + (i * (30 / 249)) for i in trading_days]  #    100  125   250 

#      
fluctuations = np.random.uniform(-1, 1, size=250)  #    -1  1
account_balance = base_balance + fluctuations  #     

#   
plt.plot(trading_days, account_balance, label='Account Balance', color='blue')
plt.xlabel('Trading Days')
plt.ylabel('Account Balance ($)')
plt.title('Trading Simulation Results with Fluctuations')

#   (Legend)    
legend = plt.legend(loc='upper left')

#     
plt.text(-3, 128, ' Win Rate: 45%\n Reward/Risk: 1:1.5\n Risk per Trade: 1.0%\n Number of Trades: 1 in day\n Risk Method: Starting Balance.', 
         fontsize=9, color='black', ha='left', va='top',
         bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=0.4'))

#    
plt.text(125, 110, '  KARIM MOHAMED \n+20 1101407933', 
         fontsize=30, color='gray', ha='center', va='center', alpha=0.2)

plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 
trading_days = range(1, 251)  # 250  
base_balance = [100 + (i * (3600 / 249)) for i in trading_days]  #    100  3700   250 

#      
fluctuations = np.random.uniform(-50, 50, size=250)  #    -3  3
account_balance = base_balance + fluctuations  #     

#      
account_balance_curved = 100 * np.exp(np.log(3700 / 100) * (np.array(trading_days) / 250))

#   
plt.plot(trading_days, account_balance_curved + fluctuations, label='Account Balance', color='blue')
plt.xlabel('Trading Days')
plt.ylabel('Account Balance ($)')
plt.title('Trading Simulation Results with Curved Fluctuations')

#   (Legend)    
legend = plt.legend(loc='upper left')

#     
plt.text(-3, 3400, ' Win Rate: 55%\n Reward/Risk: 1:1.5\n Risk per Trade: 2.0%\n Number of Trades: 2 in day\n Risk Method: dynamic.', 
         fontsize=9, color='black', ha='left', va='top',
         bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=0.4'))

#    
plt.text(125, 1700, '  KARIM MOHAMED \n+20 1101407933', 
         fontsize=30, color='gray', ha='center', va='center', alpha=0.2)

plt.grid(True)
plt.show()
```
