```python
x = [1, 2, 3, 4, 5]
y1 = [1, 4, 9, 16, 25]
y2 = [25, 20, 15, 10, 5]

plt.plot(x, y1, label="y = x^2")    # منحنى أول
plt.plot(x, y2, label="y decreasing")  # منحنى ثاني
plt.legend()  # اظهار أسماء المنحنيات
plt.show()
```
