```python
# -*- coding: utf-8 -*-
import requests
import pandas as pd
from datetime import datetime, timedelta
import time
import os
import psutil
import subprocess
import tzlocal
import MetaTrader5 as mt5

# ===================== Settings =====================
NEWS_URLS = [
    "https://nfs.faireconomy.media/ff_calendar_thisweek.json",
    "https://cdn-nfs.faireconomy.media/ff_calendar_thisweek.json",
]

# Choose impact levels to USE across the app:
# Options: ["high"], ["medium"], or ["high", "medium"]
IMPACT_LEVELS = ["high", "medium"]

# Currency filter (e.g., ["USD","EUR"]) or [] for all
CURRENCY_FILTER = ["USD", "EUR"]

# Minutes to block before/after the event time
BLOCK_BEFORE = 10
BLOCK_AFTER = 10

# Refresh news every N seconds (1 hour)
NEWS_REFRESH_INTERVAL = 3600

# How often to refresh the screen/app logic (seconds)
TICK_INTERVAL = 1

# Actions during block window
CLOSE_POSITIONS_DURING_BLOCK = True  # close all open positions once per event window
CLOSE_TERMINAL_DURING_BLOCK = True  # optional: also terminate the terminal app during block

# Terminal app (only used if CLOSE_TERMINAL_DURING_BLOCK=True)
TERMINAL_PATH = r"C:\Program Files\MetaTrader 5\terminal64.exe"
TERMINAL_PROCESS_NAME = "terminal64.exe"

# MT5 connection params (usually initialize() without login is enough if terminal is logged in)
MT5_LOGIN = None       # e.g., 12345678
MT5_PASSWORD = None    # e.g., "mypassword"
MT5_SERVER = None      # e.g., "Broker-Server"

# ===================== Time zone =====================
LOCAL_TZ = tzlocal.get_localzone()

# ===================== News helpers =====================
def fetch_news_df() -> pd.DataFrame:
    """Fetch news JSON from primary or fallback URL and return DataFrame."""
    for url in NEWS_URLS:
        try:
            r = requests.get(url, timeout=12)
            if r.status_code == 200 and r.text.strip():
                data = r.json()
                df = pd.DataFrame(data)
                return df
        except Exception:
            continue
    return pd.DataFrame()

def normalize_and_filter_today(df_raw: pd.DataFrame) -> pd.DataFrame:
    """Convert date to local tz, filter by today, impact levels, currency filter; sort by time."""
    if df_raw.empty:
        return pd.DataFrame()

    # Ensure required columns exist
    for col in ["date", "impact", "title", "country", "forecast", "previous"]:
        if col not in df_raw.columns:
            df_raw[col] = ""

    # Convert date to localized datetime
    try:
        df_raw["date"] = pd.to_datetime(df_raw["date"], utc=True, errors="coerce").dt.tz_convert(LOCAL_TZ)
    except Exception:
        # Fallback: try from ISO strings with Z replacement
        df_raw["date"] = pd.to_datetime(df_raw["date"], errors="coerce")
        df_raw["date"] = df_raw["date"].dt.tz_localize("UTC").dt.tz_convert(LOCAL_TZ)

    df = df_raw.dropna(subset=["date"]).copy()

    # Filter today (local)
    today = datetime.now(LOCAL_TZ).date()
    df = df[df["date"].dt.date == today]

    # Impact filter
    df["impact_lower"] = df["impact"].astype(str).str.lower()
    df = df[df["impact_lower"].isin([lvl.lower() for lvl in IMPACT_LEVELS])]

    # Currency filter
    if CURRENCY_FILTER:
        df = df[df["country"].astype(str).isin(CURRENCY_FILTER)]

    # Sort by time
    df = df.sort_values("date").reset_index(drop=True)
    return df

def compute_status(now_local: datetime, event_time_local: datetime) -> str:
    """Return Upcoming / Now / Passed based on BLOCK_BEFORE/BLOCK_AFTER."""
    start = event_time_local - timedelta(minutes=BLOCK_BEFORE)
    end = event_time_local + timedelta(minutes=BLOCK_AFTER)
    if now_local < start:
        return "Upcoming"
    elif start <= now_local <= end:
        return "Now"
    else:
        return "Passed"

def build_block_windows(df_today: pd.DataFrame):
    """Build block windows list from today's events that match IMPACT_LEVELS and currency filter."""
    windows = []
    for _, row in df_today.iterrows():
        event_time = row["date"]
        start = event_time - timedelta(minutes=BLOCK_BEFORE)
        end = event_time + timedelta(minutes=BLOCK_AFTER)
        windows.append((start, end, row["title"], row["country"], str(row["impact"])))
    return windows

def is_now_in_block_window(windows):
    """Check if current time falls within any window; return tuple (bool, info...)."""
    now = datetime.now(LOCAL_TZ)
    for start, end, title, currency, impact in windows:
        if start <= now <= end:
            return True, title, currency, impact, start, end
    return False, None, None, None, None, None

# ===================== Terminal app helpers (optional) =====================
def is_terminal_running() -> bool:
    for proc in psutil.process_iter(['name']):
        try:
            if proc.info.get('name') and TERMINAL_PROCESS_NAME.lower() in proc.info['name'].lower():
                return True
        except (psutil.NoSuchProcess, psutil.AccessDenied):
            continue
    return False

def close_terminal():
    closed = False
    for proc in psutil.process_iter(['name']):
        try:
            if proc.info.get('name') and TERMINAL_PROCESS_NAME.lower() in proc.info['name'].lower():
                proc.terminate()
                closed = True
        except (psutil.NoSuchProcess, psutil.AccessDenied):
            continue
    if closed:
        print("Terminal closed")

def open_terminal():
    if os.path.exists(TERMINAL_PATH):
        subprocess.Popen(TERMINAL_PATH)
        print("Terminal opened")
    else:
        print("Path not found:", TERMINAL_PATH)

# ===================== MT5 helpers =====================
def mt5_connect() -> bool:
    """Initialize MT5 connection (with optional login)."""
    # If already initialized, shutdown to cleanly re-init
    try:
        mt5.shutdown()
    except Exception:
        pass

    if not mt5.initialize():
        print("MT5 initialize failed")
        return False

    # If login details are provided, try explicit login
    if MT5_LOGIN and MT5_PASSWORD and MT5_SERVER:
        if not mt5.login(login=MT5_LOGIN, password=MT5_PASSWORD, server=MT5_SERVER):
            print("MT5 login failed")
            return False

    return True

def mt5_account_snapshot():
    """Return (balance, equity, margin, margin_free, profit) or None if not available."""
    info = mt5.account_info()
    if info is None:
        return None
    return (info.balance, info.equity, info.margin, info.margin_free, info.profit)

def mt5_close_all_positions():
    """Close all open positions via market orders."""
    positions = mt5.positions_get()
    if positions is None or len(positions) == 0:
        print("No open positions to close")
        return

    for pos in positions:
        typ = pos.type  # 0=Buy, 1=Sell
        symbol = pos.symbol
        volume = pos.volume
        ticket = pos.ticket

        # market price to hit
        tick = mt5.symbol_info_tick(symbol)
        if tick is None:
            print(f"Skipping {symbol}: no tick")
            continue

        if typ == mt5.POSITION_TYPE_BUY:
            order_type = mt5.ORDER_TYPE_SELL
            price = tick.bid
        else:
            order_type = mt5.ORDER_TYPE_BUY
            price = tick.ask

        request = {
            "action": mt5.TRADE_ACTION_DEAL,
            "symbol": symbol,
            "volume": volume,
            "type": order_type,
            "position": ticket,
            "price": price,
            "deviation": 20,
            "magic": 20250825,
            "comment": "NewsGuardian close",
            "type_time": mt5.ORDER_TIME_GTC,
            "type_filling": mt5.ORDER_FILLING_FOK,
        }
        result = mt5.order_send(request)
        if result is None:
            print(f"Close failed for {symbol} #{ticket}")
        else:
            print(f"Close {symbol} #{ticket}: retcode={result.retcode}, volume={volume}")

# ===================== UI =====================
def clear_screen():
    os.system('cls' if os.name == 'nt' else 'clear')

def print_dashboard(df_today: pd.DataFrame, last_news_update: datetime, last_action_note: str = ""):
    """Redraw dashboard: current time, account snapshot, today's events table."""
    clear_screen()
    now = datetime.now(LOCAL_TZ)
    print("=" * 100)
    print(f"News Guardian - {now.strftime('%Y-%m-%d %H:%M:%S')} (Local Time)")
    print("=" * 100)

    # MT5 account snapshot
    snap = mt5_account_snapshot()
    if snap:
        balance, equity, margin, margin_free, profit = snap
        print(f"Account: Balance={balance:.2f} | Equity={equity:.2f} | Margin={margin:.2f} | Free Margin={margin_free:.2f} | Profit={profit:.2f}")
    else:
        print("Account: not available (not connected)")

    # Filters
    currs = ", ".join(CURRENCY_FILTER) if CURRENCY_FILTER else "All"
    impacts = ", ".join([lvl.lower() for lvl in IMPACT_LEVELS])
    print(f"Filters: impact=[{impacts}] | currencies=[{currs}]")
    if last_news_update:
        print(f"News last updated: {last_news_update.strftime('%Y-%m-%d %H:%M:%S')} (local)")
    if last_action_note:
        print(f"Action: {last_action_note}")
    print("-" * 100)

    # Table header
    print(f"{'Time':<6} | {'CCY':<3} | {'Impact':<7} | {'Status':<8} | {'Title'}")
    print("-" * 100)

    if df_today.empty:
        print("No news for today with the selected filters.")
    else:
        now_local = datetime.now(LOCAL_TZ)
        for _, row in df_today.iterrows():
            time_str = row["date"].strftime('%H:%M')
            status = compute_status(now_local, row["date"])
            print(f"{time_str:<6} | {row['country']:<3} | {str(row['impact']):<7} | {status:<8} | {row['title']}")

# ===================== Main loop =====================
def main():
    print("News Guardian is running...")

    # Connect MT5 once at start
    connected = mt5_connect()
    if not connected:
        print("Proceeding without MT5 data (connection failed).")

    last_news_update = None
    df_today = pd.DataFrame()

    # To avoid repeated position-closings within the same event window:
    # remember the window currently being handled (by title + window end)
    active_block_key = None

    while True:
        try:
            now = datetime.now(LOCAL_TZ)
            need_refresh = (last_news_update is None) or ((now - last_news_update).total_seconds() > NEWS_REFRESH_INTERVAL)

            # Refresh today's news once per hour or if empty
            if need_refresh:
                raw = fetch_news_df()
                df_today = normalize_and_filter_today(raw)
                last_news_update = now

            # Build windows from today's filtered events
            windows = build_block_windows(df_today)
            in_block, title, currency, impact, start, end = is_now_in_block_window(windows)

            action_note = ""
            if in_block:
                # Make a unique key for this block window
                block_key = f"{title}|{start.isoformat()}|{end.isoformat()}"

                # Close positions once per window
                if CLOSE_POSITIONS_DURING_BLOCK and block_key != active_block_key:
                    if connected or mt5_connect():
                        try:
                            mt5_close_all_positions()
                            action_note += f"Closed positions due to: {title} ({currency} {impact})"
                        except Exception as e:
                            action_note += f"Close positions error: {e}"
                    active_block_key = block_key

                # Optionally close terminal app
                if CLOSE_TERMINAL_DURING_BLOCK and is_terminal_running():
                    close_terminal()
                    if action_note:
                        action_note += " | "
                    action_note += "Terminal closed for block window"

            else:
                # Outside block window: reset key so next window will trigger again
                active_block_key = None

                # Optionally reopen terminal app if it's not running
                if CLOSE_TERMINAL_DURING_BLOCK and not is_terminal_running():
                    open_terminal()
                    action_note = "Terminal opened (outside block window)"

            # Redraw dashboard each tick
            print_dashboard(df_today, last_news_update, last_action_note=action_note)

            time.sleep(TICK_INTERVAL)

        except KeyboardInterrupt:
            print("\nExiting by user...")
            break
        except Exception as e:
            print(f"Error: {e}")
            time.sleep(TICK_INTERVAL)

    # Cleanup
    try:
        mt5.shutdown()
    except Exception:
        pass

if __name__ == "__main__":
    main()
```
