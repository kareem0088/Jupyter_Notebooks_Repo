```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, RobustScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras import regularizers
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
from tensorflow.keras.optimizers import Adam
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import matplotlib.pyplot as plt
import seaborn as sns
import tqdm
# ========================
# Ø¥Ø¶Ø§ÙØ© Ø¯Ø§Ù„Ø© create_sliding_window Ø§Ù„Ù…ÙÙ‚ÙˆØ¯Ø©
# ========================
def create_sliding_window(X, y, window_size):
    """
    Ø¥Ù†Ø´Ø§Ø¡ Ù†ÙˆØ§ÙØ° Ø²Ù…Ù†ÙŠØ© Ù…Ù†Ø²Ù„Ù‚Ø© Ù…Ù† Ø§Ù„Ø¨ÙŠØ§Ù†Ø§Øª
    """
    X_windows = []
    y_windows = []
    
    for i in range(len(X) - window_size + 1):
        X_windows.append(X[i:i + window_size])
        y_windows.append(y[i + window_size - 1])  # Ø¢Ø®Ø± Ù‡Ø¯Ù ÙÙŠ Ø§Ù„Ù†Ø§ÙØ°Ø©
    
    return np.array(X_windows), np.array(y_windows)

# ========================
# 1. ØªØ­Ù…ÙŠÙ„ ÙˆØªØ¬Ù‡ÙŠØ² Ø§Ù„Ø¨ÙŠØ§Ù†Ø§Øª
# ========================
# ØªØ£ÙƒØ¯ Ù…Ù† ØªØ­Ù…ÙŠÙ„ Ø§Ù„Ø¨ÙŠØ§Ù†Ø§Øª Ø£ÙˆÙ„Ø§Ù‹
# df_train = pd.read_csv('your_train_data.csv')
# df_test = pd.read_csv('your_test_data.csv')

# Ø§Ù„Ù…ÙŠØ²Ø§Øª
features_train = df_train.drop(columns=['Signal','date','timestamp','future_return'])
features_test = df_test.drop(columns=['Signal','date','timestamp','future_return'])

# Ø§Ù„Ø£Ù‡Ø¯Ø§Ù
target_train = df_train['Signal']
target_test = df_test['Signal']

# Ø¥Ù†Ø´Ø§Ø¡ Ø§Ù„Ù†ÙˆØ§ÙØ° Ø§Ù„Ø²Ù…Ù†ÙŠØ©
window_size = 25
X_train_raw, y_train_raw = create_sliding_window(features_train.values, target_train.values, window_size)
X_test_raw, y_test_raw = create_sliding_window(features_test.values, target_test.values, window_size)

# ØªØ·Ø¨ÙŠÙ‚ Ø§Ù„ØªØ·Ø¨ÙŠØ¹
scaler = RobustScaler()

# ØªØ·Ø¨ÙŠØ¹ Ø§Ù„Ø¨ÙŠØ§Ù†Ø§Øª
X_train_scaled = scaler.fit_transform(X_train_raw.reshape(-1, X_train_raw.shape[-1])).reshape(X_train_raw.shape)
X_test_scaled = scaler.transform(X_test_raw.reshape(-1, X_test_raw.shape[-1])).reshape(X_test_raw.shape)

# ØªØ­ÙˆÙŠÙ„ Ø§Ù„Ø£Ù‡Ø¯Ø§Ù Ø¥Ù„Ù‰ categorical
y_train_cat = to_categorical(y_train_raw, num_classes=3)
y_test_cat = to_categorical(y_test_raw, num_classes=3)

print(f"Ø´ÙƒÙ„ Ø¨ÙŠØ§Ù†Ø§Øª Ø§Ù„ØªØ¯Ø±ÙŠØ¨: {X_train_scaled.shape}")
print(f"Ø´ÙƒÙ„ Ø¨ÙŠØ§Ù†Ø§Øª Ø§Ù„Ø§Ø®ØªØ¨Ø§Ø±: {X_test_scaled.shape}")
print(f"Ø´ÙƒÙ„ Ø£Ù‡Ø¯Ø§Ù Ø§Ù„ØªØ¯Ø±ÙŠØ¨: {y_train_cat.shape}")
print(f"Ø´ÙƒÙ„ Ø£Ù‡Ø¯Ø§Ù Ø§Ù„Ø§Ø®ØªØ¨Ø§Ø±: {y_test_cat.shape}")

# ========================
# 5. Ø¨Ù†Ø§Ø¡ Ù†Ù…ÙˆØ°Ø¬ LSTM
# ========================
model = Sequential()
model.add(LSTM(
    64,
    input_shape=(X_train_scaled.shape[1], X_train_scaled.shape[2]),
    return_sequences=False,
    dropout=0.2,
    kernel_regularizer=regularizers.l2(0.01)
))
model.add(Dense(32, activation='relu', kernel_regularizer=regularizers.l2(0.01)))
model.add(Dropout(0.2))
model.add(Dense(3, activation='softmax'))

# ========================
# 5.1 ØªØ­Ø¯ÙŠØ¯ Ù†Ø³Ø¨Ø© Ø§Ù„ØªØ¹Ù„Ù…
# ========================
optimizer = Adam(learning_rate=0.001)
model.compile(optimizer=optimizer, loss='categorical_crossentropy', metrics=['accuracy'])

# Ø¹Ø±Ø¶ Ù…Ù„Ø®Øµ Ø§Ù„Ù†Ù…ÙˆØ°Ø¬
model.summary()

# ========================
# 6.1 Ø¥Ø¹Ø¯Ø§Ø¯ Ø§Ù„ØªÙˆÙ‚Ù Ø§Ù„Ù…Ø¨ÙƒØ± ÙˆØ­ÙØ¸ Ø£ÙØ¶Ù„ Ù†Ù…ÙˆØ°Ø¬
# ========================
early_stop = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)
checkpoint = ModelCheckpoint('best_model.keras', monitor='val_loss', save_best_only=True)

# ========================
# 6. ØªØ¯Ø±ÙŠØ¨ Ø§Ù„Ù†Ù…ÙˆØ°Ø¬
# ========================
history = model.fit(
    X_train_scaled, y_train_cat,  # Ø§Ø³ØªØ®Ø¯Ø§Ù… Ø§Ù„Ù…ØªØºÙŠØ±Ø§Øª Ø§Ù„ØµØ­ÙŠØ­Ø©
    epochs=20,
    batch_size=100,
    validation_data=(X_test_scaled, y_test_cat),  # Ø§Ø³ØªØ®Ø¯Ø§Ù… Ø§Ù„Ù…ØªØºÙŠØ±Ø§Øª Ø§Ù„ØµØ­ÙŠØ­Ø©
    callbacks=[early_stop, checkpoint],
    verbose=1
)

# ========================
# 7. ØªÙ‚ÙŠÙŠÙ… ØªØ¯Ø±ÙŠØ¬ÙŠ Ø®Ø·ÙˆØ© Ø¨Ø®Ø·ÙˆØ© (Walking Forward Evaluation)
# ========================
y_true = []
y_pred = []

print("Ø¬Ø§Ø±ÙŠ Ø§Ù„ØªÙ‚ÙŠÙŠÙ… Ø§Ù„ØªØ¯Ø±ÙŠØ¬ÙŠ...")
for i in  tqdm.tqdm(range(len(X_test_scaled))):
    sample_x = X_test_scaled[i].reshape(1, X_test_scaled.shape[1], X_test_scaled.shape[2])
    true_class = np.argmax(y_test_cat[i])  # Ø§Ø³ØªØ®Ø¯Ø§Ù… y_test_cat
    prediction = model.predict(sample_x, verbose=0)
    predicted_class = np.argmax(prediction)
    
    y_true.append(true_class)
    y_pred.append(predicted_class)

walk_acc = accuracy_score(y_true, y_pred)
print(f"ğŸ§  Walking Forward Accuracy: {walk_acc:.4f}")

# ==================================================================================================================
# ØªÙ‚ÙŠÙŠÙ… Ø§Ù„Ù†Ù…ÙˆØ°Ø¬ Ø¹Ù„Ù‰ Ø§Ø®ØªØ¨Ø§Ø± Ø­Ù‚ÙŠÙ‚ÙŠ
# ==================================================================================================================
test_loss, test_accuracy = model.evaluate(X_test_scaled, y_test_cat, verbose=0)  # ØªØµØ­ÙŠØ­ Ø§Ù„Ù…ØªØºÙŠØ±Ø§Øª
print(f"ğŸ“Š Test Loss: {test_loss:.4f}")
print(f"ğŸ¯ Test Accuracy: {test_accuracy:.4f}")

# ========================
# Ø±Ø³Ù… Ù…Ù†Ø­Ù†Ù‰ Ø§Ù„Ø£Ø¯Ø§Ø¡ Ø§Ù„ÙƒØ§Ù…Ù„
# ========================
epochs_range = range(len(history.history['loss']))

plt.figure(figsize=(12, 6))

# Ø±Ø³Ù… Ø§Ù„Ø®Ø³Ø§Ø±Ø©
plt.subplot(1, 2, 1)
plt.plot(epochs_range, history.history['loss'], label='Train Loss', linestyle='--')
plt.plot(epochs_range, history.history['val_loss'], label='Validation Loss', linestyle='--')
plt.title('Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)

# Ø±Ø³Ù… Ø§Ù„Ø¯Ù‚Ø©
plt.subplot(1, 2, 2)
plt.plot(epochs_range, history.history['accuracy'], label='Train Accuracy', linestyle='-')
plt.plot(epochs_range, history.history['val_accuracy'], label='Validation Accuracy', linestyle='-')
plt.title('Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()

# ===========
# Ø¹Ù…Ù„ Ø§Ù„ØªÙ†Ø¨Ø¤Ø§Øª ÙˆØªÙ‚ÙŠÙŠÙ… Ø´Ø§Ù…Ù„
# ===========
y_pred_probs = model.predict(X_test_scaled)
y_pred = np.argmax(y_pred_probs, axis=1)
y_true = np.argmax(y_test_cat, axis=1)

# ===========
# Ù…ØµÙÙˆÙØ© Ø§Ù„Ø§Ù„ØªØ¨Ø§Ø³
# ===========
cm = confusion_matrix(y_true, y_pred)

plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
            xticklabels=[0, 1, 2], yticklabels=[0, 1, 2])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()

# ===========
# ØªÙ‚Ø±ÙŠØ± Ø§Ù„Ø£Ø¯Ø§Ø¡ Ø§Ù„ÙƒØ§Ù…Ù„
# ===========
print("\nğŸ“ˆ Classification Report:")
print("=" * 50)
print(classification_report(y_true, y_pred, digits=3))

# ===========
# Ø¥Ø­ØµØ§Ø¦ÙŠØ§Øª Ø¥Ø¶Ø§ÙÙŠØ©
# ===========
print(f"\nğŸ“Š Ø¥Ø­ØµØ§Ø¦ÙŠØ§Øª Ø¥Ø¶Ø§ÙÙŠØ©:")
print(f"Ø¹Ø¯Ø¯ Ø¹ÙŠÙ†Ø§Øª Ø§Ù„ØªØ¯Ø±ÙŠØ¨: {len(X_train_scaled)}")
print(f"Ø¹Ø¯Ø¯ Ø¹ÙŠÙ†Ø§Øª Ø§Ù„Ø§Ø®ØªØ¨Ø§Ø±: {len(X_test_scaled)}")
print(f"Ø¹Ø¯Ø¯ Ø§Ù„Ù…ÙŠØ²Ø§Øª: {X_train_scaled.shape[2]}")
print(f"Ø­Ø¬Ù… Ø§Ù„Ù†Ø§ÙØ²Ø© Ø§Ù„Ø²Ù…Ù†ÙŠØ©: {window_size}")
print(f"Ø¹Ø¯Ø¯ Ø§Ù„Ø­Ù‚Ø¨ Ø§Ù„Ù…ÙƒØªÙ…Ù„Ø©: {len(history.history['loss'])}")
```
