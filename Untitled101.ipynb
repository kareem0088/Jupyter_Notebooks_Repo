```python
import pandas as pd
import pandas_ta as ta

def generate_signals_with_tp_sl(df):
    df = df.copy()

    # حساب المتوسطات المتحركة من إغلاق Heikin Ashi
    df['EMA21'] = ta.ema(df['Close'], length=21)
    df['EMA50'] = ta.ema(df['Close'], length=50)

    # أعمدة الإشارات
    df['Signal'] = 0
    df['Entry_Price'] = None
    df['TP_Price'] = None
    df['SL_Price'] = None

    for i in range(2, len(df) - 1):  # نحتاج شمعة واحدة على الأقل بعد الدخول

        # تقاطع صاعد
        ema_cross_up = df['EMA21'].iloc[i-1] < df['EMA50'].iloc[i-1] and df['EMA21'].iloc[i] > df['EMA50'].iloc[i]
        # تقاطع هابط
        ema_cross_down = df['EMA21'].iloc[i-1] > df['EMA50'].iloc[i-1] and df['EMA21'].iloc[i] < df['EMA50'].iloc[i]

        # شرط شمعتين صعود بدون ذيول كبيرة (شراء)
        buy_signal = (
            df['Close'].iloc[i-1] > df['Open'].iloc[i-1] and
            df['Close'].iloc[i-2] > df['Open'].iloc[i-2] and
            (df['Open'].iloc[i-1] - df['Low'].iloc[i-1]) < (df['Close'].iloc[i-1] - df['Open'].iloc[i-1]) * 0.3 and
            ema_cross_up
        )

        # شرط شمعتين هبوط بدون ذيول كبيرة (بيع)
        sell_signal = (
            df['Close'].iloc[i-1] < df['Open'].iloc[i-1] and
            df['Close'].iloc[i-2] < df['Open'].iloc[i-2] and
            (df['High'].iloc[i-1] - df['Open'].iloc[i-1]) < (df['Open'].iloc[i-1] - df['Close'].iloc[i-1]) * 0.3 and
            ema_cross_down
        )

        entry_price = df['Close'].iloc[i]
        future_high = df['High'].iloc[i+1:]
        future_low = df['Low'].iloc[i+1:]

        if buy_signal:
            tp = entry_price + 5.0  # هدف 10 دولار
            sl = entry_price - 5.0   # وقف خسارة 5 دولار

            tp_hit_idx = future_high[future_high >= tp].index
            sl_hit_idx = future_low[future_low <= sl].index

            if not tp_hit_idx.empty and not sl_hit_idx.empty:
                if tp_hit_idx[0] < sl_hit_idx[0]:
                    df.at[df.index[i], 'Signal'] = 1
            elif not tp_hit_idx.empty:
                df.at[df.index[i], 'Signal'] = 1

            if df.at[df.index[i], 'Signal'] == 1:
                df.at[df.index[i], 'Entry_Price'] = entry_price
                df.at[df.index[i], 'TP_Price'] = tp
                df.at[df.index[i], 'SL_Price'] = sl

        elif sell_signal:
            tp = entry_price - 5.0
            sl = entry_price + 5.0

            tp_hit_idx = future_low[future_low <= tp].index
            sl_hit_idx = future_high[future_high >= sl].index

            if not tp_hit_idx.empty and not sl_hit_idx.empty:
                if tp_hit_idx[0] < sl_hit_idx[0]:
                    df.at[df.index[i], 'Signal'] = -1
            elif not tp_hit_idx.empty:
                df.at[df.index[i], 'Signal'] = -1

            if df.at[df.index[i], 'Signal'] == -1:
                df.at[df.index[i], 'Entry_Price'] = entry_price
                df.at[df.index[i], 'TP_Price'] = tp
                df.at[df.index[i], 'SL_Price'] = sl

    return df
```

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns

def prepare_data(df, window_size=500):
    """
    تحضير البيانات للنموذج مع النافذة المنزلقة
    """
    # نسخ البيانات لتجنب تعديل الأصل
    data = df.copy()
    
    # إزالة الأعمدة غير المطلوبة
    columns_to_remove = ['Entry_Price', 'TP_Price', 'SL_Price']
    existing_columns = [col for col in columns_to_remove if col in data.columns]
    if existing_columns:
        data = data.drop(columns=existing_columns)
    
    # فصل المتغيرات المستقلة والهدف
    target_col = 'Signal'
    feature_cols = [col for col in data.columns if col != target_col]
    
    X = data[feature_cols].values
    y = data[target_col].values
    
    # إزالة الصفوف التي تحتوي على 0 في الهدف (مع الاحتفاظ بالفهارس)
    valid_indices = y != 0
    X_filtered = X[valid_indices]
    y_filtered = y[valid_indices]
    
    # تحويل التصنيفات: 1 (بيع) -> 0, 2 (شراء) -> 1
    y_binary = np.where(y_filtered == 1, 0, 1)  # 1->0 (بيع), 2->1 (شراء)
    
    # تطبيع البيانات
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X_filtered)
    
    # إنشاء النوافذ المنزلقة
    X_sequences = []
    y_sequences = []
    
    for i in range(window_size, len(X_scaled)):
        X_sequences.append(X_scaled[i-window_size:i])
        y_sequences.append(y_binary[i])
    
    X_sequences = np.array(X_sequences)
    y_sequences = np.array(y_sequences)
    
    return X_sequences, y_sequences, scaler, feature_cols

def create_lstm_model(input_shape):
    """
    إنشاء نموذج LSTM
    """
    model = Sequential([
        # الطبقة الأولى من LSTM
        LSTM(128, return_sequences=True, input_shape=input_shape),
        Dropout(0.2),
        
        # الطبقة الثانية من LSTM
        LSTM(64, return_sequences=True),
        Dropout(0.2),
        
        # الطبقة الثالثة من LSTM
        LSTM(32, return_sequences=False),
        Dropout(0.2),
        
        # طبقات Dense
        Dense(32, activation='relu'),
        Dropout(0.3),
        Dense(16, activation='relu'),
        Dropout(0.2),
        
        # طبقة الإخراج النهائية (تصنيف ثنائي)
        Dense(1, activation='sigmoid')
    ])
    
    model.compile(
        optimizer=Adam(learning_rate=0.001),
        loss='binary_crossentropy',
        metrics=['accuracy']
    )
    
    return model

def train_model(X, y, test_size=0.2, validation_size=0.1, epochs=100, batch_size=32):
    """
    تدريب النموذج
    """
    # تقسيم البيانات
    X_temp, X_test, y_temp, y_test = train_test_split(
        X, y, test_size=test_size, random_state=42, stratify=y
    )
    
    X_train, X_val, y_train, y_val = train_test_split(
        X_temp, y_temp, test_size=validation_size/(1-test_size), 
        random_state=42, stratify=y_temp
    )
    
    print(f"حجم بيانات التدريب: {X_train.shape}")
    print(f"حجم بيانات التحقق: {X_val.shape}")
    print(f"حجم بيانات الاختبار: {X_test.shape}")
    
    # إنشاء النموذج
    model = create_lstm_model((X_train.shape[1], X_train.shape[2]))
    print(model.summary())
    
    # callbacks
    early_stopping = EarlyStopping(
        monitor='val_loss',
        patience=15,
        restore_best_weights=True
    )
    
    reduce_lr = ReduceLROnPlateau(
        monitor='val_loss',
        factor=0.5,
        patience=8,
        min_lr=1e-7
    )
    
    # تدريب النموذج
    history = model.fit(
        X_train, y_train,
        validation_data=(X_val, y_val),
        epochs=epochs,
        batch_size=batch_size,
        callbacks=[early_stopping, reduce_lr],
        verbose=1
    )
    
    return model, history, X_test, y_test

def evaluate_model(model, X_test, y_test):
    """
    تقييم النموذج
    """
    # التنبؤ
    y_pred_prob = model.predict(X_test)
    y_pred = (y_pred_prob > 0.5).astype(int).flatten()
    
    # تقرير التصنيف
    target_names = ['بيع (1)', 'شراء (2)']
    print("\nتقرير التصنيف:")
    print(classification_report(y_test, y_pred, target_names=target_names))
    
    # مصفوفة الخلط
    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
                xticklabels=target_names, yticklabels=target_names)
    plt.title('مصفوفة الخلط')
    plt.ylabel('القيم الحقيقية')
    plt.xlabel('التنبؤات')
    plt.show()
    
    return y_pred, y_pred_prob

def plot_training_history(history):
    """
    رسم تاريخ التدريب
    """
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))
    
    # دقة النموذج
    ax1.plot(history.history['accuracy'], label='دقة التدريب')
    ax1.plot(history.history['val_accuracy'], label='دقة التحقق')
    ax1.set_title('دقة النموذج')
    ax1.set_xlabel('العصر')
    ax1.set_ylabel('الدقة')
    ax1.legend()
    
    # خسارة النموذج
    ax2.plot(history.history['loss'], label='خسارة التدريب')
    ax2.plot(history.history['val_loss'], label='خسارة التحقق')
    ax2.set_title('خسارة النموذج')
    ax2.set_xlabel('العصر')
    ax2.set_ylabel('الخسارة')
    ax2.legend()
    
    plt.tight_layout()
    plt.show()

def predict_new_data(model, scaler, new_data, window_size=500):
    """
    التنبؤ على بيانات جديدة
    """
    # تطبيع البيانات الجديدة
    new_data_scaled = scaler.transform(new_data)
    
    # إنشاء النافذة
    if len(new_data_scaled) >= window_size:
        X_new = new_data_scaled[-window_size:].reshape(1, window_size, -1)
        
        # التنبؤ
        prediction_prob = model.predict(X_new)[0][0]
        prediction = 1 if prediction_prob > 0.5 else 0
        
        # تحويل إلى التصنيف الأصلي
        original_prediction = 2 if prediction == 1 else 1
        
        return original_prediction, prediction_prob
    else:
        print(f"البيانات الجديدة تحتاج على الأقل {window_size} نقطة بيانات")
        return None, None

# مثال على الاستخدام
if __name__ == "__main__":
    # تحميل البيانات (استبدل هذا بمسار ملفك الفعلي)
    # df = pd.read_csv('your_data.csv')
    
    # مثال على كيفية الاستخدام:
    
    # 1. تحضير البيانات
    X, y, scaler, feature_cols = prepare_data(df, window_size=500)
    
    # 2. تدريب النموذج
    model, history, X_test, y_test = train_model(X, y, epochs=50)
    
    # 3. رسم تاريخ التدريب
    plot_training_history(history)
    
    # 4. تقييم النموذج
    y_pred, y_pred_prob = evaluate_model(model, X_test, y_test)
    
    # 5. حفظ النموذج
    model.save('lstm_trading_model.h5')
    
    # 6. التنبؤ على بيانات جديدة
    # new_prediction, confidence = predict_new_data(model, scaler, new_data_sample)
    # print(f"التنبؤ: {'شراء' if new_prediction == 2 else 'بيع'}")
    # print(f"الثقة: {confidence:.4f}")
    
    
    print("تم إنشاء النموذج بنجاح!")
    print("\nخطوات الاستخدام:")
    print("1. قم بتحميل البيانات في متغير df")
    print("2. استخدم الدوال المتوفرة لتدريب وتقييم النموذج")
    print("3. استخدم predict_new_data للتنبؤ على بيانات جديدة")
```

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns

def prepare_data(df, window_size=50):
    """
    تحضير البيانات للنموذج مع النافذة المنزلقة
    """
    # نسخ البيانات لتجنب تعديل الأصل
    data = df.copy()
    
    # إزالة الأعمدة غير المطلوبة
    columns_to_remove = ['Entry_Price', 'TP_Price', 'SL_Price']
    existing_columns = [col for col in columns_to_remove if col in data.columns]
    if existing_columns:
        data = data.drop(columns=existing_columns)
    
    # فصل المتغيرات المستقلة والهدف
    target_col = 'Signal'
    feature_cols = [col for col in data.columns if col != target_col]
    
    X = data[feature_cols].values
    y = data[target_col].values
    
    # الاحتفاظ بجميع البيانات بدون حذف أي صفوف
    # تطبيع البيانات
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)
    
    # إنشاء النوافذ المنزلقة
    X_sequences = []
    y_sequences = []
    
    for i in range(window_size, len(X_scaled)):
        X_sequences.append(X_scaled[i-window_size:i])
        y_sequences.append(y[i])
    
    X_sequences = np.array(X_sequences)
    y_sequences = np.array(y_sequences)
    
    return X_sequences, y_sequences, scaler, feature_cols

def create_lstm_model(input_shape):
    """
    إنشاء نموذج LSTM للتصنيف الثلاثي
    """
    model = Sequential([
        # الطبقة الأولى من LSTM
        LSTM(128, return_sequences=True, input_shape=input_shape),
        Dropout(0.2),
        
        # الطبقة الثانية من LSTM
        LSTM(64, return_sequences=True),
        Dropout(0.2),
        
        # الطبقة الثالثة من LSTM
        LSTM(32, return_sequences=False),
        Dropout(0.2),
        
        # طبقات Dense
        Dense(32, activation='relu'),
        Dropout(0.3),
        Dense(16, activation='relu'),
        Dropout(0.2),
        
        # طبقة الإخراج النهائية (تصنيف ثلاثي)
        Dense(3, activation='softmax')  # 3 classes: 0, 1, 2
    ])
    
    model.compile(
        optimizer=Adam(learning_rate=0.001),
        loss='sparse_categorical_crossentropy',  # للتصنيف الثلاثي
        metrics=['accuracy']
    )
    
    return model

def train_model(X, y, test_size=0.2, validation_size=0.1, epochs=50, batch_size=32):
    """
    تدريب النموذج
    """
    # تقسيم البيانات
    X_temp, X_test, y_temp, y_test = train_test_split(
        X, y, test_size=test_size, random_state=42, stratify=y
    )
    
    X_train, X_val, y_train, y_val = train_test_split(
        X_temp, y_temp, test_size=validation_size/(1-test_size), 
        random_state=42, stratify=y_temp
    )
    
    print(f"حجم بيانات التدريب: {X_train.shape}")
    print(f"حجم بيانات التحقق: {X_val.shape}")
    print(f"حجم بيانات الاختبار: {X_test.shape}")
    
    # إحصائيات التوزيع
    unique, counts = np.unique(y_train, return_counts=True)
    print(f"\nتوزيع الفئات في التدريب:")
    for u, c in zip(unique, counts):
        if u == 0:
            print(f"  لا إشارة (0): {c}")
        elif u == 1:
            print(f"  بيع (1): {c}")
        elif u == 2:
            print(f"  شراء (2): {c}")
    
    # إنشاء النموذج
    model = create_lstm_model((X_train.shape[1], X_train.shape[2]))
    print(model.summary())
    
    # callbacks
    early_stopping = EarlyStopping(
        monitor='val_loss',
        patience=15,
        restore_best_weights=True
    )
    
    reduce_lr = ReduceLROnPlateau(
        monitor='val_loss',
        factor=0.5,
        patience=8,
        min_lr=1e-7
    )
    
    # تدريب النموذج
    history = model.fit(
        X_train, y_train,
        validation_data=(X_val, y_val),
        epochs=epochs,
        batch_size=batch_size,
        callbacks=[early_stopping, reduce_lr],
        verbose=1
    )
    
    return model, history, X_test, y_test

def evaluate_model(model, X_test, y_test):
    """
    تقييم النموذج مع التركيز على فئات البيع والشراء
    """
    # التنبؤ
    y_pred_prob = model.predict(X_test)
    y_pred = np.argmax(y_pred_prob, axis=1)
    
    # تقرير التصنيف الكامل
    target_names = ['لا إشارة (0)', 'بيع (1)', 'شراء (2)']
    print("\nتقرير التصنيف الكامل:")
    print(classification_report(y_test, y_pred, target_names=target_names))
    
    # تقرير التصنيف للبيع والشراء فقط (تجاهل 0)
    mask_non_zero = (y_test != 0) & (y_pred != 0)  # كلاهما غير صفر
    if np.sum(mask_non_zero) > 0:
        y_test_filtered = y_test[mask_non_zero]
        y_pred_filtered = y_pred[mask_non_zero]
        
        print("\nتقرير التصنيف للبيع والشراء فقط:")
        print(classification_report(y_test_filtered, y_pred_filtered, 
                                  target_names=['بيع (1)', 'شراء (2)'], labels=[1, 2]))
    
    # مصفوفة الخلط الكاملة
    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
                xticklabels=target_names, yticklabels=target_names)
    plt.title('مصفوفة الخلط - جميع الفئات')
    plt.ylabel('القيم الحقيقية')
    plt.xlabel('التنبؤات')
    plt.show()
    
    # مصفوفة الخلط للبيع والشراء فقط
    if np.sum(mask_non_zero) > 0:
        cm_filtered = confusion_matrix(y_test_filtered, y_pred_filtered, labels=[1, 2])
        plt.figure(figsize=(8, 6))
        sns.heatmap(cm_filtered, annot=True, fmt='d', cmap='Greens', 
                    xticklabels=['بيع (1)', 'شراء (2)'], yticklabels=['بيع (1)', 'شراء (2)'])
        plt.title('مصفوفة الخلط - البيع والشراء فقط')
        plt.ylabel('القيم الحقيقية')
        plt.xlabel('التنبؤات')
        plt.show()
    
    # إحصائيات إضافية
    print(f"\nإحصائيات التنبؤ:")
    unique_pred, counts_pred = np.unique(y_pred, return_counts=True)
    for u, c in zip(unique_pred, counts_pred):
        if u == 0:
            print(f"  تنبؤ لا إشارة (0): {c}")
        elif u == 1:
            print(f"  تنبؤ بيع (1): {c}")
        elif u == 2:
            print(f"  تنبؤ شراء (2): {c}")
    
    return y_pred, y_pred_prob

def plot_training_history(history):
    """
    رسم تاريخ التدريب
    """
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))
    
    # دقة النموذج
    ax1.plot(history.history['accuracy'], label='دقة التدريب')
    ax1.plot(history.history['val_accuracy'], label='دقة التحقق')
    ax1.set_title('دقة النموذج')
    ax1.set_xlabel('العصر')
    ax1.set_ylabel('الدقة')
    ax1.legend()
    
    # خسارة النموذج
    ax2.plot(history.history['loss'], label='خسارة التدريب')
    ax2.plot(history.history['val_loss'], label='خسارة التحقق')
    ax2.set_title('خسارة النموذج')
    ax2.set_xlabel('العصر')
    ax2.set_ylabel('الخسارة')
    ax2.legend()
    
    plt.tight_layout()
    plt.show()

def predict_new_data(model, scaler, new_data, window_size=50):
    """
    التنبؤ على بيانات جديدة
    """
    # تطبيع البيانات الجديدة
    new_data_scaled = scaler.transform(new_data)
    
    # إنشاء النافذة
    if len(new_data_scaled) >= window_size:
        X_new = new_data_scaled[-window_size:].reshape(1, window_size, -1)
        
        # التنبؤ
        prediction_prob = model.predict(X_new)[0]
        prediction = np.argmax(prediction_prob)
        confidence = np.max(prediction_prob)
        
        # تفسير النتيجة
        if prediction == 0:
            result = "لا إشارة"
        elif prediction == 1:
            result = "بيع"
        elif prediction == 2:
            result = "شراء"
        
        return prediction, confidence, result, prediction_prob
    else:
        print(f"البيانات الجديدة تحتاج على الأقل {window_size} نقطة بيانات")
        return None, None, None, None

def get_trading_signals_only(y_pred, y_pred_prob, threshold=0.5):
    """
    استخراج إشارات التداول فقط (بيع وشراء) مع تجاهل اللاإشارة
    """
    # الحصول على الفئة الأكثر احتمالاً للبيع والشراء فقط
    trading_signals = []
    trading_probabilities = []
    
    for i, (pred, probs) in enumerate(zip(y_pred, y_pred_prob)):
        # مقارنة احتمالات البيع والشراء فقط
        buy_prob = probs[2]   # احتمال الشراء
        sell_prob = probs[1]  # احتمال البيع
        
        # اختيار الإشارة الأقوى بين البيع والشراء
        if buy_prob > sell_prob and buy_prob > threshold:
            trading_signals.append(2)  # شراء
            trading_probabilities.append(buy_prob)
        elif sell_prob > buy_prob and sell_prob > threshold:
            trading_signals.append(1)  # بيع
            trading_probabilities.append(sell_prob)
        else:
            trading_signals.append(0)  # لا إشارة واضحة
            trading_probabilities.append(max(buy_prob, sell_prob))
    
    return np.array(trading_signals), np.array(trading_probabilities)

# مثال على الاستخدام
if __name__ == "__main__":
    # تحميل البيانات (استبدل هذا بمسار ملفك الفعلي)
    # df = pd.read_csv('your_data.csv')
    
    # مثال على كيفية الاستخدام:
    
    # 1. تحضير البيانات (يحتفظ بجميع الصفوف بما في ذلك 0)
    X, y, scaler, feature_cols = prepare_data(df, window_size=50)
    
    # 2. تدريب النموذج
    model, history, X_test, y_test = train_model(X, y, epochs=50)
    
    # 3. رسم تاريخ التدريب
    plot_training_history(history)
    
    # 4. تقييم النموذج (مع تقييم منفصل للبيع والشراء)
    y_pred, y_pred_prob = evaluate_model(model, X_test, y_test)
    
    # 5. استخراج إشارات التداول فقط
    trading_signals, trading_probs = get_trading_signals_only(y_pred, y_pred_prob, threshold=0.6)
    
    # 6. حفظ النموذج
    model.save('lstm_trading_model.h5')
    
    # 7. التنبؤ على بيانات جديدة
    # prediction, confidence, result, all_probs = predict_new_data(model, scaler, new_data_sample)
    # print(f"التنبؤ: {result}")
    # print(f"الثقة: {confidence:.4f}")
    # print(f"احتمال البيع: {all_probs[1]:.4f}")
    # print(f"احتمال الشراء: {all_probs[2]:.4f}")
    
    
    print("تم إنشاء النموذج بنجاح!")
    print("\nخصائص النموذج الجديد:")
    print("✓ يحتفظ بجميع البيانات (لا يحذف الصفوف التي تحتوي على 0)")
    print("✓ يتعامل مع التصنيف الثلاثي (0: لا إشارة، 1: بيع، 2: شراء)")
    print("✓ يركز على تقييم البيع والشراء بشكل منفصل")
    print("✓ يوفر احتمالات مفصلة لكل فئة")
    print("✓ يمكن استخراج إشارات التداول فقط حسب عتبة الثقة")
    print("\nخطوات الاستخدام:")
    print("1. قم بتحميل البيانات في متغير df")
    print("2. استخدم prepare_data() لتحضير البيانات مع الحفاظ على جميع الصفوف")
    print("3. استخدم train_model() لتدريب النموذج")
    print("4. استخدم evaluate_model() للحصول على تقييم شامل")
    print("5. استخدم get_trading_signals_only() لاستخراج إشارات التداول فقط")
    print("6. استخدم predict_new_data() للتنبؤ على بيانات جديدة")
```

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter

class TradingSignalClassifier:
    def __init__(self, window_size=500):
        """
        مصنف إشارات التداول
        
        Parameters:
        window_size (int): حجم النافذة المنزلقة
        """
        self.window_size = window_size
        self.scaler = StandardScaler()
        self.model = None
        self.feature_columns = ['Entry_Price', 'TP_Price', 'SL_Price']
        
    def create_sliding_window_features(self, df):
        """
        إنشاء المعالم باستخدام النافذة المنزلقة
        """
        features_list = []
        targets_list = []
        
        # التأكد من وجود الأعمدة المطلوبة
        required_cols = ['Signal'] + self.feature_columns
        for col in required_cols:
            if col not in df.columns:
                raise ValueError(f"العمود {col} غير موجود في البيانات")
        
        print(f"إنشاء المعالم باستخدام نافذة بحجم {self.window_size}")
        
        for i in range(self.window_size, len(df)):
            # أخذ النافذة الحالية
            window = df.iloc[i-self.window_size:i]
            target = df.iloc[i]['Signal']
            
            # تجاهل الصفوف التي تحتوي على Signal = 0
            if target == 0:
                continue
                
            # حساب المعالم الإحصائية للنافذة
            features = []
            
            for col in self.feature_columns:
                window_data = window[col].dropna()
                if len(window_data) > 0:
                    # المعالم الأساسية
                    features.extend([
                        window_data.mean(),           # المتوسط
                        window_data.std(),            # الانحراف المعياري
                        window_data.min(),            # أقل قيمة
                        window_data.max(),            # أعلى قيمة
                        window_data.median(),         # الوسيط
                        window_data.quantile(0.25),   # الربع الأول
                        window_data.quantile(0.75),   # الربع الثالث
                    ])
                    
                    # معالم الاتجاه
                    if len(window_data) > 1:
                        slope = np.polyfit(range(len(window_data)), window_data, 1)[0]
                        features.append(slope)  # ميل الاتجاه
                    else:
                        features.append(0)
                        
                    # التقلبات
                    if len(window_data) > 1:
                        volatility = window_data.pct_change().std()
                        features.append(volatility if not np.isnan(volatility) else 0)
                    else:
                        features.append(0)
                else:
                    # في حالة عدم وجود بيانات، املأ بالأصفار
                    features.extend([0] * 9)
            
            # معالم إضافية
            current_entry = df.iloc[i]['Entry_Price']
            current_tp = df.iloc[i]['TP_Price'] 
            current_sl = df.iloc[i]['SL_Price']
            
            if not (np.isnan(current_entry) or np.isnan(current_tp) or np.isnan(current_sl)):
                # نسبة المخاطرة للعائد
                risk = abs(current_entry - current_sl)
                reward = abs(current_tp - current_entry)
                rr_ratio = reward / risk if risk != 0 else 0
                
                features.extend([
                    rr_ratio,
                    current_entry,
                    current_tp,
                    current_sl,
                    current_tp - current_entry,  # المسافة للهدف
                    current_entry - current_sl,  # المسافة لوقف الخسارة
                ])
            else:
                features.extend([0] * 6)
            
            features_list.append(features)
            targets_list.append(target)
        
        return np.array(features_list), np.array(targets_list)
    
    def prepare_data(self, df):
        """
        تحضير البيانات للتدريب
        """
        print("تحضير البيانات...")
        print(f"عدد الصفوف الأصلي: {len(df)}")
        print(f"توزيع الإشارات: {Counter(df['Signal'])}")
        
        # إنشاء المعالم
        X, y = self.create_sliding_window_features(df)
        
        print(f"عدد العينات بعد النافذة المنزلقة: {len(X)}")
        print(f"توزيع الإشارات النهائي: {Counter(y)}")
        print(f"عدد المعالم: {X.shape[1] if len(X) > 0 else 0}")
        
        return X, y
    
    def train_model(self, df, model_type='random_forest', test_size=0.2):
        """
        تدريب النموذج
        
        Parameters:
        model_type (str): نوع النموذج ('random_forest', 'logistic', 'svm')
        """
        X, y = self.prepare_data(df)
        
        if len(X) == 0:
            raise ValueError("لا توجد بيانات كافية للتدريب")
        
        # تقسيم البيانات
        X_train, X_test, y_train, y_test = train_test_split(
            X, y, test_size=test_size, random_state=42, stratify=y
        )
        
        # تطبيع البيانات
        X_train_scaled = self.scaler.fit_transform(X_train)
        X_test_scaled = self.scaler.transform(X_test)
        
        # اختيار النموذج
        if model_type == 'random_forest':
            self.model = RandomForestClassifier(
                n_estimators=100,
                max_depth=10,
                random_state=42,
                class_weight='balanced'
            )
        elif model_type == 'logistic':
            self.model = LogisticRegression(
                random_state=42,
                class_weight='balanced',
                max_iter=1000
            )
        elif model_type == 'svm':
            self.model = SVC(
                kernel='rbf',
                random_state=42,
                class_weight='balanced',
                probability=True
            )
        
        print(f"تدريب النموذج: {model_type}")
        
        # تدريب النموذج
        self.model.fit(X_train_scaled, y_train)
        
        # التقييم
        y_pred = self.model.predict(X_test_scaled)
        
        print("\n=== نتائج التقييم ===")
        print(f"دقة النموذج: {accuracy_score(y_test, y_pred):.4f}")
        print("\nتقرير التصنيف:")
        print(classification_report(y_test, y_pred, 
                                  target_names=['بيع (1)', 'شراء (2)']))
        
        # التحقق المتقاطع
        cv_scores = cross_val_score(self.model, X_train_scaled, y_train, cv=5)
        print(f"\nدقة التحقق المتقاطع: {cv_scores.mean():.4f} (+/- {cv_scores.std() * 2:.4f})")
        
        # مصفوفة الخلط
        self.plot_confusion_matrix(y_test, y_pred)
        
        # أهمية المعالم (للـ Random Forest فقط)
        if model_type == 'random_forest':
            self.plot_feature_importance()
        
        return {
            'accuracy': accuracy_score(y_test, y_pred),
            'cv_scores': cv_scores,
            'classification_report': classification_report(y_test, y_pred, output_dict=True)
        }
    
    def predict(self, df, start_idx=None):
        """
        التنبؤ بالإشارات
        """
        if self.model is None:
            raise ValueError("يجب تدريب النموذج أولاً")
        
        X, _ = self.prepare_data(df)
        
        if len(X) == 0:
            return np.array([])
        
        X_scaled = self.scaler.transform(X)
        predictions = self.model.predict(X_scaled)
        probabilities = self.model.predict_proba(X_scaled)
        
        return predictions, probabilities
    
    def plot_confusion_matrix(self, y_true, y_pred):
        """
        رسم مصفوفة الخلط
        """
        plt.figure(figsize=(8, 6))
        cm = confusion_matrix(y_true, y_pred)
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                   xticklabels=['بيع (1)', 'شراء (2)'],
                   yticklabels=['بيع (1)', 'شراء (2)'])
        plt.title('مصفوفة الخلط')
        plt.ylabel('القيم الحقيقية')
        plt.xlabel('التنبؤات')
        plt.show()
    
    def plot_feature_importance(self):
        """
        رسم أهمية المعالم
        """
        if hasattr(self.model, 'feature_importances_'):
            plt.figure(figsize=(12, 8))
            importances = self.model.feature_importances_
            indices = np.argsort(importances)[-20:]  # أهم 20 معلم
            
            plt.barh(range(len(indices)), importances[indices])
            plt.title('أهمية المعالم (أهم 20)')
            plt.xlabel('الأهمية')
            plt.ylabel('رقم المعلم')
            plt.show()

# مثال على الاستخدام
def example_usage():
    """
    مثال على كيفية استخدام المصنف
    """
    # إنشاء بيانات تجريبية
    np.random.seed(42)
    n_samples = 2000
    
    df = pd.DataFrame({
        'Signal': np.random.choice([0, 1, 2], n_samples, p=[0.6, 0.2, 0.2]),
        'Entry_Price': np.random.normal(100, 10, n_samples),
        'TP_Price': np.random.normal(105, 8, n_samples),
        'SL_Price': np.random.normal(95, 8, n_samples)
    })
    
    print("مثال على استخدام مصنف إشارات التداول")
    print("=" * 50)
    
    # إنشاء المصنف
    classifier = TradingSignalClassifier(window_size=500)
    
    # تدريب النموذج
    results = classifier.train_model(df, model_type='random_forest')
    
    # التنبؤ
    predictions, probabilities = classifier.predict(df)
    
    print(f"\nتم إنتاج {len(predictions)} تنبؤ")
    print(f"توزيع التنبؤات: {Counter(predictions)}")
    
    return classifier, results

# تشغيل المثال
if __name__ == "__main__":
    classifier, results = example_usage()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout, BatchNormalization
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.losses import SparseCategoricalCrossentropy
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay

# --- 1. تحضير البيانات ---
df = df_signaled.copy()
df = df.drop(columns=['Group', 'Signal_shift'], errors='ignore')

exclude_cols = ['Signal', 'Entry_Price', 'TP_Price', 'SL_Price']
feature_cols = [col for col in df.columns if col not in exclude_cols]
X = df[feature_cols].values
y = df['Signal'].values  # قيمها: -1، 0، 1

# --- 2. إنشاء نوافذ انزلاقية ---
def create_sliding_windows(X, y, window_size=50):
    Xs, ys = [], []
    for i in range(len(X) - window_size):
        window = X[i:i + window_size]
        label = y[i + window_size]
        if label in [-1, 0, 1]:
            Xs.append(window)
            ys.append(label)
    return np.array(Xs), np.array(ys)

window_size = 50
X_windows, y_windows = create_sliding_windows(X, y, window_size)

# --- 3. تحويل y إلى (0 = sell, 1 = no signal, 2 = buy) ---
y_mapped = y_windows.copy()
y_mapped[y_windows == -1] = 0
y_mapped[y_windows == 0] = 1
y_mapped[y_windows == 1] = 2

# --- 4. تقسيم البيانات ---
split_ratio = 0.8
split_index = int(len(X_windows) * split_ratio)

X_train, X_test = X_windows[:split_index], X_windows[split_index:]
y_train, y_test = y_mapped[:split_index], y_mapped[split_index:]

# --- 5. إنشاء sample_weights بناءً على الفئة ---
# وزن الفئة 1 (No signal) = 0
class_weights = {0: 1.0, 1: 0.0, 2: 1.0}
sample_weights = np.array([class_weights[label] for label in y_train])

# --- 6. تأكيد نوع البيانات ---
X_train = X_train.astype(np.float32)
X_test = X_test.astype(np.float32)
y_train = y_train.astype(np.int32)
y_test = y_test.astype(np.int32)

# --- 7. بناء النموذج ---
model = Sequential([
    LSTM(64, return_sequences=True, input_shape=(window_size, X.shape[1])),
    BatchNormalization(),
    LSTM(32),
    BatchNormalization(),
    Dropout(0.1),
    Dense(32, activation='relu'),
    Dense(3, activation='softmax')  # ← 3 فئات
])

model.compile(loss=SparseCategoricalCrossentropy(),
              optimizer=Adam(learning_rate=1e-3),
              metrics=['accuracy'])

model.summary()

# --- 8. تدريب النموذج ---
history = model.fit(
    X_train, y_train,
    epochs=30,
    batch_size=64,
    validation_data=(X_test, y_test),
    sample_weight=sample_weights
)

# --- 9. تقييم النموذج ---
y_pred_prob = model.predict(X_test)
y_pred = np.argmax(y_pred_prob, axis=1)

print("\nClassification Report:")
print(classification_report(y_test, y_pred, target_names=['Sell (-1)', 'No Signal (0)', 'Buy (1)']))

# --- 10. مصفوفة الالتباس ---
cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['Sell (-1)', 'No Signal (0)', 'Buy (1)'])
disp.plot(cmap=plt.cm.Blues)
plt.title("Confusion Matrix")
plt.show()
```

```python
import numpy as np
def generate_signals_fast(df_final, take_profit=50, stop_loss=25, look_ahead=50):
    close = df_final['Close'].values
    high = df_final['High'].values
    low = df_final['Low'].values
    n = len(df_final)

    signals = np.zeros(n, dtype=int)

    for i in range(n - look_ahead):
        entry_price = close[i]

        future_highs = high[i+1:i+1+look_ahead]
        future_lows = low[i+1:i+1+look_ahead]

        diff_high = future_highs - entry_price
        diff_low = entry_price - future_lows

        buy_condition = (diff_high >= take_profit) & (diff_low <= stop_loss)
        sell_condition = (diff_low >= take_profit) & (diff_high <= stop_loss)

        if np.any(buy_condition):
            signals[i] = 1
        elif np.any(sell_condition):
            signals[i] = -1
        else:
            signals[i] = 0

    df_final['Signal'] = signals
    return df_final

# استخدام الدالة مع df_final

df_final = generate_signals_fast(df_final)
print(df_final[['Close', 'Signal']].tail(20))
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout, BatchNormalization
from tensorflow.keras.optimizers import Adam
from sklearn.metrics import classification_report, recall_score, confusion_matrix, ConfusionMatrixDisplay
from imblearn.over_sampling import RandomOverSampler
import tensorflow as tf

# --- focal loss ---
def focal_loss(gamma=2., alpha=0.25):
    def focal_loss_fixed(y_true, y_pred):
        epsilon = tf.keras.backend.epsilon()
        y_pred = tf.clip_by_value(y_pred, epsilon, 1. - epsilon)
        cross_entropy = -y_true * tf.math.log(y_pred) - (1 - y_true) * tf.math.log(1 - y_pred)
        weight = alpha * tf.pow(1 - y_pred, gamma) * y_true + (1 - alpha) * tf.pow(y_pred, gamma) * (1 - y_true)
        loss = weight * cross_entropy
        return tf.reduce_mean(loss)
    return focal_loss_fixed

# --- 1. تحضير البيانات ---
df = df_final.copy()
df = df.drop(columns=['Group', 'Signal_shift'], errors='ignore')

feature_cols = [col for col in df.columns if col != 'Signal']
X = df[feature_cols].values
y = df['Signal'].values

# تحويل الإشارات إلى ثنائية: -1 => 0، 1 => 1، 0 تبقى كما هي
y_binary = np.where(y == -1, 0, y)
y_binary = np.where(y == 1, 1, y_binary)

# إنشاء النوافذ الانزلاقية مع حفظ مواقع النوافذ القابلة للتصنيف فقط
def create_sliding_windows(X, y, window_size=15):
    Xs, ys, valid_flags = [], [], []
    for i in range(len(X) - window_size):
        window = X[i:i+window_size]
        label = y[i+window_size]  # الإشارة بعد آخر شمعة
        if label in [0, 1]:  # فقط لو إشارة شراء أو بيع (0 تم تحويل -1 إلى 0 سابقًا)
            Xs.append(window)
            ys.append(label)
    return np.array(Xs), np.array(ys)


window_size = 15
X_windows, y_windows, valid_flags = create_sliding_windows(X, y_binary, window_size)

X_valid = X_windows[valid_flags == 1]
y_valid = y_windows[valid_flags == 1]

split_ratio = 0.8
split_index = int(len(X_valid) * split_ratio)

X_train, X_test = X_valid[:split_index], X_valid[split_index:]
y_train, y_test = y_valid[:split_index], y_valid[split_index:]

# --- Oversampling للفئة القليلة ---
X_train_reshaped = X_train.reshape((X_train.shape[0], -1))
ros = RandomOverSampler(random_state=42)
X_train_resampled, y_train_resampled = ros.fit_resample(X_train_reshaped, y_train)
X_train = X_train_resampled.reshape((-1, window_size, len(feature_cols)))
y_train = y_train_resampled

# --- تأكيد نوع البيانات ---
X_train = X_train.astype(np.float32)
X_test = X_test.astype(np.float32)
y_train = y_train.astype(np.float32)
y_test = y_test.astype(np.float32)

# --- 2. بناء نموذج LSTM باستخدام focal loss ---
def build_model(input_shape, learning_rate=1e-3):
    model = Sequential()
    model.add(LSTM(64, return_sequences=True, input_shape=input_shape))
    model.add(BatchNormalization())
    model.add(LSTM(32))
    model.add(BatchNormalization())
    model.add(Dropout(0.1))
    model.add(Dense(32, activation='relu'))
    model.add(Dense(1, activation='sigmoid'))
    
    optimizer = Adam(learning_rate=learning_rate)
    model.compile(loss=focal_loss(gamma=2.0, alpha=0.25), optimizer=optimizer, metrics=['accuracy'])
    return model

model = build_model((window_size, len(feature_cols)), learning_rate=1e-3)
model.summary()

# --- 3. كول باك لحساب أفضل macro recall ---
class MacroRecallEarlyStopping(tf.keras.callbacks.Callback):
    def __init__(self, validation_data, patience=5):
        super().__init__()
        self.validation_data = validation_data
        self.patience = patience
        self.best_score = 0
        self.wait = 0
        self.best_weights = None
        self.recalls = []
        self.accuracies = []
        self.losses = []
        self.epoch_best = 0
        self.weights_history = []

    def on_epoch_end(self, epoch, logs=None):
        X_val, y_val = self.validation_data
        y_pred_prob = self.model.predict(X_val)
        
        best_threshold = 0.5
        best_macro_recall = 0
        
        for thresh in np.arange(0.1, 1.0, 0.1):
            y_pred = (y_pred_prob >= thresh).astype(int).flatten()
            recall_0 = recall_score(y_val, y_pred, pos_label=0)
            recall_1 = recall_score(y_val, y_pred, pos_label=1)
            macro_recall = (recall_0 + recall_1) / 2
            
            if macro_recall > best_macro_recall:
                best_macro_recall = macro_recall
                best_threshold = thresh
        
        y_pred_final = (y_pred_prob >= best_threshold).astype(int).flatten()
        recall_0 = recall_score(y_val, y_pred_final, pos_label=0)
        recall_1 = recall_score(y_val, y_pred_final, pos_label=1)
        macro_recall = best_macro_recall

        self.recalls.append(macro_recall)
        self.accuracies.append(logs.get('val_accuracy'))
        self.losses.append(logs.get('val_loss'))

        print(f"\nEpoch {epoch+1} - Best Threshold: {best_threshold:.2f} - Macro Recall: {macro_recall:.4f} - Recall Sell: {recall_0:.4f} - Recall Buy: {recall_1:.4f}")

        self.weights_history.append(self.model.get_weights())

        if macro_recall > self.best_score:
            self.best_score = macro_recall
            self.best_weights = self.model.get_weights()
            self.wait = 0
            self.epoch_best = epoch+1
        else:
            self.wait +=1
            if self.wait >= self.patience:
                print(f"Early stopping on epoch {epoch+1}")
                self.model.stop_training = True

early_stopping_callback = MacroRecallEarlyStopping(validation_data=(X_test, y_test), patience=10)

# --- 4. تدريب النموذج ---
history = model.fit(
    X_train, y_train,
    epochs=100,
    batch_size=1000,
    validation_data=(X_test, y_test),
    callbacks=[early_stopping_callback]
)

# --- 5. استرجاع أفضل وزن حسب macro recall ---
model.set_weights(early_stopping_callback.best_weights)
print(f"\nBest epoch: {early_stopping_callback.epoch_best} with Macro Recall: {early_stopping_callback.best_score:.4f}")

# --- 6. أفضل threshold على بيانات الاختبار ---
y_pred_prob = model.predict(X_test)
best_threshold = 0.5
best_macro_recall = 0

for thresh in np.arange(0.1, 1.0, 0.01):
    y_pred = (y_pred_prob >= thresh).astype(int).flatten()
    recall_0 = recall_score(y_test, y_pred, pos_label=0)
    recall_1 = recall_score(y_test, y_pred, pos_label=1)
    macro_recall = (recall_0 + recall_1) / 2
    if macro_recall > best_macro_recall:
        best_macro_recall = macro_recall
        best_threshold = thresh

print(f"Best threshold on test data: {best_threshold:.3f} with Macro Recall: {best_macro_recall:.4f}")

# --- 7. تقرير التصنيف النهائي ---
y_pred_final = (y_pred_prob >= best_threshold).astype(int).flatten()

print("\nFinal Classification Report:")
print(classification_report(y_test, y_pred_final))

# --- 8. مصفوفة الالتباس ---
cm = confusion_matrix(y_test, y_pred_final)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['Sell (-1)', 'Buy (1)'])
disp.plot(cmap=plt.cm.Blues)
plt.title('Confusion Matrix')
plt.show()

# --- 9. رسم المقاييس لكل إيبوك ---
epochs = range(1, len(early_stopping_callback.accuracies) + 1)

plt.figure(figsize=(14,5))

plt.subplot(1,3,1)
plt.plot(epochs, early_stopping_callback.accuracies, 'b-', label='Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.grid(True)
plt.legend()

plt.subplot(1,3,2)
plt.plot(epochs, early_stopping_callback.losses, 'r-', label='Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.grid(True)
plt.legend()

plt.subplot(1,3,3)
plt.plot(epochs, early_stopping_callback.recalls, 'g-', label='Validation Macro Recall')
plt.xlabel('Epoch')
plt.ylabel('Macro Recall')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()

# --- 10. اختيار الإيبوك يدويًا ---
while True:
    try:
        selected_epoch = input(f"\nأدخل رقم الإيبوك الذي تريده للتقييم اليدوي (1 إلى {len(early_stopping_callback.recalls)}), أو 'q' للخروج: ")
        if selected_epoch.lower() == 'q':
            break
        selected_epoch = int(selected_epoch)
        if selected_epoch < 1 or selected_epoch > len(early_stopping_callback.recalls):
            raise ValueError("رقم الإيبوك خارج النطاق.")
    except ValueError as e:
        print(f"خطأ: {e}")
    else:
        model.set_weights(early_stopping_callback.weights_history[selected_epoch - 1])
        y_pred_manual = model.predict(X_test)
        y_pred_manual_final = (y_pred_manual >= best_threshold).astype(int).flatten()

        print(f"\nتقييم النموذج عند الإيبوك رقم {selected_epoch} باستخدام threshold = {best_threshold:.3f}:")
        print(classification_report(y_test, y_pred_manual_final))

        cm_manual = confusion_matrix(y_test, y_pred_manual_final)
        disp_manual = ConfusionMatrixDisplay(confusion_matrix=cm_manual, display_labels=['Sell (-1)', 'Buy (1)'])
        disp_manual.plot(cmap=plt.cm.Oranges)
        plt.title(f'Confusion Matrix at Epoch {selected_epoch}')
        plt.show()
```
