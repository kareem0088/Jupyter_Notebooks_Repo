```python
import pandas as pd

# المسار الخاص بالملف
file_path = r"C:\Users\Access\Downloads\EURUSD_Ticks.csv"

# قراءة البيانات من الملف
df = pd.read_csv(file_path)

```

```python
import pandas as pd

# قراءة البيانات من الملف
data = df

# إنشاء قائمة لتخزين الشموع اليابانية
candles = []

# كل 3 أسطر يتم تحويلها إلى شمعة
for i in range(0, len(data), 3):
    subset = data.iloc[i:i+3]
    
    if len(subset) == 3:
        open_price = subset['Ask'].iloc[0]  # أول سعر
        close_price = subset['Ask'].iloc[2]  # آخر سعر
        high_price = subset['Ask'].max()  # أعلى سعر
        low_price = subset['Ask'].min()  # أقل سعر
        
        open_time = subset['Gmt time'].iloc[0]  # أول وقت
        
        # حساب الفوليوم الإجمالي
        total_volume = subset['AskVolume'].sum() + subset['BidVolume'].sum()
        
        # إضافة الشمعة إلى القائمة
        candles.append({
            'Time': open_time,
            'Open': open_price,
            'High': high_price,
            'Low': low_price,
            'Close': close_price,
            'Volume': total_volume
        })

# تحويل قائمة الشموع إلى DataFrame
candles_df = pd.DataFrame(candles)

# حفظ الشموع في ملف جديد
candles_df.to_csv(r"C:\Users\Access\Downloads\EURUSD_Ticks.csv", index=False)

print(candles_df.head())
```

```python
# تقسيم البيانات إلى مجموعات كل مجموعة تحتوي على 3 TICK
df_grouped = df.groupby(df.index // 3).agg({
    'Gmt time': 'first',           # أول توقيت في المجموعة
    'Bid': ['first', 'max', 'min', 'last'],  # Open, High, Low, Close لـ Bid
    'AskVolume': 'sum',            # حجم التداول الإجمالي لـ Ask
    'BidVolume': 'sum'             # حجم التداول الإجمالي لـ Bid
}).reset_index(drop=True)
```

```python
df_grouped
```

```python
# حساب إجمالي الحجم
df_grouped['Volume'] = df_grouped['AskVolume'] + df_grouped['BidVolume']

# إعادة تسمية الأعمدة لتكون أكثر وضوحًا
df_grouped.columns = ['Gmt time', 'Open', 'High', 'Low', 'Close', 'Ask Volume', 'Bid Volume', 'Volume']

# حذف الأعمدة غير الضرورية
df_grouped = df_grouped.drop(columns=['Ask Volume', 'Bid Volume'])
```
