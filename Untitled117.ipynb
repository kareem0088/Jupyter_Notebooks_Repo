```python
import pandas as pd
import glob
import tqdm
import warnings
warnings.filterwarnings("ignore")

# --- دالة تجميع إلى 15 دقيقة ---
def resample_to_15min(df):
    df['Gmt time'] = pd.to_datetime(df['Gmt time'])
    df.set_index('Gmt time', inplace=True)

    df_15min = df.resample('15T').agg({
        'Open': 'first',
        'High': 'max',
        'Low': 'min',
        'Close': 'last',
        'Volume': 'sum'
    }).dropna()

    df_15min.reset_index(inplace=True)
    return df_15min

# --- دالة Heikin Ashi ---
def heikin_ashi(df):
    ha_df = pd.DataFrame()
    ha_df['Gmt time'] = df['Gmt time']
    ha_df['Close'] = (df['Open'] + df['High'] + df['Low'] + df['Close']) / 4

    ha_open = [(df['Open'].iloc[0] + df['Close'].iloc[0]) / 2]
    for i in range(1, len(df)):
        ha_open.append((ha_open[i - 1] + ha_df['Close'].iloc[i - 1]) / 2)

    ha_df['Open'] = ha_open
    ha_df['High'] = pd.concat([df['High'], ha_df['Open'], ha_df['Close']], axis=1).max(axis=1)
    ha_df['Low'] = pd.concat([df['Low'], ha_df['Open'], ha_df['Close']], axis=1).min(axis=1)
    ha_df['Volume'] = df['Volume']
    return ha_df[['Gmt time', 'Open', 'High', 'Low', 'Close']]



df_path = glob.glob(r'C:\Users\Access\Documents\DATA\XAUUSD\*.csv') 

def df_merg(df_path):
    dfs = [] 
    for file in tqdm.tqdm(df_path):
        df = pd.read_csv(file)
        dfs.append(df)
        df = pd.concat(dfs,ignore_index=True) 
        df['Gmt time'] = df['Gmt time'].str.replace(':00.000','' ,regex=False) #clean time
        df['Gmt time'] = pd.to_datetime(df['Gmt time'], format='%d.%m.%Y %H:%M') 
        df = df.sort_values(by='Gmt time') #sort gmt time 
        df = df[df['Open'] >= 100].reset_index(drop=True) #delet row if have value ander 100 in Open col
        df = df[~df['Gmt time'].dt.weekday.isin([5, 6])] #delet weekend
        df = df.reset_index(drop=True) #reset index
        #pd.to_csv(r'')
        df_15min = resample_to_15min(df)
        ha_df = heikin_ashi(df_15min)
        df = ha_df
    return df

df = df_merg(df_path)

df
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras import regularizers
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
from tensorflow.keras.optimizers import Adam
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt

# ========================
# 1. تحميل وتجهيز البيانات
# ========================
# df = df_

#df['Gmt time'] = pd.to_datetime(df['Gmt time'])
#df = df.drop(columns=['Gmt time'])  # <-- إذا كنت لا تستخدم الوقت كمُدخل

#features = df.drop(columns=['Signal', 'date', ''])  # <-- غيّر 'Signal' لاسم عمود الهدف لو مختلف
features = df[['return_24h','ADX_14','DMP_14','DMN_14']]
target = df['Signal']

# ========================
# 2. مقياس الميزات
# ========================
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)

# ========================
# 3. إنشاء النوافذ المنزلقة
# ========================
def create_sliding_window(X, y, window_size=50):
    Xs, ys = [], []
    for i in range(len(X) - window_size):
        if y[i + window_size] != 0:  # نحتفظ فقط بإشارات 1 و 2
            Xs.append(X[i:i+window_size])
            ys.append(y[i + window_size])
    return np.array(Xs), np.array(ys)

window_size = 50
X, y = create_sliding_window(features_scaled, target.values, window_size=window_size)
y_cat = to_categorical(y, num_classes=3)

# ========================
# 4. تقسيم البيانات
# ========================
X_train, X_test, y_train, y_test = train_test_split(X, y_cat, test_size=0.2, shuffle=False)

# ========================
# 5. بناء نموذج LSTM
# ========================

model = Sequential()
model.add(LSTM(
    16,
    input_shape=(X.shape[1], X.shape[2]),
    return_sequences=False,
    dropout=0.2,  # لحذف بعض المدخلات عشوائيًا
    kernel_regularizer=regularizers.l2(0.01)  # تقليل الحفظ
))
model.add(Dense(8, activation='relu', kernel_regularizer=regularizers.l2(0.01)))
model.add(Dropout(0.2))  # بعد Dense
model.add(Dense(3, activation='softmax'))

# ========================
# 5.1 تحديد نسبة التعلم
# ========================
optimizer = Adam(learning_rate=0.001)
model.compile(optimizer=optimizer, loss='categorical_crossentropy', metrics=['accuracy'])

# ========================
# 6.1 إعداد التوقف المبكر وحفظ أفضل نموذج
# ========================
early_stop = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
checkpoint = ModelCheckpoint('best_model.keras', monitor='val_loss', save_best_only=True)

# ========================
# 6. تدريب النموذج
# ========================
history = model.fit(
    X_train, y_train,
    epochs=200,
    batch_size=100,
    validation_data=(X_test, y_test),
    callbacks=[early_stop, checkpoint]
)

# ========================
# 7. تقييم تدريجي خطوة بخطوة (Walking Forward Evaluation)
# ========================
y_true = []
y_pred = []

for i in range(len(X_test)):
    sample_x = X_test[i].reshape(1, X_test.shape[1], X_test.shape[2])
    true_class = np.argmax(y_test[i])
    prediction = model.predict(sample_x, verbose=0)
    predicted_class = np.argmax(prediction)
    
    y_true.append(true_class)
    y_pred.append(predicted_class)

walk_acc = accuracy_score(y_true, y_pred)
print(f"🧠 Walking Forward Accuracy: {walk_acc:.2f}")

# ==================================================================================================================
# تقييم النموذج على اختبار حقيقي
# ==================================================================================================================
test_loss, test_accuracy = model.evaluate(X_test, y_test, verbose=0)

# ========================
#  رسم منحنى الأداء الكامل
# ========================
epochs_range = range(len(history.history['loss']))

plt.figure(figsize=(12, 6))

# رسم الخسارة
plt.plot(epochs_range, history.history['loss'], label='Train Loss', linestyle='--')
plt.plot(epochs_range, history.history['val_loss'], label='Validation Loss', linestyle='--')

# رسم الدقة
plt.plot(epochs_range, history.history['accuracy'], label='Train Accuracy', linestyle='-')
plt.plot(epochs_range, history.history['val_accuracy'], label='Validation Accuracy', linestyle='-')

plt.title('Training, Validation and Test Metrics')
plt.xlabel('Epoch')
plt.ylabel('Value')
plt.legend(loc='best')
plt.grid(True)
plt.tight_layout()
plt.show()

import numpy as np
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns
import matplotlib.pyplot as plt

# ===========
#  عمل التنبؤات
# ===========
y_pred_probs = model.predict(X_test)
y_pred = np.argmax(y_pred_probs, axis=1)
y_true = np.argmax(y_test, axis=1)

# ===========
#  مصفوفة الالتباس
# ===========
cm = confusion_matrix(y_true, y_pred)

# ===========
#  رسم المصفوفة
# ===========
plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=[1, 2], yticklabels=[1, 2])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()

# ===========
#  تقرير الأداء الكامل
# ===========
print("Classification Report:\n")
print(classification_report(y_true, y_pred, digits=3))


```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, RobustScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras import regularizers
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
from tensorflow.keras.optimizers import Adam
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.utils.class_weight import compute_class_weight

# ========================
# Function to create sliding windows
# ========================
def create_sliding_window(features, target, window_size):
    """
    Create sliding windows from time series data
    """
    X, y = [], []
    for i in range(len(features) - window_size):
        X.append(features[i:i + window_size])
        y.append(target[i + window_size])
    return np.array(X), np.array(y)



# Select features and target
features = df[['return_24h', 'ADX_14', 'DMP_14', 'DMN_14']]
target = df['Signal']

# 1. Create sliding windows first (before any scaling)
X_raw, y_raw = create_sliding_window(features.values, target.values, window_size=50)

# 2. Split windows into train and test (without shuffle to maintain temporal order)
X_train_raw, X_test_raw, y_train_raw, y_test_raw = train_test_split(
    X_raw, y_raw, test_size=0.1, shuffle=False
)

# 3. Apply scaling
scaler = RobustScaler()
# Fit scaler only on training data
X_train_scaled = scaler.fit_transform(
    X_train_raw.reshape(-1, X_train_raw.shape[-1])
).reshape(X_train_raw.shape)

# Transform test data using fitted scaler
X_test_scaled = scaler.transform(
    X_test_raw.reshape(-1, X_test_raw.shape[-1])
).reshape(X_test_raw.shape)

# 4. Convert targets to categorical
y_train_cat = to_categorical(y_train_raw, num_classes=3)
y_test_cat = to_categorical(y_test_raw, num_classes=3)

# ========================
# 5. Build LSTM Model
# ========================
model = Sequential()
model.add(LSTM(
    64,
    input_shape=(X_train_scaled.shape[1], X_train_scaled.shape[2]),
    return_sequences=False,
    dropout=0.2,
    kernel_regularizer=regularizers.l2(0.0001)
))
model.add(Dense(128, activation='relu', kernel_regularizer=regularizers.l2(0.0001)))
model.add(Dropout(0.2))
model.add(Dense(3, activation='softmax'))

# ========================
# 5.1 Set learning rate and compile
# ========================
optimizer = Adam(learning_rate=0.001)
model.compile(optimizer=optimizer, loss='categorical_crossentropy', metrics=['accuracy'])

# Display model summary
model.summary()

# ========================
# 6.1 Setup early stopping and model checkpoint
# ========================
early_stop = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
checkpoint = ModelCheckpoint('best_model.keras', monitor='val_loss', save_best_only=True)

# ========================
# 6. Train the model
# ========================

# نفترض y_train هي الأصلية (ليست one-hot)
classes = np.unique(y_train_raw)
class_weights = compute_class_weight(class_weight='balanced', classes=classes, y=y_train_raw)
class_weight_dict = dict(zip(classes, class_weights))

# طباعة الأوزان
print("Class weights:", class_weight_dict)

history = model.fit(
    X_train_scaled, y_train_cat,
    epochs=20,
    batch_size=100,
    validation_data=(X_test_scaled, y_test_cat),
    callbacks=[early_stop, checkpoint],
    verbose=1,
    class_weight=class_weight_dict
)

# ========================
# 7. Walking Forward Evaluation
# ========================
y_true_walk = []
y_pred_walk = []

for i in tqdm.tqdm(range(len(X_test_scaled))):
    sample_x = X_test_scaled[i].reshape(1, X_test_scaled.shape[1], X_test_scaled.shape[2])
    true_class = np.argmax(y_test_cat[i])
    prediction = model.predict(sample_x, verbose=0)
    predicted_class = np.argmax(prediction)
    
    y_true_walk.append(true_class)
    y_pred_walk.append(predicted_class)

walk_acc = accuracy_score(y_true_walk, y_pred_walk)
print(f"🧠 Walking Forward Accuracy: {walk_acc:.4f}")

# ========================
# 8. Model Evaluation
# ========================
test_loss, test_accuracy = model.evaluate(X_test_scaled, y_test_cat, verbose=0)
print(f"📊 Test Loss: {test_loss:.4f}")
print(f"📊 Test Accuracy: {test_accuracy:.4f}")

# ========================
# 9. Plot training history
# ========================
epochs_range = range(len(history.history['loss']))

plt.figure(figsize=(12, 6))

# Plot loss
plt.subplot(1, 2, 1)
plt.plot(epochs_range, history.history['loss'], label='Train Loss', linestyle='--')
plt.plot(epochs_range, history.history['val_loss'], label='Validation Loss', linestyle='--')
plt.title('Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)

# Plot accuracy
plt.subplot(1, 2, 2)
plt.plot(epochs_range, history.history['accuracy'], label='Train Accuracy', linestyle='-')
plt.plot(epochs_range, history.history['val_accuracy'], label='Validation Accuracy', linestyle='-')
plt.title('Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()

# ========================
# 10. Detailed Evaluation
# ========================
# Make predictions
y_pred_probs = model.predict(X_test_scaled)
y_pred = np.argmax(y_pred_probs, axis=1)
y_true = np.argmax(y_test_cat, axis=1)

# Confusion Matrix
cm = confusion_matrix(y_true, y_pred)

# Plot confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
            xticklabels=['Class 0', 'Class 1', 'Class 2'], 
            yticklabels=['Class 0', 'Class 1', 'Class 2'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()

# Classification Report
print("\n" + "="*50)
print("CLASSIFICATION REPORT")
print("="*50)
print(classification_report(y_true, y_pred, digits=4))

# ========================
# 11. Prediction Confidence Analysis
# ========================
# Analyze prediction confidence
confidence_scores = np.max(y_pred_probs, axis=1)
print(f"\n📈 Average Prediction Confidence: {np.mean(confidence_scores):.4f}")
print(f"📈 Min Confidence: {np.min(confidence_scores):.4f}")
print(f"📈 Max Confidence: {np.max(confidence_scores):.4f}")

# Plot confidence distribution
plt.figure(figsize=(10, 6))
plt.hist(confidence_scores, bins=50, alpha=0.7, edgecolor='black')
plt.xlabel('Prediction Confidence')
plt.ylabel('Frequency')
plt.title('Distribution of Prediction Confidence Scores')
plt.grid(True, alpha=0.3)
plt.show()
```
