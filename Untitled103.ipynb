```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import classification_report, confusion_matrix
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout, BatchNormalization
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from tensorflow.keras.utils import to_categorical
import matplotlib.pyplot as plt
import seaborn as sns

class LSTMSignalClassifier:
    def __init__(self, window_size=60, lstm_units=50, dropout_rate=0.2):
        """
        LSTM Model للتصنيف
        
        Parameters:
        window_size: حجم النافذة المنزلقة
        lstm_units: عدد وحدات LSTM
        dropout_rate: معدل الـ dropout
        """
        self.window_size = window_size
        self.lstm_units = lstm_units
        self.dropout_rate = dropout_rate
        self.model = None
        self.scaler = StandardScaler()
        self.label_encoder = LabelEncoder()
        
    def prepare_data(self, df):
        """
        تحضير البيانات مع النافذة المنزلقة
        """
        # فصل الميزات عن الهدف (استبعاد عمود الوقت والهدف)
        feature_columns = [col for col in df.columns if col not in ['Signal', 'Gmt time']]
        X = df[feature_columns].values
        y = df['Signal'].values
        
        # تطبيق التطبيع على الميزات
        X_scaled = self.scaler.fit_transform(X)
        
        # تطبيق Label Encoding على الهدف
        y_encoded = self.label_encoder.fit_transform(y)
        
        # إنشاء النافذة المنزلقة
        X_windowed, y_windowed = self.create_sliding_window(X_scaled, y_encoded)
        
        return X_windowed, y_windowed
    
    def create_sliding_window(self, X, y):
        """
        إنشاء النافذة المنزلقة للبيانات الزمنية
        """
        X_windowed = []
        y_windowed = []
        
        for i in range(self.window_size, len(X)):
            X_windowed.append(X[i-self.window_size:i])
            y_windowed.append(y[i])
        
        return np.array(X_windowed), np.array(y_windowed)
    
    def build_model(self, input_shape, num_classes):
        """
        بناء نموذج LSTM
        """
        model = Sequential([
            # الطبقة الأولى LSTM
            LSTM(self.lstm_units, 
                 return_sequences=True, 
                 input_shape=input_shape,
                 dropout=self.dropout_rate,
                 recurrent_dropout=self.dropout_rate),
            BatchNormalization(),
            
            # الطبقة الثانية LSTM
            LSTM(self.lstm_units//2, 
                 return_sequences=False,
                 dropout=self.dropout_rate,
                 recurrent_dropout=self.dropout_rate),
            BatchNormalization(),
            
            # طبقات Dense
            Dense(32, activation='relu'),
            Dropout(self.dropout_rate),
            Dense(16, activation='relu'),
            Dropout(self.dropout_rate),
            
            # طبقة الإخراج
            Dense(num_classes, activation='softmax')
        ])
        
        model.compile(
            optimizer='adam',
            loss='sparse_categorical_crossentropy',
            metrics=['accuracy']
        )
        
        self.model = model
        return model
    
    def temporal_train_test_split(self, X, y, train_ratio=0.7, val_ratio=0.15):
        """
        تقسيم البيانات الزمنية بشكل تسلسلي (بدون عشوائية)
        """
        n_samples = len(X)
        
        # حساب نقاط التقسيم
        train_end = int(n_samples * train_ratio)
        val_end = int(n_samples * (train_ratio + val_ratio))
        
        # تقسيم البيانات تسلسليا
        X_train = X[:train_end]
        y_train = y[:train_end]
        
        X_val = X[train_end:val_end]
        y_val = y[train_end:val_end]
        
        X_test = X[val_end:]
        y_test = y[val_end:]
        
        return X_train, X_val, X_test, y_train, y_val, y_test
    
    def train(self, X_train, y_train, X_val, y_val, epochs=100, batch_size=32):
        """
        تدريب النموذج
        """
        # إعداد callbacks
        early_stopping = EarlyStopping(
            monitor='val_loss',
            patience=15,
            restore_best_weights=True
        )
        
        reduce_lr = ReduceLROnPlateau(
            monitor='val_loss',
            factor=0.2,
            patience=10,
            min_lr=0.0001
        )
        
        # تدريب النموذج
        history = self.model.fit(
            X_train, y_train,
            epochs=epochs,
            batch_size=batch_size,
            validation_data=(X_val, y_val),
            callbacks=[early_stopping, reduce_lr],
            verbose=1,
            shuffle=False  # مهم: عدم خلط البيانات الزمنية
        )
        
        return history
    
    def predict(self, X):
        """
        التنبؤ
        """
        predictions = self.model.predict(X)
        predicted_classes = np.argmax(predictions, axis=1)
        return self.label_encoder.inverse_transform(predicted_classes)
    
    def evaluate_model(self, X_test, y_test):
        """
        تقييم النموذج
        """
        predictions = self.predict(X_test)
        y_test_labels = self.label_encoder.inverse_transform(y_test)
        
        print("Classification Report:")
        print(classification_report(y_test_labels, predictions))
        
        # رسم confusion matrix
        cm = confusion_matrix(y_test_labels, predictions)
        plt.figure(figsize=(8, 6))
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
                   xticklabels=self.label_encoder.classes_,
                   yticklabels=self.label_encoder.classes_)
        plt.title('Confusion Matrix')
        plt.xlabel('Predicted')
        plt.ylabel('Actual')
        plt.show()
        
        return classification_report(y_test_labels, predictions, output_dict=True)
    
    def plot_training_history(self, history):
        """
        رسم تاريخ التدريب
        """
        fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))
        
        # Loss
        ax1.plot(history.history['loss'], label='Training Loss')
        ax1.plot(history.history['val_loss'], label='Validation Loss')
        ax1.set_title('Model Loss')
        ax1.set_xlabel('Epoch')
        ax1.set_ylabel('Loss')
        ax1.legend()
        
        # Accuracy
        ax2.plot(history.history['accuracy'], label='Training Accuracy')
        ax2.plot(history.history['val_accuracy'], label='Validation Accuracy')
        ax2.set_title('Model Accuracy')
        ax2.set_xlabel('Epoch')
        ax2.set_ylabel('Accuracy')
        ax2.legend()
        
    def plot_predictions_timeline(self, y_true, y_pred, start_idx=0, end_idx=200):
        """
        رسم الإشارات الفعلية مقابل المتوقعة على الخط الزمني
        """
        plt.figure(figsize=(15, 6))
        
        time_range = range(start_idx, min(end_idx, len(y_true)))
        
        plt.plot(time_range, y_true[start_idx:end_idx], 'b-', label='Actual Signal', alpha=0.7)
        plt.plot(time_range, y_pred[start_idx:end_idx], 'r--', label='Predicted Signal', alpha=0.7)
        
        plt.xlabel('Time Step')
        plt.ylabel('Signal Value')
        plt.title('Signal Prediction Timeline')
        plt.legend()
        plt.grid(True, alpha=0.3)
        plt.tight_layout()
        plt.show()
        
    def calculate_temporal_metrics(self, y_true, y_pred):
        """
        حساب مقاييس خاصة بالبيانات الزمنية
        """
        # دقة التتابع (accuracy of consecutive predictions)
        consecutive_accuracy = np.mean(y_true == y_pred)
        
        # دقة اتجاه التغيير
        true_changes = np.diff(y_true)
        pred_changes = np.diff(y_pred)
        direction_accuracy = np.mean(np.sign(true_changes) == np.sign(pred_changes))
        
        print(f"Consecutive Accuracy: {consecutive_accuracy:.4f}")
        print(f"Direction Change Accuracy: {direction_accuracy:.4f}")
        
        return {
            'consecutive_accuracy': consecutive_accuracy,
            'direction_accuracy': direction_accuracy
        }

# مثال على الاستخدام مع بياناتك الحقيقية
def main():
    """
    استخدام النموذج مع بياناتك الحقيقية
    """
    # تحميل البيانات (استبدل بمسار ملفك)
    df = df_new_.copy()
    
    # التأكد من الترتيب الزمني
    if 'Gmt time' in df.columns:
        df['Gmt time'] = pd.to_datetime(df['Gmt time'])
        df = df.sort_values('Gmt time').reset_index(drop=True)
    
    print(f"Dataset shape: {df.shape}")
    print(f"Signal distribution:\n{df['Signal'].value_counts()}")
    print(f"Feature columns: {[col for col in df.columns if col not in ['Signal', 'Gmt time']]}")
    
    # إنشاء وتدريب النموذج
    classifier = LSTMSignalClassifier(window_size=60, lstm_units=50)
    
    # تحضير البيانات
    X, y = classifier.prepare_data(df)
    
    print(f"Shape of X after windowing: {X.shape}")
    print(f"Shape of y after windowing: {y.shape}")
    print(f"Unique classes: {classifier.label_encoder.classes_}")
    
    # تقسيم البيانات تسلسليا (بدون عشوائية)
    X_train, X_val, X_test, y_train, y_val, y_test = classifier.temporal_train_test_split(
        X, y, train_ratio=0.7, val_ratio=0.15
    )
    
    print(f"Training set: {X_train.shape}")
    print(f"Validation set: {X_val.shape}")
    print(f"Test set: {X_test.shape}")
    
    # بناء النموذج
    model = classifier.build_model(
        input_shape=(X_train.shape[1], X_train.shape[2]),
        num_classes=len(np.unique(y))
    )
    
    print(model.summary())
    
    # تدريب النموذج
    history = classifier.train(X_train, y_train, X_val, y_val, epochs=100)
    
    # رسم تاريخ التدريب
    classifier.plot_training_history(history)
    
    # تقييم النموذج
    results = classifier.evaluate_model(X_test, y_test)
    
    # تقييم إضافي للبيانات الزمنية
    y_pred = classifier.predict(X_test)
    y_true_labels = classifier.label_encoder.inverse_transform(y_test)
    
    # حساب مقاييس زمنية إضافية
    temporal_metrics = classifier.calculate_temporal_metrics(y_true_labels, y_pred)
    
    # رسم الخط الزمني للتنبؤات
    classifier.plot_predictions_timeline(y_true_labels, y_pred, start_idx=0, end_idx=100)
    
    # حفظ النموذج
    classifier.model.save('lstm_signal_classifier.h5')
    print("Model saved as 'lstm_signal_classifier.h5'")
    
    return classifier, results, temporal_metrics

if __name__ == "__main__":
    classifier, results, temporal_metrics = main()
```

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import classification_report, confusion_matrix
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout, BatchNormalization
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from tensorflow.keras.utils import to_categorical
import matplotlib.pyplot as plt
import seaborn as sns

class LSTMSignalClassifier:
    def __init__(self, window_size=96, lstm_units=50, dropout_rate=0.2):
        """
        LSTM Model للتصنيف
        
        Parameters:
        window_size: حجم النافذة المنزلقة
        lstm_units: عدد وحدات LSTM
        dropout_rate: معدل الـ dropout
        """
        self.window_size = window_size
        self.lstm_units = lstm_units
        self.dropout_rate = dropout_rate
        self.model = None
        self.scaler = StandardScaler()
        self.label_encoder = LabelEncoder()
        
    def prepare_data(self, df):
        """
        تحضير البيانات مع النافذة المنزلقة
        """
        # فصل الميزات عن الهدف (استبعاد عمود الوقت والهدف)
        feature_columns = [col for col in df.columns if col not in ['Signal', 'Gmt time']]
        X = df[feature_columns].values
        y = df['Signal'].values
        
        # تطبيق التطبيع على الميزات
        X_scaled = self.scaler.fit_transform(X)
        
        # تطبيق Label Encoding على الهدف
        y_encoded = self.label_encoder.fit_transform(y)
        
        # إنشاء النافذة المنزلقة
        X_windowed, y_windowed = self.create_sliding_window(X_scaled, y_encoded)
        
        return X_windowed, y_windowed
    
    def create_sliding_window(self, X, y):
        """
        إنشاء النافذة المنزلقة للبيانات الزمنية
        """
        X_windowed = []
        y_windowed = []
        
        for i in range(self.window_size, len(X)):
            X_windowed.append(X[i-self.window_size:i])
            y_windowed.append(y[i])
        
        return np.array(X_windowed), np.array(y_windowed)
    
    def build_model(self, input_shape, num_classes):
        """
        بناء نموذج LSTM
        """
        model = Sequential([
            # الطبقة الأولى LSTM
            LSTM(self.lstm_units, 
                 return_sequences=True, 
                 input_shape=input_shape,
                 #dropout=self.dropout_rate,
                 #recurrent_dropout=self.dropout_rate
                ),
            BatchNormalization(),
            
            # الطبقة الثانية LSTM
            LSTM(self.lstm_units//2, 
                 return_sequences=False #,
                 #dropout=self.dropout_rate,
                 #recurrent_dropout=self.dropout_rate
                ),
            BatchNormalization(),
            
            # طبقات Dense
            Dense(32, activation='relu'),
            Dropout(self.dropout_rate),
            Dense(16, activation='relu'),
            #Dropout(self.dropout_rate),
            
            # طبقة الإخراج
            Dense(num_classes, activation='softmax')
        ])
        
        model.compile(
            optimizer='adam',
            loss='sparse_categorical_crossentropy',
            metrics=['accuracy']
        )
        
        self.model = model
        return model
    
    def temporal_train_test_split(self, X, y, train_ratio=0.7, val_ratio=0.15):
        """
        تقسيم البيانات الزمنية بشكل تسلسلي (بدون عشوائية)
        """
        n_samples = len(X)
        
        # حساب نقاط التقسيم
        train_end = int(n_samples * train_ratio)
        val_end = int(n_samples * (train_ratio + val_ratio))
        
        # تقسيم البيانات تسلسليا
        X_train = X[:train_end]
        y_train = y[:train_end]
        
        X_val = X[train_end:val_end]
        y_val = y[train_end:val_end]
        
        X_test = X[val_end:]
        y_test = y[val_end:]
        
        return X_train, X_val, X_test, y_train, y_val, y_test
    
    def train(self, X_train, y_train, X_val, y_val, epochs=10, batch_size=100):
        """
        تدريب النموذج
        """
        # إعداد callbacks
        early_stopping = EarlyStopping(
            monitor='val_loss',
            patience=5,
            restore_best_weights=True
        )
        
        reduce_lr = ReduceLROnPlateau(
            monitor='val_loss',
            factor=0.2,
            patience=10,
            min_lr=0.001
        )
        
        # تدريب النموذج
        history = self.model.fit(
            X_train, y_train,
            epochs=epochs,
            batch_size=batch_size,
            validation_data=(X_val, y_val),
            callbacks=[early_stopping, reduce_lr],
            verbose=1,
            shuffle=False  # مهم: عدم خلط البيانات الزمنية
        )
        
        return history
    
    def predict(self, X):
        """
        التنبؤ
        """
        predictions = self.model.predict(X)
        predicted_classes = np.argmax(predictions, axis=1)
        return self.label_encoder.inverse_transform(predicted_classes)
    
    def evaluate_model(self, X_test, y_test):
        """
        تقييم النموذج
        """
        predictions = self.predict(X_test)
        y_test_labels = self.label_encoder.inverse_transform(y_test)
        
        print("Classification Report:")
        print(classification_report(y_test_labels, predictions))
        
        # رسم confusion matrix
        cm = confusion_matrix(y_test_labels, predictions)
        plt.figure(figsize=(8, 6))
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
                   xticklabels=self.label_encoder.classes_,
                   yticklabels=self.label_encoder.classes_)
        plt.title('Confusion Matrix')
        plt.xlabel('Predicted')
        plt.ylabel('Actual')
        plt.show()
        
        return classification_report(y_test_labels, predictions, output_dict=True)
    
    def plot_training_history(self, history):
        """
        رسم تاريخ التدريب
        """
        fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))
        
        # Loss
        ax1.plot(history.history['loss'], label='Training Loss')
        ax1.plot(history.history['val_loss'], label='Validation Loss')
        ax1.set_title('Model Loss')
        ax1.set_xlabel('Epoch')
        ax1.set_ylabel('Loss')
        ax1.legend()
        
        # Accuracy
        ax2.plot(history.history['accuracy'], label='Training Accuracy')
        ax2.plot(history.history['val_accuracy'], label='Validation Accuracy')
        ax2.set_title('Model Accuracy')
        ax2.set_xlabel('Epoch')
        ax2.set_ylabel('Accuracy')
        ax2.legend()
        
    def plot_predictions_timeline(self, y_true, y_pred, start_idx=0, end_idx=200):
        """
        رسم الإشارات الفعلية مقابل المتوقعة على الخط الزمني
        """
        plt.figure(figsize=(15, 6))
        
        time_range = range(start_idx, min(end_idx, len(y_true)))
        
        plt.plot(time_range, y_true[start_idx:end_idx], 'b-', label='Actual Signal', alpha=0.7)
        plt.plot(time_range, y_pred[start_idx:end_idx], 'r--', label='Predicted Signal', alpha=0.7)
        
        plt.xlabel('Time Step')
        plt.ylabel('Signal Value')
        plt.title('Signal Prediction Timeline')
        plt.legend()
        plt.grid(True, alpha=0.3)
        plt.tight_layout()
        plt.show()
        
    def calculate_temporal_metrics(self, y_true, y_pred):
        """
        حساب مقاييس خاصة بالبيانات الزمنية
        """
        # دقة التتابع (accuracy of consecutive predictions)
        consecutive_accuracy = np.mean(y_true == y_pred)
        
        # دقة اتجاه التغيير
        true_changes = np.diff(y_true)
        pred_changes = np.diff(y_pred)
        direction_accuracy = np.mean(np.sign(true_changes) == np.sign(pred_changes))
        
        print(f"Consecutive Accuracy: {consecutive_accuracy:.4f}")
        print(f"Direction Change Accuracy: {direction_accuracy:.4f}")
        
        return {
            'consecutive_accuracy': consecutive_accuracy,
            'direction_accuracy': direction_accuracy
        }

# مثال على الاستخدام مع بياناتك الحقيقية
def main():
    """
    استخدام النموذج مع بياناتك الحقيقية
    """
    # تحميل البيانات (استبدل بمسار ملفك)
    df = df_new_.copy()
    
    # التأكد من الترتيب الزمني
    if 'Gmt time' in df.columns:
        df['Gmt time'] = pd.to_datetime(df['Gmt time'])
        df = df.sort_values('Gmt time').reset_index(drop=True)
    
    print(f"Dataset shape: {df.shape}")
    print(f"Signal distribution:\n{df['Signal'].value_counts()}")
    print(f"Feature columns: {[col for col in df.columns if col not in ['Signal', 'Gmt time']]}")
    
    # إنشاء وتدريب النموذج
    classifier = LSTMSignalClassifier(window_size=96, lstm_units=50)
    
    # تحضير البيانات
    X, y = classifier.prepare_data(df)
    
    print(f"Shape of X after windowing: {X.shape}")
    print(f"Shape of y after windowing: {y.shape}")
    print(f"Unique classes: {classifier.label_encoder.classes_}")
    
    # تقسيم البيانات تسلسليا (بدون عشوائية)
    X_train, X_val, X_test, y_train, y_val, y_test = classifier.temporal_train_test_split(
        X, y, train_ratio=0.7, val_ratio=0.15
    )
    
    print(f"Training set: {X_train.shape}")
    print(f"Validation set: {X_val.shape}")
    print(f"Test set: {X_test.shape}")
    
    # بناء النموذج
    model = classifier.build_model(
        input_shape=(X_train.shape[1], X_train.shape[2]),
        num_classes=len(np.unique(y))
    )
    
    print(model.summary())
    
    # تدريب النموذج
    history = classifier.train(X_train, y_train, X_val, y_val, epochs=10)
    
    # رسم تاريخ التدريب
    classifier.plot_training_history(history)
    
    # تقييم النموذج
    results = classifier.evaluate_model(X_test, y_test)
    
    # تقييم إضافي للبيانات الزمنية
    y_pred = classifier.predict(X_test)
    y_true_labels = classifier.label_encoder.inverse_transform(y_test)
    
    # حساب مقاييس زمنية إضافية
    temporal_metrics = classifier.calculate_temporal_metrics(y_true_labels, y_pred)
    
    # رسم الخط الزمني للتنبؤات
    classifier.plot_predictions_timeline(y_true_labels, y_pred, start_idx=0, end_idx=100)
    
    # حفظ النموذج
    classifier.model.save('lstm_signal_classifier_1.h5')
    print("Model saved as 'lstm_signal_classifier_1.h5'")
    
    return classifier, results, temporal_metrics

if __name__ == "__main__":
    classifier, results, temporal_metrics = main()
```

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import classification_report, confusion_matrix
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout, BatchNormalization
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from tensorflow.keras.utils import to_categorical
import matplotlib.pyplot as plt
import seaborn as sns

class LSTMSignalClassifier:
    def __init__(self, window_size=96, lstm_units=50, dropout_rate=0.2):
        """
        LSTM Model للتصنيف
        
        Parameters:
        window_size: حجم النافذة المنزلقة
        lstm_units: عدد وحدات LSTM
        dropout_rate: معدل الـ dropout
        """
        self.window_size = window_size
        self.lstm_units = lstm_units
        self.dropout_rate = dropout_rate
        self.model = None
        self.scaler = StandardScaler()
        self.label_encoder = LabelEncoder()
        
    def prepare_data(self, df):
        """
        تحضير البيانات مع النافذة المنزلقة
        """
        # فصل الميزات عن الهدف (استبعاد عمود الوقت والهدف)
        feature_columns = [col for col in df.columns if col not in ['Signal', 'Gmt time']]
        X = df[feature_columns].values
        y = df['Signal'].values
        
        # تطبيق التطبيع على الميزات
        X_scaled = self.scaler.fit_transform(X)
        
        # تطبيق Label Encoding على الهدف
        y_encoded = self.label_encoder.fit_transform(y)
        
        # إنشاء النافذة المنزلقة
        X_windowed, y_windowed = self.create_sliding_window(X_scaled, y_encoded)
        
        return X_windowed, y_windowed
    
    def create_sliding_window(self, X, y):
        """
        إنشاء النافذة المنزلقة للبيانات الزمنية
        """
        X_windowed = []
        y_windowed = []
        
        for i in range(self.window_size, len(X)):
            X_windowed.append(X[i-self.window_size:i])
            y_windowed.append(y[i])
        
        return np.array(X_windowed), np.array(y_windowed)
    
    def build_model(self, input_shape, num_classes):
        """
        بناء نموذج LSTM
        """
        model = Sequential([
            # الطبقة الأولى LSTM
            LSTM(self.lstm_units, 
                 return_sequences=True, 
                 input_shape=input_shape,
                 #dropout=self.dropout_rate,
                 #recurrent_dropout=self.dropout_rate
                ),
            BatchNormalization(),
            
            # الطبقة الثانية LSTM
            LSTM(self.lstm_units//2, 
                 return_sequences=False #,
                 #dropout=self.dropout_rate,
                 #recurrent_dropout=self.dropout_rate
                ),
            BatchNormalization(),
            
            # طبقات Dense
            Dense(32, activation='relu'),
            Dropout(self.dropout_rate),
            Dense(16, activation='relu'),
            #Dropout(self.dropout_rate),
            
            # طبقة الإخراج
            Dense(num_classes, activation='softmax')
        ])
        
        model.compile(
            optimizer='adam',
            loss='sparse_categorical_crossentropy',
            metrics=['accuracy']
        )
        
        self.model = model
        return model
    
    def temporal_train_test_split(self, X, y, train_ratio=0.7, val_ratio=0.15):
        """
        تقسيم البيانات الزمنية بشكل تسلسلي (بدون عشوائية)
        """
        n_samples = len(X)
        
        # حساب نقاط التقسيم
        train_end = int(n_samples * train_ratio)
        val_end = int(n_samples * (train_ratio + val_ratio))
        
        # تقسيم البيانات تسلسليا
        X_train = X[:train_end]
        y_train = y[:train_end]
        
        X_val = X[train_end:val_end]
        y_val = y[train_end:val_end]
        
        X_test = X[val_end:]
        y_test = y[val_end:]
        
        return X_train, X_val, X_test, y_train, y_val, y_test
    
    def train(self, X_train, y_train, X_val, y_val, epochs=10, batch_size=100):
        """
        تدريب النموذج
        """
        # إعداد callbacks
        early_stopping = EarlyStopping(
            monitor='val_loss',
            patience=5,
            restore_best_weights=True
        )
        
        reduce_lr = ReduceLROnPlateau(
            monitor='val_loss',
            factor=0.2,
            patience=10,
            min_lr=0.001
        )
        
        # تدريب النموذج
        history = self.model.fit(
            X_train, y_train,
            epochs=epochs,
            batch_size=batch_size,
            validation_data=(X_val, y_val),
            callbacks=[early_stopping, reduce_lr],
            verbose=1,
            shuffle=False  # مهم: عدم خلط البيانات الزمنية
        )
        
        return history
    
    def predict(self, X):
        """
        التنبؤ
        """
        predictions = self.model.predict(X)
        predicted_classes = np.argmax(predictions, axis=1)
        return self.label_encoder.inverse_transform(predicted_classes)
    
    def evaluate_model(self, X_test, y_test):
        """
        تقييم النموذج
        """
        predictions = self.predict(X_test)
        y_test_labels = self.label_encoder.inverse_transform(y_test)
        
        print("Classification Report:")
        print(classification_report(y_test_labels, predictions))
        
        # رسم confusion matrix
        cm = confusion_matrix(y_test_labels, predictions)
        plt.figure(figsize=(8, 6))
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
                   xticklabels=self.label_encoder.classes_,
                   yticklabels=self.label_encoder.classes_)
        plt.title('Confusion Matrix')
        plt.xlabel('Predicted')
        plt.ylabel('Actual')
        plt.show()
        
        return classification_report(y_test_labels, predictions, output_dict=True)
    
    def plot_training_history(self, history):
        """
        رسم تاريخ التدريب
        """
        fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))
        
        # Loss
        ax1.plot(history.history['loss'], label='Training Loss')
        ax1.plot(history.history['val_loss'], label='Validation Loss')
        ax1.set_title('Model Loss')
        ax1.set_xlabel('Epoch')
        ax1.set_ylabel('Loss')
        ax1.legend()
        
        # Accuracy
        ax2.plot(history.history['accuracy'], label='Training Accuracy')
        ax2.plot(history.history['val_accuracy'], label='Validation Accuracy')
        ax2.set_title('Model Accuracy')
        ax2.set_xlabel('Epoch')
        ax2.set_ylabel('Accuracy')
        ax2.legend()
        
    def plot_predictions_timeline(self, y_true, y_pred, start_idx=0, end_idx=200):
        """
        رسم الإشارات الفعلية مقابل المتوقعة على الخط الزمني
        """
        plt.figure(figsize=(15, 6))
        
        time_range = range(start_idx, min(end_idx, len(y_true)))
        
        plt.plot(time_range, y_true[start_idx:end_idx], 'b-', label='Actual Signal', alpha=0.7)
        plt.plot(time_range, y_pred[start_idx:end_idx], 'r--', label='Predicted Signal', alpha=0.7)
        
        plt.xlabel('Time Step')
        plt.ylabel('Signal Value')
        plt.title('Signal Prediction Timeline')
        plt.legend()
        plt.grid(True, alpha=0.3)
        plt.tight_layout()
        plt.show()
        
    def calculate_temporal_metrics(self, y_true, y_pred):
        """
        حساب مقاييس خاصة بالبيانات الزمنية
        """
        # دقة التتابع (accuracy of consecutive predictions)
        consecutive_accuracy = np.mean(y_true == y_pred)
        
        # دقة اتجاه التغيير
        true_changes = np.diff(y_true)
        pred_changes = np.diff(y_pred)
        direction_accuracy = np.mean(np.sign(true_changes) == np.sign(pred_changes))
        
        print(f"Consecutive Accuracy: {consecutive_accuracy:.4f}")
        print(f"Direction Change Accuracy: {direction_accuracy:.4f}")
        
        return {
            'consecutive_accuracy': consecutive_accuracy,
            'direction_accuracy': direction_accuracy
        }

# مثال على الاستخدام مع بياناتك الحقيقية
def main():
    """
    استخدام النموذج مع بياناتك الحقيقية
    """
    # تحميل البيانات (استبدل بمسار ملفك)
    df = df_new_3.copy()
    
    # التأكد من الترتيب الزمني
    if 'Gmt time' in df.columns:
        df['Gmt time'] = pd.to_datetime(df['Gmt time'])
        df = df.sort_values('Gmt time').reset_index(drop=True)
    
    print(f"Dataset shape: {df.shape}")
    print(f"Signal distribution:\n{df['Signal'].value_counts()}")
    print(f"Feature columns: {[col for col in df.columns if col not in ['Signal', 'Gmt time']]}")
    
    # إنشاء وتدريب النموذج
    classifier = LSTMSignalClassifier(window_size=96, lstm_units=50)
    
    # تحضير البيانات
    X, y = classifier.prepare_data(df)
    
    print(f"Shape of X after windowing: {X.shape}")
    print(f"Shape of y after windowing: {y.shape}")
    print(f"Unique classes: {classifier.label_encoder.classes_}")
    
    # تقسيم البيانات تسلسليا (بدون عشوائية)
    X_train, X_val, X_test, y_train, y_val, y_test = classifier.temporal_train_test_split(
        X, y, train_ratio=0.7, val_ratio=0.15
    )
    
    print(f"Training set: {X_train.shape}")
    print(f"Validation set: {X_val.shape}")
    print(f"Test set: {X_test.shape}")
    
    # بناء النموذج
    model = classifier.build_model(
        input_shape=(X_train.shape[1], X_train.shape[2]),
        num_classes=len(np.unique(y))
    )
    
    print(model.summary())
    
    # تدريب النموذج
    history = classifier.train(X_train, y_train, X_val, y_val, epochs=10)
    
    # رسم تاريخ التدريب
    classifier.plot_training_history(history)
    
    # تقييم النموذج
    results = classifier.evaluate_model(X_test, y_test)
    
    # تقييم إضافي للبيانات الزمنية
    y_pred = classifier.predict(X_test)
    y_true_labels = classifier.label_encoder.inverse_transform(y_test)
    
    # حساب مقاييس زمنية إضافية
    temporal_metrics = classifier.calculate_temporal_metrics(y_true_labels, y_pred)
    
    # رسم الخط الزمني للتنبؤات
    classifier.plot_predictions_timeline(y_true_labels, y_pred, start_idx=0, end_idx=100)
    
    # حفظ النموذج
    classifier.model.save('lstm_signal_classifier_1.h5')
    print("Model saved as 'lstm_signal_classifier_1.h5'")
    
    return classifier, results, temporal_metrics

if __name__ == "__main__":
    classifier, results, temporal_metrics = main()
```

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.ensemble import RandomForestClassifier
from sklearn.inspection import permutation_importance
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout, BatchNormalization
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from tensorflow.keras.utils import to_categorical
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr
import warnings
warnings.filterwarnings('ignore')

class LSTMSignalClassifier:
    def __init__(self, window_size=60, lstm_units=50, dropout_rate=0.2):
        """
        LSTM Model للتصنيف
        
        Parameters:
        window_size: حجم النافذة المنزلقة
        lstm_units: عدد وحدات LSTM
        dropout_rate: معدل الـ dropout
        """
        self.window_size = window_size
        self.lstm_units = lstm_units
        self.dropout_rate = dropout_rate
        self.model = None
        self.scaler = StandardScaler()
        self.label_encoder = LabelEncoder()
        
    def prepare_data(self, df):
        """
        تحضير البيانات مع النافذة المنزلقة
        """
        # فصل الميزات عن الهدف (استبعاد عمود الوقت والهدف)
        feature_columns = [col for col in df.columns if col not in ['Signal', 'Gmt time']]
        X = df[feature_columns].values
        y = df['Signal'].values
        
        # تطبيق التطبيع على الميزات
        X_scaled = self.scaler.fit_transform(X)
        
        # تطبيق Label Encoding على الهدف
        y_encoded = self.label_encoder.fit_transform(y)
        
        # إنشاء النافذة المنزلقة
        X_windowed, y_windowed = self.create_sliding_window(X_scaled, y_encoded)
        
        return X_windowed, y_windowed
    
    def create_sliding_window(self, X, y):
        """
        إنشاء النافذة المنزلقة للبيانات الزمنية
        """
        X_windowed = []
        y_windowed = []
        
        for i in range(self.window_size, len(X)):
            X_windowed.append(X[i-self.window_size:i])
            y_windowed.append(y[i])
        
        return np.array(X_windowed), np.array(y_windowed)
    
    def build_model(self, input_shape, num_classes):
        """
        بناء نموذج LSTM
        """
        model = Sequential([
            # الطبقة الأولى LSTM
            LSTM(self.lstm_units, 
                 return_sequences=True, 
                 input_shape=input_shape,
                 dropout=self.dropout_rate,
                 recurrent_dropout=self.dropout_rate),
            BatchNormalization(),
            
            # الطبقة الثانية LSTM
            LSTM(self.lstm_units//2, 
                 return_sequences=False,
                 dropout=self.dropout_rate,
                 recurrent_dropout=self.dropout_rate),
            BatchNormalization(),
            
            # طبقات Dense
            Dense(32, activation='relu'),
            Dropout(self.dropout_rate),
            Dense(16, activation='relu'),
            Dropout(self.dropout_rate),
            
            # طبقة الإخراج
            Dense(num_classes, activation='softmax')
        ])
        
        model.compile(
            optimizer='adam',
            loss='sparse_categorical_crossentropy',
            metrics=['accuracy']
        )
        
        self.model = model
        return model
    
    def temporal_train_test_split(self, X, y, train_ratio=0.7, val_ratio=0.15):
        """
        تقسيم البيانات الزمنية بشكل تسلسلي (بدون عشوائية)
        """
        n_samples = len(X)
        
        # حساب نقاط التقسيم
        train_end = int(n_samples * train_ratio)
        val_end = int(n_samples * (train_ratio + val_ratio))
        
        # تقسيم البيانات تسلسليا
        X_train = X[:train_end]
        y_train = y[:train_end]
        
        X_val = X[train_end:val_end]
        y_val = y[train_end:val_end]
        
        X_test = X[val_end:]
        y_test = y[val_end:]
        
        return X_train, X_val, X_test, y_train, y_val, y_test
    
    def train(self, X_train, y_train, X_val, y_val, epochs=100, batch_size=32):
        """
        تدريب النموذج
        """
        # إعداد callbacks
        early_stopping = EarlyStopping(
            monitor='val_loss',
            patience=15,
            restore_best_weights=True
        )
        
        reduce_lr = ReduceLROnPlateau(
            monitor='val_loss',
            factor=0.2,
            patience=10,
            min_lr=0.0001
        )
        
        # تدريب النموذج
        history = self.model.fit(
            X_train, y_train,
            epochs=epochs,
            batch_size=batch_size,
            validation_data=(X_val, y_val),
            callbacks=[early_stopping, reduce_lr],
            verbose=1,
            shuffle=False  # مهم: عدم خلط البيانات الزمنية
        )
        
        return history
    
    def predict(self, X):
        """
        التنبؤ
        """
        predictions = self.model.predict(X)
        predicted_classes = np.argmax(predictions, axis=1)
        return self.label_encoder.inverse_transform(predicted_classes)
    
    def evaluate_model(self, X_test, y_test):
        """
        تقييم النموذج
        """
        predictions = self.predict(X_test)
        y_test_labels = self.label_encoder.inverse_transform(y_test)
        
        print("Classification Report:")
        print(classification_report(y_test_labels, predictions))
        
        # رسم confusion matrix
        cm = confusion_matrix(y_test_labels, predictions)
        plt.figure(figsize=(8, 6))
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
                   xticklabels=self.label_encoder.classes_,
                   yticklabels=self.label_encoder.classes_)
        plt.title('Confusion Matrix')
        plt.xlabel('Predicted')
        plt.ylabel('Actual')
        plt.show()
        
        return classification_report(y_test_labels, predictions, output_dict=True)
    
    def plot_training_history(self, history):
        """
        رسم تاريخ التدريب
        """
        fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))
        
        # Loss
        ax1.plot(history.history['loss'], label='Training Loss')
        ax1.plot(history.history['val_loss'], label='Validation Loss')
        ax1.set_title('Model Loss')
        ax1.set_xlabel('Epoch')
        ax1.set_ylabel('Loss')
        ax1.legend()
        
        # Accuracy
        ax2.plot(history.history['accuracy'], label='Training Accuracy')
        ax2.plot(history.history['val_accuracy'], label='Validation Accuracy')
        ax2.set_title('Model Accuracy')
        ax2.set_xlabel('Epoch')
        ax2.set_ylabel('Accuracy')
        ax2.legend()
        
    def plot_predictions_timeline(self, y_true, y_pred, start_idx=0, end_idx=200):
        """
        رسم الإشارات الفعلية مقابل المتوقعة على الخط الزمني
        """
        plt.figure(figsize=(15, 6))
        
        time_range = range(start_idx, min(end_idx, len(y_true)))
        
        plt.plot(time_range, y_true[start_idx:end_idx], 'b-', label='Actual Signal', alpha=0.7)
        plt.plot(time_range, y_pred[start_idx:end_idx], 'r--', label='Predicted Signal', alpha=0.7)
        
        plt.xlabel('Time Step')
        plt.ylabel('Signal Value')
        plt.title('Signal Prediction Timeline')
        plt.legend()
        plt.grid(True, alpha=0.3)
        plt.tight_layout()
        plt.show()
        
    def calculate_temporal_metrics(self, y_true, y_pred):
        """
        حساب مقاييس خاصة بالبيانات الزمنية
        """
        # دقة التتابع (accuracy of consecutive predictions)
        consecutive_accuracy = np.mean(y_true == y_pred)
        
        # دقة اتجاه التغيير
        true_changes = np.diff(y_true)
        pred_changes = np.diff(y_pred)
        direction_accuracy = np.mean(np.sign(true_changes) == np.sign(pred_changes))
        
        print(f"Consecutive Accuracy: {consecutive_accuracy:.4f}")
        print(f"Direction Change Accuracy: {direction_accuracy:.4f}")
        
        return {
            'consecutive_accuracy': consecutive_accuracy,
            'direction_accuracy': direction_accuracy
        }
    
    def analyze_feature_importance(self, df):
        """
        تحليل تأثير الميزات على الإشارات
        """
        # استخراج الميزات والهدف
        feature_columns = [col for col in df.columns if col not in ['Signal', 'Gmt time']]
        X = df[feature_columns].values
        y = df['Signal'].values
        
        # تطبيق التطبيع
        X_scaled = self.scaler.fit_transform(X)
        
        print("=" * 60)
        print("تحليل تأثير الميزات على الإشارات")
        print("=" * 60)
        
        # 1. تحليل الارتباط (Correlation Analysis)
        self._correlation_analysis(df, feature_columns)
        
        # 2. تحليل متوسط القيم حسب الإشارة
        self._signal_mean_analysis(df, feature_columns)
        
        # 3. Random Forest Feature Importance
        self._random_forest_importance(X_scaled, y, feature_columns)
        
        # 4. Permutation Importance
        self._permutation_importance_analysis(X_scaled, y, feature_columns)
        
        return feature_columns
    
    def _correlation_analysis(self, df, feature_columns):
        """
        تحليل الارتباط بين الميزات والإشارات
        """
        print("\n1. تحليل الارتباط مع الإشارات:")
        print("-" * 40)
        
        correlations = []
        for col in feature_columns:
            corr, p_value = pearsonr(df[col], df['Signal'])
            correlations.append({
                'Feature': col,
                'Correlation': corr,
                'P_Value': p_value,
                'Effect': 'إيجابي' if corr > 0 else 'سلبي',
                'Strength': abs(corr)
            })
        
        corr_df = pd.DataFrame(correlations).sort_values('Strength', ascending=False)
        
        # عرض أهم 10 ميزات
        print("أهم 10 ميزات حسب قوة الارتباط:")
        for _, row in corr_df.head(10).iterrows():
            print(f"{row['Feature']:<20} | الارتباط: {row['Correlation']:.4f} | التأثير: {row['Effect']}")
        
        # رسم الارتباطات
        plt.figure(figsize=(12, 8))
        top_features = corr_df.head(15)
        colors = ['green' if x > 0 else 'red' for x in top_features['Correlation']]
        
        plt.barh(range(len(top_features)), top_features['Correlation'], color=colors, alpha=0.7)
        plt.yticks(range(len(top_features)), top_features['Feature'])
        plt.xlabel('معامل الارتباط')
        plt.title('تأثير الميزات على الإشارات (الارتباط)')
        plt.axvline(x=0, color='black', linestyle='-', alpha=0.3)
        plt.grid(True, alpha=0.3)
        plt.tight_layout()
        plt.show()
        
        return corr_df
    
    def _signal_mean_analysis(self, df, feature_columns):
        """
        تحليل متوسط قيم الميزات حسب كل إشارة
        """
        print("\n2. تحليل متوسط القيم حسب الإشارة:")
        print("-" * 40)
        
        signal_analysis = []
        
        for col in feature_columns:
            means = df.groupby('Signal')[col].mean()
            
            # حساب الاتجاه العام
            if len(means) == 3:  # -1, 0, 1
                trend = 'إيجابي' if means[1] > means[-1] else 'سلبي'
                diff = abs(means[1] - means[-1])
            else:
                trend = 'غير محدد'
                diff = 0
            
            signal_analysis.append({
                'Feature': col,
                'Mean_Sell': means.get(-1, 0),
                'Mean_Hold': means.get(0, 0),
                'Mean_Buy': means.get(1, 0),
                'Trend': trend,
                'Difference': diff
            })
        
        analysis_df = pd.DataFrame(signal_analysis).sort_values('Difference', ascending=False)
        
        print("أهم 10 ميزات حسب الفرق بين إشارات البيع والشراء:")
        for _, row in analysis_df.head(10).iterrows():
            print(f"{row['Feature']:<20} | البيع: {row['Mean_Sell']:.4f} | الشراء: {row['Mean_Buy']:.4f} | الاتجاه: {row['Trend']}")
        
        return analysis_df
    
    def _random_forest_importance(self, X, y, feature_columns):
        """
        حساب أهمية الميزات باستخدام Random Forest
        """
        print("\n3. أهمية الميزات - Random Forest:")
        print("-" * 40)
        
        # تدريب Random Forest
        rf = RandomForestClassifier(n_estimators=100, random_state=42)
        rf.fit(X, y)
        
        # استخراج أهمية الميزات
        importances = rf.feature_importances_
        
        # إنشاء DataFrame للنتائج
        rf_importance = pd.DataFrame({
            'Feature': feature_columns,
            'Importance': importances
        }).sort_values('Importance', ascending=False)
        
        print("أهم 10 ميزات حسب Random Forest:")
        for _, row in rf_importance.head(10).iterrows():
            print(f"{row['Feature']:<20} | الأهمية: {row['Importance']:.4f}")
        
        # رسم أهمية الميزات
        plt.figure(figsize=(12, 8))
        top_features = rf_importance.head(15)
        plt.barh(range(len(top_features)), top_features['Importance'], color='skyblue', alpha=0.7)
        plt.yticks(range(len(top_features)), top_features['Feature'])
        plt.xlabel('أهمية الميزة')
        plt.title('أهمية الميزات - Random Forest')
        plt.grid(True, alpha=0.3)
        plt.tight_layout()
        plt.show()
        
        return rf_importance
    
    def _permutation_importance_analysis(self, X, y, feature_columns):
        """
        تحليل أهمية الميزات باستخدام Permutation Importance
        """
        print("\n4. أهمية الميزات - Permutation Importance:")
        print("-" * 40)
        
        # تدريب نموذج بسيط للتحليل
        rf = RandomForestClassifier(n_estimators=50, random_state=42)
        rf.fit(X, y)
        
        # حساب Permutation Importance
        perm_importance = permutation_importance(rf, X, y, n_repeats=5, random_state=42)
        
        # إنشاء DataFrame للنتائج
        perm_df = pd.DataFrame({
            'Feature': feature_columns,
            'Importance_Mean': perm_importance.importances_mean,
            'Importance_Std': perm_importance.importances_std
        }).sort_values('Importance_Mean', ascending=False)
        
        print("أهم 10 ميزات حسب Permutation Importance:")
        for _, row in perm_df.head(10).iterrows():
            print(f"{row['Feature']:<20} | الأهمية: {row['Importance_Mean']:.4f} ± {row['Importance_Std']:.4f}")
        
        return perm_df
    
    def plot_feature_effects_summary(self, df):
        """
        رسم ملخص شامل لتأثيرات الميزات
        """
        feature_columns = [col for col in df.columns if col not in ['Signal', 'Gmt time']]
        
        # حساب الارتباطات
        correlations = []
        for col in feature_columns:
            corr, _ = pearsonr(df[col], df['Signal'])
            correlations.append(corr)
        
        # إنشاء DataFrame
        effects_df = pd.DataFrame({
            'Feature': feature_columns,
            'Correlation': correlations,
            'Effect_Type': ['إيجابي' if x > 0 else 'سلبي' for x in correlations],
            'Effect_Strength': [abs(x) for x in correlations]
        }).sort_values('Effect_Strength', ascending=True)
        
        # رسم شامل
        fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 10))
        
        # الرسم الأول: جميع الميزات
        colors = ['green' if x > 0 else 'red' for x in effects_df['Correlation']]
        ax1.barh(range(len(effects_df)), effects_df['Correlation'], color=colors, alpha=0.7)
        ax1.set_yticks(range(len(effects_df)))
        ax1.set_yticklabels(effects_df['Feature'], fontsize=8)
        ax1.set_xlabel('معامل الارتباط')
        ax1.set_title('تأثير جميع الميزات على الإشارات')
        ax1.axvline(x=0, color='black', linestyle='-', alpha=0.3)
        ax1.grid(True, alpha=0.3)
        
        # الرسم الثاني: أقوى الميزات
        top_effects = effects_df.tail(10)  # أقوى 10 ميزات
        colors2 = ['green' if x > 0 else 'red' for x in top_effects['Correlation']]
        ax2.barh(range(len(top_effects)), top_effects['Correlation'], color=colors2, alpha=0.7)
        ax2.set_yticks(range(len(top_effects)))
        ax2.set_yticklabels(top_effects['Feature'])
        ax2.set_xlabel('معامل الارتباط')
        ax2.set_title('أقوى 10 ميزات تأثيراً')
        ax2.axvline(x=0, color='black', linestyle='-', alpha=0.3)
        ax2.grid(True, alpha=0.3)
        
        plt.tight_layout()
        plt.show()
        
        return effects_df

# مثال على الاستخدام مع بياناتك الحقيقية
def main():
    """
    استخدام النموذج مع بياناتك الحقيقية
    """
    # تحميل البيانات (استبدل بمسار ملفك)
    df = df_new_4.copy()
    
    # التأكد من الترتيب الزمني
    if 'Gmt time' in df.columns:
        df['Gmt time'] = pd.to_datetime(df['Gmt time'])
        df = df.sort_values('Gmt time').reset_index(drop=True)
    
    print(f"Dataset shape: {df.shape}")
    print(f"Signal distribution:\n{df['Signal'].value_counts()}")
    print(f"Feature columns: {[col for col in df.columns if col not in ['Signal', 'Gmt time']]}")
    
    # تحليل تأثير الميزات
    print("\n" + "="*60)
    print("تحليل تأثير الميزات")
    print("="*60)
    
    classifier = LSTMSignalClassifier(window_size=60, lstm_units=50)
    
    # تحليل الميزات
    feature_columns = classifier.analyze_feature_importance(df)
    
    # رسم ملخص التأثيرات
    effects_summary = classifier.plot_feature_effects_summary(df)
    
    print("\n" + "="*60)
    print("ملخص التأثيرات:")
    print("="*60)
    print("الميزات ذات التأثير الإيجابي (تشجع على الشراء):")
    positive_features = effects_summary[effects_summary['Correlation'] > 0].sort_values('Correlation', ascending=False)
    for _, row in positive_features.head(5).iterrows():
        print(f"  • {row['Feature']}: {row['Correlation']:.4f}")
    
    print("\nالميزات ذات التأثير السلبي (تشجع على البيع):")
    negative_features = effects_summary[effects_summary['Correlation'] < 0].sort_values('Correlation', ascending=True)
    for _, row in negative_features.head(5).iterrows():
        print(f"  • {row['Feature']}: {row['Correlation']:.4f}")
    
    print("\n" + "="*60)
    print("بدء تدريب نموذج LSTM")
    print("="*60)
    
    # تحضير البيانات
    X, y = classifier.prepare_data(df)
    
    print(f"Shape of X after windowing: {X.shape}")
    print(f"Shape of y after windowing: {y.shape}")
    print(f"Unique classes: {classifier.label_encoder.classes_}")
    
    # تقسيم البيانات تسلسليا (بدون عشوائية)
    X_train, X_val, X_test, y_train, y_val, y_test = classifier.temporal_train_test_split(
        X, y, train_ratio=0.7, val_ratio=0.15
    )
    
    print(f"Training set: {X_train.shape}")
    print(f"Validation set: {X_val.shape}")
    print(f"Test set: {X_test.shape}")
    
    # بناء النموذج
    model = classifier.build_model(
        input_shape=(X_train.shape[1], X_train.shape[2]),
        num_classes=len(np.unique(y))
    )
    
    print(model.summary())
    
    # تدريب النموذج
    history = classifier.train(X_train, y_train, X_val, y_val, epochs=100)
    
    # رسم تاريخ التدريب
    classifier.plot_training_history(history)
    
    # تقييم النموذج
    results = classifier.evaluate_model(X_test, y_test)
    
    # تقييم إضافي للبيانات الزمنية
    y_pred = classifier.predict(X_test)
    y_true_labels = classifier.label_encoder.inverse_transform(y_test)
    
    # حساب مقاييس زمنية إضافية
    temporal_metrics = classifier.calculate_temporal_metrics(y_true_labels, y_pred)
    
    # رسم الخط الزمني للتنبؤات
    classifier.plot_predictions_timeline(y_true_labels, y_pred, start_idx=0, end_idx=100)
    
    # حفظ النموذج
    classifier.model.save('lstm_signal_classifier.h5')
    print("Model saved as 'lstm_signal_classifier.h5'")
    
    return classifier, results, temporal_metrics

if __name__ == "__main__":
    classifier, results, temporal_metrics = main()
```

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import classification_report, confusion_matrix
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout, BatchNormalization
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from tensorflow.keras.utils import to_categorical
import matplotlib.pyplot as plt
import seaborn as sns

class LSTMSignalClassifier:
    def __init__(self, window_size=40, lstm_units=50, dropout_rate=0.2):
        """
        LSTM Model للتصنيف
        
        Parameters:
        window_size: حجم النافذة المنزلقة
        lstm_units: عدد وحدات LSTM
        dropout_rate: معدل الـ dropout
        """
        self.window_size = window_size
        self.lstm_units = lstm_units
        self.dropout_rate = dropout_rate
        self.model = None
        self.scaler = StandardScaler()
        self.label_encoder = LabelEncoder()
        
    def prepare_data(self, df):
        """
        تحضير البيانات مع النافذة المنزلقة
        """
        # فصل الميزات عن الهدف (استبعاد عمود الوقت والهدف)
        feature_columns = [col for col in df.columns if col not in ['Signal', 'Gmt time']]
        X = df[feature_columns].values
        y = df['Signal'].values
        
        # تطبيق التطبيع على الميزات
        X_scaled = self.scaler.fit_transform(X)
        
        # تطبيق Label Encoding على الهدف
        y_encoded = self.label_encoder.fit_transform(y)
        
        # إنشاء النافذة المنزلقة
        X_windowed, y_windowed = self.create_sliding_window(X_scaled, y_encoded)
        
        return X_windowed, y_windowed
    
    def create_sliding_window(self, X, y):
        """
        إنشاء النافذة المنزلقة للبيانات الزمنية
        """
        X_windowed = []
        y_windowed = []
        
        for i in range(self.window_size, len(X)):
            X_windowed.append(X[i-self.window_size:i])
            y_windowed.append(y[i])
        
        return np.array(X_windowed), np.array(y_windowed)
    
    def build_model(self, input_shape, num_classes):
        """
        بناء نموذج LSTM
        """
        model = Sequential([
            # الطبقة الأولى LSTM
            LSTM(self.lstm_units, 
                 return_sequences=True, 
                 input_shape=input_shape,
                 #dropout=self.dropout_rate,
                 #recurrent_dropout=self.dropout_rate
                ),
            BatchNormalization(),
            
            # الطبقة الثانية LSTM
            LSTM(self.lstm_units//2, 
                 return_sequences=False #,
                 #dropout=self.dropout_rate,
                 #recurrent_dropout=self.dropout_rate
                ),
            BatchNormalization(),
            
            # طبقات Dense
            Dense(32, activation='relu'),
            #Dropout(self.dropout_rate),
            Dense(16, activation='relu'),
            #Dropout(self.dropout_rate),
            
            # طبقة الإخراج
            Dense(num_classes, activation='softmax')
        ])
        
        model.compile(
            optimizer='adam',
            loss='sparse_categorical_crossentropy',
            metrics=['accuracy']
        )
        
        self.model = model
        return model
    
    def temporal_train_test_split(self, X, y, train_ratio=0.7, val_ratio=0.15):
        """
        تقسيم البيانات الزمنية بشكل تسلسلي (بدون عشوائية)
        """
        n_samples = len(X)
        
        # حساب نقاط التقسيم
        train_end = int(n_samples * train_ratio)
        val_end = int(n_samples * (train_ratio + val_ratio))
        
        # تقسيم البيانات تسلسليا
        X_train = X[:train_end]
        y_train = y[:train_end]
        
        X_val = X[train_end:val_end]
        y_val = y[train_end:val_end]
        
        X_test = X[val_end:]
        y_test = y[val_end:]
        
        return X_train, X_val, X_test, y_train, y_val, y_test
    
    def train(self, X_train, y_train, X_val, y_val, epochs=25, batch_size=32):
        """
        تدريب النموذج
        """
        # إعداد callbacks
        early_stopping = EarlyStopping(
            monitor='val_loss',
            patience=5,
            restore_best_weights=True
        )
        
        reduce_lr = ReduceLROnPlateau(
            monitor='val_loss',
            factor=0.2,
            patience=10,
            min_lr=0.0001
        )
        
        # تدريب النموذج
        history = self.model.fit(
            X_train, y_train,
            epochs=epochs,
            batch_size=batch_size,
            validation_data=(X_val, y_val),
            callbacks=[early_stopping, reduce_lr],
            verbose=1,
            shuffle=False  # مهم: عدم خلط البيانات الزمنية
        )
        
        return history
    
    def predict(self, X):
        """
        التنبؤ
        """
        predictions = self.model.predict(X)
        predicted_classes = np.argmax(predictions, axis=1)
        return self.label_encoder.inverse_transform(predicted_classes)
    
    def evaluate_model(self, X_test, y_test):
        """
        تقييم النموذج
        """
        predictions = self.predict(X_test)
        y_test_labels = self.label_encoder.inverse_transform(y_test)
        
        print("Classification Report:")
        print(classification_report(y_test_labels, predictions))
        
        # رسم confusion matrix
        cm = confusion_matrix(y_test_labels, predictions)
        plt.figure(figsize=(8, 6))
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
                   xticklabels=self.label_encoder.classes_,
                   yticklabels=self.label_encoder.classes_)
        plt.title('Confusion Matrix')
        plt.xlabel('Predicted')
        plt.ylabel('Actual')
        plt.show()
        
        return classification_report(y_test_labels, predictions, output_dict=True)
    
    def plot_training_history(self, history):
        """
        رسم تاريخ التدريب
        """
        fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))
        
        # Loss
        ax1.plot(history.history['loss'], label='Training Loss')
        ax1.plot(history.history['val_loss'], label='Validation Loss')
        ax1.set_title('Model Loss')
        ax1.set_xlabel('Epoch')
        ax1.set_ylabel('Loss')
        ax1.legend()
        
        # Accuracy
        ax2.plot(history.history['accuracy'], label='Training Accuracy')
        ax2.plot(history.history['val_accuracy'], label='Validation Accuracy')
        ax2.set_title('Model Accuracy')
        ax2.set_xlabel('Epoch')
        ax2.set_ylabel('Accuracy')
        ax2.legend()
        
    def plot_predictions_timeline(self, y_true, y_pred, start_idx=0, end_idx=200):
        """
        رسم الإشارات الفعلية مقابل المتوقعة على الخط الزمني
        """
        plt.figure(figsize=(15, 6))
        
        time_range = range(start_idx, min(end_idx, len(y_true)))
        
        plt.plot(time_range, y_true[start_idx:end_idx], 'b-', label='Actual Signal', alpha=0.7)
        plt.plot(time_range, y_pred[start_idx:end_idx], 'r--', label='Predicted Signal', alpha=0.7)
        
        plt.xlabel('Time Step')
        plt.ylabel('Signal Value')
        plt.title('Signal Prediction Timeline')
        plt.legend()
        plt.grid(True, alpha=0.3)
        plt.tight_layout()
        plt.show()
        
    def calculate_temporal_metrics(self, y_true, y_pred):
        """
        حساب مقاييس خاصة بالبيانات الزمنية
        """
        # دقة التتابع (accuracy of consecutive predictions)
        consecutive_accuracy = np.mean(y_true == y_pred)
        
        # دقة اتجاه التغيير
        true_changes = np.diff(y_true)
        pred_changes = np.diff(y_pred)
        direction_accuracy = np.mean(np.sign(true_changes) == np.sign(pred_changes))
        
        print(f"Consecutive Accuracy: {consecutive_accuracy:.4f}")
        print(f"Direction Change Accuracy: {direction_accuracy:.4f}")
        
        return {
            'consecutive_accuracy': consecutive_accuracy,
            'direction_accuracy': direction_accuracy
        }

# مثال على الاستخدام مع بياناتك الحقيقية
def main():
    """
    استخدام النموذج مع بياناتك الحقيقية
    """
    # تحميل البيانات (استبدل بمسار ملفك)
    df = df_new_4.copy()
    
    # التأكد من الترتيب الزمني
    if 'Gmt time' in df.columns:
        df['Gmt time'] = pd.to_datetime(df['Gmt time'])
        df = df.sort_values('Gmt time').reset_index(drop=True)
    
    print(f"Dataset shape: {df.shape}")
    print(f"Signal distribution:\n{df['Signal'].value_counts()}")
    print(f"Feature columns: {[col for col in df.columns if col not in ['Signal', 'Gmt time','Open','High','Low','Close']]}")
    
    # إنشاء وتدريب النموذج
    classifier = LSTMSignalClassifier(window_size=96, lstm_units=50)
    
    # تحضير البيانات
    X, y = classifier.prepare_data(df)
    
    print(f"Shape of X after windowing: {X.shape}")
    print(f"Shape of y after windowing: {y.shape}")
    print(f"Unique classes: {classifier.label_encoder.classes_}")
    
    # تقسيم البيانات تسلسليا (بدون عشوائية)
    X_train, X_val, X_test, y_train, y_val, y_test = classifier.temporal_train_test_split(
        X, y, train_ratio=0.7, val_ratio=0.15
    )
    
    print(f"Training set: {X_train.shape}")
    print(f"Validation set: {X_val.shape}")
    print(f"Test set: {X_test.shape}")
    
    # بناء النموذج
    model = classifier.build_model(
        input_shape=(X_train.shape[1], X_train.shape[2]),
        num_classes=len(np.unique(y))
    )
    
    print(model.summary())
    
    # تدريب النموذج
    history = classifier.train(X_train, y_train, X_val, y_val, epochs=25)
    
    # رسم تاريخ التدريب
    classifier.plot_training_history(history)
    
    # تقييم النموذج
    results = classifier.evaluate_model(X_test, y_test)
    
    # تقييم إضافي للبيانات الزمنية
    y_pred = classifier.predict(X_test)
    y_true_labels = classifier.label_encoder.inverse_transform(y_test)
    
    # حساب مقاييس زمنية إضافية
    temporal_metrics = classifier.calculate_temporal_metrics(y_true_labels, y_pred)
    
    # رسم الخط الزمني للتنبؤات
    classifier.plot_predictions_timeline(y_true_labels, y_pred, start_idx=0, end_idx=100)
    
    # حفظ النموذج
    classifier.model.save('lstm_signal_classifier_1.h5')
    print("Model saved as 'lstm_signal_classifier_1.h5'")
    
    return classifier, results, temporal_metrics

if __name__ == "__main__":
    classifier, results, temporal_metrics = main()
```

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.utils.class_weight import compute_class_weight
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout, BatchNormalization
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from tensorflow.keras.utils import to_categorical
import matplotlib.pyplot as plt
import seaborn as sns

class LSTMSignalClassifier:
    def __init__(self, window_size=96, lstm_units=50, dropout_rate=0.2):
        """
        LSTM Model للتصنيف مع توزيع متوازن للأوزان
        
        Parameters:
        window_size: حجم النافذة المنزلقة
        lstm_units: عدد وحدات LSTM
        dropout_rate: معدل الـ dropout
        """
        self.window_size = window_size
        self.lstm_units = lstm_units
        self.dropout_rate = dropout_rate
        self.model = None
        self.scaler = StandardScaler()
        self.label_encoder = LabelEncoder()
        self.class_weights = None
        
    def prepare_data(self, df):
        """
        تحضير البيانات مع النافذة المنزلقة والتعامل مع الأوزان
        """
        # فصل الميزات عن الهدف (استبعاد عمود الوقت والهدف)
        feature_columns = [col for col in df.columns if col not in ['Signal', 'Gmt time']]
        X = df[feature_columns].values
        y = df['Signal'].values
        
        # تطبيق التطبيع على الميزات
        X_scaled = self.scaler.fit_transform(X)
        
        # إنشاء النافذة المنزلقة
        X_windowed, y_windowed = self.create_sliding_window(X_scaled, y)
        
        # حساب الأوزان للفئات 1 و -1 فقط
        self.calculate_balanced_weights(y_windowed)
        
        return X_windowed, y_windowed
    
    def create_sliding_window(self, X, y):
        """
        إنشاء النافذة المنزلقة للبيانات الزمنية
        """
        X_windowed = []
        y_windowed = []
        
        for i in range(self.window_size, len(X)):
            X_windowed.append(X[i-self.window_size:i])
            y_windowed.append(y[i])
        
        return np.array(X_windowed), np.array(y_windowed)
    
    def calculate_balanced_weights(self, y):
        """
        حساب الأوزان المتوازنة للفئات 1 و -1 فقط
        """
        # حساب عدد الحالات لكل فئة
        unique_values, counts = np.unique(y, return_counts=True)
        print(f"Signal distribution in windowed data:")
        for val, count in zip(unique_values, counts):
            print(f"  {val}: {count} samples")
        
        # حساب الأوزان فقط للفئات 1 و -1
        target_classes = [1, -1]
        
        # فلترة البيانات للحصول على 1 و -1 فقط
        mask_1_neg1 = np.isin(y, target_classes)
        y_filtered = y[mask_1_neg1]
        
        if len(y_filtered) == 0:
            print("Warning: No samples with values 1 or -1 found!")
            self.class_weights = {}
            return
        
        # حساب الأوزان المتوازنة
        classes = np.array(target_classes)
        weights = compute_class_weight('balanced', classes=classes, y=y_filtered)
        
        # إنشاء dictionary للأوزان (تضمين جميع الفئات الموجودة)
        self.class_weights = {}
        
        # أوزان للفئات المستهدفة (1 و -1)
        for i, cls in enumerate(target_classes):
            self.class_weights[cls] = weights[i]
        
        # وزن صفر للفئة 0 (لتجاهلها في التدريب)
        if 0 in unique_values:
            self.class_weights[0] = 0.0
        
        print(f"Class weights calculated:")
        for cls, weight in self.class_weights.items():
            print(f"  Class {cls}: {weight:.4f}")
    
    def create_sample_weights(self, y):
        """
        إنشاء أوزان العينات بناءً على أوزان الفئات
        """
        sample_weights = np.zeros_like(y, dtype=float)
        
        for cls, weight in self.class_weights.items():
            sample_weights[y == cls] = weight
        
        return sample_weights
    
    def filter_non_zero_data(self, X, y):
        """
        فلترة البيانات للاحتفاظ بالفئات 1 و -1 فقط للتقييم
        """
        mask = np.isin(y, [1, -1])
        return X[mask], y[mask]
    
    def build_model(self, input_shape, num_classes=2):
        """
        بناء نموذج LSTM (فقط للفئات 1 و -1)
        """
        model = Sequential([
            # الطبقة الأولى LSTM
            LSTM(self.lstm_units, 
                 return_sequences=True, 
                 input_shape=input_shape),
            BatchNormalization(),
            
            # الطبقة الثانية LSTM
            LSTM(self.lstm_units//2, 
                 return_sequences=False),
            BatchNormalization(),
            
            # طبقات Dense
            Dense(32, activation='relu'),
            Dropout(self.dropout_rate),
            Dense(16, activation='relu'),
            
            # طبقة الإخراج (2 فئات فقط: -1 و 1)
            Dense(num_classes, activation='softmax')
        ])
        
        model.compile(
            optimizer='adam',
            loss='sparse_categorical_crossentropy',
            metrics=['accuracy']
        )
        
        self.model = model
        return model
    
    def temporal_train_test_split(self, X, y, train_ratio=0.7, val_ratio=0.15):
        """
        تقسيم البيانات الزمنية بشكل تسلسلي
        """
        n_samples = len(X)
        
        # حساب نقاط التقسيم
        train_end = int(n_samples * train_ratio)
        val_end = int(n_samples * (train_ratio + val_ratio))
        
        # تقسيم البيانات تسلسليا
        X_train = X[:train_end]
        y_train = y[:train_end]
        
        X_val = X[train_end:val_end]
        y_val = y[train_end:val_end]
        
        X_test = X[val_end:]
        y_test = y[val_end:]
        
        return X_train, X_val, X_test, y_train, y_val, y_test
    
    def prepare_labels_for_training(self, y):
        """
        تحويل التسميات للتدريب (تحويل -1 إلى 0، و 1 إلى 1)
        """
        y_train_labels = np.copy(y)
        y_train_labels[y == -1] = 0  # -1 becomes class 0
        y_train_labels[y == 1] = 1   # 1 remains class 1
        y_train_labels[y == 0] = -999  # علامة للتجاهل
        
        return y_train_labels
    
    def train(self, X_train, y_train, X_val, y_val, epochs=10, batch_size=100):
        """
        تدريب النموذج مع الأوزان المتوازنة
        """
        # تحضير التسميات للتدريب
        y_train_encoded = self.prepare_labels_for_training(y_train)
        y_val_encoded = self.prepare_labels_for_training(y_val)
        
        # إنشاء أوزان العينات للتدريب
        sample_weights_train = self.create_sample_weights(y_train)
        sample_weights_val = self.create_sample_weights(y_val)
        
        # فلترة البيانات للتدريب (إزالة الفئة 0)
        train_mask = y_train_encoded != -999
        val_mask = y_val_encoded != -999
        
        X_train_filtered = X_train[train_mask]
        y_train_filtered = y_train_encoded[train_mask]
        sample_weights_train_filtered = sample_weights_train[train_mask]
        
        X_val_filtered = X_val[val_mask]
        y_val_filtered = y_val_encoded[val_mask]
        sample_weights_val_filtered = sample_weights_val[val_mask]
        
        print(f"Training data after filtering:")
        print(f"  X_train shape: {X_train_filtered.shape}")
        print(f"  y_train distribution: {np.bincount(y_train_filtered.astype(int))}")
        print(f"  X_val shape: {X_val_filtered.shape}")
        print(f"  y_val distribution: {np.bincount(y_val_filtered.astype(int))}")
        
        # إعداد callbacks
        early_stopping = EarlyStopping(
            monitor='val_loss',
            patience=5,
            restore_best_weights=True
        )
        
        reduce_lr = ReduceLROnPlateau(
            monitor='val_loss',
            factor=0.2,
            patience=10,
            min_lr=0.0001
        )
        
        # تدريب النموذج مع الأوزان
        history = self.model.fit(
            X_train_filtered, y_train_filtered,
            epochs=epochs,
            batch_size=batch_size,
            validation_data=(X_val_filtered, y_val_filtered, sample_weights_val_filtered),
            sample_weight=sample_weights_train_filtered,
            callbacks=[early_stopping, reduce_lr],
            verbose=1,
            shuffle=False
        )
        
        return history
    
    def predict(self, X, y_original=None):
        """
        التنبؤ (مع فلترة للفئات 1 و -1 فقط)
        """
        if y_original is not None:
            # فلترة البيانات للفئات 1 و -1 فقط
            X_filtered, y_filtered = self.filter_non_zero_data(X, y_original)
            if len(X_filtered) == 0:
                return np.array([]), np.array([])
                
            predictions = self.model.predict(X_filtered)
            predicted_classes = np.argmax(predictions, axis=1)
            
            # تحويل النتائج إلى التسميات الأصلية
            predicted_labels = np.copy(predicted_classes)
            predicted_labels[predicted_classes == 0] = -1  # class 0 -> -1
            predicted_labels[predicted_classes == 1] = 1   # class 1 -> 1
            
            return predicted_labels, y_filtered
        else:
            predictions = self.model.predict(X)
            predicted_classes = np.argmax(predictions, axis=1)
            predicted_labels = np.copy(predicted_classes)
            predicted_labels[predicted_classes == 0] = -1
            predicted_labels[predicted_classes == 1] = 1
            return predicted_labels
    
    def evaluate_model(self, X_test, y_test):
        """
        تقييم النموذج (فقط على الفئات 1 و -1)
        """
        predictions, y_test_filtered = self.predict(X_test, y_test)
        
        if len(predictions) == 0:
            print("No samples with classes 1 or -1 found in test set!")
            return {}
        
        print(f"Test set after filtering: {len(y_test_filtered)} samples")
        print(f"Test distribution: {np.bincount(y_test_filtered, minlength=3)}")
        
        print("Classification Report:")
        print(classification_report(y_test_filtered, predictions, target_names=['Class -1', 'Class 1']))
        
        # رسم confusion matrix
        cm = confusion_matrix(y_test_filtered, predictions)
        plt.figure(figsize=(8, 6))
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
                   xticklabels=['Predicted -1', 'Predicted 1'],
                   yticklabels=['Actual -1', 'Actual 1'])
        plt.title('Confusion Matrix (Classes -1 and 1 only)')
        plt.xlabel('Predicted')
        plt.ylabel('Actual')
        plt.show()
        
        return classification_report(y_test_filtered, predictions, output_dict=True)
    
    def plot_training_history(self, history):
        """
        رسم تاريخ التدريب
        """
        fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))
        
        # Loss
        ax1.plot(history.history['loss'], label='Training Loss')
        ax1.plot(history.history['val_loss'], label='Validation Loss')
        ax1.set_title('Model Loss')
        ax1.set_xlabel('Epoch')
        ax1.set_ylabel('Loss')
        ax1.legend()
        
        # Accuracy
        ax2.plot(history.history['accuracy'], label='Training Accuracy')
        ax2.plot(history.history['val_accuracy'], label='Validation Accuracy')
        ax2.set_title('Model Accuracy')
        ax2.set_xlabel('Epoch')
        ax2.set_ylabel('Accuracy')
        ax2.legend()
        
        plt.tight_layout()
        plt.show()
    
    def plot_predictions_timeline(self, y_true, y_pred, start_idx=0, end_idx=200):
        """
        رسم الإشارات الفعلية مقابل المتوقعة على الخط الزمني
        """
        plt.figure(figsize=(15, 6))
        
        time_range = range(start_idx, min(end_idx, len(y_true)))
        
        plt.plot(time_range, y_true[start_idx:end_idx], 'b-', label='Actual Signal', alpha=0.7, linewidth=2)
        plt.plot(time_range, y_pred[start_idx:end_idx], 'r--', label='Predicted Signal', alpha=0.7, linewidth=2)
        
        plt.xlabel('Time Step')
        plt.ylabel('Signal Value')
        plt.title('Signal Prediction Timeline (Classes -1 and 1 only)')
        plt.legend()
        plt.grid(True, alpha=0.3)
        plt.tight_layout()
        plt.show()
        
    def calculate_temporal_metrics(self, y_true, y_pred):
        """
        حساب مقاييس خاصة بالبيانات الزمنية
        """
        # دقة التتابع
        consecutive_accuracy = np.mean(y_true == y_pred)
        
        # دقة اتجاه التغيير
        if len(y_true) > 1:
            true_changes = np.diff(y_true)
            pred_changes = np.diff(y_pred)
            # تجنب القسمة على صفر
            valid_changes = (true_changes != 0) | (pred_changes != 0)
            if np.any(valid_changes):
                direction_accuracy = np.mean(np.sign(true_changes[valid_changes]) == np.sign(pred_changes[valid_changes]))
            else:
                direction_accuracy = 0.0
        else:
            direction_accuracy = 0.0
        
        print(f"Consecutive Accuracy: {consecutive_accuracy:.4f}")
        print(f"Direction Change Accuracy: {direction_accuracy:.4f}")
        
        return {
            'consecutive_accuracy': consecutive_accuracy,
            'direction_accuracy': direction_accuracy
        }

# مثال على الاستخدام مع بياناتك الحقيقية
def main():
    """
    استخدام النموذج مع بياناتك الحقيقية
    """
    # تحميل البيانات (استبدل بمسار ملفك)
    df = df_new_4.copy()
    
    # التأكد من الترتيب الزمني
    if 'Gmt time' in df.columns:
        df['Gmt time'] = pd.to_datetime(df['Gmt time'])
        df = df.sort_values('Gmt time').reset_index(drop=True)
    
    print(f"Dataset shape: {df.shape}")
    print(f"Signal distribution:\n{df['Signal'].value_counts()}")
    print(f"Feature columns: {[col for col in df.columns if col not in ['Signal', 'Gmt time']]}")
    
    # إنشاء وتدريب النموذج
    classifier = LSTMSignalClassifier(window_size=96, lstm_units=50)
    
    # تحضير البيانات
    X, y = classifier.prepare_data(df)
    
    print(f"Shape of X after windowing: {X.shape}")
    print(f"Shape of y after windowing: {y.shape}")
    
    # تقسيم البيانات تسلسليا
    X_train, X_val, X_test, y_train, y_val, y_test = classifier.temporal_train_test_split(
        X, y, train_ratio=0.7, val_ratio=0.15
    )
    
    print(f"Training set: {X_train.shape}")
    print(f"Validation set: {X_val.shape}")
    print(f"Test set: {X_test.shape}")
    
    # بناء النموذج
    model = classifier.build_model(
        input_shape=(X_train.shape[1], X_train.shape[2]),
        num_classes=2  # فقط فئتان: -1 (->0) و 1 (->1)
    )
    
    print(model.summary())
    
    # تدريب النموذج
    history = classifier.train(X_train, y_train, X_val, y_val, epochs=10)
    
    # رسم تاريخ التدريب
    classifier.plot_training_history(history)
    
    # تقييم النموذج
    results = classifier.evaluate_model(X_test, y_test)
    
    # تقييم إضافي للبيانات الزمنية
    y_pred, y_true_filtered = classifier.predict(X_test, y_test)
    
    if len(y_pred) > 0:
        # حساب مقاييس زمنية إضافية
        temporal_metrics = classifier.calculate_temporal_metrics(y_true_filtered, y_pred)
        
        # رسم الخط الزمني للتنبؤات
        classifier.plot_predictions_timeline(y_true_filtered, y_pred, start_idx=0, end_idx=100)
    else:
        temporal_metrics = {}
        print("No predictions to evaluate!")
    
    # حفظ النموذج
    classifier.model.save('lstm_signal_classifier_balanced.h5')
    print("Model saved as 'lstm_signal_classifier_balanced.h5'")
    
    return classifier, results, temporal_metrics

if __name__ == "__main__":
    classifier, results, temporal_metrics = main()
```
